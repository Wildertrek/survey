## Enneagram Model (EM)<a class="anchor" id="PTMD_page_26"></a>

[Back to Top](#PTMD_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">


**The Enneagram Model (EM)** is a holistic and integrative personality model that categorizes individuals into nine distinct personality types, each characterized by a unique set of traits, motivations, and behaviors. These types are interconnected through adjacencies, which influence how individuals may exhibit traits from neighboring types. Here is a summary of the key aspects of the Enneagram Model:

**1. Types:** The Enneagram Model consists of nine primary personality types, each with a specific name and a set of factors that define them:
   - *Type 1 - Perfectionist:* Known for perfectionism, resentment, and anger.
   - *Type 2 - Helper:* Characterized by helpfulness, the need for appreciation, and pride.
   - *Type 3 - Achiever:* Focused on achievement orientation, adaptability, and vanity.
   - *Type 4 - Individualist:* Defined by emotional sensitivity, individualism, and envy.
   - *Type 5 - Investigator:* Characterized by intellectual curiosity, observation, and avarice.
   - *Type 6 - Loyalist:* Known for loyalty, anxiety, and suspicion.
   - *Type 7 - Enthusiast:* Characterized by enthusiasm, planning, and gluttony.
   - *Type 8 - Challenger:* Focused on assertiveness, ego, and lust.
   - *Type 9 - Peacemaker:* Characterized by perseverance, skepticism, and peacefulness.

**2. Factors:** Each personality type is associated with specific factors, which represent key traits or characteristics. These factors are further described by adjectives:
   - For example, the factor "Perfectionism" associated with Type 1 is described by adjectives like "Detail-Oriented," "Disciplined," and "Inflexible."

**3. Adjacencies:** The Enneagram Model incorporates the concept of adjacencies, where each personality type is connected to three neighboring types. These adjacencies influence how individuals may exhibit traits from neighboring types and can provide insights into their behavior.

The Enneagram Model serves as a comprehensive tool for understanding personality dynamics and personal development. It is often used in various contexts, including psychology, counseling, and self-improvement, to help individuals gain insight into their motivations, behaviors, and areas for growth. Additionally, it provides a framework for understanding how individuals of different types interact and relate to one another, promoting better communication and empathy in relationships.

The Enneagram Model (EM) does not have a single author or a single definitive source. It is a complex and multifaceted system that has evolved over centuries, with contributions from various cultures, spiritual traditions, and contemporary psychologists and authors. 

The Enneagram is rooted in ancient wisdom and has been influenced by multiple religious and philosophical traditions, including Christianity, Sufism, and more. It gained popularity in the West through the works of various authors and teachers who adapted and synthesized its teachings for modern contexts.

Some influential figures and authors associated with the development and popularization of the Enneagram in the Western world include:

1. **George Ivanovich Gurdjieff:** A mystic and spiritual teacher who is often credited with introducing the Enneagram symbol to the West in the early 20th century. His teachings formed the basis for much of the modern Enneagram system.

2. **Oscar Ichazo:** A Chilean teacher who developed his own interpretation of the Enneagram in the mid-20th century. His work laid the foundation for the psychological and personality-focused Enneagram system.

3. **Claudio Naranjo:** A psychiatrist and author who integrated the Enneagram into his psychotherapeutic work. He is known for adapting the Enneagram to psychological and therapeutic contexts.

4. **Don Richard Riso and Russ Hudson:** Authors of "The Wisdom of the Enneagram" and "Personality Types," among other books. They have made significant contributions to the modern understanding of the Enneagram as a tool for personal growth and self-awareness.

5. **Helen Palmer:** Author of "The Enneagram" and a prominent Enneagram teacher who has played a key role in popularizing the system in the West.

It's important to note that the Enneagram is a dynamic and evolving field, and there are many other teachers, authors, and organizations that have contributed to its development and dissemination. Different Enneagram schools and teachers may emphasize different aspects of the system, leading to a diverse range of interpretations and applications.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client SEMus", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client SEMus", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response SEMus", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response SEMus", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction SEMus", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction SEMus", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

# **Step 4: Create EM Embeddings** <a class="anchor" id="EM_page_3"></a>

[Back to Top](#EM_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
from datetime import datetime, timezone
import pandas as pd
from openai import OpenAI, OpenAIError
import mlflow
from dotenv import load_dotenv
import ast

# ─── 0) Init ───────────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("EM")

# ─── 1) Helper: read OpenAI key ────────────────────────────────────────────────
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    path = os.path.expanduser(filepath)
    with open(path, 'r') as f:
        key = f.read().strip()
    if not key:
        raise ValueError(f"API key file at '{path}' is empty.")
    return key

# ─── 2) Initialize OpenAI client ───────────────────────────────────────────────
OPENAI_KEY = os.getenv("OPENAI_API_KEY") or get_openai_api_key_from_file()
client = OpenAI(api_key=OPENAI_KEY)

# ─── 3) Load EM dataset ─────────────────────────────────────────────────────────
dataset_path = '../Datasets/em.csv'
em_df = pd.read_csv(dataset_path)

# Ensure Adjacencies is read as a list
if em_df['Adjacencies'].dtype == object:
    em_df['Adjacencies'] = em_df['Adjacencies'].apply(ast.literal_eval)

# ─── 4) Embedding helper ───────────────────────────────────────────────────────
def get_embedding(text: str, model="text-embedding-3-small"):
    resp = client.embeddings.create(input=[text], model=model)
    return resp.data[0].embedding

# ─── 5) Build prompt including adjacencies ─────────────────────────────────────
def make_prompt(row):
    adj = ", ".join(row['Adjacencies'])
    return (
        f"{row['Type']} ({row['Name']}) | "
        f"{row['Factor']} | {row['Adjective']} | "
        f"{row['Synonym']} | {row['Verb']} | {row['Noun']} | "
        f"Adjacent types: {adj}"
    )

# ─── 6) Generate & Log Embeddings ─────────────────────────────────────────────
with mlflow.start_run(run_name="Generate_EM_Embeddings"):
    mlflow.log_param("model", "text-embedding-3-small")
    mlflow.log_param("dataset", dataset_path)

    start_ts = datetime.now(timezone.utc)
    mlflow.log_param("start_time", start_ts.isoformat())

    if em_df.empty:
        raise ValueError("EM dataframe is empty!")

    em_df['Embedding'] = em_df.apply(lambda r: get_embedding(make_prompt(r)), axis=1)

    end_ts = datetime.now(timezone.utc)
    mlflow.log_param("end_time", end_ts.isoformat())

    # Ensure output directory
    out_dir = '../Embeddings'
    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, 'em_embeddings.csv')

    # Write all columns plus embeddings
    em_df.to_csv(
        out_path,
        columns=[
            'Type','Name','Factor','Adjective','Synonym',
            'Verb','Noun','Adjacencies','Embedding'
        ],
        index=False
    )
    mlflow.log_artifact(out_path, artifact_path="embeddings")

    # Log stats
    mlflow.log_param("num_rows", em_df.shape[0])
    mlflow.log_param("num_columns", em_df.shape[1] + 1)  # +1 for embedding

print("✅ EM embedding generation complete.")
````

````python
import os
import ast
from datetime import datetime, timezone

import joblib
import mlflow
from openai import OpenAI, OpenAIError
from dotenv import load_dotenv

# ─── Setup ─────────────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("EM_Inference")

# ─── 0) Configuration ──────────────────────────────────────────────────────────
# Choose which target you want to predict: 'Factor' or 'Type'
TARGET = 'Factor'   # ← or 'Type'

# Map TARGET to the correct encoder & model
ENCODER_PATHS = {
    'Factor': "../Models/em_factor_encoder.pkl",
    'Type':   "../Models/em_type_encoder.pkl"
}
MODEL_PATHS = {
    'Factor': "../Models/em_rf_factor.pkl",
    'Type':   "../Models/em_rf_type.pkl"
}

# ─── 1) Initialize OpenAI client ──────────────────────────────────────────────
def _get_api_key_from_file(path="~/openai_api_key.txt"):
    p = os.path.expanduser(path)
    with open(p, "r") as f:
        key = f.read().strip()
    if not key:
        raise ValueError(f"No API key found in {p}")
    return key

api_key = os.getenv("OPENAI_API_KEY") or _get_api_key_from_file()
client  = OpenAI(api_key=api_key)

# ─── 2) Embedding helper ───────────────────────────────────────────────────────
def get_embedding(text: str, model: str = "text-embedding-3-small") -> list:
    try:
        resp = client.embeddings.create(input=[text], model=model)
        return resp.data[0].embedding
    except OpenAIError as e:
        mlflow.log_param("openai_error", str(e))
        raise

# ─── 3) Inference run ──────────────────────────────────────────────────────────
with mlflow.start_run(run_name="EM_Inference_New_Data"):
    mlflow.log_param("step", "inference")
    mlflow.log_param("target", TARGET)
    mlflow.log_param("start_time", datetime.now(timezone.utc).isoformat())

    # 3a) Record which artifacts we'll load
    encoder_path = ENCODER_PATHS[TARGET]
    model_path   = MODEL_PATHS[TARGET]
    mlflow.log_param("encoder_path", encoder_path)
    mlflow.log_param("model_path", model_path)

    # 3b) Load encoder & model
    le  = joblib.load(encoder_path)
    mlflow.log_param("label_encoder_loaded", True)
    clf = joblib.load(model_path)
    mlflow.log_param("model_loaded", True)

    # 3c) Prediction helper
    def predict_text(text: str) -> str:
        emb  = get_embedding(text)
        code = clf.predict([emb])[0]
        return le.inverse_transform([code])[0]

    # 3d) Example inputs — you can swap these out or add new ones
        # 3d) Enneagram-focused example inputs
    examples = {
        "Perfectionism":          "I review every detail of my project multiple times to make sure it’s absolutely flawless.",
        "Helpfulness":            "I volunteer to help colleagues even when it means staying late to ensure they succeed.",
        "Achievement Orientation":"I set ambitious goals each week and track my progress until I hit every milestone.",
        "Emotional Sensitivity":  "I often find myself deeply moved by art or music, feeling every nuance of the emotion.",
        "Assertiveness":          "I speak up confidently in meetings and push for decisions even when others hesitate."
    }

    mlflow.log_param("n_examples", len(examples))

    print(f"\n--- EM Inference ({TARGET}) Examples ---")
    for idx, (label, text) in enumerate(examples.items(), start=1):
        try:
            pred = predict_text(text)
            print(f"{idx}. [{label}] “{text}” → Predicted {TARGET}: {pred}")
            mlflow.log_param(f"example_{idx}_text", text)
            mlflow.log_param(f"example_{idx}_predicted_{TARGET}", pred)
        except Exception as e:
            print(f"❌ Error on example '{label}': {e}")
            mlflow.log_param(f"example_{idx}_error", str(e))

    mlflow.log_param("end_time", datetime.now(timezone.utc).isoformat())

print("✅ EM inference run complete.")
````

````python
import os
import ast
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timezone
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, StratifiedKFold, LeaveOneOut, cross_val_score
import mlflow
from dotenv import load_dotenv

# ─── Setup MLflow & Environment ───────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("EM_Train_Eval_CV")

# ─── Configuration ────────────────────────────────────────────────────────────
# Choose prediction target: 'Factor' or 'Type'
TARGET = 'Factor'  # ← change to 'Type' to train on Enneagram types

ENCODER_PATHS = {
    'Factor': '../Models/em_factor_encoder.pkl',
    'Type':   '../Models/em_type_encoder.pkl'
}
MODEL_OUT_PATHS = {
    'Factor': '../Models/em_rf_factor.pkl',
    'Type':   '../Models/em_rf_type.pkl'
}

# ─── 1) Load embeddings ────────────────────────────────────────────────────────
embeddings_path = '../Embeddings/em_embeddings.csv'
df = pd.read_csv(
    embeddings_path,
    converters={
        'Embedding': lambda s: np.array(ast.literal_eval(s))
                        if isinstance(s, str) else np.array(s)
    }
)
X = np.stack(df['Embedding'].values)
print(f"✅ Loaded {len(df)} EM embeddings from {embeddings_path}")

# ─── 2) Load LabelEncoder ─────────────────────────────────────────────────────
encoder_path = ENCODER_PATHS[TARGET]
le = joblib.load(encoder_path)
print(f"✅ Loaded '{TARGET}' encoder from: {encoder_path}")

# ─── 3) Prepare features & labels ─────────────────────────────────────────────
y = df[TARGET].values
y_encoded = le.transform(y)
print(f"{TARGET} classes ({len(le.classes_)}): {list(le.classes_)}")

# ─── 4) Train/test split ──────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.4,
    random_state=42,
    stratify=y_encoded
)
print(f"Train/Test split: {X_train.shape[0]}/{X_test.shape[0]} samples")

# ─── 5) MLflow run: train, evaluate, cross-validate ───────────────────────────
with mlflow.start_run(run_name=f"EM_RF_{TARGET}_TrainEvalCV"):
    mlflow.log_param("target", TARGET)
    mlflow.log_param("n_samples", len(df))
    mlflow.log_param("n_classes", len(le.classes_))
    mlflow.log_artifact(encoder_path, artifact_path="encoder")
    mlflow.log_param("test_size", 0.4)
    mlflow.log_param("random_state", 42)
    mlflow.log_param("start_time", datetime.now(timezone.utc).isoformat())

    # ─── Train RandomForest ─────────────────────────────────────────────────
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)
    mlflow.sklearn.log_model(clf, artifact_path=f"model_rf_{TARGET.lower()}")

    # ─── Evaluate on hold-out set ────────────────────────────────────────────
    y_pred = clf.predict(X_test)
    acc    = accuracy_score(y_test, y_pred)
    mlflow.log_metric("test_accuracy", acc)
    print(f"🔍 Test accuracy: {acc:.3f}")

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.title(f"EM Confusion Matrix ({TARGET})")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.tight_layout()
    cm_img = f"em_confusion_matrix_{TARGET.lower()}.png"
    plt.savefig(cm_img)
    plt.close()
    mlflow.log_artifact(cm_img, artifact_path="evaluation")
    print(f"✅ Confusion matrix saved: {cm_img}")

    # Classification report
    report_dict = classification_report(
        y_test, y_pred,
        target_names=le.classes_,
        output_dict=True,
        zero_division=0
    )
    report_df = pd.DataFrame(report_dict).transpose()
    report_csv = f"em_classification_report_{TARGET.lower()}.csv"
    report_df.to_csv(report_csv, index=True)
    mlflow.log_artifact(report_csv, artifact_path="evaluation")
    print(f"✅ Classification report saved: {report_csv}")

    # ─── Cross-Validation ────────────────────────────────────────────────────
    counts = pd.Series(y_encoded).value_counts()
    if counts.min() < 5:
        cv = LeaveOneOut()
        mlflow.log_param("cv_method", "LeaveOneOut")
    else:
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        mlflow.log_param("cv_method", "StratifiedKFold5")

    mlflow.log_param("n_splits", cv.get_n_splits(X, y_encoded))

    cv_scores = cross_val_score(
        clf, X, y_encoded,
        cv=cv,
        scoring='accuracy',
        n_jobs=-1
    )
    mean_cv = cv_scores.mean()
    std_cv  = cv_scores.std()
    mlflow.log_metric("cv_mean_accuracy", mean_cv)
    mlflow.log_metric("cv_std_accuracy", std_cv)
    print(f"▶ CV accuracy: {mean_cv:.3f} ± {std_cv:.3f}")

    mlflow.log_param("end_time", datetime.now(timezone.utc).isoformat())

print("✅ EM training, evaluation, and CV complete.")
````

````python
#!/usr/bin/env python3
import os
import ast
import numpy as np
import joblib
from openai import OpenAI, OpenAIError
from dotenv import load_dotenv

# ─── 1) Bootstrap OpenAI client ────────────────────────────────────────────────
load_dotenv(override=True)

def _get_api_key_from_file(path="~/openai_api_key.txt"):
    p = os.path.expanduser(path)
    try:
        key = open(p, "r").read().strip()
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at {p}")
    if not key:
        raise ValueError(f"No API key found in {p}")
    return key

api_key = os.getenv("OPENAI_API_KEY") or _get_api_key_from_file()
client  = OpenAI(api_key=api_key)

# ─── 2) Load artifacts ─────────────────────────────────────────────────────────
ENCODER_PATH = '../Models/em_factor_encoder.pkl'
MODEL_PATH   = '../Models/em_rf_factor.pkl'
try:
    label_encoder = joblib.load(ENCODER_PATH)
    clf           = joblib.load(MODEL_PATH)
    print("✅ Loaded EM factor encoder and RF model.")
except Exception as e:
    print(f"❌ Failed to load artifacts: {e}")
    exit(1)

# ─── 3) Inference helpers ───────────────────────────────────────────────────────
def get_embedding(text: str, model: str = "text-embedding-3-small") -> list:
    try:
        resp = client.embeddings.create(input=[text], model=model)
        return resp.data[0].embedding
    except OpenAIError as e:
        raise RuntimeError(f"OpenAI embedding error: {e}")

def predict_em_factor(text: str) -> str:
    emb  = get_embedding(text)
    code = clf.predict([emb])[0]
    return label_encoder.inverse_transform([code])[0]

# ─── 4) Test on EM-specific examples ────────────────────────────────────────────
examples = {
    "Perfectionism":            "I meticulously review every detail of my work until it’s flawless.",
    "Helpfulness":              "I offer my time and energy freely to support others, even at personal cost.",
    "Achievement Orientation":  "I set ambitious goals and will persist until I’ve reached them.",
    "Emotional Sensitivity":    "I often feel deeply moved by others’ emotions and reflect on why I feel that way.",
    "Assertiveness":            "I speak up confidently in meetings and push for decisions when no one else will.",
}

print("\n--- EM Model Predictions on New Examples ---\n")
for factor, sample in examples.items():
    try:
        pred = predict_em_factor(sample)
        print(f"{factor}\n  Input: “{sample}”\n  → Predicted EM Factor: {pred}\n")
    except Exception as ex:
        print(f"❌ Error on '{factor}': {ex}\n")
````

````python
#!/usr/bin/env python3
import os
import pandas as pd
from dotenv import load_dotenv
from py2neo import Graph, Node, Relationship

# ─── 1) Load env & connect to Neo4j ────────────────────────────────────────────
load_dotenv()
raw_uri = os.getenv("NEO4J_URI", "")
if raw_uri.startswith("neo4j://"):
    bolt_uri = "bolt://" + raw_uri[len("neo4j://"):]
else:
    bolt_uri = raw_uri

user = os.getenv("NEO4J_USERNAME")
pwd  = os.getenv("NEO4J_PASSWORD")
if not all([bolt_uri, user, pwd]):
    raise RuntimeError("Make sure NEO4J_URI, NEO4J_USERNAME & NEO4J_PASSWORD are set")

graph = Graph(bolt_uri, auth=(user, pwd))
print("🔗 Connected to Neo4j via", bolt_uri)

# ─── 2) Clear existing data ────────────────────────────────────────────────────
graph.run("MATCH (n) DETACH DELETE n")
print("🗑️  Cleared all existing nodes and relationships")

# ─── 3) Load EM CSV ───────────────────────────────────────────────────────────
em_csv = "../Datasets/em.csv"
df = pd.read_csv(em_csv)
print(f"✅ Loaded {len(df)} rows from {em_csv}")

# ─── 4) Build graph ────────────────────────────────────────────────────────────
tx = graph.begin()

# caches to avoid duplicates
domains, factors, adjs, syns, verbs, nouns = {}, {}, {}, {}, {}, {}

for _, row in df.iterrows():
    # helper to safely extract and strip text or return ''
    def get_text(col):
        val = row.get(col)
        return val.strip() if pd.notna(val) and isinstance(val, str) else ""

    d = get_text("Domain")
    f = get_text("Factor")
    a = get_text("Adjective")
    s = get_text("Synonym")
    v = get_text("Verb")
    n = get_text("Noun")

    # Domain
    if d and d not in domains:
        domains[d] = Node("Domain", name=d)
        tx.create(domains[d])

    # Factor
    if f and f not in factors:
        factors[f] = Node("Factor", name=f)
        tx.create(factors[f])
        tx.create(Relationship(domains[d], "HAS_FACTOR", factors[f]))

    # Adjective
    if a and a not in adjs:
        adjs[a] = Node("Adjective", text=a)
        tx.create(adjs[a])
        tx.create(Relationship(factors[f], "HAS_ADJECTIVE", adjs[a]))

    # Synonym
    if s and s not in syns:
        syns[s] = Node("Synonym", text=s)
        tx.create(syns[s])
        if a:
            tx.create(Relationship(adjs[a], "HAS_SYNONYM", syns[s]))

    # Verb
    if v and v not in verbs:
        verbs[v] = Node("Verb", text=v)
        tx.create(verbs[v])
        if a:
            tx.create(Relationship(adjs[a], "HAS_VERB", verbs[v]))

    # Noun
    if n and n not in nouns:
        nouns[n] = Node("Noun", text=n)
        tx.create(nouns[n])
        if a:
            tx.create(Relationship(adjs[a], "HAS_NOUN", nouns[n]))

tx.commit()
print("🎉 Completed building the EM taxonomy graph in Neo4j.")
````

````python
#!/usr/bin/env python3
import os
import ast
import pandas as pd
from dotenv import load_dotenv
from py2neo import Graph, Node, Relationship

# ─── Load env & set up connection ─────────────────────────────────────────────
load_dotenv(override=True)
raw_uri = os.getenv("NEO4J_URI", "")
bolt_uri = raw_uri.replace("neo4j://", "bolt://") if raw_uri.startswith("neo4j://") else raw_uri
user     = os.getenv("NEO4J_USERNAME")
pwd      = os.getenv("NEO4J_PASSWORD")
if not all([bolt_uri, user, pwd]):
    raise RuntimeError("Please set NEO4J_URI, NEO4J_USERNAME & NEO4J_PASSWORD")
graph = Graph(bolt_uri, auth=(user, pwd))
print("🔗 Connected to Neo4j via", bolt_uri)

# ─── Clear existing graph ────────────────────────────────────────────────────
graph.run("MATCH (n) DETACH DELETE n")
print("🗑️  Cleared all existing nodes and relationships")

# ─── Load EM dataset ─────────────────────────────────────────────────────────
df = pd.read_csv(
    '../Datasets/em.csv',
    converters={'Adjacencies': lambda s: ast.literal_eval(s) if isinstance(s, str) else s}
)
print(f"✅ Loaded {len(df)} rows from ../Datasets/em.csv")

tx = graph.begin()

# ─── 1) Create Type nodes ────────────────────────────────────────────────────
type_nodes = {}
for t, grp in df.groupby('Type'):
    name = grp['Name'].iat[0]
    node = Node("Type", code=t, name=name)
    tx.merge(node, "Type", "code")
    type_nodes[t] = node

# ─── 2) Create adjacency relationships ────────────────────────────────────────
for t, node in type_nodes.items():
    neighbors = df[df['Type']==t]['Adjacencies'].iat[0]  # list of codes
    for neigh in neighbors:
        neigh_node = type_nodes.get(neigh)
        if neigh_node:
            rel = Relationship(node, "ADJACENT_TO", neigh_node)
            tx.merge(rel)

# ─── 3) Create Factors, Adjectives, Synonyms, Verbs, Nouns ───────────────────
# Use dictionaries to avoid duplicate Node creations
factor_nodes    = {}
adjective_nodes = {}
synonym_nodes   = {}
verb_nodes      = {}
noun_nodes      = {}

for _, row in df.iterrows():
    tcode = row['Type']
    fac   = row['Factor']
    adj   = row['Adjective']
    syn   = row['Synonym'] or ""
    verb  = row['Verb']    or ""
    noun  = row['Noun']    or ""

    # Factor
    if fac not in factor_nodes:
        fn = Node("Factor", name=fac)
        tx.merge(fn, "Factor", "name")
        # link Type -> Factor
        tx.merge(Relationship(type_nodes[tcode], "HAS_FACTOR", fn))
        factor_nodes[fac] = fn
    else:
        fn = factor_nodes[fac]

    # Adjective
    if adj not in adjective_nodes:
        an = Node("Adjective", name=adj)
        tx.merge(an, "Adjective", "name")
        tx.merge(Relationship(fn, "HAS_ADJECTIVE", an))
        adjective_nodes[adj] = an
    else:
        an = adjective_nodes[adj]

    parent = an

    # Synonym
    if syn:
        if syn not in synonym_nodes:
            sn = Node("Synonym", name=syn)
            tx.merge(sn, "Synonym", "name")
            synonym_nodes[syn] = sn
        else:
            sn = synonym_nodes[syn]
        tx.merge(Relationship(parent, "HAS_SYNONYM", sn))
        parent = sn

    # Verb
    if verb:
        if verb not in verb_nodes:
            vn = Node("Verb", name=verb)
            tx.merge(vn, "Verb", "name")
            verb_nodes[verb] = vn
        else:
            vn = verb_nodes[verb]
        tx.merge(Relationship(parent, "HAS_VERB", vn))
        parent = vn

    # Noun
    if noun:
        if noun not in noun_nodes:
            nn = Node("Noun", name=noun)
            tx.merge(nn, "Noun", "name")
            noun_nodes[noun] = nn
        else:
            nn = noun_nodes[noun]
        tx.merge(Relationship(parent, "HAS_NOUN", nn))

# ─── Commit everything ────────────────────────────────────────────────────────
graph.commit(tx)
print("🎉 Completed building the EM taxonomy graph in Neo4j.")
````

## Process for Building the EM Personality Model

1. **Data Preparation**

   * Define taxonomy (`Factor \to Adjective \to Synonym/Verb/Noun`).
   * Normalize lists and flatten to CSV (`EM.csv`).

2. **Embedding Generation**

   * Load `EM.csv` into pandas.
   * Initialize OpenAI client and MLflow.
   * Generate embeddings via `client.embeddings.create(...)` for full text prompt.
   * Save `EM_embeddings.csv` with `Embedding` column.

3. **Label Encoding**

   * Load embeddings CSV.
   * Fit `LabelEncoder` on `Factor` column.
   * Save encoder (`EM_label_encoder.pkl`).
   * Visualize mapping with bar chart and log to MLflow.

4. **Model Training**

   * Load embeddings and label encoder.
   * Split into train/test (stratify when possible).
   * Train `RandomForestClassifier`.
   * Log metrics, artifacts, and model to MLflow.

5. **Model Evaluation**

   * Reload model and encoder.
   * Split and predict on test set.
   * Compute accuracy, classification report, confusion matrix.
   * Log evaluation artifacts to MLflow.
   * Perform cross-validation (LOO if class imbalance).

6. **Inference on New Data**

   * Load model & encoder.
   * Define `predict_factor(text)` helper that requests embedding and predicts.
   * Test on meaningful example sentences.
   * Log predictions to MLflow.

7. **Visualization & Clustering**

   * Train RF on full dataset.
   * In‑sample evaluation & plot confusion matrix.
   * PCA into 2D & scatter by encoded class.
   * K‑Means clustering on embeddings, save `EM_clustered_embeddings.csv`.
   * Plot clusters in PCA space.

8. **Neo4j Graph Construction**

   * Clear existing graph: `MATCH (n) DETACH DELETE n`.
   * Ingest `EM.csv`, create nodes for Factor, Adjective, Synonym, Verb, Noun.
   * Create `:RELATES_TO` relationships following taxonomy hierarchy.

9. **Cypher Query to View Graph**

   ```cypher
   MATCH p = (f:Factor)-[:HAS_ADJECTIVE]->(a:Adjective)
             -[:HAS_SYNONYM|HAS_VERB|HAS_NOUN]->(x)
   RETURN p LIMIT 100
   ```
