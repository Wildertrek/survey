**Theories of Emotional Intelligence (TEI)** encompass various models that aim to conceptualize and measure emotional intelligence, a critical aspect of human psychology related to the recognition, understanding, management, and utilization of emotions. These models provide frameworks for understanding and improving emotional intelligence in individuals. Here, we summarize some of the key models within TEI:

**1. Ability Model:**
   - *Perceiving Emotions:* This facet focuses on recognizing emotions in oneself and others, interpreting nonverbal cues, and identifying emotions from facial expressions.
   - *Using Emotions:* It involves harnessing emotions to facilitate thinking, using emotions to enhance problem-solving, and employing emotional reasoning.
   - *Understanding Emotions:* This component revolves around comprehending emotional language, understanding the causes and consequences of emotions, and recognizing emotional transitions.
   - *Managing Emotions:* Managing emotions encompasses regulating one's emotions, responding adaptively to emotional situations, and exercising emotional self-control.

**2. Trait Model:**
   - *Self-Awareness:* This trait involves being aware of one's emotions and their impact, recognizing emotional strengths and weaknesses, and identifying emotional triggers.
   - *Emotional Resilience:* It relates to bouncing back from adversity, effectively handling stress and setbacks, and maintaining a positive outlook.
   - *Empathy:* Empathy entails understanding and sharing others' emotions, showing compassion and concern, and engaging in perspective-taking.
   - *Emotional Intelligence Quotient (EQ):* EQ measures and assesses emotional intelligence, quantifies emotional competencies, and evaluates emotional self-awareness.

**3. Mixed Model:**
   - *Self-Awareness:* Similar to the Trait Model, it emphasizes recognizing one's emotions and their effects, understanding strengths and weaknesses, and engaging in emotional self-assessment.
   - *Social Awareness:* Social awareness involves perceiving and understanding others' emotions, empathizing with others, and exhibiting social perceptiveness.
   - *Self-Management:* This component centers on managing one's emotions, adapting to changing circumstances, and effectively managing anger and stress.
   - *Relationship Management:* Relationship management encompasses building and maintaining relationships, influencing and inspiring others, and effectively resolving conflicts.

**4. Integrated Model:**
   - *Intrapersonal Intelligence:* Intrapersonal intelligence includes self-reflection, self-regulation, and self-motivation.
   - *Interpersonal Intelligence:* It involves empathy, social skills, and effective communication.
   - *Adaptability:* Adaptability encompasses flexibility, resilience, and the ability to cope with change.
   - *Emotional Competence:* Emotional competence comprises emotional self-awareness, emotional regulation, and empathy in relationships.

These theories of emotional intelligence provide valuable frameworks for understanding and enhancing individuals' emotional competencies. They find applications in various domains, including personal development, leadership, education, and psychology. Researchers and practitioners use these models to assess, train, and promote emotional intelligence, ultimately leading to improved emotional well-being and more effective interpersonal relationships.

**Bartle Types (TEI)**, also known as Bartle's Player Types, is a categorization system that classifies players of multiplayer online games based on their gaming preferences and behaviors. It was developed by Richard Bartle, a British video game designer and researcher. Bartle identified four primary player types in online games, each with distinct characteristics and motivations:

1. **Achievers:** Achievers are highly competitive and goal-oriented players. They seek to excel in the game by accumulating achievements, points, or other measurable rewards. They enjoy mastering the game's mechanics and demonstrating their skills to others. Achievers are often persistent in their pursuit of in-game success.

2. **Explorers:** Explorers are curious and adventurous players who enjoy discovering and uncovering the game's secrets. They thrive on exploration, innovation, and experimentation. Explorers are analytical thinkers who seek to understand the game world thoroughly and may focus on uncovering hidden elements or solving mysteries.

3. **Socializers:** Socializers are players who value social interaction and collaboration with others. They derive enjoyment from building relationships, forming alliances, and communicating with fellow players. Socializers are often friendly, empathetic, and cooperative, and they view the game as a means to connect with others.

4. **Killers:** Killers are competitive and aggressive players who thrive on conflict and competition. They enjoy challenging other players and may engage in combat, player-versus-player (PvP) interactions, or strategic gameplay. Killers are motivated by the thrill of defeating opponents and exerting dominance.

**Practical Applications of Bartle Types (TEI):**

- **Game Design:** Game developers use the Bartle Types framework to design and balance games that cater to different player preferences. By understanding what motivates each player type, developers can create more engaging and enjoyable gaming experiences.

- **Community Building:** Online gaming communities and guilds often use Bartle Types to understand the dynamics within their groups. Recognizing the diverse player types can help leaders foster a positive and inclusive gaming environment.

- **Player Engagement:** Game publishers and operators can use Bartle Types to tailor in-game content, events, and rewards to different player segments, increasing player retention and satisfaction.

- **Market Research:** Researchers and marketers in the gaming industry use Bartle Types to study player behavior and preferences, enabling them to make informed decisions about game development and marketing strategies.

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key STEIus", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"✅ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key STEIus", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key STEIus", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("🔑 API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("🤖 OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("✅ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"⚠️ Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"❌ Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"❌ Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"⚠️ Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client STEIus", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client STEIus", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response STEIus", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response STEIus", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction STEIus", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction STEIus", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import os
from datetime import datetime, timezone
import pandas as pd
from openai import OpenAI, OpenAIError
import mlflow
from dotenv import load_dotenv

# ─── 0) Init ───────────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("TEI")

# ─── 1) Helper: read OpenAI key ────────────────────────────────────────────────
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    path = os.path.expanduser(filepath)
    try:
        with open(path, 'r') as f:
            key = f.read().strip()
        if not key:
            raise ValueError(f"API key file at '{path}' is empty.")
        return key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{path}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from '{path}': {e}")

# ─── 2) Initialize OpenAI client ───────────────────────────────────────────────
try:
    OPENAI_KEY = os.getenv("OPENAI_API_KEY") or get_openai_api_key_from_file()
    client = OpenAI(api_key=OPENAI_KEY)
    print("✅ OpenAI client initialized.")
except Exception as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    exit(1)

# ─── 3) Load TEI dataset ─────────────────────────────────────────────────────────
dataset_path = '../Datasets/tei.csv'
try:
    tei_df = pd.read_csv(dataset_path)
    print(f"✅ Loaded TEI dataset: {dataset_path} ({len(tei_df)} rows)")
except FileNotFoundError:
    raise FileNotFoundError(f"TEI dataset not found at {dataset_path}")

# ─── 4) Embedding helper ───────────────────────────────────────────────────────
def get_embedding(text: str, model="text-embedding-3-small"):
    try:
        resp = client.embeddings.create(input=[text], model=model)
        return resp.data[0].embedding
    except OpenAIError as e:
        print(f"❌ OpenAI API error for text '{text[:30]}...': {e}")
        raise

# ─── 5) Generate & Log Embeddings ─────────────────────────────────────────────
with mlflow.start_run(run_name="Generate_TEI_Embeddings"):
    mlflow.log_param("model", "text-embedding-3-small")
    mlflow.log_param("dataset", dataset_path)

    start_ts = datetime.now(timezone.utc)
    mlflow.log_param("start_time", start_ts.isoformat())
    print(f"⏳ Started embeddings at {start_ts.isoformat()}")

    if tei_df.empty:
        raise ValueError("Loaded TEI dataframe is empty.")

    # Build prompt string from each row
    tei_df['Embedding'] = tei_df.apply(
        lambda r: get_embedding(
            f"{r['Domain']} | {r['Factor']} | {r['Adjective']} | "
            f"{r['Synonym']} | {r['Verb']} | {r['Noun']}"
        ),
        axis=1
    )

    end_ts = datetime.now(timezone.utc)
    mlflow.log_param("end_time", end_ts.isoformat())
    print(f"✅ Finished embeddings at {end_ts.isoformat()} (duration: {end_ts - start_ts})")

    # Save embeddings CSV
    out_path = '../Embeddings/tei_embeddings.csv'
    tei_df.to_csv(
        out_path,
        columns=['Domain','Factor','Adjective','Synonym','Verb','Noun','Embedding'],
        index=False
    )
    mlflow.log_artifact(out_path, artifact_path="embeddings")
    print(f"💾 Embeddings saved to {out_path}")

    # Log run stats
    mlflow.log_param("num_rows",    tei_df.shape[0])
    mlflow.log_param("num_columns", tei_df.shape[1])
    mlflow.log_param("embedding_dim", len(tei_df['Embedding'].iloc[0]))

    print("\nSample embeddings:\n", tei_df.head())

print("✅ TEI embedding generation complete.")
````

````python
import os
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv

# ─── Setup MLflow ─────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("TEI")

# ─── 1) Load embeddings dataset ───────────────────────────────────────────────
embeddings_csv_path = '../Embeddings/tei_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded: {embeddings_csv_path} ({len(df)} rows)")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}")
    exit(1)

# ─── 2) Fit LabelEncoder on the 'Factor' column ───────────────────────────────
label_encoder = LabelEncoder()
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])
n_classes = len(label_encoder.classes_)
print(f"✅ Encoded {n_classes} unique TEI factors")

# ─── 3) Save the encoder to disk ───────────────────────────────────────────────
os.makedirs("../Models", exist_ok=True)
label_encoder_path = "../Models/tei_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# ─── 4) Visualization helper ───────────────────────────────────────────────────
def visualize_label_encoder(le, artifact_path="visualization"):
    classes = le.classes_
    codes   = le.transform(classes)

    plt.figure(figsize=(8, 4))
    sns.barplot(x=classes, y=codes)
    plt.xlabel("TEI Factor")
    plt.ylabel("Encoded Value")
    plt.title("TEI Factors → Encoded Mapping")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    out_file = "tei_label_encoder_mapping.png"
    plt.savefig(out_file)
    plt.close()
    print(f"✅ Plot saved to {os.path.abspath(out_file)}")
    mlflow.log_artifact(out_file, artifact_path=artifact_path)

# ─── 5) Log everything in MLflow ────────────────────────────────────────────────
with mlflow.start_run(run_name="TEI_LabelEncoding_Visualization"):
    mlflow.log_param("step", "label_encoding")
    mlflow.log_param("embeddings_csv", embeddings_csv_path)
    mlflow.log_param("num_rows", df.shape[0])
    mlflow.log_param("num_unique_factors", n_classes)
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Preview mapping
    preview = (
        df[['Factor', 'Factor_Encoded']]
        .drop_duplicates()
        .sort_values('Factor_Encoded')
        .reset_index(drop=True)
    )
    print("\nFactor → Encoded preview:\n", preview)

    # Full mapping dict
    mapping = {
        factor: code
        for factor, code in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))
    }
    print("\nFull mapping:")
    for factor, code in mapping.items():
        print(f"  {factor} → {code}")
    mlflow.log_dict(mapping, "label_encoder/mapping.json")

    # Generate & log bar chart
    visualize_label_encoder(label_encoder)

print("✅ TEI label encoding & visualization complete.")
````

````python
import os
import ast
import numpy as np
import pandas as pd
import joblib
import mlflow
from datetime import datetime
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv

# ─── Setup MLflow ─────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("TEI")

# ─── 1) Load embeddings dataset ───────────────────────────────────────────────
embeddings_csv_path = '../Embeddings/tei_embeddings.csv'
try:
    df = pd.read_csv(
        embeddings_csv_path,
        converters={'Embedding': lambda s: np.array(ast.literal_eval(s)) if isinstance(s, str) else np.array(s)}
    )
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}")
    exit(1)

# ─── 2) Load the pre-fitted label encoder ───────────────────────────────────────
label_encoder_path = "../Models/tei_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"✅ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
    exit(1)

# ─── 3) Prepare features and labels ───────────────────────────────────────────
X = np.stack(df['Embedding'].values)      # shape: (n_samples, embedding_dim)
y = df['Factor'].values                   # TEI factor labels
y_encoded = label_encoder.transform(y)    # numeric labels

# ─── 4) Split into train/test ─────────────────────────────────────────────────
counts = pd.Series(y_encoded).value_counts()
stratify_param = y_encoded if counts.min() >= 2 else None

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.4,
    random_state=42,
    shuffle=True,
    stratify=stratify_param
)

print(f"Train samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
if stratify_param is None:
    print("⚠️  Skipped stratification because at least one class has only 1 sample.")

# ─── 5) Start MLflow run ──────────────────────────────────────────────────────
with mlflow.start_run(run_name="TEI_RF_Training"):
    # Log hyperparameters
    mlflow.log_param("model_type", "RandomForestClassifier")
    mlflow.log_param("n_estimators", 100)
    mlflow.log_param("test_size", 0.4)
    mlflow.log_param("random_state", 42)
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    start_ts = datetime.now()
    mlflow.log_param("start_time", start_ts.isoformat())
    print(f"⏳ Training started at {start_ts}")

    # Train
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)

    end_ts = datetime.now()
    mlflow.log_param("end_time", end_ts.isoformat())
    print(f"✅ Training finished at {end_ts} (Duration: {end_ts - start_ts})")

    # Predict & evaluate
    y_pred = clf.predict(X_test)
    acc   = accuracy_score(y_test, y_pred)
    mlflow.log_metric("accuracy", acc)
    print(f"🔍 Test accuracy: {acc:.4f}")

    # Only keep labels actually present in the test set
    present = np.unique(np.concatenate([y_test, y_pred]))
    names   = [label_encoder.classes_[i] for i in present]

    # Classification report
    report_dict = classification_report(
        y_test, y_pred,
        labels=present,
        target_names=names,
        output_dict=True,
        zero_division=0
    )
    report_df = pd.DataFrame(report_dict).transpose()
    report_path = "tei_classification_report.csv"
    report_df.to_csv(report_path, index=True)
    mlflow.log_artifact(report_path, artifact_path="metrics")
    print(f"📊 Classification report saved to {report_path}")

    # Confusion matrix
    cm    = confusion_matrix(y_test, y_pred, labels=present)
    cm_df = pd.DataFrame(cm, index=names, columns=names)
    cm_csv = "tei_confusion_matrix.csv"
    cm_df.to_csv(cm_csv, index=True)
    mlflow.log_artifact(cm_csv, artifact_path="metrics")
    print(f"📊 Confusion matrix saved to {cm_csv}")

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
    plt.title("TEI RandomForest Confusion Matrix")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.tight_layout()
    cm_img = "tei_confusion_matrix.png"
    plt.savefig(cm_img)
    plt.close()
    mlflow.log_artifact(cm_img, artifact_path="metrics")
    print(f"🖼️ Saved confusion matrix plot to {cm_img}")

    # Save trained model
    os.makedirs("../Models", exist_ok=True)
    model_path = "../Models/tei_rf_model.pkl"
    joblib.dump(clf, model_path)
    mlflow.log_artifact(model_path, artifact_path="models")
    print(f"💾 Trained model saved to {model_path}")

    # Final printout
    print("\nClassification Report:\n",
          classification_report(
              y_test, y_pred,
              labels=present,
              target_names=names,
              zero_division=0
          ))
    print("\nConfusion Matrix:\n", cm_df)

print("✅ TEI model training complete.")
````

````python
import os
import ast
from datetime import datetime, timezone

import joblib
import mlflow
from openai import OpenAI, OpenAIError
from dotenv import load_dotenv

# ─── Setup ─────────────────────────────────────────────────────────────────────
load_dotenv(override=True)
if os.getenv("MLFLOW_TRACKING_URI"):
    mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("TEI_Inference")

# ─── Initialize OpenAI client ─────────────────────────────────────────────────
def _get_api_key_from_file(path="~/openai_api_key.txt"):
    p = os.path.expanduser(path)
    with open(p, "r") as f:
        key = f.read().strip()
    if not key:
        raise ValueError(f"No API key found in {p}")
    return key

api_key = os.getenv("OPENAI_API_KEY") or _get_api_key_from_file()
client  = OpenAI(api_key=api_key)

# ─── Paths ─────────────────────────────────────────────────────────────────────
ENCODER_PKL = "../Models/tei_label_encoder.pkl"
MODEL_PKL   = "../Models/tei_rf_model.pkl"

# ─── Embedding helper ───────────────────────────────────────────────────────────
def get_embedding(text: str, model: str = "text-embedding-3-small") -> list:
    try:
        resp = client.embeddings.create(input=[text], model=model)
        return resp.data[0].embedding
    except OpenAIError as e:
        mlflow.log_param("openai_error", str(e))
        raise

# ─── Inference run ──────────────────────────────────────────────────────────────
with mlflow.start_run(run_name="TEI_Inference_New_Data"):
    # Log start time
    mlflow.log_param("step", "inference")
    mlflow.log_param("start_time", datetime.now(timezone.utc).isoformat())

    # Record artifacts
    mlflow.log_artifact(ENCODER_PKL, artifact_path="artifacts")
    mlflow.log_artifact(MODEL_PKL,   artifact_path="artifacts")

    # Load encoder & model
    label_encoder = joblib.load(ENCODER_PKL)
    mlflow.log_param("label_encoder_loaded", True)
    clf = joblib.load(MODEL_PKL)
    mlflow.log_param("model_loaded", True)

    # Prediction helper
    def predict_factor(text: str) -> str:
        emb  = get_embedding(text)
        code = clf.predict([emb])[0]
        return label_encoder.inverse_transform([code])[0]

    # TEI example inputs
    test_texts = {
        "Self-Awareness":       "I notice my mood shifts throughout the day and reflect on why I feel that way.",
        "Empathy":              "When a friend is upset, I intuitively sense their emotions and offer genuine comfort.",
        "Emotional Regulation": "In stressful situations, I pause and choose how to respond rather than react impulsively.",
        "Social Skills":        "I build rapport quickly, adapting my tone and language to different people.",
        "Stress Resilience":    "When facing setbacks, I recover my optimism and keep pursuing my goals."
    }
    mlflow.log_param("n_test_texts", len(test_texts))

    print("\n--- Inference on TEI Examples ---")
    for idx, (label, txt) in enumerate(test_texts.items(), start=1):
        try:
            pred = predict_factor(txt)
            print(f"{idx}. [{label}] “{txt}” → Predicted Factor: {pred}")
            mlflow.log_param(f"text_{idx}", txt)
            mlflow.log_param(f"expected_factor_{idx}", label)
            mlflow.log_param(f"predicted_factor_{idx}", pred)
        except Exception as e:
            print(f"❌ Error on example '{label}': {e}")
            mlflow.log_param(f"error_{idx}", str(e))

    # Log end time
    mlflow.log_param("end_time", datetime.now(timezone.utc).isoformat())

print("✅ TEI inference run complete.")
````

````python
import os
import ast
import numpy as np
import pandas as pd
import joblib
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, StratifiedKFold, LeaveOneOut, cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns

# ─── 1) Load embeddings safely ─────────────────────────────────────────────────
embeddings_path = '../Embeddings/tei_embeddings.csv'
df = pd.read_csv(
    embeddings_path,
    converters={'Embedding': lambda s: np.array(ast.literal_eval(s)) if isinstance(s, str) else np.array(s)}
)
print(f"✅ Loaded {len(df)} TEI embeddings from {embeddings_path}")

# ─── 2) Prepare features & labels ──────────────────────────────────────────────
X = np.stack(df['Embedding'].values)
y = df['Factor'].values
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print(f"TEI factors ({len(le.classes_)}): {list(le.classes_)}")

# ─── 3) Train/test split ───────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.4,
    random_state=42,
    stratify=y_encoded
)
print(f"Train/Test split: {X_train.shape[0]}/{X_test.shape[0]} samples")

# ─── 4) Train RandomForest ────────────────────────────────────────────────────
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
print("✅ Trained RandomForest on TEI factors")

# ─── 5) Evaluate on hold-out set ───────────────────────────────────────────────
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test accuracy: {acc:.3f}")

# Confusion matrix plot
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title("TEI Confusion Matrix (Test Set)")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.tight_layout()
plt.savefig("TEI_confusion_matrix.png")
plt.close()
print("✅ Confusion matrix saved to TEI_confusion_matrix.png")

# Classification report
print("\nClassification Report:")
print(classification_report(
    y_test, y_pred,
    target_names=le.classes_,
    zero_division=0
))

# ─── 6) Cross-Validation ───────────────────────────────────────────────────────
class_counts = pd.Series(y_encoded).value_counts()
min_count = class_counts.min()
if min_count < 5:
    print(f"⚠️  Least-populated class has only {min_count} samples, switching to Leave-One-Out CV")
    cv = LeaveOneOut()
else:
    print(f"▶ Using 5-fold StratifiedKFold CV")
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    clf, X, y_encoded,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1
)
print(f"\nCV accuracy scores (first 10): {np.round(cv_scores[:10], 3)}{'…' if len(cv_scores)>10 else ''}")
print(f"Mean CV accuracy: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
````

````python
#!/usr/bin/env python3
import os
import ast
import numpy as np
import pandas as pd
import joblib
from openai import OpenAI, OpenAIError
from dotenv import load_dotenv

# ─── 1) Bootstrap OpenAI client ────────────────────────────────────────────────
load_dotenv(override=True)

def _get_api_key_from_file(path="~/openai_api_key.txt"):
    p = os.path.expanduser(path)
    try:
        key = open(p, "r").read().strip()
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at {p}")
    if not key:
        raise ValueError(f"No API key found in {p}")
    return key

api_key = os.getenv("OPENAI_API_KEY") or _get_api_key_from_file()
client  = OpenAI(api_key=api_key)

# ─── 2) Load TEI embeddings ─────────────────────────────────────────────────────
EMB_PATH = '../Embeddings/tei_embeddings.csv'
try:
    df = pd.read_csv(
        EMB_PATH,
        converters={
            'Embedding': lambda s: np.array(ast.literal_eval(s)) 
                              if isinstance(s, str) else np.array(s)
        }
    )
    print(f"✅ Loaded {len(df)} rows from {EMB_PATH}")
except Exception as e:
    print(f"❌ Failed to load embeddings: {e}")
    exit(1)

# ─── 3) Load artifacts ─────────────────────────────────────────────────────────
ENCODER_PATH = '../Models/tei_label_encoder.pkl'
MODEL_PATH   = '../Models/tei_rf_model.pkl'
try:
    label_encoder = joblib.load(ENCODER_PATH)
    clf           = joblib.load(MODEL_PATH)
    print("✅ Loaded label encoder and RandomForest model.")
except Exception as e:
    print(f"❌ Failed to load artifacts: {e}")
    exit(1)

# ─── 4) Inference helpers ───────────────────────────────────────────────────────
def get_embedding(text: str, model: str = "text-embedding-3-small") -> list:
    try:
        resp = client.embeddings.create(input=[text], model=model)
        return resp.data[0].embedding
    except OpenAIError as e:
        raise RuntimeError(f"OpenAI embedding error: {e}")

def predict_tei_factor(text: str) -> str:
    emb  = get_embedding(text)
    code = clf.predict([emb])[0]
    return label_encoder.inverse_transform([code])[0]

# ─── 5) Test on representative TEI examples ─────────────────────────────────────
examples = {
    "Perceiving Emotions":  "I notice subtle changes in people's facial expressions and tone.",
    "Using Emotions":       "I harness my feelings to solve problems and make decisions.",
    "Understanding Emotions":"I analyze why I feel a certain way and anticipate how emotions evolve.",
    "Managing Emotions":    "I regulate my mood and adaptively respond to stressful situations.",
}

print("\n--- TEI Model Predictions on New Examples ---\n")
for factor, sample in examples.items():
    try:
        pred = predict_tei_factor(sample)
        print(f"{factor}\n  Input: “{sample}”\n  → Predicted TEI Factor: {pred}\n")
    except Exception as ex:
        print(f"❌ Error on '{factor}': {ex}\n")
````

````python
#!/usr/bin/env python3
import os
import pandas as pd
from dotenv import load_dotenv
from py2neo import Graph, Node, Relationship

# ─── 1) Load env & connect to Neo4j ────────────────────────────────────────────
load_dotenv()
raw_uri = os.getenv("NEO4J_URI", "")
if raw_uri.startswith("neo4j://"):
    bolt_uri = "bolt://" + raw_uri[len("neo4j://"):]
else:
    bolt_uri = raw_uri

user = os.getenv("NEO4J_USERNAME")
pwd  = os.getenv("NEO4J_PASSWORD")
if not all([bolt_uri, user, pwd]):
    raise RuntimeError("Make sure NEO4J_URI, NEO4J_USERNAME & NEO4J_PASSWORD are set")

graph = Graph(bolt_uri, auth=(user, pwd))
print("🔗 Connected to Neo4j via", bolt_uri)

# ─── 2) Clear existing data ────────────────────────────────────────────────────
graph.run("MATCH (n) DETACH DELETE n")
print("🗑️  Cleared all existing nodes and relationships")

# ─── 3) Load TEI CSV ───────────────────────────────────────────────────────────
tei_csv = "../Datasets/tei.csv"
df = pd.read_csv(tei_csv)
print(f"✅ Loaded {len(df)} rows from {tei_csv}")

# ─── 4) Build graph ────────────────────────────────────────────────────────────
tx = graph.begin()

# caches to avoid duplicates
domains, factors, adjs, syns, verbs, nouns = {}, {}, {}, {}, {}, {}

for _, row in df.iterrows():
    # helper to safely extract and strip text or return ''
    def get_text(col):
        val = row.get(col)
        return val.strip() if pd.notna(val) and isinstance(val, str) else ""

    d = get_text("Domain")
    f = get_text("Factor")
    a = get_text("Adjective")
    s = get_text("Synonym")
    v = get_text("Verb")
    n = get_text("Noun")

    # Domain
    if d and d not in domains:
        domains[d] = Node("Domain", name=d)
        tx.create(domains[d])

    # Factor
    if f and f not in factors:
        factors[f] = Node("Factor", name=f)
        tx.create(factors[f])
        tx.create(Relationship(domains[d], "HAS_FACTOR", factors[f]))

    # Adjective
    if a and a not in adjs:
        adjs[a] = Node("Adjective", text=a)
        tx.create(adjs[a])
        tx.create(Relationship(factors[f], "HAS_ADJECTIVE", adjs[a]))

    # Synonym
    if s and s not in syns:
        syns[s] = Node("Synonym", text=s)
        tx.create(syns[s])
        if a:
            tx.create(Relationship(adjs[a], "HAS_SYNONYM", syns[s]))

    # Verb
    if v and v not in verbs:
        verbs[v] = Node("Verb", text=v)
        tx.create(verbs[v])
        if a:
            tx.create(Relationship(adjs[a], "HAS_VERB", verbs[v]))

    # Noun
    if n and n not in nouns:
        nouns[n] = Node("Noun", text=n)
        tx.create(nouns[n])
        if a:
            tx.create(Relationship(adjs[a], "HAS_NOUN", nouns[n]))

tx.commit()
print("🎉 Completed building the TEI taxonomy graph in Neo4j.")
````

````python
#!/usr/bin/env python3
import os
import ast
import pandas as pd
from dotenv import load_dotenv
from py2neo import Graph, Node, Relationship

# ─── Load env & set up connection ─────────────────────────────────────────────
load_dotenv()
raw = os.getenv("NEO4J_URI", "")
bolt_uri = raw.replace("neo4j://", "bolt://") if raw.startswith("neo4j://") else raw
user     = os.getenv("NEO4J_USERNAME")
pwd      = os.getenv("NEO4J_PASSWORD")
if not all([bolt_uri, user, pwd]):
    raise RuntimeError("Please set NEO4J_URI, NEO4J_USERNAME & NEO4J_PASSWORD")
graph = Graph(bolt_uri, auth=(user, pwd))
print("🔗 Connected to Neo4j via", bolt_uri)

# ─── Clear existing graph ────────────────────────────────────────────────────
graph.run("MATCH (n) DETACH DELETE n")
print("🗑️  Cleared all existing nodes and relationships")

# ─── Load TEI dataset ─────────────────────────────────────────────────────────
df = pd.read_csv(
    '../Datasets/tei.csv',
    converters={'Embedding': lambda s: ast.literal_eval(s) if isinstance(s, str) else s}
)
print(f"✅ Loaded {len(df)} rows from ../Datasets/tei.csv")

# ─── Build taxonomy graph ────────────────────────────────────────────────────
tx = graph.begin()
domains, factors, adjectives = {}, {}, {}

for _, row in df.iterrows():
    dom, fac = row['Domain'], row['Factor']
    adj      = row['Adjective']
    syn      = row['Synonym'] if pd.notna(row['Synonym']) else ""
    verb     = row['Verb']    if pd.notna(row['Verb'])    else ""
    noun     = row['Noun']    if pd.notna(row['Noun'])    else ""

    # Domain node
    if dom not in domains:
        domains[dom] = Node("Domain", name=dom)
        tx.create(domains[dom])

    # Factor node
    if fac not in factors:
        factors[fac] = Node("Factor", name=fac)
        tx.create(factors[fac])
        tx.create(Relationship(domains[dom], "HAS_FACTOR", factors[fac]))

    # Adjective node
    if adj not in adjectives:
        adjectives[adj] = Node("Adjective", name=adj)
        tx.create(adjectives[adj])
        tx.create(Relationship(factors[fac], "HAS_ADJECTIVE", adjectives[adj]))

    parent = adjectives[adj]

    # Synonym node (if any)
    if syn:
        syn_node = Node("Synonym", name=syn)
        tx.merge(syn_node, "Synonym", "name")
        tx.create(Relationship(parent, "HAS_SYNONYM", syn_node))
        parent = syn_node

    # Verb node (if any)
    if verb:
        verb_node = Node("Verb", name=verb)
        tx.merge(verb_node, "Verb", "name")
        tx.create(Relationship(parent, "HAS_VERB", verb_node))
        parent = verb_node

    # Noun node (if any)
    if noun:
        noun_node = Node("Noun", name=noun)
        tx.merge(noun_node, "Noun", "name")
        tx.create(Relationship(parent, "HAS_NOUN", noun_node))

# Commit everything
graph.commit(tx)
print("🎉 Completed building the TEI taxonomy graph in Neo4j.")
````

## Process for Building the TEI Personality Model

1. **Data Preparation**

   * Define taxonomy (`Factor \to Adjective \to Synonym/Verb/Noun`).
   * Normalize lists and flatten to CSV (`TEI.csv`).

2. **Embedding Generation**

   * Load `TEI.csv` into pandas.
   * Initialize OpenAI client and MLflow.
   * Generate embeddings via `client.embeddings.create(...)` for full text prompt.
   * Save `TEI_embeddings.csv` with `Embedding` column.

3. **Label Encoding**

   * Load embeddings CSV.
   * Fit `LabelEncoder` on `Factor` column.
   * Save encoder (`TEI_label_encoder.pkl`).
   * Visualize mapping with bar chart and log to MLflow.

4. **Model Training**

   * Load embeddings and label encoder.
   * Split into train/test (stratify when possible).
   * Train `RandomForestClassifier`.
   * Log metrics, artifacts, and model to MLflow.

5. **Model Evaluation**

   * Reload model and encoder.
   * Split and predict on test set.
   * Compute accuracy, classification report, confusion matrix.
   * Log evaluation artifacts to MLflow.
   * Perform cross-validation (LOO if class imbalance).

6. **Inference on New Data**

   * Load model & encoder.
   * Define `predict_factor(text)` helper that requests embedding and predicts.
   * Test on meaningful example sentences.
   * Log predictions to MLflow.

7. **Visualization & Clustering**

   * Train RF on full dataset.
   * In‑sample evaluation & plot confusion matrix.
   * PCA into 2D & scatter by encoded class.
   * K‑Means clustering on embeddings, save `TEI_clustered_embeddings.csv`.
   * Plot clusters in PCA space.

8. **Neo4j Graph Construction**

   * Clear existing graph: `MATCH (n) DETACH DELETE n`.
   * Ingest `TEI.csv`, create nodes for Factor, Adjective, Synonym, Verb, Noun.
   * Create `:RELATES_TO` relationships following taxonomy hierarchy.

9. **Cypher Query to View Graph**

   ```cypher
   MATCH p = (f:Factor)-[:HAS_ADJECTIVE]->(a:Adjective)
             -[:HAS_SYNONYM|HAS_VERB|HAS_NOUN]->(x)
   RETURN p LIMIT 100
   ```
