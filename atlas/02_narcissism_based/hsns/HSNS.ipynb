# Hypersensitive Narcissism Scale (HSNS)

## Structure and Development  
The **Hypersensitive Narcissism Scale (HSNS)** is a brief self-report measure designed to assess *covert* or *vulnerable* narcissism ([Hypersensitive Narcissism Scale - Wikipedia](https://en.wikipedia.org/wiki/Hypersensitive_Narcissism_Scale#:~:text=The%20Hypersensitive%20Narcissism%20Scale%20,point%20scale)) ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=The%20HSNS%20is%20a%2010,Hendin%20%26%20Cheek%2C%201997)). It was developed by Holly M. Hendin and Jonathan M. Cheek in 1997 ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=A%20new%20measure%20of%20hypersensitive,Results%20support%20P)) as a refinement of earlier narcissism measures. The HSNS consists of **10 items**, each rated on a five-point Likert scale (from “very uncharacteristic” to “very characteristic”), yielding a total score from 10 to 50 ([Hypersensitive Narcissism Scale - Wikipedia](https://en.wikipedia.org/wiki/Hypersensitive_Narcissism_Scale#:~:text=The%20Hypersensitive%20Narcissism%20Scale%20,point%20scale)) ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=The%20HSNS%20is%20a%2010,Hendin%20%26%20Cheek%2C%201997)). Administration is straightforward: respondents indicate agreement with statements reflecting feelings and behaviors (e.g. *“My feelings are easily hurt by ridicule or the slighting remarks of others.”*) that tap a narcissistic personality marked by sensitivity and defensiveness. 

**Theoretical foundation:** The HSNS was inspired by the distinction between **overt (grandiose)** and **covert (vulnerable)** narcissism in personality theory ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). Classic psychodynamic thinkers like Kernberg (1975) and Kohut (1977) proposed that narcissism has an outward, flamboyant form and a more inward, hypersensitive form ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=Narcissism%20is%20defined%20as%20a,measurably%20differentiated%2C%20and%20not%20correlated)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). Wink (1991) empirically confirmed two “faces” of narcissism – **Grandiosity-Exhibitionism** (overt) and **Vulnerability-Sensitivity** (covert) – which were found to be essentially uncorrelated in normal populations ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). Despite their different interpersonal styles, both types share an underlying **sense of entitlement** and **grandiose fantasies** ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=scales,measures%20that%20were%20studied%20by)). Hendin and Cheek (1997) built on this framework to create a scale specifically capturing the *covert/vulnerable* narcissistic traits often overlooked by existing measures ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=A%20new%20measure%20of%20hypersensitive,Results%20support%20P)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,Results%20support%20P)). They started with **H. A. Murray’s (1938) Narcism Scale**, a set of face-valid narcissism items developed in an early study of Harvard undergraduates ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=DSM,via%20his%20exploratory%20study%20of)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=the%20less%20well%20known%20MMPI,via%20his%20exploratory%20study%20of)). Using an empirical approach, they selected the subset of 10 items that best correlated with a composite index of **“covert narcissism”** derived from the MMPI (combining two MMPI-based narcissism scales reflecting hypersensitivity and narcissistic vulnerability) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=in%20which%20they%20developed%20an,Hypersensitivity%20Scale%29%20as)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=The%20composite%20MMPI,%CE%B1%205)). The resulting HSNS captures a personality profile of someone who is self-absorbed, easily hurt by criticism, anxious in social contexts, and defensive or resentful in response to others’ demands ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,worrying%20about%20other%20people%27s%20troubles)) ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,for%20my%20time%20and%20sympathy)). In the authors’ words, *“we view the covert ‘face’ of narcissism as assessed by the HSNS to be one of many facets of the higher order construct that Maslow (1942) labeled psychological insecurity and that has been more recently called negative emotionality”* ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,%E2%80%A6%E2%80%8B)). Thus, the HSNS is theoretically grounded in the idea of a narcissism variant characterized by **insecurity, sensitivity, and introversion**, rather than arrogance or exhibitionism. 

**Traits and subscales:** The HSNS was originally presented as a **unidimensional scale**, meaning all 10 items together indicate a single covert narcissism trait ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=The%20Hypersensitive%20Narcissism%20Scale%20,The%20two)). Key themes measured by the items include **hypersensitivity to criticism**, **social anxiety or self-consciousness**, **feelings of inadequacy or shame**, **resentment of others**, and **self-absorbed attitudes** ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,worrying%20about%20other%20people%27s%20troubles)) ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,for%20my%20time%20and%20sympathy)). For example, items probe tendencies like being easily hurt or offended, feeling that one is different from others, disliking to share credit, or being secretly annoyed by others’ requests for support ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,worrying%20about%20other%20people%27s%20troubles)) ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,for%20my%20time%20and%20sympathy)). There are no official subscales in the original HSNS; it yields a single score representing the overall level of *hypersensitive (vulnerable) narcissism*. However, subsequent research has found that the HSNS items can be grouped into two related facets: one reflecting **hypersensitivity/neuroticism** and the other reflecting **egocentrism/entitlement** ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=from%20three%20large%20samples%20,research%20reconciles%20the%20HSNS%20with)). These correspond to the idea that vulnerable narcissism has both an **internalizing, defensive side** (e.g. feeling insecure, anxious and easily wounded) and an **antagonistic, self-centered side** (e.g. being preoccupied with oneself, unwilling to empathize or share glory). Later analyses (discussed below) have labeled these facets *“Oversensitivity”* vs. *“Egocentrism”* ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=from%20three%20large%20samples%20,research%20reconciles%20the%20HSNS%20with)). In practical terms, the HSNS is typically used as a single scale, but awareness of these subcomponents helps in interpreting the nuanced traits it measures. 

## History and Timeline  
**1938 – Murray’s Narcism Scale:** Psychologist Henry A. Murray included a **Narcism scale** in his seminal work *Explorations in Personality* (1938). Murray’s concept of narcissism was remarkably prescient: he viewed narcissism as a blend of **vulnerability and self-absorption**, noting that narcissistic individuals may at times appear **grandiose and attention-seeking**, yet also experience **anxious self-preoccupation, hypersensitivity, feelings of neglect, and persecution** ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=DSM,via%20his%20exploratory%20study%20of)). Murray’s Narcism Scale items reflected these dual dynamics (some items tapping overt egotism and others covert defensiveness). This early measure fell out of use but later served as the basis for the HSNS. 

**1970s–1980s – Narcissism in Personality Research:** In the wake of Narcissistic Personality Disorder entering the DSM-III (1980), researchers developed tools to measure narcissistic traits in non-clinical samples ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=The%20clinical%20diagnostic%20criteria%20for,Grisham%2C%20Trotter%2C%20%26%20Biderman%2C%201984)). The **Narcissistic Personality Inventory (NPI)**, a measure of *grandiose* narcissism, was introduced by Raskin & Hall (1979) and refined by Raskin & Terry (1988). Studies soon showed that the NPI (overt narcissism) had low correlations with certain clinical narcissism scales that seemed to capture a more *hypersensitive* form ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=The%20clinical%20diagnostic%20criteria%20for,Grisham%2C%20Trotter%2C%20%26%20Biderman%2C%201984)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=principal,based)). For example, Wink (1991) found the NPI and a Narcissistic Personality Disorder scale were uncorrelated, suggesting two distinct dimensions of narcissism ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). This underscored the need for a dedicated measure of **covert/vulnerable narcissism**. Raskin & Terry (1988) specifically pointed to Murray’s long-forgotten Narcism Scale as a promising instrument for the covert side ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=DSM,via%20his%20exploratory%20study%20of)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=the%20less%20well%20known%20MMPI,via%20his%20exploratory%20study%20of)). 

**1991 – Two “Faces” of Narcissism:** **Paul Wink (1991)** published a influential paper empirically splitting narcissistic traits into two factors across multiple samples ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). He labeled them **Grandiosity-Exhibitionism** (the flamboyant, self-aggrandizing face measured by NPI and similar scales) and **Vulnerability-Sensitivity** (a shy, defensive form closer to clinical descriptions of narcissistic insecurity) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=psychodynamic%20accounts%20%28e,form%20have%20become%20accepted%20and)). Wink showed that people could score high on one and not the other, supporting the idea that overt and covert narcissism are independent. Notably, Wink observed that *both types share an underlying narcissistic core of entitlement and fantasy*, even if their outward behaviors differ ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=scales,measures%20that%20were%20studied%20by)). This work set the stage for developing covert narcissism measures, directly inspiring Hendin & Cheek. 

**1997 – HSNS Development:** **Hendin and Cheek (1997)** published the HSNS in the *Journal of Research in Personality* ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=ARTICLE%20NO,10%20items)). To create it, they reexamined **Murray’s (1938) Narcism Scale** by administering it to college students and correlating each item with a composite criterion of **covert narcissism** ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=in%20which%20they%20developed%20an,Hypersensitivity%20Scale%29%20as)). The criterion was a 35-item amalgam of two MMPI-based scales – the **Narcissistic Personality Disorder Scale (NPDS)** and a **Narcissism-Hypersensitivity scale** – identified by Wink (1991) as measuring the vulnerable narcissistic syndrome ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=principal,based)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=The%20composite%20MMPI,%CE%B1%205)). Out of Murray’s item pool, **10 items** showed strong correlations with this composite and were chosen for the HSNS, which demonstrated good internal consistency (Cronbach’s α ≈ .75) in three college samples ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=A%20new%20measure%20of%20hypersensitive,Results%20support%20P)). Crucially, Hendin & Cheek found the HSNS **correlated near zero with the NPI** (grandiose narcissism measure), while mirroring the MMPI composite’s pattern of correlations with the Big Five traits ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)). These results validated that HSNS indeed tapped a different “face” of narcissism. The authors urged personality researchers to assess **both overt and covert narcissism** in the future, as each relates to different personality profiles and outcomes ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=Personality%20Inventory%2C%20which%20assesses%20overt,narcissistic%20tendencies%20in%20the%20normal)). 

**2003 – Validation in Personality Disorders Research:** The concept of vulnerable narcissism gained further support in clinical research. For example, Dickinson & Pincus (2003) provided an **interpersonal analysis of grandiose vs. vulnerable narcissism**, highlighting that the vulnerable type is marked by **shame, low self-esteem, and sensitivity to criticism** ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=13.%20)). Although their study focused on pathological narcissism broadly, it reinforced the distinct characteristics that instruments like the HSNS measure (vulnerable narcissism was associated with feelings of inadequacy and interpersonal avoidance, unlike the domineering grandiose type) ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=13.%20)). Such findings helped legitimize the HSNS’s target construct within psychopathology circles. 

**2008 – Psychometric Review (Arble, 2008):** A master’s thesis by E. P. **Arble (2008)** took a closer look at the HSNS’s reliability and factor structure ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=Recent%20research%20has%20suggested%20that,The)) ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=validity%2C%20producing%20a%20pattern%20of,psychometric%20properties%20and%20clinical%20utility)). Using data from 298 adults, Arble found the HSNS had **high internal reliability** and clear discriminant validity – its pattern of correlations with other traits fit the theoretical profile of covert narcissism (e.g. positively related to neuroticism and introversion, and uncorrelated with overt narcissism) ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=completed%20the%20measure%2C%20as%20well,psychometric%20properties%20and%20clinical%20utility)). However, **factor analysis suggested the HSNS might not be purely unidimensional** ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=validity%2C%20producing%20a%20pattern%20of,psychometric%20properties%20and%20clinical%20utility)). Some items clustered separately, hinting at multiple components within the scale. The relationship between HSNS and NPI in Arble’s sample was also “somewhat unclear” (i.e. not strongly negative or positive) ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=validity%2C%20producing%20a%20pattern%20of,psychometric%20properties%20and%20clinical%20utility)). The thesis concluded that HSNS is a valid measure of covert narcissism, but noted its **shortcomings (e.g. possible multidimensionality)** and suggested a **revision or expansion** to improve its psychometric properties ([
"Evaluating the psychometric properties of the hypersensitive narcissis" by Eamonn Patrick Arble
](http://commons.emich.edu/theses/236/#:~:text=assessing%20a%20unitary%20construct%2C%20and,psychometric%20properties%20and%20clinical%20utility)). This was an early call to refine the HSNS for better precision. 

**2009 – Italian Validation (Fossati et al., 2009):** Researchers led by **Andrea Fossati** conducted a study titled *“Tracking the hypersensitive dimension in narcissism: Reliability and validity of the HSNS.”* This was published in *Personality and Mental Health (2009)* ([
	Fossati, A., Borroni, S., Grazioli, F., et al. (2009) Tracking the Hypersensitive Dimension in Narcissism Reliability and Validity of the Hypersensitive Narcissism Scale. Personality & Mental Health, 3, 235-247.  - References - Scientific Research Publishing
](https://www.scirp.org/(S(351jmbntvnsjt1aadkozje))/reference/referencespapers?referenceid=2223600#:~:text=Fossati%2C%20A,247.%20https%3A%2F%2Fdoi.org%2F10.1002%2Fpmh.92)). They translated the HSNS into Italian and administered it to clinical and nonclinical samples. Fossati’s team confirmed that the HSNS was reliable and valid in another culture, and notably they found evidence of a **two-factor structure** ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=construction%20article%2C%20the%2010%20items,%282009%3B%20Egocentrism%20and%20Oversensitivity)). Their factor analysis of the 10 items yielded two interpretable components: one reflecting **Egocentrism** (self-centered, indifferent or resentful toward others) and the other reflecting **Oversensitivity to Judgment** (insecure, hypersensitive, and shy) ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)). This two-factor interpretation aligned with the theorized facets of vulnerable narcissism and provided an empirical basis for treating HSNS as having subdimensions. Fossati et al. still supported using the total HSNS score, but their work spurred further research into the scale’s internal structure ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)). 

**2013 – Expanded HSNS (Hendin, Cheek & Wink, 2013):** To address the HSNS’s brevity, its original authors collaborated on an expanded version. At the Association for Research in Personality (ARP) 2013 conference, **Cheek, Hendin, & Wink** presented the **“Maladaptive Covert Narcissism Scale (MCNS)”**, a **23-item expanded HSNS** ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=The%20purpose%20of%20the%20present,1997)). Using a large internet sample (Amazon MTurk, N=420), they selected additional items to cover more content without sacrificing the focus on vulnerable narcissism ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=The%20purpose%20of%20the%20present,1997)). The 23-item MCNS achieved **α ≈ .89** (much improved over the original .75) and correlated r≈.65 with the MMPI composite of covert narcissism, compared to r≈.55 for the 10-item HSNS ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=Cheek%2C%201997,the%20new%20scale%20represents%20a)). In a second college sample, the new scale showed similarly strong reliability and valid correlations ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=expanded%20version%20of%20the%20Maladaptive,An%20expanded%20version%20of)). The authors reported that the MCNS retained the same pattern of Big Five correlations observed by Hendin & Cheek (1997), suggesting it measured the same construct but more robustly ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=with%20the%20MMPI%20composite%20measure,The%20Maladaptive%20Covert%20Narcissism%20Scale)). They concluded that the expanded scale was a “significant improvement” for assessing maladaptive *covert narcissism* ([(PDF) An Expanded Version of the Hypersensitive Narcissism Scale (The Maladaptive Covert Narcissism Scale)](https://www.researchgate.net/publication/281624968_An_Expanded_Version_of_the_Hypersensitive_Narcissism_Scale_The_Maladaptive_Covert_Narcissism_Scale#:~:text=MMPI%20measure%20of%20covert%20narcissism%2C,Research%20in%20Personality%2C%20Charlotte%2C%20NC)). (As of 2013, this was a conference poster; it signaled ongoing refinement, though the MCNS is not yet as widely used as the original HSNS in publications.) 

**2014 – Comparative Studies:** By the 2010s, the HSNS had become a **widely used index of vulnerable narcissism** in research, often alongside grandiose measures for comparison. For instance, Miller et al. (2014) conducted a comprehensive comparison of popular narcissism scales (including NPI, HSNS, the Pathological Narcissism Inventory, and others) and their relations to external criteria ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=scales%20%28i,ratings%20on%20the%20Big%20Five)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=narcissism%2C%20including%20the%20HSNS%2C%20PNI,4%05%20measure%20of%20NPD)). They found that HSNS consistently clustered with other **“vulnerable” narcissism measures** and showed little overlap with “grandiose” measures, reinforcing the distinction. Around the same time, a Polish version (Czarna et al., 2014) was used to link vulnerable narcissism with emotional intelligence and other traits ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=To%20assess%20vulnerable%20narcissism%20we,My)). Such studies extended the HSNS’s validity evidence across diverse populations and contexts (e.g., linking high HSNS scores to greater neuroticism, lower extraversion, and unique outcomes like different patterns of emotional processing). 

**2015 – Large Sample Factor Analysis (Cheek et al., 2015):** Jonathan Cheek and colleagues followed up on the structural question in a 2015 ARP conference paper titled “Factor Structure of the HSNS.” They analyzed HSNS responses from two very large English-speaking samples (total N≈28,800) ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=researchers%20studying%20an%20Italian%20translation,components%20should%20be%20explored%20further)). Using more sophisticated tests, they again found **two factors** fit the data better than one ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)). Cheek et al. named the two subscales **“Self-Absorption”** and **“Hypersensitivity,”** noting these were conceptually equivalent to the factors identified by Fossati et al. (2009) – i.e. an *Egocentrism* factor and an *Oversensitivity* factor, respectively ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)). They recommended that researchers “explore these two components further in ongoing work on revising and expanding the HSNS” ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=factors%20to%20rotate%20indicated%20two,item%20HSNS)). This work, along with other psychometric studies, solidified the understanding that while the HSNS total score is useful, its item content is multifaceted. 

**2020s – Ongoing Research:** The HSNS continues to be studied and utilized. A recent effort by Schneider et al. (2023/2025) created a **German adaptation of the HSNS** and conducted thorough construct validation ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=The%20Hypersensitive%20Narcissism%20Scale%20,to%20map%20the%20proposed%20HSNS)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=from%20three%20large%20samples%20,research%20reconciles%20the%20HSNS%20with)). Across thousands of participants, they confirmed the **two correlated dimensions** of **Oversensitivity** and **Egocentrism** within the HSNS and mapped these onto broader personality and psychopathology models ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=from%20three%20large%20samples%20,research%20reconciles%20the%20HSNS%20with)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). They found that the **Oversensitivity subscale** (comprising the emotionally fragile, self-critical aspects) predicted *internalizing problems* like anxiety, depression, and low self-esteem, aligning with “neurotic narcissism” ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=respect%20to%20various%20personality%20traits%2C,%28PsycInfo%20Database)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). The **Egocentrism subscale** (comprising the selfish, indifferent aspects) was associated with *antagonism*, low agreeableness, and externalizing tendencies (e.g. hostility), aligning it with the more *entitled/hostile* side of narcissism ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=respect%20to%20various%20personality%20traits%2C,%28PsycInfo%20Database)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). This study, published in *Psychological Assessment* in 2025, helps reconcile the HSNS with other multidimensional narcissism frameworks and underscores its continued relevance ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=Egocentrism%20overlaps%20with%20antagonistic%20narcissism%2C,to%20its%20utility%20to%20capture)). Meanwhile, cross-cultural validations (e.g. a Chinese version in 2024) have demonstrated that the HSNS reliably measures the same construct abroad, with good internal consistency (~0.78) and test-retest reliability ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=Results%20The%20exploratory%20factor%20analysis,validity%20reached%20a%20significant%20level)) ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=consistency%20reliability%20of%20the%20scale,validity%20reached%20a%20significant%20level)). These modern studies show the HSNS has stood the test of time, while also benefiting from refinements in interpretation (recognition of subfactors) and application (use in diverse samples). 

## Academic Research Findings and Comparisons  
**Personality profile of HSNS:** Individuals who score high on the HSNS tend to have a distinctive personality profile. Consistent with the “vulnerable” narcissism characterization, HSNS scores correlate **positively with Neuroticism** and **negatively with Extraversion and Agreeableness** on the Big Five dimensions ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=Narcissism%20Scale%2C%20and%20the%20NPI,2.18)). For example, Hendin & Cheek (1997) reported HSNS had a substantial **positive correlation with Neuroticism (r ≈ +.51)** and moderate **negative correlations with Extraversion (r ≈ –.28)** and **Agreeableness (r ≈ –.44) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=Narcissism%20Scale%2C%20and%20the%20NPI,2.18))**. This indicates that hypersensitive narcissists are typically anxious, emotionally volatile, introverted, and antagonistic (skeptical or critical of others). In contrast, HSNS scores showed little relationship to Conscientiousness or Openness (small negatives) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=Narcissism%20Scale%2C%20and%20the%20NPI,2.18)), suggesting that vulnerable narcissism is not defined by those domains. These traits diverge sharply from the profile of grandiose narcissists, who are often low in Neuroticism and high in Extraversion. Indeed, the **HSNS correlates near zero (or even slightly negatively) with the Narcissistic Personality Inventory (NPI)** ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,Results%20support%20P)), the classic measure of overt narcissism. This near-zero correlation, found in multiple studies, emphasizes that **HSNS captures a different form of narcissism** than the NPI – one more aligned with emotional insecurity than with bold self-confidence ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)). Moreover, HSNS correlates **positively with measures of introversion, anxiety, and vulnerability**. For instance, it has been linked to **low self-esteem and proneness to shame** ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=Narcissism%20Scale,to%20completely%20respond%20to%20the)), consistent with clinical observations that vulnerable narcissists privately feel inadequate even as they harbor grandiose wishes ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=13.%20)). One study noted that *rejection sensitivity* is a hallmark of hypersensitive narcissists – they have an intense fear of rejection and criticism, leading to social withdrawal and high rumination ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=%3B%20hypersensitive%20narcissism%20is%20also,10)). As a result, high-HSNS individuals often come across as **defensive, shy/avoidant, and easily hurt** in interpersonal situations ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=manner%20,2011%20%2C%20%2040)). They crave recognition to bolster their weak self-worth, but if they *feel* undervalued, they may withdraw or react with silent resentment rather than overt aggression ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=manner%20,2011%20%2C%20%2040)). This pattern can lead to difficulties in relationships and mental health. Research has found, for example, that **vulnerable narcissism (HSNS) is a risk factor for depression** – it correlates with greater depressive symptoms and may predict higher depression severity in those prone to mood disorders ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=researchers%20are%20devoting%20their%20energies,sensitivity%20is%20the%20most%20fundamental)). It is also associated with **anxiety and social insecurity**, and some evidence links it to health impacts (one study noted chronic hypersensitive narcissistic tendencies might elevate stress-related health risks like cardiovascular issues) ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=%3B%20hypersensitive%20narcissism%20is%20also,in%20social%20relationships%20raise%20the)) ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=narcissism,11)). On the interpersonal side, vulnerable narcissists do not typically lash out with open aggression unless their ego is provoked; however, when their self-image is challenged, they can display **hostile or aggressive responses** rooted in their fragile self-esteem (sometimes termed “narcissistic rage”) ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=research%20support%20that%20narcissism%20is,esteem%20and%20uncertainty%20of)). In summary, academic findings paint HSNS high-scorers as **emotionally reactive, privately self-important individuals who lack the outward confidence of grandiose narcissists**. They struggle with inner shame and insecurity, yet maintain a self-focus and entitlement that can strain their relationships. 

**Use in research settings:** The HSNS has been widely used in both **personality psychology** and **clinical psychology** research as the go-to measure of **vulnerable narcissism** in non-clinical samples. Its brevity and clear focus make it convenient for large surveys and for inclusion alongside other measures. Researchers often administer the HSNS in tandem with the NPI (grandiose narcissism) to investigate how the two forms of narcissism relate differently to outcomes. For example, studies have examined how each type relates to emotional intelligence, relationship behaviors, social media usage, aggression, etc. A consistent finding is that **HSNS (vulnerable narcissism) relates to internalizing problems and inhibited behavior**, whereas NPI (grandiose narcissism) relates to externalizing behaviors and confidence. One study on emotional intelligence found that vulnerable narcissism (HSNS) was associated with lower self-reported emotional ability and more distress in handling emotions, whereas grandiose narcissism showed an opposite or nil pattern ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=manner%20,2011%20%2C%20%2040)). In personality disorder research, the HSNS has been used as a proxy for covert narcissistic traits, helping to identify individuals who might not appear narcissistic on the surface (since they may not overtly brag or seek attention) but still harbor narcissistic needs and vulnerabilities. Notably, **clinical samples** have been studied with the HSNS as well. For instance, substance abuse treatment populations have been assessed, finding that HSNS scores can predict certain maladaptive narcissistic features in those clients ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=narcissism%20is%20a%20major%20predictive,At%20the%20same%20time%2C%20hypersensitive)). Such findings suggest hypersensitive narcissism may complicate treatment, as these individuals might struggle with criticism or feedback during therapy. Overall, the HSNS has proven useful for exploring the *“shy” or covert manifestations of narcissism* in varied contexts, from college student personality surveys to psychiatric assessments, providing a fuller picture of narcissistic personality beyond the grandiose stereotype. Its continued use in contemporary research (often alongside multi-facet instruments) attests to its value in capturing an important dimension of personality. 

**Comparisons with other narcissism models:** The HSNS primarily targets what is now commonly called **vulnerable narcissism**, and it has been directly compared with other narcissism inventories: 

- **NPI (Narcissistic Personality Inventory):** The NPI is the classic measure of *grandiose narcissism* – characterized by assertiveness, high self-esteem, exhibitionism, and dominance. As noted, HSNS and NPI scores tend to be uncorrelated or even slightly negatively correlated ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,Results%20support%20P)). Empirically, they form distinct factors. The NPI correlates with traits like high Extraversion and low Neuroticism (opposite of HSNS) ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=Grandiose%20narcissism%20is%20characterized%20by,agreeableness%2C%20and%20positively%20with%20extraversion)) ([
            Vulnerable and Grandiose Narcissism Are Differentially Associated With Ability and Trait Emotional Intelligence - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC6120976/#:~:text=omnipotence,22%3B%20Zajenkowski%20and%20Czarna%2C%202015)). When factor analyses combine NPI and HSNS items, they separate cleanly into different components (one for overt, one for covert narcissism). Thus, HSNS measures aspects of narcissism that the NPI misses – namely **insecurity, defensiveness, and introverted self-absorption**, which is why incorporating HSNS can improve our ability to predict outcomes like interpersonal anxiety or depression that NPI alone cannot. 

- **PNI (Pathological Narcissism Inventory):** The PNI (Pincus et al., 2009) is a longer instrument that assesses both grandiose and vulnerable narcissistic traits, yielding two broad factors: **Narcissistic Grandiosity** and **Narcissistic Vulnerability**. The Vulnerability factor of the PNI covers things like contingent self-esteem, hiding of the self, entitlement rage, and dejection – all conceptually similar to HSNS content. In fact, HSNS shows a **strong positive correlation with the PNI’s vulnerable narcissism dimension** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=match%20at%20L312%20FFNI%20Vulnerable%2C,30%20facets%20of%20the%20NEO)). Studies have found these two measures largely converge; for example, one report noted the intercorrelation of HSNS with PNI Vulnerability (and with another vulnerable measure) was about r = .70 ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=match%20at%20L312%20FFNI%20Vulnerable%2C,30%20facets%20of%20the%20NEO)). Both correlate with Neuroticism and negative affect, and both diverge from PNI’s grandiose facets. Essentially, HSNS can be seen as a *brief unidimensional proxy* for what the PNI’s vulnerable factor captures. However, the PNI, being more comprehensive, also captures some clinical nuance (like dysregulation and grandiose fantasies) that HSNS does not directly ask about. Researchers have observed that HSNS correlates highly with **self-reported shame, neediness, and depressive affects**, which overlap a great deal with PNI Vulnerability ([Validity and reliability of the Hypersensitive Narcissism Scale (HSNS) in China | medRxiv](https://www.medrxiv.org/content/10.1101/2024.09.28.24312621v1.full-text#:~:text=13.%20)). Conversely, HSNS has minimal correlation to the PNI’s Grandiosity factor (which includes exploitative tendencies and fantasies of power). Therefore, HSNS fits neatly into the **nomological network** of narcissism as essentially measuring the *vulnerable pole* of narcissistic personality ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). 

- **FFNI (Five Factor Narcissism Inventory):** The FFNI (Glover et al., 2012) is a trait-based measure that breaks narcissism into 15 lower-level facets aligned with the Five-Factor Model, and also yields composite scores for **Grandiose** and **Vulnerable** narcissism. In comparisons, the HSNS correlates very strongly with the **FFNI Vulnerable** score and its constituent facets (like **Distrust, Shame, Need for Admiration**, etc.), which reflect high Neuroticism and low Agreeableness facets of narcissism ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=FFNI%20G%20PNI%20G%20NPI,47)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=narcissism%2C%20including%20the%20HSNS%2C%20PNI,4%05%20measure%20of%20NPD)). Miller et al. (2014) reported that HSNS’s profile across the NEO-PI facets was almost identical to the FFNI’s vulnerable narcissism profile ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=FFNI%20G%20PNI%20G%20NPI,52)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=FFNI%20G%20PNI%20G%20NPI,47)). For instance, both show high correlations with NEO facets such as Anxiety, Depression, Self-Consciousness, and low trust/modesty. HSNS also converges with FFNI on predicting **internalizing psychopathology** (like borderline tendencies or depressive symptoms) and on showing little relation to the FFNI Grandiose facets like authoritativeness or exhibitionism ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=FFNI%20G%20PNI%20G%20NPI,52)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=narcissism%2C%20including%20the%20HSNS%2C%20PNI,4%05%20measure%20of%20NPD)). This convergence supports the idea that whether one uses item-level factor modeling (FFNI) or the older rationally derived HSNS, one taps into the same underlying “neurotic, resentful narcissism” construct. In short, HSNS behaves as a subset of the FFNI’s vulnerable narcissism traits, particularly capturing what some have called the “negative emotionality” or *neurotic narcissism* component ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). 

- **NARQ (Narcissistic Admiration and Rivalry Questionnaire):** The NARQ (Back et al., 2013) divides grandiose narcissism into two strategies: **Admiration** (self-enhancing, charm, assertiveness) and **Rivalry** (self-protection, defensiveness, aggression toward threats). While the NARQ primarily measures overt narcissism, its **Rivalry component** shares some features with vulnerable narcissism (e.g. hostility in response to ego threat, distrust of others). Research finds that HSNS has little in common with **Admiration** (which requires confidence and social boldness that hypersensitive narcissists lack), but may show a modest positive correlation with **Rivalry** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=scales%20%28i,ratings%20on%20the%20Big%20Five)). This is because someone high in vulnerable narcissism can also be antagonistic or distrustful: HSNS’s *Egocentrism* facet (entitled self-focus) overlaps conceptually with the **antagonistic, low-agreeableness** aspect of Rivalry ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). For example, both HSNS and NARQ-Rivalry associate with being quick to feel slighted and to derogate others (albeit the HSNS individual might do so more covertly). However, the overlap is not complete; NARQ-Rivalry still assumes an outward assertive antagonism, whereas HSNS egocentrism is coupled with avoidant defensiveness. In the structural analysis by Schneider et al. (2025), the HSNS Egocentrism dimension mapped onto “antagonistic narcissism” and externalizing tendencies, which aligns with constructs like Rivalry ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=respect%20to%20various%20personality%20traits%2C,%28PsycInfo%20Database)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). Meanwhile, HSNS Oversensitivity mapped onto internalizing tendencies (which NARQ does not capture at all). So, one might say HSNS fills the **vulnerable niche** that NARQ leaves open. In practice, using HSNS alongside NARQ can show that not all narcissistic rivalry is loud – some of it is quiet and privately harbored. 

In summary, across these comparisons, the HSNS consistently emerges as a measure of the **“shy, hypersensitive” variant of narcissism**, complementing other instruments that capture the “bold, exploitative” variant. Its correlations with the **vulnerable scales** of multi-dimensional measures (FFNI-V, PNI-V) are high, while correlations with **grandiose scales** (NPI, NARQ-Admiration) are near zero or negative ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=match%20at%20L615%20narcissism%2C%20including,4%05%20measure%20of%20NPD)) ([Untitled](http://www.sakkyndig.com/psykologi/artvit/hendin1997.pdf#:~:text=HSNS%20and%20the%20MMPI,narcissistic%20tendencies%20in%20the%20normal)). This pattern affirms that the HSNS is an efficient way to index the vulnerable narcissism construct in research. Modern work, such as Schneider et al. (2025), further indicates that parsing HSNS into its two subcomponents can align it even more closely with those other models – with *HSNS-Oversensitivity* corresponding to what some call “neurotic narcissism” or the emotional insecurity aspect, and *HSNS-Egocentrism* corresponding to the antagonistic entitlement aspect of narcissism ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=from%20three%20large%20samples%20,research%20reconciles%20the%20HSNS%20with)) ([The Hypersensitive Narcissism Scale: Updated and extended construct validation in community samples using a newly constructed German version - PubMed](https://pubmed.ncbi.nlm.nih.gov/39786844/#:~:text=HSNS,2025%20APA%2C%20all%20rights%20reserved)). Thus, the HSNS has proven academically valuable both on its own and in comparative studies, deepening our understanding of narcissistic personality by highlighting the **vulnerable side of narcissism** that earlier models did not quantify. 

## Dataset Creation: HSNS Trait Structure and Descriptors  
Based on the HSNS and its identified facets, we can organize its content into a structured format of trait descriptors. Recent research suggests two main **factors** within the HSNS – which we will call **Oversensitivity** and **Egocentrism** – each encompassing several characteristics ([(PDF) Factor Structure of the Hypersensitive Narcissism Scale](https://www.researchgate.net/publication/336665679_Factor_Structure_of_the_Hypersensitive_Narcissism_Scale#:~:text=present%20research%20was%20to%20explore,item%20HSNS)). Below is a breakdown of each factor with descriptive **adjectives**, some **synonyms** (or related phrases), and representative **verbs** (behaviors) and **nouns** (abstract traits or tendencies) associated with that facet of hypersensitive narcissism:

- **Factor: Oversensitivity** (Hypersensitive, Vulnerable Narcissism facet)  
  - **Adjectives:** *sensitive*, *insecure*, *defensive*, *anxious*, *self-conscious*, *introverted*, *ashamed* (easily feels shame), *touchy* (easily offended).  
  - **Synonyms/Related Terms:** *vulnerable*, *thin-skinned*, *easily hurt*, *fragile ego*, *over-sensitive*, *prone to self-doubt*, *emotional*. (These terms all convey a tendency to feel hurt, rejected or anxious about one’s worth.)  
  - **Verbs (Behaviors):** *overthink slights*, *take things personally*, *withdraw from criticism*, *ruminate* (over perceived offenses), *avoid the spotlight*, *feel hurt*, *crave reassurance*, *blame self*. For example, someone high in Oversensitivity might **interpret** innocuous remarks as personal attacks and **withdraw** to protect themselves ([Hypersensitive Narcissism Scale (HSNS)](https://www.psytoolkit.org/survey-library/narcism-hsns.html#:~:text=,for%20my%20time%20and%20sympathy)). They often **avoid** situations where they might be judged, and **brood** over past criticisms.

````python
import os
import pandas as pd
import mlflow

# Set the MLflow tracking URI
mlflow.set_tracking_uri("http://127.0.0.1:5000/")

# Set up HSNS experiment
experiment_name = "HSNS"
mlflow.set_experiment(experiment_name)

hsns_dataset = {
    'Oversensitivity': {
        'Hypersensitive': {
            'Synonyms': ['Thin-skinned', 'Easily hurt', 'Fragile', 'Reactive'],
            'Verbs': ['React strongly', 'Take offense', 'Feel hurt', 'Internalize criticism'],
            'Nouns': ['Sensitivity', 'Fragility', 'Insecurity', 'Vulnerability']
        },
        'Insecure': {
            'Synonyms': ['Uncertain', 'Self-doubting', 'Anxious', 'Self-conscious'],
            'Verbs': ['Question oneself', 'Doubt oneself', 'Worry', 'Second guess'],
            'Nouns': ['Insecurity', 'Self-doubt', 'Anxiety', 'Uncertainty']
        },
        'Fragile': {
            'Synonyms': ['Delicate', 'Vulnerable', 'Easily upset', 'Sensitive'],
            'Verbs': ['Break down', 'Withdraw', 'Get upset', 'Feel overwhelmed'],
            'Nouns': ['Fragility', 'Vulnerability', 'Sensitivity', 'Weakness']
        },
        'Defensive': {
            'Synonyms': ['Protective', 'Guarded', 'Reactive', 'Defensive'],
            'Verbs': ['Defend oneself', 'Protect self-esteem', 'Deflect criticism', 'Guard emotions'],
            'Nouns': ['Defensiveness', 'Guardedness', 'Protection', 'Self-defense']
        },
        'Shame-prone': {
            'Synonyms': ['Embarrassed', 'Self-conscious', 'Humiliated', 'Guilt-ridden'],
            'Verbs': ['Feel ashamed', 'Avoid exposure', 'Feel embarrassed', 'Hide self'],
            'Nouns': ['Shame', 'Embarrassment', 'Humiliation', 'Self-consciousness']
        },
        'Anxious': {
            'Synonyms': ['Nervous', 'Apprehensive', 'Worried', 'Uneasy'],
            'Verbs': ['Worry', 'Stress', 'Feel nervous', 'Fret'],
            'Nouns': ['Anxiety', 'Nervousness', 'Stress', 'Unease']
        }
    },
    'Egocentrism': {
        'Self-centered': {
            'Synonyms': ['Self-focused', 'Egotistical', 'Self-absorbed', 'Self-interested'],
            'Verbs': ['Prioritize oneself', 'Ignore others', 'Focus inward', 'Dominate conversations'],
            'Nouns': ['Self-absorption', 'Egocentrism', 'Self-interest', 'Self-focus']
        },
        'Entitled': {
            'Synonyms': ['Demanding', 'Expectant', 'Privileged', 'Presumptuous'],
            'Verbs': ['Demand', 'Expect special treatment', 'Claim privileges', 'Assert rights'],
            'Nouns': ['Entitlement', 'Privilege', 'Expectation', 'Demands']
        },
        'Self-absorbed': {
            'Synonyms': ['Narcissistic', 'Self-focused', 'Preoccupied', 'Egotistical'],
            'Verbs': ['Obsess over oneself', 'Disregard others', 'Indulge in self-thought', 'Self-promote'],
            'Nouns': ['Self-absorption', 'Narcissism', 'Egotism', 'Self-indulgence']
        },
        'Resentful': {
            'Synonyms': ['Bitter', 'Grudging', 'Vindictive', 'Spiteful'],
            'Verbs': ['Harbor grudges', 'Resent', 'Envy', 'Hold bitterness'],
            'Nouns': ['Resentment', 'Bitterness', 'Spite', 'Vindictiveness']
        },
        'Withdrawn': {
            'Synonyms': ['Reserved', 'Introverted', 'Isolated', 'Distant'],
            'Verbs': ['Withdraw', 'Distance oneself', 'Retreat', 'Isolate'],
            'Nouns': ['Withdrawal', 'Isolation', 'Introversion', 'Detachment']
        },
        'Hostile': {
            'Synonyms': ['Antagonistic', 'Angry', 'Aggressive', 'Confrontational'],
            'Verbs': ['Express anger', 'Confront', 'Argue', 'Lash out'],
            'Nouns': ['Hostility', 'Aggression', 'Anger', 'Confrontation']
        }
    }
}

# Create a list to store the dataset
data = []
for factor, attributes in hsns_dataset.items():
    for adjective, details in attributes.items():
        synonyms = details['Synonyms']
        verbs = details['Verbs']
        nouns = details['Nouns']
        # Iterate through each corresponding synonym, verb, and noun
        for synonym, verb, noun in zip(synonyms, verbs, nouns):
            data.append((factor, adjective, synonym, verb, noun))

# Create a DataFrame from the data list
hsns_df = pd.DataFrame(data, columns=['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun'])

# Define the dataset path
dataset_path = "../Datasets/hsns.csv"

# Start an MLflow run for dataset creation
with mlflow.start_run(run_name="HSNS Dataset Creation"):
    # Log dataset metadata
    mlflow.log_param("Experiment", "HSNS")
    mlflow.log_param("Total Rows", len(hsns_df))
    mlflow.log_param("Total Factors", len(hsns_dataset))
    mlflow.log_param("Data Schema", str(hsns_df.columns.tolist()))
    
    # Save dataset to CSV
    hsns_df.to_csv(dataset_path, index=False)
    mlflow.log_artifact(dataset_path, artifact_path="data")  # Logs the dataset to MLflow

    # Evaluation: Basic dataset inspection
    print(f"✅ HSNS dataset created and saved to {dataset_path}")
    print("First 5 rows of the dataset:")
    print(hsns_df.head())
    print("\nDataset shape:", hsns_df.shape)
    print("\nNumber of unique Factors:", hsns_df['Factor'].nunique())
    print("Unique Factors:", hsns_df['Factor'].unique())

    # Confirm MLflow tracking
    experiment = mlflow.get_experiment_by_name(experiment_name)
    if experiment:
        print(f"📌 MLflow tracking active under experiment '{experiment_name}' (ID: {experiment.experiment_id})")
    else:
        print("❌ Failed to track experiment in MLflow.")
````

# **Step 2: API key setup** <a class="anchor" id="hsns_page_2"></a>

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key Status", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"✅ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key Status", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key Status", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("🔑 API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("🤖 OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("✅ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"⚠️ Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"❌ Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"❌ Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"⚠️ Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved hsns dataset from a CSV file
dataset_path = '../Datasets/hsns.csv'
try:
    hsns_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'hsns_df' is empty before processing
        if hsns_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", hsns_df.shape[0])
            print(f"📊 Processing {hsns_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        hsns_df['Embedding'] = hsns_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/hsns_embeddings.csv'
        hsns_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = hsns_df.shape[0]
        num_columns = hsns_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in hsns_df.columns and not hsns_df['Embedding'].empty:
            embedding_length = len(hsns_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(hsns_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/hsns_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/hsns_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "hsns_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

# **Step 6: Create Model** <a class="anchor" id="hsns_page_3"></a>

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/hsns_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/hsns_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"✅ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"❌ Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="hsns Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "hsns_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"📊 Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "hsns_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"📊 Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "hsns_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"🖼️ Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/hsns_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"💾 Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"✅ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/hsns_label_encoder.pkl"
model_path = "../Models/hsns_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"✅ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"❌ Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"❌ An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"❌ Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Model testing (inference on new data) complete.")
````

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("HSNS")

# Load the dataset
dataset_path = "../Embeddings/hsns_embeddings.csv"
hsns_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in hsns_df.columns:
    hsns_df['Embedding'] = hsns_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the HSNS schema in Neo4j
def create_hsns_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with HSNS labels
        graph.run("MERGE (f:HSNS_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:HSNS_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:HSNS_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:HSNS_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:HSNS_Noun {name: $noun})", noun=noun)
        
        # Create relationships with HSNS namespacing
        graph.run("""
            MATCH (f:HSNS_Factor {name: $factor}), (a:HSNS_Adjective {name: $adjective})
            MERGE (f)-[:HSNS_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:HSNS_Adjective {name: $adjective}), (s:HSNS_Synonym {name: $synonym})
            MERGE (a)-[:HSNS_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:HSNS_Adjective {name: $adjective}), (v:HSNS_Verb {name: $verb})
            MERGE (a)-[:HSNS_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:HSNS_Adjective {name: $adjective}), (n:HSNS_Noun {name: $noun})
            MERGE (a)-[:HSNS_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:HSNS_Factor {name: $factor}), (a:HSNS_Adjective {name: $adjective})
                MERGE (e:HSNS_Embedding {value: $embedding})
                MERGE (f)-[:HSNS_HAS_EMBEDDING]->(e)
                MERGE (a)-[:HSNS_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create HSNS Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create HSNS Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the HSNS schema
        create_hsns_schema(graph, hsns_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", hsns_df.shape[0])
        mlflow.log_param("Number of Columns", hsns_df.shape[1])
        
        # Print the DataFrame
        print(hsns_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
