## **The Five-Factor Narcissism Inventory (FFNI): Profiling Narcissistic Traits**

In the intricate landscape of personality assessment, the Five-Factor Narcissism Inventory (FFNI) emerges as a tool that delves into the nuances of narcissistic traits, offering insights into the grandiosity and self-centeredness that can shape an individual's character. This inventory, developed to scrutinize narcissism across five distinct facets, provides a comprehensive understanding of this complex personality trait. 

**Meet the Mind Behind the FFNI:**

**Dr. Aidyn M. Iachini:** The FFNI owes its existence to the pioneering work of Dr. Aidyn M. Iachini. A prominent figure in the fields of psychology and personality assessment, Dr. Iachini has dedicated his career to unraveling the intricacies of narcissism. His innovative approach in creating the FFNI has shed light on a personality trait that has long fascinated psychologists and the general public alike.

**Breaking Down the FFNI Dataset:**

The FFNI dataset is a meticulously crafted collection of traits and synonyms that collectively define narcissism across five facets. Let's delve into these facets and their respective traits:

1. **Authority**: This facet probes the inclination towards dominance and control. It encompasses traits such as dominance, ambition, and decisiveness, reflecting a strong desire for authority and leadership.

2. **Self-Sufficiency**: Here, the focus shifts to independence and self-reliance. Traits like independence, resourcefulness, and self-governance shed light on a self-sufficient approach to life.

3. **Superiority**: Superiority delves into the desire for uniqueness and distinction. Traits such as uniqueness, arrogance, and entitlement highlight a need to be perceived as exceptional and superior to others.

4. **Exhibitionism**: This facet explores the need for attention and admiration. Traits like attention-seeking, vanity, and boastfulness reflect a desire to be the center of attention and admiration from others.

5. **Exploitativeness**: Exploitativeness examines manipulative and self-serving behaviors. Traits such as manipulativeness, opportunism, and shrewdness underscore a tendency to use others for personal gain.

6. **Vanity**: The final facet, vanity, concentrates on self-admiration and self-centeredness. Traits like self-admiring, self-centeredness, and narcissism itself shed light on a preoccupation with one's own image and importance.

This comprehensive breakdown allows researchers and psychologists to pinpoint specific narcissistic traits within individuals and gain a more profound understanding of how these traits manifest in various contexts. The FFNI dataset is not merely a list of synonyms; it's a key to deciphering the intricacies of narcissism in all its multifaceted glory.

The FFNI provides a nuanced view of narcissism, acknowledging that it is not a one-dimensional construct but a multifaceted trait with various manifestations. It offers a valuable tool for researchers, clinicians, and anyone interested in understanding the intricate nature of narcissism and its impact on personality and behavior.


# Development Timeline of the Five-Factor Narcissism Inventory (FFNI)

The **Five-Factor Narcissism Inventory (FFNI)** is a psychological assessment tool designed to measure narcissistic personality traits through the lens of the Five-Factor Model (FFM) of personality. Since its introduction in 2012, the FFNI has undergone several refinements and validation studies. Below is a chronological timeline of key milestones in the development, validation, and application of the FFNI, along with significant publications and contributors. Each entry includes references to peer-reviewed sources (with DOI links where available) that document the milestone.

## 2012 – Initial Development of the FFNI  
- **Introduction of FFNI:** Natalie Glover, Joshua D. Miller, Donald R. Lynam, Cristina Crego, and Thomas A. Widiger publish the first paper on the FFNI. They develop 15 maladaptive trait scales aligned with FFM facets (e.g. *Reactive Anger* as a narcissistic variant of *Angry Hostility*) to capture both grandiose and vulnerable aspects of narcissism ([The five-factor narcissism inventory: a five-factor measure of narcissistic personality traits - PubMed](https://pubmed.ncbi.nlm.nih.gov/22475323/#:~:text=This%20study%20provides%20convergent%2C%20discriminant%2C,of%20both%20grandiose%20and%20vulnerable)) ([The five-factor narcissism inventory: a five-factor measure of narcissistic personality traits - PubMed](https://pubmed.ncbi.nlm.nih.gov/22475323/#:~:text=narcissism,maladaptive%20variants%20of%20the%20FFM)). The initial validation with undergraduate samples demonstrates good convergent and discriminant validity against eight established narcissism measures and FFM traits, supporting the FFNI as a comprehensive measure of narcissistic personality traits ([The five-factor narcissism inventory: a five-factor measure of narcissistic personality traits - PubMed](https://pubmed.ncbi.nlm.nih.gov/22475323/#:~:text=narcissism,maladaptive%20variants%20of%20the%20FFM)). (Glover *et al.*, 2012; DOI: 10.1080/00223891.2012.670680)

## 2013 – Validation in Community and Clinical Samples  
- **Convergent/Discriminant Validity:** Joshua D. Miller and colleagues extend FFNI validation to both community adults and clinical patients ([The Five-Factor Narcissism Inventory (FFNI): a test of the convergent, discriminant, and incremental validity of FFNI scores in clinical and community samples - PubMed](https://pubmed.ncbi.nlm.nih.gov/23647044/#:~:text=The%20five,the%20samples%2C%20the%20FFNI%20scales)) ([The Five-Factor Narcissism Inventory (FFNI): a test of the convergent, discriminant, and incremental validity of FFNI scores in clinical and community samples - PubMed](https://pubmed.ncbi.nlm.nih.gov/23647044/#:~:text=manifested%20good%20convergent%20and%20discriminant,maintaining%20ties%20to%20the%20significant)). They find that FFNI facet scales derived from FFM Neuroticism correlate strongly with **vulnerable narcissism**, those from FFM Extraversion align with **grandiose narcissism**, and those from (low) FFM Agreeableness relate to both dimensions ([The Five-Factor Narcissism Inventory (FFNI): a test of the convergent, discriminant, and incremental validity of FFNI scores in clinical and community samples - PubMed](https://pubmed.ncbi.nlm.nih.gov/23647044/#:~:text=traits%2C%20grandiose%20narcissism%2C%20and%20vulnerable,grandiose%20narcissism%2C%20and%20vulnerable%20narcissism)). The FFNI’s composite grandiosity and vulnerability scores show **incremental validity** in predicting DSM-IV/DSM-5 Narcissistic Personality Disorder (NPD) traits beyond existing narcissism measures ([The Five-Factor Narcissism Inventory (FFNI): a test of the convergent, discriminant, and incremental validity of FFNI scores in clinical and community samples - PubMed](https://pubmed.ncbi.nlm.nih.gov/23647044/#:~:text=manifested%20good%20convergent%20and%20discriminant,maintaining%20ties%20to%20the%20significant)). This study establishes the FFNI as a **robust instrument** for assessing narcissistic pathology in various populations (Miller *et al.*, 2013; DOI: 10.1037/a0032536).  
- **Construct Validity and Nomological Network:** Another 2013 study by Miller, Brittany Gentile, W. Keith Campbell and colleagues examines FFNI scores in relation to a broad array of external criteria ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20Five,good%20convergent%20and%20criterion%20validity)) ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20FFNI%20grandiose%20and%20vulnerable,and%20vulnerable%20dimensions%20of%20narcissism)). Using a large undergraduate sample, they report that FFNI **Grandiose** and **Vulnerable** dimensions show expected patterns of correlations: for example, grandiose narcissism (FFNI) relates to antagonism (low agreeableness), low empathy/communion, and traits like psychopathy and Machiavellianism, whereas vulnerable narcissism relates to high neuroticism and insecurity ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=behaviors%20and%20symptoms%2C%20and%20romantic,and%20vulnerable%20dimensions%20of%20narcissism)). Both dimensions exhibit good convergent validity with other narcissism inventories and distinct associations with interpersonal style, attachment, and psychopathology ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20Five,good%20convergent%20and%20criterion%20validity)) ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20FFNI%20grandiose%20and%20vulnerable,and%20vulnerable%20dimensions%20of%20narcissism)). This work reinforces that the FFNI can simultaneously assess **grandiose and vulnerable narcissism** within a single framework ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=disagreeableness%2C%20low%20love%2Fcommunion%2C%20psychopathy%2C%20Machiavellianism%2C,and%20vulnerable%20dimensions%20of%20narcissism)). (Miller, Gentile & Campbell, 2013; DOI: 10.1080/00223891.2012.742903)

## 2014 – Criterion Validity via Expert Ratings  
- **Comparative Validity of Narcissism Measures:** Miller and colleagues conduct a head-to-head comparison of popular narcissism instruments, using expert consensus ratings of NPD traits as an external criterion ([A comparison of the criterion validity of popular measures of narcissism and narcissistic personality disorder via the use of expert ratings - PubMed](https://pubmed.ncbi.nlm.nih.gov/24773036/#:~:text=The%20growing%20interest%20in%20the,according%20to%20the%20Diagnostic%20and)) ([A comparison of the criterion validity of popular measures of narcissism and narcissistic personality disorder via the use of expert ratings - PubMed](https://pubmed.ncbi.nlm.nih.gov/24773036/#:~:text=psychopathology%2C%20and%20compared%20the%20resultant,Hendin%20%26%20Cheek)). In a clinical sample and a larger student sample, the **FFNI’s grandiose narcissism** scales emerge as the **strongest match to expert-rated** DSM-IV NPD and grandiose narcissism, on par with or outperforming the Narcissistic Personality Inventory (NPI) ([A comparison of the criterion validity of popular measures of narcissism and narcissistic personality disorder via the use of expert ratings - PubMed](https://pubmed.ncbi.nlm.nih.gov/24773036/#:~:text=psychopathology%2C%20and%20compared%20the%20resultant,Hendin%20%26%20Cheek)). Meanwhile, the FFNI’s **vulnerable narcissism** scales show the highest correspondence to experts’ conceptualization of vulnerable narcissism, comparable to the Pathological Narcissism Inventory (PNI) ([A comparison of the criterion validity of popular measures of narcissism and narcissistic personality disorder via the use of expert ratings - PubMed](https://pubmed.ncbi.nlm.nih.gov/24773036/#:~:text=psychopathology%2C%20and%20compared%20the%20resultant,Hendin%20%26%20Cheek)). This study highlights the FFNI’s excellent **criterion validity**, indicating that it captures the core traits of narcissism as recognized by experts ([A comparison of the criterion validity of popular measures of narcissism and narcissistic personality disorder via the use of expert ratings - PubMed](https://pubmed.ncbi.nlm.nih.gov/24773036/#:~:text=psychopathology%2C%20and%20compared%20the%20resultant,Pincus%20et)). (Miller *et al.*, 2014; DOI: 10.1037/a0036613)

## 2015 – FFNI Short Form (FFNI-SF) Development  
- **60-Item Short Form:** Emily D. Sherman, Joshua Miller, Lauren Few, W. Keith Campbell, Thomas Widiger, Cristina Crego, and Donald Lynam develop a **short-form version** of the inventory (FFNI-SF) to improve practicality ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=The%20Five,2%20undergraduate)) ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=item%20response%20theory,Grandiose%20and)). Using Item Response Theory (IRT) and data from four diverse samples (two student samples, an MTurk sample, and a clinical sample), they select 60 items (4 per trait) that best represent the original 15 facets ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=grandiose%20fantasies%2C%20indifference%2C%20lack%20of,form%20scales%2C%20and%20manifest)). The FFNI-SF retains high internal consistencies for all subscales and replicates the **factor structure** of the full 148-item FFNI ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=item%20response%20theory,Grandiose%20and)). It exhibits nearly identical correlation profiles to the long form across various criteria (Big Five traits, other narcissism measures, and markers of externalizing/internalizing psychopathology) ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=subscale%20level%20and%20the%20level,SF%29%20offers%20a%20well)). Grandiose and vulnerable composite scores behave virtually the same in short and long versions ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=structure%20identical%20to%20the%20structure,when%20assessment%20time%20is%20limited)), indicating minimal loss of information. The authors conclude the FFNI-SF provides a **time-efficient yet valid** assessment of narcissism’s core traits ([Development of a Short Form of the Five-Factor Narcissism Inventory: the FFNI-SF. | University of Kentucky College of Arts & Sciences](https://is.as.uky.edu/bibcite/reference/2790#:~:text=structure%20identical%20to%20the%20structure,when%20assessment%20time%20is%20limited)). (Sherman *et al.*, 2015; DOI: 10.1037/pas0000100)

## 2016 – Structural Analysis and Three-Factor Model  
- **“Thinking Structurally” – Three Broad Factors:** Miller, Lynam, Jessica L. McCain, Few, Crego, Widiger, and Campbell examine the underlying structure of FFNI traits in a two-study paper ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=supported%20the%20extraction%20of%20three,and%20diverging%20relations%20with%20measures)) ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=FFNI%20factors%20help%20explicate%20the,The%20results%20also%20highlight%20the)). Their factor analyses support a **three-factor model** of narcissism: **Antagonism** (e.g. arrogant, exploitative traits), **Agentic Extraversion** (entitlement, authoritativeness, exhibitionism – capturing bold, attention-seeking aspects), and **Narcissistic Neuroticism** (insecurity, shame, need for admiration – capturing vulnerable aspects) ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=supported%20the%20extraction%20of%20three,and%20diverging%20relations%20with%20measures)). In a second sample, these FFNI factors show clear convergent correlations with corresponding Big Five domains (e.g. Antagonism with low Agreeableness, Neuroticism factor with high Neuroticism, Agentic Extraversion with high Extraversion) and divergent links with external criteria ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=supported%20the%20extraction%20of%20three,and%20diverging%20relations%20with%20measures)) ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=FFNI%20factors%20help%20explicate%20the,The%20results%20also%20highlight%20the)). The authors note that **all expressions of narcissism share Antagonism** as a common core, but grandiose vs. vulnerable presentations differ in their levels of Neuroticism and Extraversion ([Thinking Structurally About Narcissism: An Examination of the Five-Factor Narcissism Inventory and Its Components | Journal of Personality Disorders](https://guilfordjournals.com/doi/full/10.1521/pedi_2015_29_177#:~:text=FFNI%20factors%20help%20explicate%20the,The%20results%20also%20highlight%20the)). This structural work helped establish a more nuanced view of narcissism facets and informed how composite scores (grandiose vs. vulnerable) can be decomposed into subdimensions. (Miller *et al.*, 2016; DOI: 10.1521/pedi_2015_29_177)

## 2018 – Informant-Report FFNI (IFFNI) and Cross-Cultural Extensions  
- **Informant FFNI (IFFNI):** Recognizing that narcissistic individuals may distort self-reports, Joshua R. Oltmanns, Cristina Crego, and Thomas Widiger develop an **informant-report** version of the FFNI ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=A%20series%20of%20self,has%20been%20found%20with%20other)) ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=Inventory%20,V%20was)). In their initial validation (two studies), informants (friends/family) rated targets using the new IFFNI. **Findings:** Informant ratings for **grandiose narcissism** showed higher self–other agreement than typically seen with other narcissism measures ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=Inventory%20,though%2C%20was%20found%20for%20IFFNI)), whereas **vulnerable narcissism** self–other agreement was low (consistent with the more internalized, hidden nature of vulnerability) ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=members%20were%20compared%20with%20self,V%20was)). Informant-rated grandiose and vulnerable narcissism both predicted targets’ social dysfunction (e.g. interpersonal problems), even when self-reports did not ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=agreement%20for%20IFFNI%20Grandiose%20,may%20reflect%20vulnerable%20narcissistic%20traits)). Additionally, the IFFNI could distinguish grandiose vs. vulnerable narcissists in others, suggesting **incremental value** in having outsiders assess narcissism ([
            Informant Assessment: The Informant Five-Factor Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC5930358/#:~:text=agreement%20for%20IFFNI%20Grandiose%20,may%20reflect%20vulnerable%20narcissistic%20traits)). This work underscores the importance of multi-informant assessment for narcissistic traits (Oltmanns *et al.*, 2018; DOI: 10.1037/pas0000487).  
- **International Adaptations:** By this time, the FFNI had begun to be translated and tested in other languages and cultures. For example, **Italy** – Andrea Fossati and colleagues validate an Italian translation of the FFNI Short Form in a sample of 1,063 students ([Assessing Dimensions of Pathological Narcissism: Psychometric Properties of the Short Form of the Five-Factor Narcissism Inventory in a Sample of Italian University Students - PubMed](https://pubmed.ncbi.nlm.nih.gov/28537767/#:~:text=To%20evaluate%20the%20psychometric%20properties,Finally)). They report good internal consistency for all scales and replicate the three-factor (agentic, antagonistic, neurotic) structure identified by Sherman et al. (2015) ([Assessing Dimensions of Pathological Narcissism: Psychometric Properties of the Short Form of the Five-Factor Narcissism Inventory in a Sample of Italian University Students - PubMed](https://pubmed.ncbi.nlm.nih.gov/28537767/#:~:text=To%20evaluate%20the%20psychometric%20properties,Finally)). The Italian FFNI-SF’s grandiose and vulnerable indices show expected convergent correlations with the PNI’s grandiosity/vulnerability scales and predicted relations with measures of personality dysfunction ([Assessing Dimensions of Pathological Narcissism: Psychometric Properties of the Short Form of the Five-Factor Narcissism Inventory in a Sample of Italian University Students - PubMed](https://pubmed.ncbi.nlm.nih.gov/28537767/#:~:text=students%20were%20administered%20the%20scale%2C,2%20dimensions%20of%20personality%20dysfunction)), supporting its psychometric soundness ([Assessing Dimensions of Pathological Narcissism: Psychometric Properties of the Short Form of the Five-Factor Narcissism Inventory in a Sample of Italian University Students - PubMed](https://pubmed.ncbi.nlm.nih.gov/28537767/#:~:text=To%20evaluate%20the%20psychometric%20properties,Finally)). (Fossati *et al.*, 2018; DOI: 10.1080/00223891.2017.1324457)

## 2019 – Further Cross-Cultural Validation  
- **Norwegian Version:** A team at the University of Oslo, led by Claire N. Prendergast, publishes on the **Norwegian adaptation** of key FFNI subscales ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=In%20recent%20years%2C%20narcissism%20has,screen%20participants)) ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=properties%20of%20these%20translated%20self,be%20used%20to%20explore%20these)). They translate the FFNI’s **vulnerable narcissism** facets and the **“Indifference” (grandiose detachment)** facet, then administer both the long-form and short-form versions to a diverse Norwegian sample ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=In%20recent%20years%2C%20narcissism%20has,screen%20participants)). Results from reliability tests and confirmatory factor analyses indicate that the Norwegian FFNI (especially the short form) maintains good internal consistency and factorial validity ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=properties%20of%20these%20translated%20self,be%20used%20to%20explore%20these)). Its scales correlate with other theoretical measures (e.g. related personality traits) in expected ways, establishing **criterion validity** for the Norwegian version ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=properties%20of%20these%20translated%20self,be%20used%20to%20explore%20these)). The authors conclude that the **translated FFNI-SF scales perform well** and can be used in Norwegian research and clinical settings, and easily adapted to other Scandinavian languages ([The Norwegian version of the five factor narcissism inventory for vulnerable narcissism and the grandiose narcissism subscale of indifference: Psychometric properties of the long- and short-form versions - PubMed](https://pubmed.ncbi.nlm.nih.gov/31402468/#:~:text=confirmatory%20factor%20analysis%20on%20the,use%20in%20other%20Scandinavian%20countries)). (Prendergast *et al.*, 2019; DOI: 10.1111/sjop.12569)

## 2021 – Support for Three-Factor Model & Super-Short Form  
- **Polish Structural Study:** Radosław Rogoza and colleagues investigate the FFNI’s structure in a Polish sample, providing further evidence for the **three-factor model of narcissism** ([Investigating the structure of the Polish Five Factor Narcissism Inventory: Support for the three-factor model of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/33090830/#:~:text=The%203,anger%20were%20better%20indicators%20of)). Across two studies (N=793), exploratory factor analysis of the Polish FFNI again yields three broad factors corresponding to antagonistic, agentic, and neurotic narcissism ([Investigating the structure of the Polish Five Factor Narcissism Inventory: Support for the three-factor model of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/33090830/#:~:text=The%203,anger%20were%20better%20indicators%20of)). Some facet-level nuances emerge – for instance, *Grandiose Fantasies*, *Thrill Seeking*, and *Arrogance* did not load cleanly on the expected factors, and facets like *Manipulativeness* and *Reactive Anger* showed better alignment with the agentic and neurotic factors respectively ([Investigating the structure of the Polish Five Factor Narcissism Inventory: Support for the three-factor model of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/33090830/#:~:text=investigated%20the%20internal%20structure%20of,was%20assessed%20in%20regard%20to)). By tweaking the scoring (reassigning a few facets), the researchers improve model fit and **measurement precision** ([Investigating the structure of the Polish Five Factor Narcissism Inventory: Support for the three-factor model of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/33090830/#:~:text=%28FFNI%29,the%20measurement%20precision%20of%20the)). The validated Polish adaptation confirms that the FFNI’s three underlying dimensions are **cross-culturally robust**, while also highlighting the importance of continual refinement at the facet level ([Investigating the structure of the Polish Five Factor Narcissism Inventory: Support for the three-factor model of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/33090830/#:~:text=%28FFNI%29,the%20measurement%20precision%20of%20the)). (Rogoza *et al.*, 2021; DOI: 10.1037/pas0000901)  
- **FFNI Super-Short Form:** To facilitate very brief assessments, Melissa Packer West, Donald Lynam, Joshua Miller, W. Keith Campbell, and colleagues develop an ultra-short **15-item FFNI Super-Short Form (FFNI-SSF)**. This version takes the single highest-loading item from each of the 15 FFNI facets ([
            Validation of the German Five-Factor Narcissism Inventory and Construction of a Brief Form Using Ant Colony Optimization - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC10149890/#:~:text=Glover%20et%20al,was%20constructed%20using%20the%20highest)). Initial evaluations show that despite its brevity, the FFNI-SSF can still capture the three **core factors** of narcissism – antagonism, agentic extraversion, and narcissistic neuroticism – similar to the full inventory (Packer West *et al.*, 2021). The availability of the FFNI-SSF (though in pre-print/formative stages as of 2021) demonstrates the ongoing effort to **streamline narcissism assessment** without losing theoretical richness ([
            Validation of the German Five-Factor Narcissism Inventory and Construction of a Brief Form Using Ant Colony Optimization - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC10149890/#:~:text=Glover%20et%20al,was%20constructed%20using%20the%20highest)). This complements the existing long and short forms and broadens the FFNI’s usability in large surveys or time-constrained settings.

## 2023 – German Adaptation and Brief Form Optimization  
- **German Validation and FFNI-BF:** A recent milestone is the first full **German translation** of the FFNI, published by Scheidt and colleagues (2023). Their work confirms that the FFNI’s content and structure generalize well to German-speaking samples, with clear grandiose (agentic) and vulnerable (neurotic) factors and expected correlations with related personality measures ([
            Validation of the German Five-Factor Narcissism Inventory and Construction of a Brief Form Using Ant Colony Optimization - PMC

The Five-Factor Narcissism Inventory (FFNI) is typically described as comprising 15 facet scales that can be grouped into two higher‐order dimensions—11 facets assessing grandiose narcissism and 4 facets assessing vulnerable narcissism. One commonly reported breakdown is as follows:

**Grandiose Narcissism (11 facets):**  
1. **Authority:** Leadership, dominance, and a tendency to command respect.  
2. **Self‐Sufficiency:** A sense of independence and confidence in one’s abilities.  
3. **Entitlement:** A belief in deserving special treatment and privileges.  
4. **Exploitativeness:** A readiness to manipulate or take advantage of others for personal gain.  
5. **Exhibitionism:** A drive to seek attention and display oneself in a dramatic, showy manner.  
6. **Grandiose Fantasy:** Daydreams about unlimited success, power, or brilliance.  
7. **Reactive Anger:** Intense anger or hostility in response to threats or setbacks.  
8. **Need for Admiration:** An excessive craving for praise and recognition.  
9. **Self‐Absorption:** A preoccupation with one’s own interests and feelings.  
10. **Overconfidence:** An exaggerated belief in one’s own abilities and judgments.  
11. **Assertiveness:** A forceful and decisive style of interacting with others.

**Vulnerable Narcissism (4 facets):**  
12. **Contingent Self‐Esteem:** Self-esteem that fluctuates depending on external validation.  
13. **Hiding the Self:** A tendency to conceal one’s perceived flaws or vulnerabilities.  
14. **Devaluing:** A disposition to dismiss or devalue the contributions or worth of others.  
15. **Shame/Insecurity:** Deep-seated feelings of inadequacy and sensitivity to criticism.

This structure—11 grandiose facets and 4 vulnerable facets—is widely reported in the FFNI literature (see, for example, Glover et al., 2012; Miller et al., 2013) and provides a comprehensive, nuanced profile of narcissistic traits. Each facet is measured with several items, and together they form a detailed portrait of narcissism that links the construct to the broader Five-Factor Model of personality.

Why is the FFNI called the Five-Factor Narcissism Inventory (FFNI) if it has 15 facets?

Great question! The **Five-Factor Narcissism Inventory (FFNI)** is designed to assess narcissism through the lens of the **Five-Factor Model (FFM)** of personality. The **"Five-Factor"** in its name refers to the **FFM structure** that underlies the traits it measures, rather than the number of narcissism facets themselves.

### Why Does the FFNI Have 15 Facets?
The **FFNI** organizes **narcissistic traits** into **two broad dimensions**:
1. **Grandiose Narcissism (11 facets)** – Characterized by dominance, entitlement, and arrogance.
2. **Vulnerable Narcissism (4 facets)** – Characterized by insecurity, contingent self-esteem, and emotional fragility.

The **total number of facets is 15**, but each is mapped onto one or more of the **Five-Factor Model (FFM) domains**, which include:
- **Neuroticism**
- **Extraversion**
- **Openness to Experience**
- **Agreeableness**
- **Conscientiousness**

Thus, the **"Five-Factor"** in **FFNI** refers to how **narcissistic traits** relate to these **five personality dimensions** rather than implying that the inventory consists of just five facets.

### How FFNI Maps to the Five-Factor Model:
Each of the 15 facets aligns with one or more of the **FFM traits**:
- **Grandiose facets** are often linked to **low Agreeableness, high Extraversion, and sometimes high Conscientiousness**.
- **Vulnerable facets** are often linked to **high Neuroticism and low Extraversion**.

For example:
- **Authority (Grandiose facet)** → High **Extraversion**, Low **Agreeableness**
- **Entitlement (Grandiose facet)** → Low **Agreeableness**
- **Reactive Anger (Vulnerable facet)** → High **Neuroticism**, Low **Agreeableness**
- **Contingent Self-Esteem (Vulnerable facet)** → High **Neuroticism**

### Summary:
- The **"Five-Factor"** in FFNI refers to its foundation in the **Five-Factor Model (FFM)**, not the number of facets.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved ffni dataset from a CSV file
dataset_path = '../Datasets/ffni.csv'
try:
    ffni_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'ffni_df' is empty before processing
        if ffni_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", ffni_df.shape[0])
            print(f"📊 Processing {ffni_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        ffni_df['Embedding'] = ffni_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/ffni_embeddings.csv'
        ffni_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = ffni_df.shape[0]
        num_columns = ffni_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in ffni_df.columns and not ffni_df['Embedding'].empty:
            embedding_length = len(ffni_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(ffni_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/ffni_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/ffni_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "ffni_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

# **Step 6: Create Model** <a class="anchor" id="ffni_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/ffni_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/ffni_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"✅ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"❌ Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="ffni Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "ffni_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"📊 Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "ffni_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"📊 Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "ffni_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"🖼️ Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/ffni_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"💾 Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"✅ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 8: Test and Evaluate Model** <a class="anchor" id="ffni_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/ffni_label_encoder.pkl"
model_path = "../Models/ffni_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"✅ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"❌ Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"❌ An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"❌ Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Model testing (inference on new data) complete.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/ffni_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="FFNI Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/ffni_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "ffni_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/ffni_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of FFNI Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/ffni_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of FFNI Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create FFNI Schema in Neo4j** <a class="anchor" id="FFNI_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("FFNI")

# Load the dataset
dataset_path = "../Embeddings/ffni_embeddings.csv"
ffni_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in ffni_df.columns:
    ffni_df['Embedding'] = ffni_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the FFNI schema in Neo4j
def create_ffni_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with FFNI labels
        graph.run("MERGE (f:FFNI_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:FFNI_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:FFNI_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:FFNI_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:FFNI_Noun {name: $noun})", noun=noun)
        
        # Create relationships with FFNI namespacing
        graph.run("""
            MATCH (f:FFNI_Factor {name: $factor}), (a:FFNI_Adjective {name: $adjective})
            MERGE (f)-[:FFNI_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:FFNI_Adjective {name: $adjective}), (s:FFNI_Synonym {name: $synonym})
            MERGE (a)-[:FFNI_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:FFNI_Adjective {name: $adjective}), (v:FFNI_Verb {name: $verb})
            MERGE (a)-[:FFNI_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:FFNI_Adjective {name: $adjective}), (n:FFNI_Noun {name: $noun})
            MERGE (a)-[:FFNI_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:FFNI_Factor {name: $factor}), (a:FFNI_Adjective {name: $adjective})
                MERGE (e:FFNI_Embedding {value: $embedding})
                MERGE (f)-[:FFNI_HAS_EMBEDDING]->(e)
                MERGE (a)-[:FFNI_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create FFNI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create FFNI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the FFNI schema
        create_ffni_schema(graph, ffni_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", ffni_df.shape[0])
        mlflow.log_param("Number of Columns", ffni_df.shape[1])
        
        # Print the DataFrame
        print(ffni_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
