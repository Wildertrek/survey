# Inventory of Pathological Narcissism (IPN) 

**Note:** The **Inventory of Pathological Narcissism (IPN)** is commonly known in the literature as the **Pathological Narcissism Inventory (PNI)** ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). It is a multidimensional self-report measure designed to assess pathological narcissism, encompassing both **grandiose** and **vulnerable** aspects of narcissistic personality pathology. Below, we explore its historical development, structural components, comparisons with other narcissism models (PNI, NPI, FFNI, DT4, MCMI), and provide a structured dataset of its factors with descriptive terms.

## Historical Development of the IPN (PNI) 

**Origins (When, How, Why):** The IPN/PNI was developed in the late 2000s by Aaron L. Pincus and colleagues, with the foundational paper published in 2009 ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). It emerged to resolve two key problems in narcissism research: **(a)** ambiguity between “pathological” vs. “normal” narcissism, and **(b)** the insufficient scope of existing measures ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)). Traditional narcissism scales (like the Narcissistic Personality Inventory, NPI) mainly captured **grandiose traits** in normal populations, leaving a gap in assessing the more **vulnerable, maladaptive** expressions seen in clinical narcissistic personality disorder ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)) ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). To bridge this gap, Pincus et al. conducted a series of studies to derive and validate the IPN/PNI, ensuring it covered the full spectrum of narcissistic pathology ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=pathological%20narcissism%20vs,The%20PNI)).

**Foundational Research:** The seminal article *“Initial construction and validation of the Pathological Narcissism Inventory”* by Pincus et al. (2009) documented the development of the IPN/PNI ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=pathological%20narcissism%20vs,The%20PNI)). This research used large samples to identify underlying factors of pathological narcissism and confirmed a seven-factor structure via factor analysis ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). Subsequent theoretical work by Pincus and Lukowitsky (2010) highlighted how the DSM criteria for Narcissistic Personality Disorder (NPD) had a “criterion problem,” failing to capture covert/vulnerable narcissistic features ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)). The IPN/PNI was intended as a solution, informed by clinical theories (Kernberg’s and Kohut’s concepts of narcissistic pathology) and empirical findings on “covert” narcissism (e.g. hypersensitivity, shame). Early validation studies showed the IPN/PNI had expected correlations: it correlated **negatively** with self-esteem and empathy, and **positively** with shame, interpersonal distress, aggression, and borderline personality organization ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=narcissistic%20vulnerability%20%28Contingent%20Self,homicidal%20ideation%2C%20and%20several%20aspects)). These patterns reinforced that the IPN was measuring maladaptive narcissism rather than normative self-confidence. 

**Intended Clinical Applications:** The IPN/PNI was designed for both research **and clinical contexts**. Clinically, it provides a more nuanced assessment of narcissistic pathology than existing tools. Because it assesses **grandiose** *and* **vulnerable** narcissism, clinicians can identify patients who might present with arrogance and entitlement or, alternatively, with defensiveness, shame, and fluctuating self-worth. The IPN/PNI has been shown to predict clinically relevant outcomes. For example, in a clinical sample it was significantly associated with **suicidality** (e.g. self-harm and suicide attempts) and other severe symptoms ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=Grandiose%20PNI%20scales%20were%20associated,several%20aspects%20of%20psychotherapy%20utilization)). Pincus et al. reported that higher IPN scores related to greater **parasuicidal behavior, homicidal ideation, and intensive psychotherapy utilization** ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=correlated%20negatively%20with%20self,several%20aspects%20of%20psychotherapy%20utilization)), indicating its utility in flagging at-risk individuals. Later studies have found the PNI’s “vulnerability” component in particular correlates with depression and anxiety, whereas its “grandiosity” component relates more to antagonistic interpersonal styles ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=behavior%20%28r%20%3D%20)) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=correlated%20negatively%20with%20self,several%20aspects%20of%20psychotherapy%20utilization)). Overall, the IPN was created to improve the assessment of NPD in both **research (to integrate clinical and social/personality perspectives) and practice (to inform diagnosis and treatment)** ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)).

## Structure & Factors of the IPN 

**Format and Scales:** The IPN/PNI is a self-report questionnaire with **52 items** ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=pathological%20narcissism%20vs,The%20PNI)). Respondents rate each item on a Likert-type scale (e.g. from *0/1 = not at all like me* to *5/6 = very much like me*) indicating how well each statement describes them ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=The%20PNI%20consists%20of%2052,under%20grandiose%20and%20vulnerable%20narcissism)) ([PAULSEN - DISSERTATION - FINAL VERSION](https://conservancy.umn.edu/bitstreams/2f65ec68-a04b-453d-add7-8d7898839d27/download#:~:text=Pathological%20Narcissism%20Inventory%20,pathological%20narcissism%2C%20each%20associated%20with)). The inventory yields **seven clinical subscales**, each representing a distinct dimension of pathological narcissism ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). These seven traits serve as **primary factors** of the IPN, and they roll up into two higher-order factors: **Narcissistic Grandiosity** and **Narcissistic Vulnerability** ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=Pathological%20Narcissism%20Inventory%20,concept%20with%20limited)). Below is a breakdown of the IPN’s factors and their content:

- **Narcissistic Grandiosity** – This higher-order domain reflects overt, outward narcissistic expressions: an **inflated self-image**, aggressiveness in asserting superiority, and lack of empathy. The IPN includes four grandiosity-related subscales ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)):  
  - *Entitlement Rage (ER):* Pronounced anger or frustration when one’s **entitled expectations** are not met ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=1)). Individuals high in ER feel they **deserve special treatment** and react with rage or resentment when others don’t accord them the respect or privileges they expect. *(Example item: “It irritates me when people don't notice how good a person I am.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=1)).  
  - *Exploitativeness (EXP):* A tendency to **manipulate or take advantage** of others to achieve one’s own goals ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=2)). High EXP is marked by using relationships for personal gain, deceit, and a lack of remorse for exploiting others. *(Example item: “I can make anyone believe anything I want them to.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=2)).  
  - *Grandiose Fantasy (GF):* **Fantasies of unlimited success, power, brilliance, or beauty** ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=3)). This facet involves an often unrealistic belief in one’s greatness or potential, with frequent daydreams about admiration and achievement. *(Example item: “I often fantasize about being admired for my accomplishments.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=3)).  
  - *Self-Sacrificing Self-Enhancement (SSSE):* Engaging in apparently **altruistic or self-sacrificing acts primarily to bolster one’s own ego** or gain recognition. This somewhat paradoxical facet captures people who may help or give to others, but in a way that is ultimately *self*-enhancing (seeking praise or moral superiority). *(Example item: not explicitly given above, but e.g. “I try to help others so that they will admire me.”)*. **Note:** SSSE is conceptually grandiose (it’s a self-enhancement strategy), but it’s a **covert form of grandiosity**; one study noted SSSE was the only grandiose facet not positively linked to impulsive, disinhibited behavior ([Development of a Brief Version of the Pathological Narcissism ...](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=Development%20of%20a%20Brief%20Version,Disinhibition%2C%20whereas%20both%20composites)) (suggesting it’s a more controlled, image-managing grandiosity).

- **Narcissistic Vulnerability** – This higher-order domain reflects covert or “fragile” narcissistic features: an **insecure self-concept**, sensitivity to criticism, and defensive behaviors. The IPN includes three vulnerability-related subscales ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)):  
  - *Contingent Self-Esteem (CSE):* A fragile sense of self-worth that is **highly dependent on others’ approval and admiration** ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=1.%20Contingent%20Self)). Those high in CSE experience extreme swings in self-esteem; they feel good about themselves only when receiving praise, but feel worthless or ashamed when validation is absent. *(Example item: “When people don't notice me, I start to feel bad about myself.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=1.%20Contingent%20Self)).  
  - *Hiding the Self (HS):* The tendency to **conceal one’s weaknesses or needs** for fear of being judged or seen as inferior ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=2)). High HS individuals are socially avoidant and defensive; they withdraw or put up a facade to prevent others from seeing their perceived flaws. *(Example item: “I try to hide my weaknesses so that others don't judge me.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=2)).  
  - *Devaluing (DEV):* **Diminishing or belittling others** as a defensive reaction when one’s esteem needs are not met ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=3)). This facet involves projecting blame or scorn onto others (e.g. viewing others as *inferior* or unworthy) to maintain a sense of superiority or to cope with inner shame. *(Example item: “I tend to view others as inferior to me.”)* ([Pathological Narcissism Inventory (PNI) — Moriel Mental Health ||| Riverside California](https://www.morielmentalhealth.com/blog/pathological-narcissism-inventory-pni#:~:text=3)).

**Scoring System:** Each of the seven subscale scores is computed by summing or averaging the item ratings for that facet. These subscales serve as **clinical scales**, meaning higher scores indicate greater presence of that maladaptive trait. The IPN can also yield composite scores: typically a **Grandiosity composite** (combining ER, EXP, GF, SSSE) and a **Vulnerability composite** (combining CSE, HS, DEV) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=Pathological%20Narcissism%20Inventory%20,concept%20with%20limited)). Researchers sometimes use an overall total score as well (an index of overall narcissistic pathology) ([PAULSEN - DISSERTATION - FINAL VERSION](https://conservancy.umn.edu/bitstreams/2f65ec68-a04b-453d-add7-8d7898839d27/download#:~:text=the%20two%20general%20components%20of,Sample%201%2C%20the%20total%20score)). Pincus et al. (2009) reported good internal consistencies for all subscales (α ≈ .75–.93) and the total score (α ≈ .95) ([PAULSEN - DISSERTATION - FINAL VERSION](https://conservancy.umn.edu/bitstreams/2f65ec68-a04b-453d-add7-8d7898839d27/download#:~:text=%28Contingent%20Self,In%20Sample%202%2C%20the%20total)). 

The structure has been replicated in independent samples and cultures ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=behavior%20%28r%20%3D%20)). For instance, Tritt et al. (2010) confirmed the 7-factor structure and showed that vulnerability facets (CSE, HS, DEV) strongly relate to depressive and anxious temperament, whereas grandiosity facets (especially GF and EXP) relate to more outgoing but volatile temperament ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=Several%20recent%20studies%20have%20extended,relating%20to%20the%20core%20affective)). Such findings reinforce the **two-factor model** within the IPN and validate that it captures the intended **vulnerable vs. grandiose** distinction in narcissism ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=to%20these%20variables%20in%20analyses,34%20Hendin%20%26%20Cheek)).

## Comparative Analysis: IPN vs. Other Narcissism Models

The IPN/PNI has become an important tool in understanding narcissism because it intentionally straddles the worlds of normal and pathological narcissism. Here we compare the IPN to several other models and measures of narcissism – highlighting overlaps in constructs and unique contributions of each:

- **Pathological Narcissism Inventory (PNI):** The term PNI refers to the same instrument as the IPN. Any comparison between “IPN and PNI” would be tautological – they are identical measures developed by Pincus et al. (the name “Pathological Narcissism Inventory” is the official one used in publications) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). If the user intended PNI to mean something else, it’s worth clarifying that in contemporary psychology, PNI = IPN. In this context, we treat them interchangeably. That said, the **PNI/IPN** can be contrasted with older conceptions of **“pathological narcissism.”** Before the PNI’s development, clinicians described narcissistic patients with terms like **grandiose vs. vulnerable narcissist** (or overt vs. covert) ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=Although%20the%20DSM,11)), but there was no widely accepted multi-facet self-report measure. The PNI filled this gap by providing specific scales for those manifestations. **Unique aspects of IPN/PNI:** it was the first comprehensive measure to capture **both polarities** of narcissism in one instrument ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=of%20narcissism%20has%20led%20to,Narcissism%20Scale%3B%20Hendin%20%26%20Cheek)). It goes beyond simple entitlement or vanity (as in the NPI) by including **internalized feelings** (shame, contingent self-esteem) and defensive behaviors (hiding, devaluation). In research, the PNI’s grandiosity scales correlate with other **antagonism** traits (like vindictiveness and dominance), while the vulnerability scales correlate with **internalizing distress** (e.g. anxiety, cold withdrawal) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=correlated%20negatively%20with%20self,several%20aspects%20of%20psychotherapy%20utilization)). This dual profile aligns with theories that narcissistic personality disorder can present with either outward arrogance or hypersensitivity and depression. The PNI thus uniquely bridges personality psychology and clinical diagnosis, addressing the “insufficient scope” of prior measures ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)).

- **Narcissistic Personality Inventory (NPI):** The NPI is the most established measure of **“normal” or subclinical narcissism**, developed by Raskin & Hall (1979) to reflect DSM-III narcissistic traits in a non-clinical format. It consists of forced-choice items assessing vanity, entitlement, leadership, and exhibitionism. **Overlap:** The NPI and IPN share content on **grandiose narcissism** – for example, both assess **entitlement** and a sense of superiority. In fact, studies have found that NPI scores correlate **strongly** with clinical narcissism measures; one study reported *r* = 0.66 between the NPI and the MCMI Narcissistic Personality Disorder scale ([Validity of the Narcissistic Personality Inventory (NPI) in a psychiatric sample - PubMed](https://pubmed.ncbi.nlm.nih.gov/6746921/#:~:text=Investigated%20the%20validity%20of%20the,construct%20validity%20of%20the%20NPI)), indicating that individuals scoring high on trait narcissism often show narcissistic personality features. Another study noted the NPI’s scores had a high correspondence (74% agreement) with NPD diagnoses on the MCMI in a patient sample ([Validity of the Narcissistic Personality Inventory (NPI) in a psychiatric sample - PubMed](https://pubmed.ncbi.nlm.nih.gov/6746921/#:~:text=Investigated%20the%20validity%20of%20the,construct%20validity%20of%20the%20NPI)). This suggests the **core construct** (grandiose self-importance) is captured by both. **Differences:** The crucial difference is one of **scope and context**. The NPI primarily measures **grandiose, “adaptive” narcissism** (e.g. confidence, leadership, exhibitionism) in the general population, and does **not** capture distress or dysfunction ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). Most people who score high on the NPI are not clinically disordered – as one source puts it, *“the NPI measures ‘normal’ narcissism… most people who score very high on the NPI do not have NPD”* ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). In fact, the NPI tends to correlate with high self-esteem and extraversion, and *negatively* with neuroticism (i.e. less anxiety) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Rosenthal%20%26%20Hooley%2C%202010%29,development%20of%20new%20measures%20that)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=grandiosity%20as%20a%20fa%C3%A7ade%20or,In)). By contrast, the IPN/PNI’s **vulnerability scales** correlate with low self-esteem and high neuroticism (e.g. PNI total scores correlated r ≈ –.37 with self-esteem, and r ≈ +.55 with shame) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=%28Pincus%2C%20in%20press%20%29,the%20PNI%20was%20predictive%20of)), which the NPI entirely misses. Additionally, the IPN includes maladaptive expressions of entitlement (e.g. rage) and exploitative tendencies causing interpersonal problems, whereas the NPI includes some items that might reflect more benign pride or leadership. In summary, the IPN encompasses **NPI-like grandiosity** but adds a whole other dimension of **insecurity and psychopathology**. The two instruments often show only moderate correlation (since NPI taps mostly the grandiose component of PNI). Researchers Thomas et al. (2012) found PNI “grandiosity” and “vulnerability” composites had different patterns of external correlates, and they argued that using the PNI in clinical contexts provides criterion validity that the NPI lacks ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=utility%20of%20the%20Pathological%20Narcissism,a%20host%20of%20criterion%20variables)) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=Accumulating%20evidence%20supports%20the%20reliability,Within)). In practical terms: **NPI** is ideal for assessing narcissistic traits in say, a college sample or the workplace (where narcissism might manifest as boldness or arrogance but not severe distress), whereas **IPN/PNI** is better for assessing if someone has a narcissistic *personality pathology* that might impact their mental health or therapy.

- **Five-Factor Narcissism Inventory (FFNI):** The FFNI (by Glover, Miller, Lynam, et al., 2012) is another multidimensional narcissism model. It was developed slightly after the PNI, amid recognition that narcissism has multiple facets and that those facets could be mapped onto the **Five-Factor Model (FFM)** of personality ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=of%20narcissism%20has%20led%20to,Narcissism%20Scale%3B%20Hendin%20%26%20Cheek)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,In%20addition%2C%20there%20are)). The FFNI is a **148-item** inventory measuring **15 traits** of narcissism, each corresponding to a maladaptive extreme of an FFM domain or facet ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)). For example, it includes traits like *Reactive Anger* (high Neuroticism facet), *Dominance* and *Exhibitionism* (extreme Extraversion facets), *Manipulativeness* (low Agreeableness), *Vanity* (high trait Narcissism), *Thrill-Seeking* (low Agreeableness/high Openness mix), etc ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=of%20narcissism%20has%20led%20to,Narcissism%20Scale%3B%20Hendin%20%26%20Cheek)). **Overlap:** The FFNI’s content overlaps substantially with the IPN. Many FFNI facets parallel PNI facets – e.g. FFNI “Entitlement” and “Exploitativeness” correspond to similar entitled and exploitative tendencies as in the IPN. Both models cover **grandiose fantasies, arrogance, entitlement, exploitative behavior, and emotional reactivity**. Indeed, both aim to capture **grandiose and vulnerable narcissism** (the FFNI has higher-order scales for Grandiose vs. Vulnerable narcissism, much like the PNI) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,95%29%20and%20vulnerable)). In an expert-rating study, the FFNI and PNI showed very similar profiles in how well they matched the DSM-5 trait prototypical profile of NPD ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Rosenthal%20%26%20Hooley%2C%202010%29,development%20of%20new%20measures%20that)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,as%20measures%20that%20capture%20DSM%E2%80%93IV%E2%80%93TR)), indicating both tap the pathological construct well. **Unique aspects:** The FFNI is grounded in the **FFM (Big Five)** framework, which means it conceptualizes narcissistic traits as concoctions of extreme Big Five traits ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=with%20basic%20personality%20traits%20%28e,Narcissistic)). For instance, FFNI “Distinguishing” (a facet) might reflect extremely low Agreeableness plus high Openness manifesting as uniqueness-seeking. This FFM lens allows researchers to see how narcissism relates to general personality structure – something the IPN does not explicitly do. Additionally, the FFNI includes a facet called *Grandiose Fantasies* (similar to PNI’s GF) and a facet called *Need for Admiration* that explicitly links to narcissistic admiration–rivalry concepts. The FFNI’s developers also created a short form and have done item-response analysis to ensure each facet is unidimensional ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=match%20at%20L769%20Five,Grandiose%20Exhibitionism%3B%20EE%20%02%20Entitlement%2FExploitativeness)). In summary, **FFNI vs. IPN**: Both are **comprehensive**, but the IPN is **clinically oriented** (derived from clinical theory and meant to signal pathology), whereas the FFNI is **structurally oriented** (derived from general personality traits). They tend to be highly correlated and often used complementarily. One study noted that the NPI (a purely grandiose measure) had the highest convergence with expert ratings of DSM NPD, but measures like PNI and FFNI were needed to capture the **vulnerable features** of narcissism that experts also recognize ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=of%20narcissism%20has%20led%20to,Narcissism%20Scale%3B%20Hendin%20%26%20Cheek)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Rosenthal%20%26%20Hooley%2C%202010%29,development%20of%20new%20measures%20that)).

- **Dark Tetrad / Dark Triad (DT4) Narcissism:** The “Dark Triad” refers to the overlapping trio of socially aversive traits – **Narcissism, Machiavellianism, and Psychopathy** – often extended to a “Dark Tetrad” by adding **Sadism** ([Dark Tetrad | Psychology Today](https://www.psychologytoday.com/us/basics/dark-tetrad#:~:text=The%20Dark%20Tetrad%2C%20also%20known,of%20the%20Dark%20Triad%20construct)). In these models (Paulhus & Williams, 2002; Buckels et al., 2013), **narcissism** is included as the **“socially charming but self-absorbed”** dark trait. Typically, Dark Triad researchers measure narcissism with either the NPI or a brief derived scale, focusing on **grandiose narcissism** only. For example, the Dark Tetrad is described as including *“Narcissism: a belief that one is special, gifted, obsessed with oneself and superior to others”* ([Dark Tetrad | Psychology Today](https://www.psychologytoday.com/us/basics/dark-tetrad#:~:text=People%20with%20Dark%20Tetrad%20traits,on%20measures%20of%20the%20following)). This definition mirrors the **IPN’s grandiose aspects** (especially Grandiose Fantasy and Exploitativeness). **Overlap:** IPN’s grandiosity facets (ER, EXP, GF, SSSE) overlap with the **antagonistic, entitled, and egocentric** features that narcissism contributes to the Dark Triad/Tetrad. All Dark Tetrad personalities share a tendency to view others as means to an end ([Dark Tetrad | Psychology Today](https://www.psychologytoday.com/us/basics/dark-tetrad#:~:text=Created%20with%20Sketch)), which aligns with the IPN’s Exploitativeness and lack of empathy. Additionally, Dark Triad measures of narcissism (like the Narcissistic Admiration and Rivalry Questionnaire’s Admiration component, or the Short Dark Triad’s narcissism subscale) capture **attention-seeking, grandiose self-view, and low modesty**, which correspond to IPN’s high self-enhancement and entitlement. **Unique aspects:** The Dark Triad conceptualization does **not account for narcissistic vulnerability** at all – it treats narcissism purely as a **“bright on the outside, dark on the inside”** construct (self-enhanced outwardly, callous underneath). Vulnerable narcissistic traits (e.g. shame, neediness) would actually *contradict* the typical Dark Triad profile, which is characterized by *low* Neuroticism and high self-esteem in narcissism ([Correlates of the Dark Tetrad - PubMed](https://pubmed.ncbi.nlm.nih.gov/38502994/#:~:text=This%20study%20explored%20demographic%2C%20ideological%2C,Trait)) ([Dark Tetrad | Psychology Today](https://www.psychologytoday.com/us/basics/dark-tetrad#:~:text=Those%20with%20high%20Dark%20Tetrad,of%20the%20Dark%20Tetrad%20traits)). Thus, the IPN’s vulnerable facets are unique and *not* represented in Dark Triad models. Another difference is that Dark Triad research often emphasizes **outcomes like aggression, impulse control, or workplace behavior**, linking narcissism to things like risk-taking and competitiveness ([Correlates of the Dark Tetrad - PubMed](https://pubmed.ncbi.nlm.nih.gov/38502994/#:~:text=personality%20test%20,with%20all%20four%20dark%20traits)). The IPN, in contrast, is used to predict outcomes in clinical domains (like suicide risk or interpersonal dysfunction). In summary, **Dark Triad narcissism** can be seen as a **subset** of IPN’s grandiose narcissism. Narcissism in DT4 overlaps chiefly with the **exploitive, arrogant, self-aggrandizing** components of IPN (and indeed, high IPN grandiosity would likely correspond to high Dark Triad narcissism). But IPN covers additional internal dynamics (e.g. contingent self-esteem) that Dark Triad models deliberately leave out.

- **Millon Clinical Multiaxial Inventory (MCMI) – Narcissism Scale:** The MCMI is a widely used clinical inventory (Theodore Millon’s instrument) that provides diagnostic indices for personality disorders, including Narcissistic Personality Disorder. The MCMI’s narcissism scale assesses traits like **egotism, grandiosity, lack of empathy, and a sense of entitlement** consistent with DSM criteria for NPD. **Overlap:** Conceptually, the MCMI’s narcissism scale covers similar ground to the IPN’s **grandiose domain** – a person who scores high would likely exhibit arrogance, **self-importance, and exploitation of others**. Empirically, as noted above, the MCMI narcissism scale correlates strongly with trait narcissism measures. In one validation, **NPI and MCMI narcissism correlated r = .66** in a psychiatric sample ([Validity of the Narcissistic Personality Inventory (NPI) in a psychiatric sample - PubMed](https://pubmed.ncbi.nlm.nih.gov/6746921/#:~:text=Investigated%20the%20validity%20of%20the,construct%20validity%20of%20the%20NPI)), suggesting they tap a common core of narcissism (despite one being a “normal” trait test and the other a diagnostic test). Another study found the MCMI Narcissistic scale had significant overlap with the **IPN/PNI** as well, especially on entitlement/exploitative aspects (since both were designed with DSM NPD features in mind) ([Validity of the Narcissistic Personality Inventory (NPI) in a psychiatric sample - PubMed](https://pubmed.ncbi.nlm.nih.gov/6746921/#:~:text=Investigated%20the%20validity%20of%20the,construct%20validity%20of%20the%20NPI)) ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=developed%20by%20Theodore%20Millon%20,79)). **Unique aspects:** The **MCMI is a clinical **screening tool**** – it’s intended to *diagnose or classify* someone as having Narcissistic Personality or not ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). It provides a single scale score (with a cutoff indicating likely NPD) rather than a nuanced facet profile. The IPN, by contrast, is **dimensional** and gives a multi-faceted profile (one can see which aspects of narcissism are pronounced). Another difference is that the MCMI is theory-driven by Millon’s model; Millon identified subtypes of narcissist (e.g. *“confident–egotistic”* which is a moderate form, versus *“narcissistic (pathological)”* vs. an even more extreme *“turbulent” variant introduced in MCMI-IV) ([[PDF] MCMI®-IV Scales | Pearson Assessments](https://www.pearsonassessments.com/content/dam/school/global/clinical/us/assets/mcmi-iv/mcmi-iv_scales.pdf#:~:text=%5BPDF%5D%20MCMI%C2%AE,provides%20clinicians%20with%20a)) ([[PDF] MCMI-IV comprehensive brochure - Pearson Clinical Assessment](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=%5BPDF%5D%20MCMI,broader%20range%20of%20personality)). The IPN doesn’t label subtypes, but one could imagine a profile with high grandiosity/low vulnerability on the IPN corresponding to Millon’s *“oblivious” (grandiose) narcissist*, and a profile with high vulnerability could correspond to a *“fragile” narcissist*. In practical use, the IPN is often used alongside or after broad measures like the MCMI to **quantify the severity** of narcissistic traits or to track changes. One notable overlap in construct: **Both MCMI and IPN treat narcissism as fundamentally maladaptive**. This is in contrast to the NPI where high scores can sometimes reflect high self-esteem without pathology. In fact, researchers have noted the **NPI and MCMI are well correlated, but the NPI detects “normal” narcissism while the MCMI indicates disordered narcissism** ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). The IPN falls closer to the MCMI’s purpose because it was calibrated on pathological expressions (it *negatively* correlates with adaptive traits like empathy and positively with pathology like identity diffusion ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=fit%20data%20from%20a%20large,43))). However, unlike the MCMI, the IPN is not used alone for diagnosis – rather, it complements clinical interviews and broader inventories. 

**Summary of Comparisons:** In essence, the IPN/PNI distinguishes itself by covering the **full spectrum** of narcissistic phenomena. It overlaps with traditional narcissism measures on **grandiose traits** (e.g. entitlement, exploitation, need for admiration) – these are common threads seen in NPI, FFNI, Dark Triad models, and diagnostic tools. What the IPN uniquely adds is the assessment of **vulnerable narcissism** (feelings of insecurity, hiding, shame). Other models each have their niche: the **NPI** excels at efficient assessment of *grandiose narcissism* in non-clinical settings ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)); the **FFNI** provides a deep trait-profile grounded in general personality structure ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)); the **Dark Triad/Tetrad** situates narcissism among other “dark” traits and highlights its shared antagonistic features ([Dark Tetrad | Psychology Today](https://www.psychologytoday.com/us/basics/dark-tetrad#:~:text=Those%20with%20high%20Dark%20Tetrad,of%20the%20Dark%20Tetrad%20traits)); the **MCMI** aligns with psychiatric diagnosis and underscores narcissism’s clinical severity ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=The%20Millon%20Clinical%20Multiaxial%20Inventory,79)). The IPN intersects with all of these, but fills an important gap by explicitly measuring the internal suffering and defensive fragility that often accompanies pathological narcissism ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=and%20to%20view%20oneself%20as,2009)) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=narcissistic%20vulnerability%20%28Contingent%20Self,homicidal%20ideation%2C%20and%20several%20aspects)). Modern research often uses multiple measures to capture narcissism – for example, using the IPN for a rich clinical picture and the NPI or NARQ (Admiration/Rivalry Questionnaire) for distinguishing interpersonal presentation styles. Notably, all these measures show **overlapping constructs** like **Entitlement** (present in IPN, NPI, FFNI, and as a key feature in NPD criteria and Dark Triad), and **Exploitativeness/Manipulativeness** (IPN EXP, FFNI Manipulativeness, a core trait in Machiavellianism as well). However, the IPN’s inclusion of **Contingent Self-Esteem and Hiding the Self** is fairly unique, tapping into the private emotional life of narcissistic individuals that other instruments often neglect.

## IPN Factors Dataset: Descriptive Terms 

Below is a structured breakdown of the **IPN’s seven primary factors** (subscales), along with descriptive adjectives, synonyms, action verbs, and related nouns for each. This can serve as a lexicon for understanding and identifying each facet of pathological narcissism:

- **Entitlement Rage** – When expectations of special treatment are unmet, leading to anger.  
  - *Adjectives*: **entitled**, **resentful**, **irritable**, **demanding**, arrogant.  
  - *Synonyms*: **indignant**, **aggrieved**, bitter, frustrated.  
  - *Verbs*: **demand**, **rant**, **lash out**, sulk, complain.  
  - *Nouns*: **entitlement**, **anger**, **rage**, resentment, frustration.  

- **Exploitativeness** – Tendency to use or manipulate others for personal gain.  
  - *Adjectives*: **manipulative**, **exploitative**, **cunning**, conniving, self-centered.  
  - *Synonyms*: **opportunistic**, **deceitful**, scheming, unscrupulous.  
  - *Verbs*: **exploit**, **manipulate**, **deceive**, take advantage (of), dominate.  
  - *Nouns*: **exploitation**, **manipulation**, deceit, selfishness, opportunism.  

- **Grandiose Fantasy** – Fantasies of unlimited success, power, or beauty; seeing oneself as extraordinary.  
  - *Adjectives*: **grandiose**, **fantasizing**, unrealistic, megalomaniacal, self-aggrandizing.  
  - *Synonyms*: **inflated**, delusional (of grandeur), aspirational (to an extreme), over-ambitious.  
  - *Verbs*: **fantasize**, **daydream**, imagine (greatness), idealize (oneself).  
  - *Nouns*: **fantasy**, **delusions of grandeur**, daydreams, ambition, glory.  

- **Self-Sacrificing Self-Enhancement** – Gaining self-esteem by acts that seem generous or selfless.  
  - *Adjectives*: **ingratiating**, **patronizing** (in help), ostentatiously generous, **self-aggrandizing** (under guise of helping), martyr-like.  
  - *Synonyms*: **self-serving** (altruism), **pretentious humble**, fake-generous, benevolent (for show).  
  - *Verbs*: **help** (to be admired), **sacrifice** (for attention), boast (subtly), **indulge** (in public kindness).  
  - *Nouns*: **grandiosity** (hidden in altruism), **martyrdom**, **self-enhancement**, show of generosity, patronage.  

- **Contingent Self-Esteem** – Self-worth that depends on external validation and admiration.  
  - *Adjectives*: **insecure**, **approval-seeking**, needy, sensitive (to criticism), dependent.  
  - *Synonyms*: **validation-seeking**, **self-doubting**, emotionally fragile, approval-dependent.  
  - *Verbs*: **seek approval**, **fish for compliments**, crave (praise), **feel inferior** (when not admired).  
  - *Nouns*: **insecurity**, **need for approval**, **self-doubt**, sensitivity, shame.  

- **Hiding the Self** – Concealing one’s shortcomings or true self for fear of judgment or rejection.  
  - *Adjectives*: **secretive**, **guarded**, defensive, **self-conscious**, withdrawn.  
  - *Synonyms*: **protective** (of self-image), **evasive**, closet (one’s flaws), introverted (for safety).  
  - *Verbs*: **hide**, **conceal**, cover up, withdraw, suppress (true feelings).  
  - *Nouns*: **self-concealment**, **mask**, **facade**, defense, privacy.  

- **Devaluing** – Dismissing others as worthless or inferior to maintain one’s own superiority.  
  - *Adjectives*: **dismissive**, **contemptuous**, critical, belittling, superior (acting).  
  - *Synonyms*: **scornful**, **disdainful**, derogatory, **patronizing**.  
  - *Verbs*: **devalue**, **belittle**, **mock**, criticize, denigrate.  
  - *Nouns*: **devaluation**, **contempt**, **scorn**, disdain, criticism.  

Each factor’s descriptors above paint a portrait of how that facet of narcissism might manifest in personality and behavior. For example, a person scoring high on **Exploitativeness** might be described as *manipulative* or *opportunistic*, seen *deceiving* others for personal gain; whereas someone high on **Hiding the Self** might appear *guarded* or *defensive*, often *withdrawing* from situations that threaten to expose their insecurities. This structured dataset is useful for psychologists in both research (e.g. creating item pools or interpreting factor profiles) and practice (e.g. explaining test results to clients or trainees in relatable terms).

##  Steps for IPN Personality Modeling Workflow 

1.  **Step 1: Create the Personality Dataset**

    * **Purpose:** This initial step involves defining and generating the dataset that represents the Eysenck Personality Model (IPN).
    * **Actions:**
        * Define the IPN dataset structure (e.g., using a dictionary to represent factors, adjectives, synonyms, etc.).
        * Generate the dataset by organizing the data into a Pandas DataFrame.
        * Save the dataset to a CSV file.
        * Log dataset metadata (e.g., number of rows, factors, data schema) and the dataset file itself as an artifact in MLflow.
    * **Importance:** This step creates the raw data that will be used for embedding generation and model training.

2.  **Step 2: API Key Handling and Initialization**

    * **Purpose:** This critical step ensures that your OpenAI API key is securely loaded and the OpenAI client is initialized. This sets the foundation for using the OpenAI API in subsequent steps.
    * **Actions:**
        * Load the OpenAI API key from a secure location (e.g., a file in the user's home directory).
        * Validate the API key (e.g., check for existence, emptiness, and potentially a basic API call).
        * Initialize the OpenAI client (`client`).
        * Log the API key handling process and its outcome in MLflow.
    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

3.  **Step 3: Test Embedding API**

    * **Purpose:** This step verifies that the OpenAI Embedding API is accessible and functioning correctly.
    * **Actions:**
        * Use the initialized OpenAI client (`client`) to make a test call to the Embedding API (e.g., by embedding a sample text).
        * Check the API response for validity.
        * Log the API call details and the outcome (success or failure) in MLflow.
    * **Importance:** This step ensures that you can successfully generate embeddings before proceeding to the next step.

4.  **Step 4: Create Embeddings for the Dataset**

    * **Purpose:** This step generates numerical representations (embeddings) for the text data in the IPN dataset using the OpenAI Embedding API.
    * **Actions:**
        * Load the IPN dataset (created in Step 2).
        * Use the OpenAI client (`client`) to generate embeddings for the relevant text fields (e.g., combining factor, adjective, synonym, verb, noun).
        * Add the generated embeddings as a new column in the Pandas DataFrame.
        * Save the DataFrame with embeddings to a new CSV file.
        * Log embedding generation parameters (e.g., embedding model used), statistics (e.g., embedding length), and the embeddings file as an artifact in MLflow.
    * **Importance:** This step transforms the text data into a numerical format that can be used for machine learning models.

5.  **Step 5: Create and Visualize a Label Encoder**

    * **Purpose:** This step prepares the categorical labels (personality factors) for model training by encoding them into numerical values and provides a visualization of this encoding.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Initialize a `LabelEncoder` from scikit-learn.
        * Fit the `LabelEncoder` to the 'Factor' column to create the mapping between personality factors and numerical codes.
        * Transform the 'Factor' column using the fitted `LabelEncoder` to create a new 'Factor_Encoded' column.
        * Save the fitted `LabelEncoder` object.
        * Generate a visualization (e.g., a bar chart) to show the mapping between original factors and encoded values.
        * Save the visualization as an image file.
        * Log the label encoder object and the visualization as artifacts in MLflow.
        * Log the mapping between original factors and encoded values as a dictionary in MLflow.
    * **Importance:** This step prepares the target variable for model training and provides a clear representation of the encoding.

6.  **Step 6: Create our IPN Model (Model Training and Evaluation)**

    * **Purpose:** This step trains a machine learning model on the generated embeddings to predict personality factors and evaluates its performance.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Prepare the data for model training:
            * Extract the embeddings as features (`X`).
            * Encode the 'Factor' column using the loaded `LabelEncoder` to get the target variable (`y`).
            * Split the data into training and testing sets.
        * Initialize a machine learning model (e.g., `RandomForestClassifier`).
        * Train the model on the training data.
        * Make predictions on the test data.
        * Evaluate the model's performance using appropriate metrics (e.g., accuracy, classification report, confusion matrix).
        * Generate visualizations of the evaluation results (e.g., confusion matrix plot).
        * Save the trained model.
        * Log model training parameters (e.g., hyperparameters), evaluation metrics, visualizations, and the trained model as artifacts in MLflow.
    * **Importance:** This step is the core of the machine learning process, where the model learns to predict personality factors from the embeddings.

7.  **Step 7: Model Testing (Inference on New Data)**

    * **Purpose:** This step demonstrates how to use the trained model to predict personality factors for new, unseen text inputs.
    * **Actions:**
        * Load the trained model (saved in Step 6).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Define a function that:
            * Takes new text as input.
            * Generates an embedding for the new text using the OpenAI API.
            * Uses the loaded model to predict the personality factor.
            * Uses the loaded `LabelEncoder` to decode the numerical prediction back to the original factor name.
        * Provide example new text inputs.
        * Use the function to predict personality factors for the example texts.
        * Print the predictions.
        * Log the test inputs and predictions in MLflow.
    * **Importance:** This step demonstrates the practical application of the trained model for making predictions on new data.

8.  **Step 8: Model Application, Visualization, and Analysis**

    * **Purpose:** This step provides additional visualization and analysis of the data and model.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Apply PCA for dimensionality reduction and visualization of the embeddings.
        * Generate and log PCA plots to visualize the embedding distribution.
        * Perform K-Means clustering on the embeddings to identify potential groupings or clusters of similar personality traits.
        * Add cluster labels to the dataset and save the clustered data.
        * Log clustering parameters (e.g., number of clusters) and the clustered data as artifacts in MLflow.
    * **Importance:** This step offers valuable insights into the data and model:
        * PCA visualization helps understand the distribution of embeddings.
        * Clustering can reveal underlying patterns in the data.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved ipn dataset from a CSV file
dataset_path = '../Datasets/ipn.csv'
try:
    ipn_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'ipn_df' is empty before processing
        if ipn_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", ipn_df.shape[0])
            print(f"📊 Processing {ipn_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        ipn_df['Embedding'] = ipn_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/ipn_embeddings.csv'
        ipn_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = ipn_df.shape[0]
        num_columns = ipn_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in ipn_df.columns and not ipn_df['Embedding'].empty:
            embedding_length = len(ipn_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(ipn_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/ipn_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/ipn_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "ipn_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import joblib
import mlflow
from openai import OpenAI
import os
from dotenv import load_dotenv
import ast  # Import ast for literal_eval

# Load the dataset containing embeddings
embeddings_csv_path = '../Embeddings/ipn_embeddings.csv'

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array using ast.literal_eval.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    try:
        return np.array(ast.literal_eval(embedding_str))
    except (ValueError, SyntaxError):
        return np.array()  # Or handle the error as appropriate

with mlflow.start_run(run_name="Step 9: Model Application, Visualization, and Analysis") as run:
    try:
        mlflow.log_param("Step", "Model Application, Visualization, and Analysis")

        # Define paths
        label_encoder_path = "../Models/ipn_label_encoder.pkl"
        model_path = "../Models/ipn_rf_model.pkl"

        mlflow.log_param("Embeddings CSV", embeddings_csv_path)
        mlflow.log_param("Label Encoder", label_encoder_path)
        mlflow.log_param("Model", model_path)

        # -------------------- Load Data and Artifacts --------------------

        # Load the dataset containing embeddings
        try:
            df = pd.read_csv(embeddings_csv_path)
            print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Embeddings dataset not found at: {embeddings_csv_path}"
            )
            print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}")
            raise

        # Apply the parsing function
        df['Embedding'] = df['Embedding'].apply(parse_embedding)

        # Print for debugging
        print("\nDataFrame with Embeddings:")
        print(df.head())
        print("\nShape of DataFrame:")
        print(df.shape)
        print("\nType of 'Embedding' column:")
        print(df['Embedding'].dtype)
        print("\nFirst Embedding:")
        if not df['Embedding'].empty:
            print(df['Embedding'].iloc[0])
        else:
            print("No embeddings found in DataFrame")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # -------------------- PCA Visualization --------------------

        mlflow.log_param("Visualization", "PCA")

        try:
            # Stack embeddings
            X = np.stack(df['Embedding'].values)

            # Print shape of X
            print("\nShape of X:")
            print(X.shape)

            # Encode Factor
            y_encoded = label_encoder.transform(df['Factor'].values)

            # Use PCA for dimensionality reduction
            pca = PCA(n_components=2)
            X_pca = pca.fit_transform(X)

            # Create the visualization
            plt.figure(figsize=(12, 8))
            scatter = plt.scatter(
                X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis'
            )
            plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
            plt.xlabel("PCA Component 1")
            plt.ylabel("PCA Component 2")
            plt.title("PCA of IPN Embeddings")

            # Save and log the plot
            pca_plot_path = "ipn_embeddings_pca.png"
            plt.savefig(pca_plot_path)
            mlflow.log_artifact(pca_plot_path, artifact_path="visualization")
            plt.close()  # Close the plot to release memory

            print(f"✅ PCA visualization saved to: {pca_plot_path}")

        except Exception as e:
            mlflow_error = f"An error occurred during PCA visualization: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ {mlflow_error}")
            raise

        # -------------------- K-Means Clustering --------------------

        mlflow.log_param("Clustering", "K-Means")

        try:
            n_clusters = 5  # You can parameterize this
            mlflow.log_param("n_clusters", n_clusters)

            # Stack embeddings (if not already done)
            if "X" not in locals():
                X = np.stack(df["Embedding"].values)

            # Cluster embeddings
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            clusters = kmeans.fit_predict(X)

            # Add cluster information to the DataFrame and save it
            df["Cluster"] = clusters
            clustered_csv_path = "ipn_clustered_embeddings.csv"
            df.to_csv(clustered_csv_path, index=False)
            mlflow.log_artifact(clustered_csv_path, artifact_path="clustering")

            print(
                f"✅ K-Means clustering performed. Clustered data saved to: {clustered_csv_path}"
            )

        except Exception as e:
            mlflow_error = f"An error occurred during K-Means clustering: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ {mlflow_error}")
            raise

        # -------------------- Model Application/Inference --------------------

        mlflow.log_param("Model Application", "Inference")

        try:
            # Load the trained model
            loaded_clf = joblib.load(model_path)

            def predict_factor(new_text):
                """
                Predicts the personality factor for new text using the trained model
                and the saved label encoder.

                Args:
                    new_text: The text to predict the personality factor for.

                Returns:
                    The predicted personality factor.
                """
                try:
                    # Get the embedding for the new text
                    new_embedding = get_embedding(new_text)

                    # Predict the factor
                    predicted_factor_encoded = loaded_clf.predict([new_embedding])

                    # Decode the prediction using the label encoder
                    predicted_factor = label_encoder.inverse_transform(
                        predicted_factor_encoded
                    )[0]

                    return predicted_factor

                except Exception as e:
                    print(f"❌ An error occurred during prediction: {e}")
                    raise

            # Example usage:
            new_text = "I love going to parties and meeting new people."
            predicted_factor = predict_factor(new_text)
            print(f"Predicted Factor for '{new_text}': {predicted_factor}")
            mlflow.log_param("New Text", new_text)
            mlflow.log_param("Predicted Factor", predicted_factor)

        except FileNotFoundError as e:
            mlflow_error = f"Model not found at: {model_path}"
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ Error: {e}")
            raise

        except Exception as e:
            mlflow_error = f"An error occurred during prediction: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ An error occurred during prediction: {e}")
            raise

    except Exception as e:
        # Catch any unexpected errors
        mlflow_error = f"An unexpected error occurred: {e}"
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", mlflow_error)
        print(f"❌ {mlflow_error}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Step 10: Save Model Visualization and Analysis complete.")
````

# **Step 13: Create IPN Schema in Neo4j** <a class="anchor" id="IPN_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("IPN")

# Load the dataset
dataset_path = "../Embeddings/ipn_embeddings.csv"
ipn_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in ipn_df.columns:
    ipn_df['Embedding'] = ipn_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the IPN schema in Neo4j
def create_ipn_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with IPN labels
        graph.run("MERGE (f:IPN_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:IPN_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:IPN_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:IPN_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:IPN_Noun {name: $noun})", noun=noun)
        
        # Create relationships with IPN namespacing
        graph.run("""
            MATCH (f:IPN_Factor {name: $factor}), (a:IPN_Adjective {name: $adjective})
            MERGE (f)-[:IPN_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:IPN_Adjective {name: $adjective}), (s:IPN_Synonym {name: $synonym})
            MERGE (a)-[:IPN_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:IPN_Adjective {name: $adjective}), (v:IPN_Verb {name: $verb})
            MERGE (a)-[:IPN_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:IPN_Adjective {name: $adjective}), (n:IPN_Noun {name: $noun})
            MERGE (a)-[:IPN_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:IPN_Factor {name: $factor}), (a:IPN_Adjective {name: $adjective})
                MERGE (e:IPN_Embedding {value: $embedding})
                MERGE (f)-[:IPN_HAS_EMBEDDING]->(e)
                MERGE (a)-[:IPN_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create IPN Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create IPN Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the IPN schema
        create_ipn_schema(graph, ipn_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", ipn_df.shape[0])
        mlflow.log_param("Number of Columns", ipn_df.shape[1])
        
        # Print the DataFrame
        print(ipn_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
