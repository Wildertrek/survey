# Dark Triad and Dark Tetrad Personality Models: History and Development

## Origins and Development of the Dark Triad (DT3)  
The **Dark Triad** refers to three closely related, socially aversive personality traits: **Machiavellianism**, **subclinical narcissism**, and **subclinical psychopathy** ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=The%20dark%20triad%20is%20a,8)). These are considered ‚Äúdark‚Äù due to their malevolent qualities (e.g. manipulativeness, callousness) even though they manifest in non-clinical populations ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=The%20dark%20triad%20is%20a,8)). Research shows that while the three traits overlap, they are distinct constructs ([The Dark Triad Of Personality - Narcissism, Machiavellianism, And Psychopathy 8](http://www.sakkyndig.com/psykologi/artvit/paulhus2002.pdf#:~:text=psychopaths%20exhibited%20self,of%20Machiavellianism%E2%80%94in%20short%2C%20the%20manip%02ulative)). All three share a common core of callous, manipulative interpersonal style ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=Ten%20years%20ago%20Paulhus%20and,located%20in%20mainstream%20structural%20models)), often aligning with low Agreeableness and low Honesty‚ÄìHumility in broad personality models ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=A%20factor%20analysis%20%20found,18)) ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=indis)).

**Key historical milestones in Dark Triad research include:**  

- **1998:** McHoskey et al. argue that narcissism, Machiavellianism, and psychopathy may be *interchangeable* in normal (non-criminal) samples ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=In%201998%2C%20John%20McHoskey%2C%20William,While%20some%20psychologists)). This provoked debate about whether these traits are truly distinct or just reflections of a single ‚Äúantisocial‚Äù factor.  

- **2002:** Delroy Paulhus and Kevin Williams formally coin the term **‚ÄúDark Triad‚Äù** and encourage studying the three traits in tandem ([Introducing the Short Dark Triad (SD3) - Daniel N. Jones, Delroy L ...](https://journals.sagepub.com/doi/10.1177/1073191113514105#:~:text=Introducing%20the%20Short%20Dark%20Triad,three%20traits%20in%20tandem%3A)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=In%201998%2C%20John%20McHoskey%2C%20William,While%20some%20psychologists)). In their seminal paper, they demonstrate that Machiavellianism, narcissism, and psychopathy‚Äîmeasured via established scales (e.g. Christie & Geis‚Äôs Mach-IV for Machiavellianism, Raskin & Hall‚Äôs NPI for narcissism, and a subclinical psychopathy scale)‚Äîare *overlapping but distinct* personality constructs ([The Dark Triad Of Personality - Narcissism, Machiavellianism, And Psychopathy 8](http://www.sakkyndig.com/psykologi/artvit/paulhus2002.pdf#:~:text=psychopaths%20exhibited%20self,of%20Machiavellianism%E2%80%94in%20short%2C%20the%20manip%02ulative)). This work established the Dark Triad as a collective concept in personality psychology.  

- **2008‚Äì2010:** Research on the Dark Triad grows rapidly. Studies begin mapping the triad traits onto broad models (e.g. Big Five correlations) and exploring their behavioral outcomes. For example, all three traits are found to correlate with antagonism (low agreeableness) and to predict an exploitative, ‚Äúfast life‚Äù mating strategy (e.g. more casual sex and mate poaching) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=personality%20traits%20have%20an%20accelerated,114%20%5D%20a)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=P.%20%28February%202009%29.%20,S2CID%C2%A012854051)). Importantly, short measures are introduced to facilitate research: Jonason & Webster (2010) developed the **‚ÄúDirty Dozen‚Äù** 12-item scale to assess all three traits concisely ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=A%20number%20of%20measures%20have,perceived%20as%20low%20in%20deceptiveness)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=24.%20,PMID%C2%A020528068)). Such tools made Dark Triad assessment quick and practical, spurring further studies.  

- **2013:** A decade into its study, the Dark Triad‚Äôs literature is robust. Furnham, Richards, & Paulhus (2013) published a **10-year review**, noting dozens of studies and over 350 citations of the original 2002 paper ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=Ten%20years%20ago%20Paulhus%20and,located%20in%20mainstream%20structural%20models)). They highlight that the triad traits, while conceptually distinct, share a *‚Äúcommon core of callous-manipulation‚Äù* ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=Ten%20years%20ago%20Paulhus%20and,located%20in%20mainstream%20structural%20models)). By this time, the Dark Triad had been examined in contexts ranging from workplace behavior to social media, demonstrating its broad relevance. Researchers also developed the 27-item **Short Dark Triad (SD3)** scale (Jones & Paulhus, 2014) to improve on earlier measures, allotting 9 items per trait for balanced coverage ([A brief measure of dark personality traits (SD3) - PsyToolkit](https://www.psytoolkit.org/survey-library/short-dark-triad.html#:~:text=A%20brief%20measure%20of%20dark,Assessment%2C%2021)).  

- **Ongoing:** The Dark Triad framework continues to inform psychological research. High scores on Dark Triad traits are linked to greater risk of unethical or harmful behavior ‚Äì for instance, higher likelihood of criminal activity, aggression, and organizational misbehavior ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=High%20scores%20in%20these%20traits,17)). Such individuals tend to be less empathetic, less agreeable, and more prone to use others for personal gain ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=High%20scores%20in%20these%20traits,17)). At the same time, studies note that these traits can confer short-term advantages (e.g. obtaining leadership positions or short-term mating success) at the expense of others ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=personal%20gain%20defines%20the%20negative,others%20or%20society%20at%20large)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=High%20scores%20in%20these%20traits,17)). The Dark Triad concept thus remains a popular tool for understanding the ‚Äúdark side‚Äù of personality in various domains of life.

## Expansion to the Dark Tetrad (DT4)  
As research evolved, psychologists considered whether other ‚Äúdark‚Äù traits should join the triad. **Sadism** ‚Äì the tendency to derive pleasure from inflicting pain or humiliation on others ‚Äì emerged as a prime candidate for a fourth element, leading to the idea of a **Dark Tetrad** ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=tions%20to%20our%20taxonomy%20of,example%2C%20has%20already%20been%20included)). Several researchers noted that sadistic tendencies satisfy the same core criterion of callousness and malevolence that defines the Dark Triad ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=indis)). Importantly, studies showed that **everyday sadism** (subclinical sadistic tendencies in normal populations) has unique predictive power beyond the original triad traits ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Several%20researchers%20have%20suggested%20that,51)). Key developments in the Dark Tetrad‚Äôs emergence include:

- **2009:** Chabrol et al. (2009) conduct one of the first studies incorporating **sadistic personality traits** alongside Machiavellianism, narcissism, and psychopathy. In a sample of adolescents, they found that sadism contributed to delinquent behaviors independently of the other three traits ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=48.%20,020)). This early evidence suggested that sadism captured an added dimension of ‚Äúdark‚Äù behavior (e.g. pro-active cruelty) not fully explained by Machiavellian manipulation, narcissistic entitlement, or psychopathic impulsivity.

- **2013:** Buckels, Jones, and Paulhus provide compelling support for including sadism. In a *Psychological Science* study, they demonstrate **behavioral confirmation of everyday sadism**: individuals high in sadism were willing to expend extra effort to harm innocent others (e.g. killing bugs in a lab task) purely for personal enjoyment, whereas those high in the Dark Triad *without* sadistic tendencies did not exhibit such gratuitous cruelty ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). The authors concluded that sadism encompasses *‚Äúdistinctly cruel traits not covered by the rest of the dark triad‚Äù*, warranting its addition to form a ‚ÄúDark Tetrad‚Äù ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). By the early 2010s, the term **Dark Tetrad** gained traction, referring to Machiavellianism, narcissism, psychopathy, **plus sadism** as the fourth dark trait ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=tions%20to%20our%20taxonomy%20of,example%2C%20has%20already%20been%20included)).

- **2015:** Research by Meƒëedoviƒá and Petroviƒá (2015) formally examined the **structure of the Dark Tetrad**. Their findings indicated that sadism is related to the other Dark Triad traits yet also stands as an independent factor in personality space ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Several%20researchers%20have%20suggested%20that,51)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). Sadism showed incremental validity, predicting antisocial and delinquent behaviors beyond what Machiavellianism, narcissism, and psychopathy could explain ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Furthermore%2C%20sadism%20predicted%20delinquent%20behavior,51)). This solidified the notion that the tetrad provides a more comprehensive view of dark personality. Around this time, the Dark Tetrad concept also spurred the development of integrated measures ‚Äì for example, Paulhus and colleagues later introduced a **Short Dark Tetrad (SD4)** 28-item scale to simultaneously assess all four traits. Factor analyses of the SD4 confirm a clear four-factor solution (one for each trait), with only modest inter-correlations among the traits ([Screening for Dark Personalities: The Short Dark Tetrad (SD4) ](https://www.erinbuckels.com/uploads/EJPA.SD4.with.pagination.pdf#:~:text=The%20first%20four%20factors%20were,However%2C%20a)). In other words, Machiavellianism, narcissism, psychopathy, and sadism can be measured as distinct (but related) dimensions in a person‚Äôs dark personality profile ([Screening for Dark Personalities: The Short Dark Tetrad (SD4) ](https://www.erinbuckels.com/uploads/EJPA.SD4.with.pagination.pdf#:~:text=The%20first%20four%20factors%20were,However%2C%20a)).

## Comparing the Dark Triad and Dark Tetrad  
The Dark Triad (DT3) and Dark Tetrad (DT4) models are similar in that they encompass socially aversive, antagonistic traits, but the DT4 explicitly **adds sadism** to the mix. This addition expands the model in meaningful ways:  

- **Unique Cruelty Component:** Sadism represents a *proactive cruelty* component ‚Äì an intrinsic enjoyment of harming others ‚Äì which is not explicitly captured by narcissism, Machiavellianism, or psychopathy. Psychopathy does involve callousness and antisocial behavior, but research shows that only those high in sadism actively seek opportunities to inflict suffering for pleasure ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). By including sadism, the Dark Tetrad addresses this dimension of cruelty. Studies have found that sadistic tendencies predict **unprovoked aggression** and willingness to hurt others even when it‚Äôs effortful, beyond what psychopathic impulsivity alone would predict ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=psychopathy%20and%20antisocial%20behavior%20,51)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). In essence, DT4 covers the full spectrum of interpersonal malevolence: from manipulative deceit (Machiavellianism) and grandiose entitlement (narcissism) to reckless callousness (psychopathy) and sadistic brutality (everyday sadism).

- **Incremental Predictive Power:** The inclusion of sadism in DT4 offers **incremental validity** in explaining negative behaviors. For example, sadism has been shown to predict delinquency and harmful behaviors *separately* from the other dark traits ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Furthermore%2C%20sadism%20predicted%20delinquent%20behavior,51)). One study noted that sadism was a stronger predictor of certain forms of abuse and criminal recidivism than psychopathic traits were ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=triad%20traits%20when%20evaluating%20high,51)). Thus, DT4 can account for outcomes (like gratuitous violence or vicarious enjoyment of others‚Äô pain) that DT3 might under-predict. In practical terms, a Dark Tetrad assessment may better identify individuals at risk for the most extreme forms of antisocial conduct.  

- **Structural Considerations:** Introducing a fourth factor does not drastically change the moderate inter-correlations seen among Dark Triad traits, but it adds nuance. Machiavellianism, narcissism, and psychopathy already tended to cluster around a shared core of callous, self-serving behavior ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=Ten%20years%20ago%20Paulhus%20and,located%20in%20mainstream%20structural%20models)). Sadism overlaps with this core (it is also correlated with antagonism and lack of empathy) ([(PDF) The Dark Triad of Personality: A 10Year Review](https://www.researchgate.net/publication/235914370_The_Dark_Triad_of_Personality_A_10Year_Review#:~:text=indis)), yet it remains distinguishable. Empirical factor analyses confirm that sadism emerges as its own factor when all four traits are assessed ([Screening for Dark Personalities: The Short Dark Tetrad (SD4) ](https://www.erinbuckels.com/uploads/EJPA.SD4.with.pagination.pdf#:~:text=The%20first%20four%20factors%20were,However%2C%20a)). Interestingly, certain sadism items (especially those about *direct physical harm*) can cross-load with psychopathy, reflecting their common ground in violence ([Screening for Dark Personalities: The Short Dark Tetrad (SD4) ](https://www.erinbuckels.com/uploads/EJPA.SD4.with.pagination.pdf#:~:text=A%20remaining%20challenge%20for%20the,%E2%80%9CI%20enjoy%20watching%20violent)). Even so, careful scale development (focusing on vicarious or playful cruelty items for sadism) can achieve clear separation ([Screening for Dark Personalities: The Short Dark Tetrad (SD4) ](https://www.erinbuckels.com/uploads/EJPA.SD4.with.pagination.pdf#:~:text=Following%20Johnson%20et%20al,number%20of%20physical%20sadism%20items)). Overall, DT3 and DT4 are organized similarly, but DT4 provides a **broader framework** by recognizing an additional facet of dark personality.

In summary, the Dark Tetrad builds upon the Dark Triad by explicitly accounting for sadistic tendencies, thereby offering a more complete portrait of an individual‚Äôs potential for malevolence. Researchers adopting the DT4 model argue that it *‚Äúthrows light on the dark triad‚Äù* by covering a facet of evil delight that the triad alone only hinted at ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Studies%20on%20how%20sadists%20gain,52)). That said, not all scholars fully agree on the necessity of a tetrad ‚Äì some note the overlap between sadism and psychopathy is considerable, and they caution against proliferating traits without strong justification. Nonetheless, the DT4 has gained recognition as a useful expansion when the focus is on **active cruelty and pleasure in others‚Äô suffering**, which extends the scope of Dark Triad research.

## Comparison with Narcissism-Focused Models (NPI, DT4, FFNI)  
It‚Äôs important to distinguish the Dark Triad/Tetrad‚Äôs treatment of **narcissism** from models that focus exclusively on narcissistic personality features. The Dark Triad‚Äôs narcissism component is typically assessed as *grandiose narcissism* in non-clinical form (often via the Narcissistic Personality Inventory), whereas other models provide a more fine-grained or clinical look at narcissism. Below is a comparison:  

- **Narcissistic Personality Inventory (NPI):** The NPI is a widely used self-report measure of **grandiose narcissism** developed by Raskin & Hall (1979). It captures traits like vanity, entitlement, dominance, and self-confidence. The Dark Triad paradigm generally uses NPI (or similar) to represent narcissism ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Assessment%20of%20narcissism%20required%20clinical,what%20many%20argue%20the%20NPI)), meaning DT3 research centers on *grandiose* narcissistic traits (charismatic but egotistical behavior). Notably, NPI-based narcissism is considered a **subclinical** form ‚Äì narcissism within the normal population, not necessarily reaching the level of narcissistic personality disorder ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=The%20dark%20triad%20is%20a,8)). This aligns with the DT3 approach of studying non-pathological ‚Äúeveryday‚Äù manifestations of dark traits. However, the NPI (and by extension, Dark Triad narcissism) does *not* assess narcissistic **vulnerability** (e.g. insecurity, shame), focusing instead on the overt, confident aspects of narcissism ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Assessment%20of%20narcissism%20required%20clinical,what%20many%20argue%20the%20NPI)). Thus, the Dark Triad‚Äôs view of narcissism is narrower compared to clinical conceptions.

- **Pathological Narcissism Inventory (DT4):** The DT4 was developed by Pincus et al. (2009) to address the ‚Äúinsufficient scope‚Äù of earlier narcissism measures ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)). It is a 52-item inventory measuring seven dimensions across two broad themes: **narcissistic grandiosity** (e.g. Exploitativeness, Entitlement Rage, Grandiose Fantasy) and **narcissistic vulnerability** (e.g. Contingent Self-Esteem, Hiding the Self, feelings of inadequacy) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=derivation%20and%20validation%20of%20the,associated%20with%20vindictive%2C%20domineering%2C%20intrusive)). In contrast to the NPI, the DT4 captures the darker, more sensitive side of narcissism ‚Äì including emotional fragility and deep-seated insecurity behind the grandiose facade ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=DT4%20is%20a%2052,avoidant%2C%20and%20exploitable%20interpersonal%20problems)). When comparing to DT models, the Dark Triad‚Äôs narcissism facet corresponds mostly to the grandiosity portion of the DT4 and hardly taps the vulnerable aspects. In fact, vulnerable narcissism (as measured by DT4) tends to correlate with traits like anxiety or even introversion, which are quite distinct from the callous, socially bold profile of Machiavellians and psychopaths ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=The%20vulnerable%20dark%20triad%20comprises,54)). Some researchers have even identified a separate **‚ÄúVulnerable Dark Triad‚Äù** cluster involving vulnerable narcissism, secondary (emotional) psychopathy, and borderline personality features ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=)) ‚Äì highlighting that the Dark Triad (in its classic form) does not encompass all narcissistic phenomena. In summary, the DT4 provides a broader, clinical perspective on narcissism that goes beyond the DT3‚Äôs focus, including dysfunction and internal distress that the Dark Triad‚Äôs narcissism measure would miss.

- **Five-Factor Narcissism Inventory (FFNI):** The FFNI (Glover, Miller, Lynam et al., 2012) is a comprehensive model that parses narcissism into **15 trait facets**, mapped onto the Five-Factor Model of personality ([Five Factor Narcissism Inventory, FFNI-SF](https://psytests.org/darktriad/ffniSen.html#:~:text=Online%20version%20of%20the%20Five,either%20or%20both%20narcissism%20dimensions)). Each facet represents a maladaptive variant of a basic personality trait that is relevant to narcissism ‚Äì for example, facets include *Manipulativeness* (low Agreeableness), *Entitlement*, *Authoritativeness* (high Extraversion), *Grandiosity*, as well as *Vulnerability* (high Neuroticism) and *Distrust* ([Five Factor Narcissism Inventory, FFNI-SF](https://psytests.org/darktriad/ffniSen.html#:~:text=Online%20version%20of%20the%20Five,either%20or%20both%20narcissism%20dimensions)). The FFNI thereby assesses both grandiose and vulnerable narcissistic tendencies within a unified framework. In relation to the Dark Triad/Tetrad, the FFNI‚Äôs narcissism profile is much more granular. The Dark Triad‚Äôs single ‚Äúnarcissism‚Äù score (usually via NPI or a short scale) would correspond to a combination of several FFNI facets, primarily those reflecting grandiose narcissism (like authoritativeness, exploitative tendencies, lack of empathy). What the Dark Triad doesn‚Äôt capture is the differentiation among those facets or the inclusion of vulnerability facets ‚Äì whereas the FFNI will show, for instance, how much of a narcissist‚Äôs personality is driven by **attention-seeking and charm** versus **anger and shame**. Another distinction is context: DT3/DT4 situates narcissism alongside Machiavellianism and psychopathy, emphasizing what narcissism has in common with those (e.g. antagonism). The FFNI, by contrast, situates narcissism in the broader personality space and dissects its internal composition. Therefore, a researcher interested specifically in narcissistic personality pathology might prefer DT4 or FFNI for depth, while one interested in narcissism as part of an **overall dark personality profile** may use the Dark Triad approach for breadth.

**In summary**, the Dark Triad/Tetrad models incorporate a *narrower, subclinical* view of narcissism (focusing on grandiose, overt traits) as one element among several antagonistic traits. Tools like the NPI align with this approach. On the other hand, specialized narcissism inventories (DT4, FFNI) provide a more **nuanced and expansive** assessment of narcissistic traits, including internal vulnerability and multiple facets. These models are complementary: the Dark Triad offers a broader lens on how narcissism interacts with other dark traits, whereas the DT4 and FFNI zoom in on the complexities within narcissism itself. Notably, all these models contribute to understanding narcissism‚Äôs role in maladaptive behavior. For example, grandiose narcissism (as in NPI/Dark Triad) has been tied to exploitativeness in relationships, while vulnerable narcissism (captured by DT4/FFNI facets) relates more to psychological distress and defensive aggression ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=DT4%20is%20a%2052,avoidant%2C%20and%20exploitable%20interpersonal%20problems)). By comparing DT3/DT4 with narcissism-specific models, researchers can better appreciate what each framework highlights: the Dark Triad/Tetrad underscores the *interpersonal impact* of narcissism as part of a constellation of toxic traits, whereas the FFNI and DT4 illuminate the *intrapersonal complexity* and clinical significance of narcissism in its own right.

## Relevance and Ongoing Research  
Both the Dark Triad and Dark Tetrad models have proven highly influential in psychological research. They offer a concise way to study ‚Äúdark‚Äù personality features and have been applied in many areas: **organizational psychology** (to understand toxic leadership and workplace deviance), **social psychology** (e.g. examining internet trolling, bullying, or cheating behaviors), **clinical psychology** (for insights into subclinical antisocial traits), and **evolutionary psychology** (exploring mating strategies and social competition). High DT scores are linked not only to antisocial outcomes but also to certain adaptive advantages in competitive environments, which makes these traits intriguing from an evolutionary standpoint ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=personal%20gain%20defines%20the%20negative,others%20or%20society%20at%20large)) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=High%20scores%20in%20these%20traits,17)). The Dark Tetrad, with its inclusion of sadism, is especially relevant in contexts of proactive aggression and cruelty (such as cyberharassment or violent criminality). As research progresses, scholars continue to refine these models ‚Äì for instance, investigating how the Dark Triad traits develop over the lifespan ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=A%20factor%20analysis%20%20found,18)), how they might be mitigated or channeled, and how they compare to ‚Äúlight‚Äù or positive personality traits in new models (e.g. the proposed ‚ÄúLight Triad‚Äù of empathy and compassion as a counterpoint) ([Dark triad - Wikipedia](https://en.wikipedia.org/wiki/Dark_triad#:~:text=Light%20triad)). 

In conclusion, the Dark Triad and Dark Tetrad serve as valuable frameworks for encapsulating the most antagonistic aspects of personality. The addition of the fourth trait in DT4 reflects the field‚Äôs effort to capture the full range of everyday evil, highlighting that beyond vanity, deceit, and callousness lies an even darker enjoyment of cruelty. Together with specialized narcissism measures like NPI, DT4, and FFNI, these models enrich our understanding of narcissistic and antisocial personalities, ensuring both breadth and depth in the study of the darker corners of human personality.

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key Status", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"‚úÖ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key Status", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key Status", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("üîë API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("ü§ñ OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("‚úÖ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"‚ö†Ô∏è Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"‚ö†Ô∏è Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/dt4_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/dt4_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="dt4 Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "dt4_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"üìä Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "dt4_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"üìä Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "dt4_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"üñºÔ∏è Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/dt4_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"üíæ Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"‚úÖ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/dt4_label_encoder.pkl"
model_path = "../Models/dt4_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"‚úÖ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"‚ùå Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"‚ùå An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"‚ùå Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n‚úÖ Model testing (inference on new data) complete.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/dt4_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="DT4 Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/dt4_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "dt4_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/dt4_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of DT4 Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/dt4_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of DT4 Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create DT4 Schema in Neo4j** <a class="anchor" id="DT4_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("DT4")

# Load the dataset
dataset_path = "../Embeddings/dt4_embeddings.csv"
dt4_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in dt4_df.columns:
    dt4_df['Embedding'] = dt4_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the DT4 schema in Neo4j
def create_dt4_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with DT4 labels
        graph.run("MERGE (f:DT4_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:DT4_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:DT4_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:DT4_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:DT4_Noun {name: $noun})", noun=noun)
        
        # Create relationships with DT4 namespacing
        graph.run("""
            MATCH (f:DT4_Factor {name: $factor}), (a:DT4_Adjective {name: $adjective})
            MERGE (f)-[:DT4_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:DT4_Adjective {name: $adjective}), (s:DT4_Synonym {name: $synonym})
            MERGE (a)-[:DT4_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:DT4_Adjective {name: $adjective}), (v:DT4_Verb {name: $verb})
            MERGE (a)-[:DT4_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:DT4_Adjective {name: $adjective}), (n:DT4_Noun {name: $noun})
            MERGE (a)-[:DT4_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:DT4_Factor {name: $factor}), (a:DT4_Adjective {name: $adjective})
                MERGE (e:DT4_Embedding {value: $embedding})
                MERGE (f)-[:DT4_HAS_EMBEDDING]->(e)
                MERGE (a)-[:DT4_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create DT4 Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create DT4 Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the DT4 schema
        create_dt4_schema(graph, dt4_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", dt4_df.shape[0])
        mlflow.log_param("Number of Columns", dt4_df.shape[1])
        
        # Print the DataFrame
        print(dt4_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
