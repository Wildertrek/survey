# Pathological Narcissism Inventory (PNI)  

The **Pathological Narcissism Inventory (PNI)** is a self-report questionnaire developed to assess *pathological* narcissism, capturing both **narcissistic grandiosity** and **narcissistic vulnerability** in one measure ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=consistently%20describe%20variations%20in%20the,the%20diagnostic%20emphasis%20on%20overt)) ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=grandiosity%2C%20DSM%20NPD%20has%20been,could%20lead%20to%20a%20more)). It was created by Aaron Pincus and colleagues in the late 2000s to address gaps in existing narcissism scales. The PNI provides a more clinically oriented assessment of narcissistic personality pathology compared to traditional “normal” narcissism measures ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)). It consists of multiple subscales that reflect maladaptive self-enhancement strategies, entitlement, and underlying emotional insecurity. Below is a timeline of the PNI’s development and refinements, followed by a comparison of the PNI with two other key narcissism inventories: the Narcissistic Personality Inventory (NPI) and the Five Factor Narcissism Inventory (FFNI).  

## Development and Timeline of the PNI  

- **2008 – Theoretical Foundations:** Nicole M. Cain, Aaron Pincus, and Emily Ansell publish a review highlighting two phenotypic themes of pathological narcissism: **grandiose** narcissistic behaviors versus **vulnerable** self-esteem and affect. They note that clinical theories for decades described both overt grandiosity and hypersensitive vulnerability, whereas DSM criteria and common measures focused mainly on grandiosity ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=consistently%20describe%20variations%20in%20the,the%20diagnostic%20emphasis%20on%20overt)) ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=grandiosity%2C%20DSM%20NPD%20has%20been,could%20lead%20to%20a%20more)). This article underscored the need for a comprehensive measure covering both facets of narcissism.  

- **2009 – PNI Introduction:** Pincus et al. introduce the **Pathological Narcissism Inventory**, a 52-item self-report measure, in *Psychological Assessment*. Through four studies, they establish the PNI’s factor structure and initial validity ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=existing%20narcissism%20measures,The%20PNI)) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,and%20vulnerable%20PNI%20scales%20were)). The PNI contains **7 subscales** spanning two higher-order dimensions: narcissistic **Grandiosity** (comprising *Entitlement Rage, Exploitativeness, Grandiose Fantasy,* and *Self-Sacrificing Self-Enhancement*) and narcissistic **Vulnerability** (comprising *Contingent Self-Esteem, Hiding the Self,* and *Devaluing*) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,and%20vulnerable%20PNI%20scales%20were)). This structure was supported by confirmatory factor analysis, confirming that the PNI captures the dual nature of pathological narcissism.  

- **2012 – Validity Evidence:** Follow-up research provides **criterion validity** and clinical utility data for the PNI. For example, Lukowitsky et al. (2012) found that the PNI’s grandiosity and vulnerability scales showed distinct patterns of correlations with personality traits and psychopathology, supporting their convergent and discriminant validity ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=In%20this%20study%20we%20evaluated,clinicians%20without%20specific%20training%20in)) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=pathological%20narcissism%20as%20well%20as,and%20utility%20of%20the%20narcissistic)). Clinicians (both those experienced with narcissism and those without specialized training) could meaningfully predict how PNI grandiosity vs. vulnerability relate to other traits, underscoring the PNI’s clinical interpretability ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=traits,clinicians%E2%80%99%20predictions%20that%20was%20more)) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=fully%20elaborated%20by%20examining%20systematic,as%20measured%20by%20the%20PNI)). This period established that PNI grandiosity and vulnerability are **reliably distinguishable constructs** with different external correlates, aligning with expert conceptions of narcissistic personality.  

- **2013 – Continued Research:** By the early 2010s, accumulating evidence supported the PNI’s **psychometric soundness** and clinical relevance. Pincus (2013) reviewed growing findings that PNI scores are reliable and valid ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=Pincus%20and%20colleagues%20created%20the,Ellison%2C%20Levy%2C%20Cain%2C%20Ansell%2C)). Studies showed that high PNI scores are associated with a range of clinically meaningful outcomes: for instance, total PNI and its facets correlate negatively with self-esteem and empathy (as expected in pathological narcissism) and positively with shame, interpersonal distress, aggression, and borderline personality features ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=narcissism%20spanning%20problems%20with%20narcissistic,avoidant%2C%20and%20exploitable%20interpersonal%20problems)). Narcissistic **grandiosity** (via PNI) was linked to interpersonally exploitative, domineering, and vindictive tendencies, whereas narcissistic **vulnerability** was linked to internalized distress, social avoidance, and feelings of inadequacy ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20PNI%20structure%20was%20validated,homicidal%20ideation%2C%20and%20several%20aspects)). Importantly, in clinical samples the PNI also showed ties to severe outcomes like self-harm, suicidality, and even homicidal ideation ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=and%20overly,several%20aspects%20of%20psychotherapy%20utilization)). These findings reinforced the PNI’s utility in capturing the maladaptive and distressing aspects of narcissistic personality disorder (NPD) that earlier measures were missing.  

- **2015 – Brief PNI (B-PNI):** To improve practicality, a **short form** of the inventory was developed. Wright and colleagues created the **Brief PNI**, a 28-item version, by selecting the best-performing items from each of the 7 facets ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=undergraduate%2C%20community%2C%20and%20clinical%20populations,loss%20of%20information%2C%20which%20may)). The B-PNI preserves the original’s factor structure (all seven subscales and the two broad composites) and demonstrates reliability and validity comparable to the full 52-item PNI ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=undergraduate%2C%20community%2C%20and%20clinical%20populations,assess%20the%20various%20facets%20of)) ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=pathological%20narcissism%20effectively%20and%20without,pathological%20narcissism%20in%20future%20work)). This refinement made it easier to include a comprehensive narcissism measure in research and clinical assessments without sacrificing content coverage.  

- **2022 – State Version:** Recognizing that narcissistic expressions can fluctuate, Engyel et al. (2022) introduced a **state version** of the PNI for experience sampling. The **PNI-S** is a momentary assessment tool measuring **transient narcissistic states** (both grandiose and vulnerable) in daily life ([Frontiers | Momentarily narcissistic? Development of a short, state version of the Pathological Narcissism Inventory applicable in momentary assessment](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2022.992271/full#:~:text=Background%3A%20Narcissism%20viewed%20as%20a,states%2C%20applicable%20in%20momentary%20assessment)) ([Frontiers | Momentarily narcissistic? Development of a short, state version of the Pathological Narcissism Inventory applicable in momentary assessment](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2022.992271/full#:~:text=the%20factor%20structure%20and%20associations,n%20%3D%20123)). Initial studies found that the state PNI had acceptable factor structure and correlated appropriately with other narcissism measures at both the within-person (state) and between-person (trait) levels ([Frontiers | Momentarily narcissistic? Development of a short, state version of the Pathological Narcissism Inventory applicable in momentary assessment](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2022.992271/full#:~:text=the%20factor%20structure%20and%20associations,n%20%3D%20123)). This development extends the PNI’s utility to studying how narcissistic feelings and behaviors ebb and flow over time.  

*(The PNI has also been translated and validated in multiple languages and populations, further supporting its robustness. For instance, researchers have examined its cross-cultural measurement invariance and found it broadly reliable across different groups. These efforts underscore the PNI’s status as a key instrument for pathological narcissism.)*

**Structure and Subscales:** As shown above, each inventory has a distinct structure. The **NPI** has a seven-factor structure identified by Raskin & Terry (1988), which largely taps traits of *grandiose narcissism* (e.g. leadership/authority, exhibitionism, entitlement) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=,example%20of%20one%20is%20below)). The **PNI**, in contrast, includes seven facets divided into two broader domains, explicitly modeling the two faces of narcissism – it assesses not only grandiose exploitative tendencies but also vulnerable aspects like fragile self-esteem and concealment of weaknesses ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,and%20vulnerable%20PNI%20scales%20were)). The **FFNI** provides an even more fine-grained trait profile with 15 facets; these were developed by mapping narcissistic characteristics onto the well-established Big Five trait framework ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20Five,good%20convergent%20and%20criterion%20validity)). Notably, both the PNI and FFNI yield separate **grandiosity** and **vulnerability** scores, whereas the NPI predominantly yields an overall narcissism score (or at best, a division into components of grandiosity) and does **not** have a built-in vulnerable narcissism scale.  

**Theoretical Foundations:** Each measure’s design reflects different theoretical origins. The **NPI** was originally based on the diagnostic criteria for Narcissistic Personality Disorder in DSM-III, but intended for use in non-clinical populations ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=initial%20development%2C%20the%20NPI%20has,1)). It emerged from social/personality psychology to measure narcissistic traits like vanity and entitlement in the general population, often seen as an index of “normal” narcissism (sometimes even adaptive confidence) rather than pathology ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)). In contrast, the **PNI** was grounded in clinical psychology and personality disorder theory. It draws on psychodynamic concepts (e.g., Kohut’s *narcissistic vulnerability* and Kernberg’s *pathological grandiosity*) to capture the maladaptive, distressing manifestations of narcissism ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=consistently%20describe%20variations%20in%20the,the%20diagnostic%20emphasis%20on%20overt)) ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=grandiosity%2C%20DSM%20NPD%20has%20been,could%20lead%20to%20a%20more)). The **FFNI** is rooted in the Five-Factor Model of personality – its developers conceptualized narcissism as a constellation of extreme traits (especially extremely low Agreeableness, plus aspects of high Extraversion and high Neuroticism) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)). This FFM grounding means the FFNI is theory-driven from a trait perspective, attempting to integrate clinical narcissism with general personality structure ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20Five,good%20convergent%20and%20criterion%20validity)). In summary, **NPI** reflects a *personality-social psychology* approach (emphasizing grandiose traits), **PNI** reflects a *clinical psychology* approach (emphasizing pathology in both grandiose and vulnerable domains), and **FFNI** reflects a *trait-model integration* approach (placing narcissism in the context of the Big Five).  

## Differences and Overlaps in Measurement  

Despite measuring the same broad construct of narcissism, the PNI, NPI, and FFNI differ in both **content emphasis** and **assessment format**, with some overlaps in what they cover:

- **Grandiose Narcissism:** All three instruments assess narcissistic *grandiosity*, but with different nuance. **Entitlement** and **Exploitativeness** appear in each measure. For example, the NPI has an *Entitlement* subscale and items about exploiting others, the PNI includes *Exploitativeness* and *Entitlement Rage* facets, and the FFNI has trait facets for Entitlement, Exploitative tendencies, and Lack of Empathy ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)). These overlapping elements indicate consensus on core grandiose features (a sense of superiority and willingness to use others). The **NPI**, however, also contains some arguably more *benign* grandiosity facets like *Authority* (leadership/assertiveness) and *Self-Sufficiency*, which reflect confidence and independence. The PNI and FFNI tend not to frame these as positively – for instance, the FFNI’s “Authoritativeness” facet is couched in a maladaptive context (bossy, domineering tendencies), and the PNI doesn’t have a direct “leadership” scale at all (as extreme self-confidence is not necessarily viewed as pathological unless paired with entitlement or aggression). Additionally, the NPI uniquely has a *Vanity* subscale (concern with one’s physical appearance and superiority) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=,example%20of%20one%20is%20below)), which the PNI and FFNI do not explicitly separate – vanity might be partly captured under PNI’s grandiose fantasy or FFNI’s exhibitionism, but it’s less emphasized. Thus, **NPI covers some socially bold or superficial aspects of grandiosity more than the clinically oriented PNI**, which focuses on the darker grandiose traits (rageful entitlement, fantasies of unlimited success, exploiting others).  

- **Vulnerable Narcissism:** A key difference is that the **NPI lacks measures of narcissistic vulnerability**, whereas **PNI and FFNI explicitly include them**. The PNI introduced facets like *Contingent Self-Esteem* (instability of self-worth depending on others’ validation), *Hiding the Self* (avoiding exposure of imperfections), and *Devaluing* others when disappointed ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,and%20vulnerable%20PNI%20scales%20were)) – all of which capture the fragile, defensive side of narcissism. Similarly, the FFNI includes vulnerability facets such as *Shame* and *Need for Admiration* (indicating insecurity and a desperate desire for approval), as well as *Reactive Anger* and *Distrust* which reflect hypersensitivity to criticism or rejection ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)). These vulnerable aspects overlap conceptually with PNI’s content (e.g. both PNI and FFNI assess proneness to shame and anger when one’s ego needs aren’t met). The NPI, by contrast, was not designed to tap feelings of inadequacy or internal distress – its forced-choice items typically contrast an overtly narcissistic statement with a modest one, so respondents high in vulnerability (who might agree with self-critical or anxious statements) do not score highly on NPI. This is a fundamental distinction: **PNI and FFNI capture both “faces” of narcissism (grandiose and vulnerable)**, whereas **NPI primarily captures the grandiose face** ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)).

- **Format and Response Style:** The NPI’s format is another notable difference. The NPI uses a **forced-choice** format (each item forces the respondent to choose between a narcissistic statement and a non-narcissistic one) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=,example%20of%20one%20is%20below)). This can influence how results are obtained – it measures narcissism in a relatively *normative* way, possibly reducing social desirability bias for agreeing with egotistical statements (since one must choose one of two options). The PNI and FFNI use **Likert-type rating scales**, asking respondents to rate agreement with statements on a continuum. Likert scales (e.g., 0–5 or 1–5) allow capturing the degree of trait presence and tend to be more sensitive to distress and internal experiences. For instance, a PNI item like *“When others don’t notice me, I start to feel worthless”* can be rated from “not at all” to “very much,” directly tapping a vulnerable feeling that has no equivalent in the binary NPI format. The Likert format of PNI and FFNI is well-suited to clinical samples (who may strongly endorse maladaptive feelings), while the NPI’s format is geared toward detecting relative narcissistic tendencies in the general population. 

- **Unique Facet Content:** Each inventory includes some unique facets that the others do not. The **PNI** has a facet called *Self-Sacrificing Self-Enhancement*, which describes doing things for others or appearing altruistic as a way to receive praise or bolster one’s self-image. This facet is a nuanced combination of grandiosity and vulnerability (a covert way to seek admiration) and is not explicitly represented in the NPI or FFNI. The **FFNI**, on the other hand, introduces facets like *Thrill-Seeking* (a tendency toward risky, excitement-seeking behavior) and *Indifference* (a lack of concern for others’ needs or feelings, conceptually akin to low empathy) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Distrust,Acclaim%20Seeking%2C%20and%20Thrill%20Seeking)) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)). While thrill-seeking is not a classic narcissistic trait per se, it correlates with the impulsivity sometimes seen in narcissistic personalities (especially overlapping with antisocial tendencies). *Indifference* (particularly when reversed, indicating hypersensitivity) ties into the FFNI’s vulnerable dimension and has some overlap with the PNI’s *Hiding the Self* or *Contingent Self-Esteem*. The **NPI** uniquely covers *Self-Sufficiency* (believing one can handle things solo and not rely on others) – PNI/FFNI do not isolate this trait, although an extreme form of self-sufficiency might be reflected in low help-seeking or distrust in those measures. In sum, there is considerable **overlap in core narcissistic themes** (entitlement, exploitation, exhibitionism) across the three inventories, but **PNI and FFNI expand into internalized insecurity** which NPI omits, and each measure includes a few distinct sub-traits aligned with its theoretical orientation.  

## Psychometric Properties and Validity  

**Reliability (Internal Consistency):** All three inventories generally show good internal consistency for their total scores, with more variation at the subscale level. The **NPI-40** (the most common NPI form) typically yields a reliable total narcissism score (Cronbach’s alphas often in the .80s), though some of its seven subscale scores can be less reliable due to having few items per facet (e.g., the original Entitlement subscale has only a couple of items). The PNI’s scales were designed to be broader; Pincus et al. (2009) reported adequate to good alphas for the seven facets (many in the .70–.90 range) and strong reliability for the higher-order grandiosity and vulnerability composites. The **FFNI**, with 148 items, provides very robust reliability – its facet scales typically have high internal consistencies, and the composite scores for grandiose and vulnerable narcissism are highly reliable (benefiting from many items aggregating). In short, **NPI has the brevity advantage**, but its subscale fidelity is debated, whereas **PNI and FFNI, being longer and multi-faceted, achieve solid reliability** for both overall and specific dimensions.  

**Factor Structure:** The structure of these measures has been studied extensively. The NPI’s factor structure has been a subject of debate – while Raskin & Terry (1988) proposed seven factors ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=4.%20,3514.54.5.890)), later research found that a shorter **three-factor** or even a two-factor (leadership/authority vs. grandiose exhibitionism/entitlement) model might better fit the data in some cases ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=match%20at%20L229%20Given%20the,10)). This inconsistency has led to critiques of the NPI’s structural validity ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=predominantly%20%E2%80%9Cnormal%E2%80%9D%20or%20%E2%80%9Cadaptive%E2%80%9D%20narcissism,Fraley%2C%20in%20press%3B%20%2028)). The PNI’s seven-factor structure (with two higher-order factors) has been supported by confirmatory factor analysis in the original study ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,and%20vulnerable%20PNI%20scales%20were)) and subsequent research, including tests of measurement invariance. The two broad PNI dimensions (grandiosity and vulnerability) are usually moderately correlated but distinguishable, which mirrors theory. The FFNI was developed with an intended structure (15 facets grouped into three FFM domains and two narcissism dimensions) and research generally supports this organization; factor analyses confirm the grouping of FFNI facets into a large **Antagonism** factor (capturing the shared core of narcissism – egocentric, callous traits) and separable aspects corresponding to grandiose (e.g. high Extraversion, low Neuroticism facets) versus vulnerable (high Neuroticism facets) narcissism ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Total%20Score%20%3D%20sum%20of,Acclaim%20Seeking%2C%20and%20Thrill%20Seeking)). Overall, **PNI and FFNI demonstrate well-defined multifaceted structures aligning with theoretical expectations**, whereas the NPI’s factor structure, while conceptually divided into facets, is sometimes treated uni-dimensionally due to ongoing debates about its exact composition.  

**Convergent and Discriminant Validity:** Each inventory correlates with external variables in ways that reflect its focus. The **NPI** shows strong links with normal personality traits such as high **Extraversion** and low **Agreeableness** (especially low modesty and high dominance) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=Rhodewalt%20%26%20Morf.,strong%20construct%20validity%20and%20ecological)). NPI scores tend to correlate with high self-esteem and low neuroticism (since vulnerable individuals score low on NPI), indicating that high NPI scorers often exhibit confident, outgoing personalities. For example, studies have found NPI scores associated with leadership emergence and extraverted behavior, as well as aggressiveness when ego-threatened (consistent with a grandiose profile). However, NPI has **weaker associations with pathology** – it often shows only modest correlations with measures of psychological distress, depression, or anxiety. In fact, the NPI has been criticized for *weak convergent validity with clinical narcissism*: many who score high on NPI do not necessarily have the impairments of NPD ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)). Researchers have noted that some NPI items capture relatively normative traits (e.g., *“I like to take responsibility for making decisions”* might correlate with healthy leadership qualities). Thus, the NPI can be seen as measuring the **socially assertive and confident** side of narcissism more than the troubled, clinical side ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)).  

In contrast, the **PNI** shows a very different pattern: it correlates with a range of maladaptive outcomes. PNI *Grandiosity* scores correlate positively with traits like aggression, lack of empathy, domineering interpersonal styles, and other personality disorder features (like antisocial and histrionic tendencies) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=narcissism%20spanning%20problems%20with%20narcissistic,avoidant%2C%20and%20exploitable%20interpersonal%20problems)). PNI *Vulnerability* scores correlate with high neuroticism, anxiety, depression, low self-esteem, shame, and interpersonal insecurity ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=narcissism%20spanning%20problems%20with%20narcissistic,avoidant%2C%20and%20exploitable%20interpersonal%20problems)) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=associated%20with%20cold%2C%20socially%20avoidant%2C,several%20aspects%20of%20psychotherapy%20utilization)). Importantly, PNI total and subscale scores have demonstrated links to clinical indicators such as self-harm, suicidal ideation, and poorer therapy outcomes ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=and%20overly,several%20aspects%20of%20psychotherapy%20utilization)). These correlations affirm that the PNI is tapping the **pathological expressions** of narcissism – individuals who score high on PNI tend to experience significant internal distress and interpersonal dysfunction. Notably, the PNI’s inclusion of vulnerability allows it to correlate with symptoms that the NPI would miss (e.g. PNI vulnerability correlates with emotional lability and depressive feelings). At the same time, PNI grandiosity still converges with other measures of grandiose narcissism (it does correlate positively with NPI scores and with observer ratings of narcissistic traits), but it provides a more nuanced picture by also relating to underlying distress. Overall, the PNI exhibits strong **convergent validity** with constructs like shame, entitlement, and interpersonal exploitation, and good **discriminant validity** in separating grandiose vs. vulnerable correlates ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=In%20this%20study%20we%20evaluated,clinicians%20without%20specific%20training%20in)) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=pathological%20narcissism%20as%20well%20as,These%20results)). Researchers have concluded that PNI’s two components map well onto the clinical understanding of narcissism’s dual nature ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=traits,clinicians%E2%80%99%20predictions%20that%20was%20more)) ([
            Evidence for the criterion validity and clinical utility of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC3354008/#:~:text=fully%20elaborated%20by%20examining%20systematic,as%20measured%20by%20the%20PNI)).  

The **FFNI** being newer, has been subject to intensive validation by its authors and others. As intended, FFNI scores align with **Five-Factor Model** traits: for instance, FFNI grandiose narcissism correlates with very low Agreeableness (high Antagonism) and high facets of Extraversion (like assertiveness), while FFNI vulnerable narcissism correlates with very low Agreeableness plus high Neuroticism (e.g., anxiety, depressiveness) and low facets of Extraversion (like low warmth) ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=behaviors%20and%20symptoms%2C%20and%20romantic,and%20vulnerable%20dimensions%20of%20narcissism)). In studies, the FFNI grandiose dimension correlates strongly with other grandiose narcissism measures (such as the NPI and the Narcissistic Grandiosity Scale), and the FFNI vulnerable dimension correlates strongly with other vulnerability measures (such as the PNI vulnerability and the Hypersensitive Narcissism Scale) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=measures%20of%20grandiose%20and%20vulnerable,Hendin%20%26%20Cheek)). One study using expert ratings found that the **FFNI’s grandiose scale** was among the **best at capturing DSM-defined narcissistic personality** (grandiose presentation), outperforming even some clinical instruments, while the **FFNI’s vulnerable scale** matched expert profiles of vulnerable NPD nearly as well as the PNI ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=measures%20of%20grandiose%20and%20vulnerable,Hendin%20%26%20Cheek)). This indicates excellent criterion validity: the FFNI can mimic how experts rate narcissistic individuals on fundamental traits. Moreover, FFNI facets show predictable relations to external criteria: e.g., *Manipulativeness, Exploitativeness,* and *Entitlement* facets correlate with observed manipulative behaviors and low empathy; *Shame* facet correlates with self-reported internalizing distress; *Thrill-Seeking* correlates with impulsivity and risk-taking behavior, etc. The FFNI thus demonstrates a **comprehensive validity** profile – it correlates with interpersonal dysfunction (due to antagonism facets), with psychopathology (vulnerability facets linking to depression/anxiety), and with aggression/risk-taking (some grandiose facets). Its incremental validity has also been tested; FFNI can explain variance in narcissistic personality pathology above and beyond what the NPI captures, thanks to inclusion of vulnerability.  

**Clinical Utility:** In practice, the choice between these inventories can depend on the context. For **general population or social psychology research**, the NPI remains popular – it is brief and has a large research base linking it to outcomes like leadership, relationship conflict, and self-enhancement biases. However, for **clinical assessment or research on personality disorders**, the PNI and FFNI are often preferred because they assess the maladaptive range of narcissism. The PNI is particularly useful for identifying individuals who not only display arrogant, entitled behaviors but also suffer from feelings of inadequacy and emotional turmoil. The FFNI is useful when a researcher or clinician wants to understand a narcissistic individual’s broad personality trait profile (for instance, how much of their narcissism is driven by antagonism vs. neurotic insecurity, etc.). Both the PNI and FFNI have been used to explore narcissism’s connections to other disorders (like borderline personality disorder, depression, or aggression), and both show strong **construct validity** in those areas ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=interpretations%20%28see%20Pincus%2C%202013%20%29,the%20length%20of%20the%20PNI)) ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=behaviors%20and%20symptoms%2C%20and%20romantic,and%20vulnerable%20dimensions%20of%20narcissism)). The NPI, while not a diagnostic tool, has predictive utility in areas such as workplace behavior, social media activity, and generational trend studies (e.g., it famously has been used to argue about increasing narcissism in younger generations). Each instrument has been validated in its sphere, but their differing focus means results are not always interchangeable – for example, a person scoring high on NPI might not score high on PNI if they lack the vulnerable tendencies, and vice versa.  

In summary, the **PNI, NPI, and FFNI** each contribute a unique perspective on narcissism: 

- The **NPI** offers a window into the **grandiose self-promotional side** of narcissism often seen in everyday social behavior, grounded in trait psychology but with noted limitations in capturing pathology ([
            Development of a Brief Version of the Pathological Narcissism Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC4659764/#:~:text=amount%20of%20previous%20research%20has,2013)). 

- The **PNI** provides a **comprehensive clinical picture**, measuring the oscillation between egotistical grandiosity and wounded vulnerability that characterizes pathological narcissism ([Narcissism at the crossroads: phenotypic description of pathological narcissism across clinical theory, social/personality psychology, and psychiatric diagnosis - PubMed](https://pubmed.ncbi.nlm.nih.gov/18029072/#:~:text=consistently%20describe%20variations%20in%20the,the%20diagnostic%20emphasis%20on%20overt)) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=interpersonal%20distress%2C%20aggression%2C%20and%20borderline,several%20aspects%20of%20psychotherapy%20utilization)). 

- The **FFNI** frames narcissism in terms of **basic personality traits**, allowing integration with general personality assessment and highlighting how extreme antagonism underlies both narcissistic presentations ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=FFNI%20factors%20%28Miller%20et%20al,Seeking%2C%20Authoritativeness%2C%20Grandiose%20Fantasies%2C%20and)) ([A test of the construct validity of the Five-Factor Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/23186210/#:~:text=The%20Five,good%20convergent%20and%20criterion%20validity)). 

When comparing across them, researchers have found that **NPI and FFNI grandiosity scales align best with “classic” narcissistic grandiosity**, whereas **PNI (and FFNI) vulnerability scales are superior for assessing the covert, vulnerable narcissistic features** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=measures%20of%20grandiose%20and%20vulnerable,Hendin%20%26%20Cheek)). All three measures are well-supported by academic research, but their use should be tailored to the aspect of narcissism under investigation – whether it’s the outward bravado, the inward fragility, or the broad trait profile. Each has proven psychometrically sound within its scope, and together they paint a fuller picture of the complex narcissism construct.

The **Pathological Narcissism Inventory (PNI)** consists of two overarching dimensions (**Grandiose Narcissism** and **Vulnerable Narcissism**) comprising **seven** primary factors or facets:

### **Grandiose Narcissism Factors:**
1. **Exploitativeness**
   - Tendency to manipulate or exploit others for personal gain.

2. **Grandiose Fantasy**
   - Preoccupation with fantasies of unlimited success, power, brilliance, or ideal love.

3. **Self-Sacrificing Self-Enhancement**
   - Enhancing one's own self-esteem or image through acts of sacrifice or martyrdom, positioning oneself as uniquely noble or heroic.

### **Vulnerable Narcissism Factors:**
4. **Contingent Self-Esteem**
   - Self-esteem that fluctuates significantly, depending heavily on external validation and others' approval.

5. **Hiding the Self**
   - The inclination to conceal personal weaknesses or vulnerabilities due to fear of criticism, rejection, or exposure.

6. **Devaluing**
   - A defensive tendency to minimize or diminish the value of others when threatened or insecure.

7. **Entitlement Rage**
   - Intense anger or hostility experienced when perceived entitlements or expectations are unmet or challenged.

##  Steps for PNI Personality Modeling Workflow 

1.  **Step 1: Create the Personality Dataset**

    * **Purpose:** This initial step involves defining and generating the dataset that represents the Eysenck Personality Model (PNI).
    * **Actions:**
        * Define the PNI dataset structure (e.g., using a dictionary to represent factors, adjectives, synonyms, etc.).
        * Generate the dataset by organizing the data into a Pandas DataFrame.
        * Save the dataset to a CSV file.
        * Log dataset metadata (e.g., number of rows, factors, data schema) and the dataset file itself as an artifact in MLflow.
    * **Importance:** This step creates the raw data that will be used for embedding generation and model training.

2.  **Step 2: API Key Handling and Initialization**

    * **Purpose:** This critical step ensures that your OpenAI API key is securely loaded and the OpenAI client is initialized. This sets the foundation for using the OpenAI API in subsequent steps.
    * **Actions:**
        * Load the OpenAI API key from a secure location (e.g., a file in the user's home directory).
        * Validate the API key (e.g., check for existence, emptiness, and potentially a basic API call).
        * Initialize the OpenAI client (`client`).
        * Log the API key handling process and its outcome in MLflow.
    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

3.  **Step 3: Test Embedding API**

    * **Purpose:** This step verifies that the OpenAI Embedding API is accessible and functioning correctly.
    * **Actions:**
        * Use the initialized OpenAI client (`client`) to make a test call to the Embedding API (e.g., by embedding a sample text).
        * Check the API response for validity.
        * Log the API call details and the outcome (success or failure) in MLflow.
    * **Importance:** This step ensures that you can successfully generate embeddings before proceeding to the next step.

4.  **Step 4: Create Embeddings for the Dataset**

    * **Purpose:** This step generates numerical representations (embeddings) for the text data in the PNI dataset using the OpenAI Embedding API.
    * **Actions:**
        * Load the PNI dataset (created in Step 2).
        * Use the OpenAI client (`client`) to generate embeddings for the relevant text fields (e.g., combining factor, adjective, synonym, verb, noun).
        * Add the generated embeddings as a new column in the Pandas DataFrame.
        * Save the DataFrame with embeddings to a new CSV file.
        * Log embedding generation parameters (e.g., embedding model used), statistics (e.g., embedding length), and the embeddings file as an artifact in MLflow.
    * **Importance:** This step transforms the text data into a numerical format that can be used for machine learning models.

5.  **Step 5: Create and Visualize a Label Encoder**

    * **Purpose:** This step prepares the categorical labels (personality factors) for model training by encoding them into numerical values and provides a visualization of this encoding.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Initialize a `LabelEncoder` from scikit-learn.
        * Fit the `LabelEncoder` to the 'Factor' column to create the mapping between personality factors and numerical codes.
        * Transform the 'Factor' column using the fitted `LabelEncoder` to create a new 'Factor_Encoded' column.
        * Save the fitted `LabelEncoder` object.
        * Generate a visualization (e.g., a bar chart) to show the mapping between original factors and encoded values.
        * Save the visualization as an image file.
        * Log the label encoder object and the visualization as artifacts in MLflow.
        * Log the mapping between original factors and encoded values as a dictionary in MLflow.
    * **Importance:** This step prepares the target variable for model training and provides a clear representation of the encoding.

6.  **Step 6: Create our PNI Model (Model Training and Evaluation)**

    * **Purpose:** This step trains a machine learning model on the generated embeddings to predict personality factors and evaluates its performance.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Prepare the data for model training:
            * Extract the embeddings as features (`X`).
            * Encode the 'Factor' column using the loaded `LabelEncoder` to get the target variable (`y`).
            * Split the data into training and testing sets.
        * Initialize a machine learning model (e.g., `RandomForestClassifier`).
        * Train the model on the training data.
        * Make predictions on the test data.
        * Evaluate the model's performance using appropriate metrics (e.g., accuracy, classification report, confusion matrix).
        * Generate visualizations of the evaluation results (e.g., confusion matrix plot).
        * Save the trained model.
        * Log model training parameters (e.g., hyperparameters), evaluation metrics, visualizations, and the trained model as artifacts in MLflow.
    * **Importance:** This step is the core of the machine learning process, where the model learns to predict personality factors from the embeddings.

7.  **Step 7: Model Testing (Inference on New Data)**

    * **Purpose:** This step demonstrates how to use the trained model to predict personality factors for new, unseen text inputs.
    * **Actions:**
        * Load the trained model (saved in Step 6).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Define a function that:
            * Takes new text as input.
            * Generates an embedding for the new text using the OpenAI API.
            * Uses the loaded model to predict the personality factor.
            * Uses the loaded `LabelEncoder` to decode the numerical prediction back to the original factor name.
        * Provide example new text inputs.
        * Use the function to predict personality factors for the example texts.
        * Print the predictions.
        * Log the test inputs and predictions in MLflow.
    * **Importance:** This step demonstrates the practical application of the trained model for making predictions on new data.

8.  **Step 8: Model Application, Visualization, and Analysis**

    * **Purpose:** This step provides additional visualization and analysis of the data and model.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Apply PCA for dimensionality reduction and visualization of the embeddings.
        * Generate and log PCA plots to visualize the embedding distribution.
        * Perform K-Means clustering on the embeddings to identify potential groupings or clusters of similar personality traits.
        * Add cluster labels to the dataset and save the clustered data.
        * Log clustering parameters (e.g., number of clusters) and the clustered data as artifacts in MLflow.
    * **Importance:** This step offers valuable insights into the data and model:
        * PCA visualization helps understand the distribution of embeddings.
        * Clustering can reveal underlying patterns in the data.

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key Status", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"✅ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key Status", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key Status", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("🔑 API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("🤖 OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("✅ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"⚠️ Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"❌ Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"❌ Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"⚠️ Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved pni dataset from a CSV file
dataset_path = '../Datasets/pni.csv'
try:
    pni_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'pni_df' is empty before processing
        if pni_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", pni_df.shape[0])
            print(f"📊 Processing {pni_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        pni_df['Embedding'] = pni_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/pni_embeddings.csv'
        pni_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = pni_df.shape[0]
        num_columns = pni_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in pni_df.columns and not pni_df['Embedding'].empty:
            embedding_length = len(pni_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(pni_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/pni_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/pni_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "pni_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

# **Step 6: Create Model** <a class="anchor" id="pni_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/pni_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/pni_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"✅ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"❌ Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="pni Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "pni_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"📊 Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "pni_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"📊 Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "pni_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"🖼️ Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/pni_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"💾 Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"✅ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 8: Test and Evaluate Model** <a class="anchor" id="pni_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/pni_label_encoder.pkl"
model_path = "../Models/pni_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"✅ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"❌ Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"❌ An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"❌ Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Model testing (inference on new data) complete.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/pni_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="PNI Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/pni_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "pni_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/pni_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of PNI Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/pni_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of PNI Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create PNI Schema in Neo4j** <a class="anchor" id="PNI_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("PNI")

# Load the dataset
dataset_path = "../Embeddings/pni_embeddings.csv"
pni_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in pni_df.columns:
    pni_df['Embedding'] = pni_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the PNI schema in Neo4j
def create_pni_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with PNI labels
        graph.run("MERGE (f:PNI_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:PNI_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:PNI_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:PNI_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:PNI_Noun {name: $noun})", noun=noun)
        
        # Create relationships with PNI namespacing
        graph.run("""
            MATCH (f:PNI_Factor {name: $factor}), (a:PNI_Adjective {name: $adjective})
            MERGE (f)-[:PNI_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:PNI_Adjective {name: $adjective}), (s:PNI_Synonym {name: $synonym})
            MERGE (a)-[:PNI_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:PNI_Adjective {name: $adjective}), (v:PNI_Verb {name: $verb})
            MERGE (a)-[:PNI_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:PNI_Adjective {name: $adjective}), (n:PNI_Noun {name: $noun})
            MERGE (a)-[:PNI_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:PNI_Factor {name: $factor}), (a:PNI_Adjective {name: $adjective})
                MERGE (e:PNI_Embedding {value: $embedding})
                MERGE (f)-[:PNI_HAS_EMBEDDING]->(e)
                MERGE (a)-[:PNI_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create PNI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create PNI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the PNI schema
        create_pni_schema(graph, pni_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", pni_df.shape[0])
        mlflow.log_param("Number of Columns", pni_df.shape[1])
        
        # Print the DataFrame
        print(pni_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
