## Narcissistic Admiration and Rivalry Questionnaire (NARQ)  <a class="anchor" id="PTMD_page_7"></a>

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">


A detailed history of the Narcissistic Admiration and Rivalry Questionnaire (NARQ), including its structure, development, and key academic and psychological research papers. Compares NARQ with other prominent narcissism models such as FFNI, PNI, and NPI, highlighting similarities, differences, and unique contributions.



# Development and Timeline of the NARQ  
The **Narcissistic Admiration and Rivalry Questionnaire (NARQ)** was introduced by Mitja D. Back and colleagues in 2013 as part of the Narcissistic Admiration and Rivalry Concept (NARC) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=introduced%20models%20of%20grandiose%20narcissism%2C,as%20a%20mix%20of%20both)). This model arose to capture the **“bright” (agentic) and “dark” (antagonistic)** sides of grandiose narcissism within a single framework ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=introduced%20models%20of%20grandiose%20narcissism%2C,as%20a%20mix%20of%20both)). Back et al. (2013) developed the NARQ as a two-dimensional measure corresponding to these two pathways: **Narcissistic Admiration** (self-promoting, assertive strategies) and **Narcissistic Rivalry** (self-protective, antagonistic strategies) ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=Abstract)). The initial questionnaire (in both German and English) was validated across **seven studies**, demonstrating a clear two-factor structure, good reliability, and distinct behavioral outcomes for each dimension ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=of%207%20studies%2C%20we%20validated,esteem)) ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). Notably, NARQ’s two scales were positively correlated yet showed markedly different “nomological networks” (i.e. unique patterns of traits and outcomes) ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=of%207%20studies%2C%20we%20validated,esteem)). For example, Admiration and Rivalry predicted interpersonal reactions and relationship outcomes in opposite ways, highlighting the importance of separating these facets of narcissism ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). Furthermore, the NARQ’s predictive validity was strong – across studies it **outperformed the Narcissistic Personality Inventory (NPI)** in predicting outcomes like social status and relationship dynamics ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). 

After its introduction, the NARQ gained traction in research. Key timeline milestones include the development of a **short form (NARQ-S)** to facilitate use in large surveys and time-constrained settings. Back and colleagues introduced the NARQ-S with 6 items (3 per dimension) while maintaining the two-factor structure ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=,strive%20for%20supremacy%2C%20and%20aggressiveness)). The short scale was validated as reliable and was incorporated into longitudinal panel studies (e.g. the German SOEP in 2018) ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=The%20Narcissistic%20Admiration%20and%20Rivalry,year%20intervals)) ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=,strive%20for%20supremacy%2C%20and%20aggressiveness)). The NARQ has also been adapted internationally: by 2016–2020 it had been translated and validated in **multiple languages** (Polish, Italian, Spanish, etc.) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=increasingly%20being%20applied%20in%20different,being%20measured%20and%20that%20different)), enabling cross-cultural research. Recent refinements include domain-specific extensions – for instance, a **Domain-Specific NARQ (D-NARQ)** was developed to assess admiration and rivalry in specific life domains (intellectual, social, communal, physical), with 72 items capturing narcissistic processes in each domain ([A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PubMed](https://pubmed.ncbi.nlm.nih.gov/34085540/#:~:text=Research%20on%20grandiose%20narcissism%20distinguishes,We%20investigated%20the%20psychometric%20properties)). These developments illustrate how the NARQ has been refined and expanded since its 2013 origin, while its core concept (differentiating narcissistic self-enhancement vs. self-defense) remains central.  

# Academic Foundations and Validation  
**Purpose and Theoretical Basis:** The NARQ was designed to address limitations in existing narcissism measures by explicitly measuring two complementary processes through which narcissists maintain a grandiose self-image ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=We%20present%20a%20process%20model,dimensional)). Narcissistic **Admiration** reflects an *agentic self-enhancement* strategy – pursuing social admiration, attention, and status to reinforce a grandiose self-view. In contrast, Narcissistic **Rivalry** reflects an *antagonistic self-protection* strategy – reacting to ego threats by devaluing others and asserting superiority to defend the self ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=We%20present%20a%20process%20model,dimensional)) ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=arrogance,to%206)). This two-dimensional approach was grounded in prior theory suggesting narcissism has both an **agentic (socially bold, dominant)** side and an **antagonistic (entitled, aggressive)** side ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=introduced%20models%20of%20grandiose%20narcissism%2C,as%20a%20mix%20of%20both)). By capturing both, the NARQ’s purpose is to disentangle the “bright” facets of narcissism (e.g. charisma, confidence) from the “dark” facets (e.g. hostility, entitlement) within the grandiose narcissism spectrum ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=follows%20from%20literature%20characterizing%20grandiose,with%20positive%20consequences%20such%20as)) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=positive%20self,Campbell%20%26%20Campbell%2C%202009)). 

**Methodology:** The NARQ is a **self-report questionnaire** originally consisting of 18 Likert-scale items (rated 1–6 from “not at all” to “agree completely”) ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=)). There are **two primary scales** – Admiration and Rivalry – each measured by 9 items in the full version ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=)). Item content was developed based on the NARC theory to cover the **affective-motivational, cognitive, and behavioral** aspects of each narcissism dimension ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=exclusively%20on%20grandiose%20narcissism%20and,While%20creating%20and)). For example, items were written to reflect grandiose thinking, striving motives, and interpersonal behaviors associated with each process. Back et al. (2013) conducted extensive validation, including exploratory and confirmatory factor analyses, to confirm that a two-factor model (with three sub-facets per factor) fit the data well ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=of%207%20studies%2C%20we%20validated,showed%20unique%20relations%20to%20the)). The initial paper reported that the NARQ showed the expected **two-dimensional, multifaceted structure** with each subscale loading appropriately on Admiration vs. Rivalry ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=of%207%20studies%2C%20we%20validated,showed%20unique%20relations%20to%20the)). 

**Validation Evidence:** Across validation studies, the NARQ has demonstrated strong **psychometric properties**. Both Admiration and Rivalry scales showed high internal consistencies (reliability alphas in the .80s) and test–retest stability over time ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=they%20have%20markedly%20different%20nomological,admiration%20and%20rivalry%2C%20the%202)). The measure also showed good self–other agreement: in one study, informant ratings of a person’s narcissistic admiration and rivalry correlated significantly with that person’s self-reports, indicating these traits are observable by others ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=they%20have%20markedly%20different%20nomological,admiration%20and%20rivalry%2C%20the%202)). Importantly, the NARQ has shown rich patterns of **convergent and discriminant validity**. The two NARQ dimensions correlate in expected ways with other narcissism measures and personality constructs ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=96%29,For%20all)): Admiration relates positively to traditional grandiose narcissism measures and traits like extraversion and self-esteem, whereas Rivalry correlates with antagonism (low agreeableness), envy, and potentially some vulnerability indicators (e.g. fragile self-esteem) ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=96%29,For%20all)). Back et al. (2013) found each NARQ dimension had **unique relations** with external criteria that older unidimensional measures could not distinguish ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=96%29,For%20all)). For instance, NARQ Admiration showed ties to **social boldness**, assertive self-enhancement, and lesser neuroticism (indicative of genuine high self-confidence), while Rivalry was uniquely linked to **hostility**, interpersonal distrust, and defensive anger ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=96%29,For%20all)). 

Crucially, the two dimensions predict different **outcomes and behaviors**. Admiration, the agentic facet, is associated with short-term interpersonal advantages – e.g. higher momentary self-esteem and initial popularity upon meeting new people ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=outcomes%20of%20narcissism%2C%20the%20agentic,whereas%20the%20antagonistic%20aspects%20are)). Rivalry, the antagonistic facet, predicts negative consequences over time – e.g. increasing peer dislike, relationship conflicts, and aggressive responses when challenged ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Nestler%2C%20%26%20Back%2C%202015%3B%20being,2016%3B%20see%20also)). These differences were evident in experimental and longitudinal studies: Admiration led to being perceived as confident and charming in group interactions, whereas Rivalry led to being seen as arrogant or aggressive, undermining social relationships ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). Notably, by modeling these processes separately, the NARQ was able to **predict outcomes more powerfully than the one-dimensional NPI** in many cases ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). For example, in Back et al.’s work, combining Admiration and Rivalry scores explained behavior in friendships, romantic relationships, and group settings better than the total NPI score ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). This underscores the NARQ’s value in research: it provides a more nuanced assessment of grandiose narcissism that aligns with theoretical distinctions and improves predictive accuracy in psychological and social outcomes. Reputable psychology assessments (e.g. in **Psychological Assessment** journal) have since confirmed that the NARQ and its short form are **reliable and valid** for both **convenience and representative samples** ([Validation of the Narcissistic Admiration and Rivalry Questionnaire ...](https://www.researchgate.net/publication/310801881_Validation_of_the_Narcissistic_Admiration_and_Rivalry_Questionnaire_Short_Scale_NARQ-S_in_Convenience_and_Representative_Samples#:~:text=Validation%20of%20the%20Narcissistic%20Admiration,antagonistic%20aspects%20of%20grandiose%20narcissism)). In sum, a robust body of academic work supports the NARQ’s purpose (differentiating narcissistic strategies) and its methodology/validity as a sound measure of those constructs ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=they%20have%20markedly%20different%20nomological,admiration%20and%20rivalry%2C%20the%202)) ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). 

# Structure and Components of the NARQ  
**Overall Structure:** The NARQ specifically measures **grandiose narcissism** in a bifurcated way, consisting of two higher-order factors: **Narcissistic Admiration** and **Narcissistic Rivalry** ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=We%20present%20a%20process%20model,dimensional)). These two dimensions are positively correlated (as both stem from a common narcissistic core of grandiose self-importance) but are distinguishable in content and outcomes ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=of%207%20studies%2C%20we%20validated,esteem)). Each dimension is further **multi-faceted**, with three theorized subcomponents that capture different psychological processes ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=NARC%20states%20that%20each%20of,facets%20of%20each%20of%20the)). In the full NARQ, each subscale is assessed by multiple items (the original measure has 18 items total: 9 Admiration, 9 Rivalry) ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=)). The short version (NARQ-S) uses one representative item per subfacet, retaining the same structure ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=,strive%20for%20supremacy%2C%20and%20aggressiveness)). The table below outlines the NARQ’s key components and their implications:

- **Narcissistic Admiration (Agentic Dimension):** An **assertive self-enhancement strategy** aimed at gaining admiration and bolstering the ego by showcasing one’s greatness ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=We%20present%20a%20process%20model,and)). This dimension reflects the “bright” side of narcissism – charm, confidence, and ambition for status. Its subscales are:  
  - **Grandiosity (Grandiose Fantasies)** – a *cognitive* facet involving inflated self-perceptions and fantasies of unlimited success, beauty, or power ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=NARC%20states%20that%20each%20of,facets%20of%20each%20of%20the)). High grandiosity means the individual truly believes in their exceptional status (“I am extraordinary”) and engages in positive self-thoughts. Psychologically, this bolsters self-esteem and can make the person appear confident or even inspiring to others.  
  - **Striving for Uniqueness** – an *affective–motivational* facet reflecting the desire to be distinct and celebrated ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=NARC%20states%20that%20each%20of,facets%20of%20each%20of%20the)). It captures a narcissist’s strong need to stand out as *special* and to be recognized for unique talents or qualities. This drive for uniqueness motivates achievement-oriented behaviors and exaggeration of one’s accomplishments to attract praise. Emotionally, it is tied to pride in oneself and intolerance of being “average.”  
  - **Charmingness** – a *behavioral* facet involving the use of charisma, charm, and socially confident behaviors to attract and captivate others ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=NARC%20states%20that%20each%20of,facets%20of%20each%20of%20the)). It represents how narcissistic admiration is enacted interpersonally. Those high in this facet are often witty, entertaining, or flirtatious as a means to win admiration. The psychological implication is that charm is used as a tool to affirm one’s self-image through others’ positive reactions.  

  **Implications of Admiration:** Narcissistic Admiration tends to yield **short-term interpersonal benefits**. Because it involves self-assured and socially bold behavior, it can lead to positive first impressions. Research has shown that the agentic aspects of narcissism (like those in Admiration) correlate with **higher state self-esteem and initial popularity** in new social contexts ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=outcomes%20of%20narcissism%2C%20the%20agentic,whereas%20the%20antagonistic%20aspects%20are)). Individuals high in Admiration may be seen as leaders or attractive at first glance (e.g., they often emerge as group leaders and can charm potential romantic partners) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=outcomes%20of%20narcissism%2C%20the%20agentic,Oltmanns%2C%20Friedman%2C%20Fiedler%2C%20%26%20Turkheimer)). However, this comes purely from the *grandiose* side of narcissism and does not account for any self-doubt – Admiration operates as genuine (if inflated) self-confidence.

- **Narcissistic Rivalry (Antagonistic Dimension):** A **self-defense and dominance strategy** aimed at protecting the ego and asserting superiority by belittling others and reacting aggressively to threats ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=We%20present%20a%20process%20model,and)). This dimension represents the “dark” side of narcissism – arrogance, defensiveness, and hostility. Its subscales are:  
  - **Devaluation of Others** – a *cognitive* facet involving negative appraisals of other people ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=facets,behavioral%20facet%3B%20for%20more)). Here, the narcissist maintains superiority in their mind by viewing others as inferior, flawed, or unworthy. This might manifest as constant criticism of others or dismissing others’ successes. Psychologically, devaluation protects the narcissist from feeling inadequate by comparison, but damages empathy and respect in relationships.  
  - **Striving for Supremacy** – an *affective–motivational* facet reflecting a strong drive to be better than others at all costs ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=facets,behavioral%20facet%3B%20for%20more)). It denotes competitive dominance: the person is emotionally invested in *winning*, being seen as the best, and cannot tolerate others outshining them. This facet is fueled by feelings of envy and insecurity – the narcissist feels good only when on top, leading to hostile competitiveness.  
  - **Aggressiveness** – a *behavioral* facet involving overt or covert aggression toward others, especially when the narcissist’s ego is threatened ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=cognitive%20facet,behavioral%20facet%3B%20for%20more)). This includes retaliatory behaviors, confrontation, and even social aggression (insults, exclusion) to defend one’s status. Essentially, when challenged or criticized, high Rivalry individuals lash out to restore dominance. This facet captures the interpersonal cost of narcissism’s dark side, as it often results in conflict and intimidation.  

  **Implications of Rivalry:** Narcissistic Rivalry is associated with **negative interpersonal consequences, especially long-term**. While it may help the narcissist fend off immediate threats to their ego, over time it erodes social bonds. Studies show antagonistic aspects of narcissism predict outcomes like **fragile self-esteem, aggressive retaliation, and relationship problems** ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Nestler%2C%20%26%20Back%2C%202015%3B%20being,2016%3B%20see%20also)). For example, individuals high in Rivalry might initially gain a position of power, but peers’ opinions of them grow increasingly negative with prolonged exposure ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=2004%3B%20dating%20success%2C%20Wurst%20et,also%20Campbell%20%26%20Campbell%2C%202009)). Rivalry is linked to higher likelihood of conflicts, such as friendship ruptures and romantic relationship dissatisfaction (even higher divorce rates in the long run) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=2004%3B%20dating%20success%2C%20Wurst%20et,also%20Campbell%20%26%20Campbell%2C%202009)). In essence, Rivalry entails a self-defeating pattern: it protects the narcissist’s self-view in the moment but alienates others, leading to social isolation or reputational damage over time. It represents the core of narcissistic antagonism – entitlement and lack of empathy – which is central to the interpersonal **“dark side”** of narcissism. 

In summary, the NARQ’s structure of Admiration vs. Rivalry (each with distinct subcomponents) provides a nuanced look at grandiose narcissism. Admiration captures how narcissists **shine** – through grandiose thinking, uniqueness, and charm – explaining their charismatic leadership and initial success. Rivalry captures how narcissists **fight** – through derogation, supremacy seeking, and aggression – explaining the toxicity and downfall that often accompany narcissistic personalities. This bifurcated structure has important psychological implications: it helps explain the paradox of narcissism (socially seductive yet ultimately corrosive) by assigning those outcomes to different facets. Researchers recommend examining **both** NARQ dimensions rather than a single “narcissism score” to fully understand a narcissistic individual’s likely behaviors and self-regulatory style ([Narcissistic Admiration and Rivalry Questionnaire Short Scale (NARQ-S) — SOEPcompanion 1.0.0 documentation](http://companion.soep.de/Survey%20Design/Narcissistic%20Admiration%20and%20Rivalry%20Questionnaire%20Short%20Scale%20%28NARQ-S%29.html#:~:text=cognitive%2C%20and%20behavioral%20processes,effects%20of)). 

# Comparison with Other Narcissism Models  

Multiple instruments exist to measure narcissism. The NARQ (Admiration–Rivalry) introduced a **two-process approach** within grandiose narcissism, which can be contrasted with other prominent narcissism models: the Narcissistic Personality Inventory (NPI), the Pathological Narcissism Inventory (PNI), and the Five-Factor Narcissism Inventory (FFNI). Below is an overview of how NARQ compares to these measures, including similarities, differences, and the unique insights contributed by the NARQ.

**NARQ vs. NPI (Narcissistic Personality Inventory):** The NPI is the most classic narcissism measure, originally developed by Raskin & Hall (1979) and updated by Raskin & Terry (1988) to assess traits associated with Narcissistic Personality Disorder criteria ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=The%20NPI%20was%20developed%20to,criteria%20of%20narcissistic%20personality%20disorder)). It consists of 40 forced-choice items (pairs of statements where one is the narcissistic response) and primarily captures **grandiose narcissism**. The NPI has several subscale facets – commonly a three-factor structure is reported: **Leadership/Authority**, **Grandiose Exhibitionism**, and **Entitlement/Exploitativeness** ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=)). In practice, however, the NPI is often treated as unidimensional or split into a **leadership/exhibitionism** component (agentic) and an **entitlement** component (antagonistic). This highlights a key issue: the NPI’s content is *skewed toward agentic traits* (e.g. confidence, assertiveness, vanity) with only a small portion assessing antagonism (the Entitlement/Exploitativeness facet) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Narcissistic%20Personality%20Inventory%20,of%20scores%20on%20some%20of)). In fact, researchers have criticized the NPI for its **ambiguous factor structure and poor coverage of maladaptive aspects** of narcissism ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Narcissistic%20Personality%20Inventory%20,of%20scores%20on%20some%20of)). For example, the NPI’s one explicit antagonistic facet (entitlement) often shows low reliability (alpha ~0.45) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Narcissistic%20Personality%20Inventory%20,of%20scores%20on%20some%20of)), and many “narcissistic” NPI items reflect relatively adaptive or neutral traits (leadership, high self-confidence) rather than pathology. 

- **Similarities:** Both NARQ and NPI aim to measure **grandiose narcissism** (neither is designed to capture the “vulnerable” narcissism phenotype of shame and hypersensitivity). Scores on NARQ Admiration correlate strongly with NPI scores since both reflect the grandiose self-promotional aspect. In fact, Admiration aligns conceptually with NPI’s Leadership/Exhibitionism facets (agentic narcissism), while Rivalry corresponds to the NPI’s Entitlement/Exploitativeness facet (antagonistic narcissism) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=admiration%20and%20rivalry%20with%20NPI,rivalry%20showed%20moderate%20to%20strong)). Thus, NARQ covers much of the same trait space as the NPI, but in a more structured way. Both measures have been widely used in non-clinical populations and show positive relations with extraversion and self-esteem (for their grandiose components). 

- **Differences:** The NARQ provides a **clear separation** between narcissistic self-enhancement and aggression, whereas the NPI mixes various traits into one index. NARQ’s two distinct scales allow researchers to **differentiate the “socially bold” vs. “toxic” elements** of narcissism easily, something the NPI struggles with. For example, a high NPI score could mean a person is predominantly self-confident, or it could mean they are highly entitled – NARQ would disentangle this by giving separate Admiration and Rivalry scores. Additionally, the NPI’s forced-choice format and heterogeneous content have been cited for psychometric weaknesses; by contrast, NARQ uses straightforward Likert ratings and was built from the ground up to map onto a coherent narcissism process model ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=The%20most%20frequently%20used%20measure,and%20unreliability%20of%20scores)). Another difference is predictive validity: Back et al. (2013) found that dividing narcissism into Admiration and Rivalry (NARQ) enhanced prediction of outcomes compared to the one-dimensional NPI ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). For instance, NARQ Rivalry specifically predicted aggressive responses to provocation and friendship deterioration, which the NPI total score did not catch, while NARQ Admiration predicted leadership emergence beyond what NPI could ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)). Thus, NARQ offers **greater nuance and accuracy**. In sum, NARQ’s innovation over the NPI is the **explicit modeling of narcissism’s dual nature** – it isolates the entitlement/hostility component that the NPI lumps imperfectly with grandiosity. This provides unique insight: researchers can examine how the *same* narcissistic individual might simultaneously charm others yet behave callously, by looking at their NARQ profile (high Admiration + high Rivalry), which the single NPI score would obscure.

**NARQ vs. PNI (Pathological Narcissism Inventory):** The PNI (Pincus et al., 2009) was developed to assess narcissism as understood in clinical contexts – capturing both **grandiose** and **vulnerable** expressions of narcissistic pathology. It contains 52 Likert-scale items and yields seven first-order factors that cluster into two higher-order dimensions: **Narcissistic Grandiosity** and **Narcissistic Vulnerability** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Pathological%20Narcissism%20Inventory%20%28PNI%29,92%29%20dimensions%20in%20the)). The facets of the PNI cover a broad range of narcissistic tendencies, including some that overlap with NARQ’s content and others that reflect internalized distress: for example, **Exploitativeness**, **Grandiose Fantasy**, and **Self-Sacrificing Self-Enhancement** tap the grandiose, outward-reaching side of narcissism (manipulating others, fantasizing about greatness, showing off altruism for praise), while **Contingent Self-Esteem**, **Hiding the Self**, **Entitlement Rage**, and **Devaluation** tap the vulnerable, defensive side (sensitivity to criticism, feelings of shame or inadequacy, anger when entitlement needs aren’t met, and distrustful devaluation of others) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=vouchers%20worth%2050%E2%82%AC,381%2C%20for%20exploitativeness)) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=esteem%3B%20t%2887%29%20%3D%201,for%20entitlement%2C%20or%20their%20mean)). In short, the PNI paints narcissism as a combination of **external grandiosity** and **internal vulnerability** – where a person may display arrogance and entitlement but underneath struggle with insecurity and emotional dysregulation. 

- **Differences:** The key difference is scope. The NARQ focuses **exclusively on grandiose narcissism** (agentic vs antagonistic social behaviors) and *does not measure* the vulnerable aspects of narcissism ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=Like%20the%20NARQ%20,For%20narcissistic%20rivalry%2C%20these%20facets)). In fact, the NARQ was intentionally limited to “domain-unspecific grandiose narcissism” in its design ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=between%20self,grandiose%20narcissism%20items%20capture%20in)). By contrast, the PNI was designed to capture **pathological narcissism in its entirety**, encompassing the classic grandiosity (which corresponds to NARQ Admiration and parts of Rivalry) *and* the often-overlooked vulnerability (e.g. feelings of inadequacy, depression, interpersonal fear). Therefore, PNI includes many internally focused items (e.g. “I get angry when criticized”) that have no parallel in NARQ. Another difference is complexity: the PNI’s seven facets provide a detailed profile (which can be cumbersome for research), whereas NARQ’s structure is more parsimonious, with two broad factors that are easier to interpret. Also, PNI is considered a **clinical tool** – it was validated on both community and clinical samples, and its “Pathological” label implies use in identifying narcissistic personality disorder features. NARQ, on the other hand, is geared toward **normal-range individual differences** in narcissistic traits, not for diagnosis ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=Like%20the%20NARQ%20,For%20narcissistic%20rivalry%2C%20these%20facets)). 

- **Similarities:** There is overlap in the grandiose content. PNI Grandiosity shares features with Narcissistic Admiration (e.g. grandiose fantasies, exploitative tendencies for self-enhancement) and PNI contains antagonistic facets (Entitlement Rage, Devaluation) that correspond to Narcissistic Rivalry’s core of reacting aggressively and derogating others. Empirically, NARQ Rivalry correlates with PNI Entitlement and exploitativeness, while NARQ Admiration correlates with PNI’s grandiose fantasies and self-enhancement facets ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=96%29,For%20all)). Both measures acknowledge a **duality in narcissism** – but in PNI it is framed as *grandiose vs. vulnerable*, whereas in NARQ it is *assertive vs. antagonistic* (with both being aspects of grandiose narcissism). A unique insight from NARQ in comparison is that even *within* the grandiose form of narcissism (the part that is outwardly confident), there is an internal split between prosocial charm and aggressive defense. The PNI tends to combine all aggression with vulnerability (e.g. “Entitlement rage” in PNI is grouped as a vulnerability symptom), while NARQ situates aggression (Rivalry) as part of an alternate grandiose strategy. This highlights how NARQ’s lens views narcissistic anger as an interpersonal power move rather than a purely neurotic reaction. Researchers have noted that **NARQ Rivalry is associated not only with low agreeableness but also with some neurotic traits (e.g. hostile reactive emotions)**, bridging a gap between the PNI’s vulnerability and grandiosity constructs ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=vulnerable%20dimensions%20from%20the%20FFNI,anxiety%2C%20depression%2C%20borderline%20per%02sonality)). In summary, NARQ contributes a streamlined, **process-oriented view of grandiose narcissism** to complement the PNI’s broader trait coverage. In practice, NARQ is often used when the focus is on **social behaviors and outcomes** of narcissism, whereas PNI is used to understand the **clinical psychology** of narcissism, including internal suffering.

**NARQ vs. FFNI (Five-Factor Narcissism Inventory):** The FFNI (Glover et al., 2012) is a newer assessment that approaches narcissism from the perspective of general personality traits. It contains 148 items measuring 15 facet-level traits that are theoretically narcissistic expressions of the Big Five personality dimensions ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)). Each FFNI facet corresponds to an extreme or maladaptive variant of a normal trait. For example, facets include **Reactive Anger**, **Shame**, **Indifference**, **Need for Admiration**, **Exhibitionism**, **Authoritativeness**, etc., which map onto higher-order dimensions roughly aligning with narcissistic **grandiosity** (e.g. assertiveness, exhibitionism, lack of empathy) and **vulnerability** (e.g. neurotic shame, anger) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,95%29%20and%20vulnerable)). The FFNI can be scored to yield an overall narcissism score, as well as broad **Grandiose** and **Vulnerable narcissism** composites and the 15 specific facet scores ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,95%29%20and%20vulnerable)). Essentially, it attempts to integrate various elements of narcissism into the well-established Five-Factor Model (FFM) of personality.

- **Similarities:** Both NARQ and FFNI recognize **narcissism as a multi-faceted construct** with distinct subcomponents. In particular, FFNI includes facets tapping antagonistic interpersonal style (e.g. Manipulativeness, Distrust, Arrogance) and ones tapping agentic extraversion (e.g. Exhibitionism, Authoritativeness), which parallel NARQ’s Rivalry and Admiration content. For instance, an FFNI facet like *Entitlement* or *Lack of Empathy* is conceptually related to Narcissistic Rivalry’s other-derogation, while *Exhibitionism* and *Grandiose Fantasies* are related to Narcissistic Admiration’s self-enhancement. Both measures therefore differentiate the *assertive* and *antagonistic* tendencies within narcissism, rather than treating it as one undifferentiated trait. Moreover, like NARQ, the FFNI is intended for *normal to subclinical populations* (not just diagnosed NPD), and both have been used in research on outcomes like aggression or leadership to see which specific narcissistic facets drive those behaviors ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=vulnerable%20dimensions%20from%20the%20FFNI,anxiety%2C%20depression%2C%20borderline%20per%02sonality)).

- **Differences:** The major difference is **breadth vs. focus**. The FFNI is a comprehensive, granular inventory – it provides a **facet-level profile** across 15 traits, which gives very detailed information but at the cost of length and complexity ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)). NARQ, in contrast, is succinct and **process-focused**. It homes in on two central ingredients of grandiose narcissism (agentic admiration and antagonistic rivalry) and measures those in a streamlined way. So while FFNI might tell you a person scores high on, say, “Exhibitionism” and “Manipulativeness” but low on “Shame,” the NARQ will more simply tell you they have high Admiration and perhaps moderate Rivalry. The FFNI also explicitly covers **vulnerable narcissistic traits** (like Shame, Need for Acceptance, etc.) as separate facets, whereas NARQ leaves out vulnerability altogether ([
            A Process × Domain Assessment of Narcissism: The Domain-Specific Narcissistic Admiration and Rivalry Questionnaire - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC9434250/#:~:text=narcissistic%20personality%20disorder,behavioral%20facet%3B%20for%20more)). As a result, FFNI’s structure aligns with a three-fold model (agentic, antagonistic, neurotic aspects) if one groups its facets, whereas NARQ strictly provides a two-fold model (agentic vs antagonistic) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=Pathological%20Narcissism%20Inventory%20%28B,invariance%20and%20mean%20differences%20for)) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=NPI%20,Based%20on%20prior)). Another difference is theoretical framing: NARQ is framed in terms of **interpersonal strategy** – it emphasizes how narcissistic people act in social contexts to manage their self-image (either by lifting themselves up or putting others down). FFNI is framed in terms of **basic personality traits** – it shows how narcissism is essentially a configuration of extreme trait scores (e.g. extremely low Agreeableness + high Extraversion + high Neuroticism in certain facets). Thus, the NARQ offers the unique insight of narcissism as a **self-regulatory process** (something the FFNI itself does not highlight, since FFNI treats narcissism as a static trait profile). Researchers have noted that *“in comparison to other existing measures of grandiose narcissism, the NARQ is specifically interested in the interpersonal processes”* that underlie narcissistic behavior ([The Unified Narcissism Scale-Revised: Testing Incremental Validity ...](https://www.tandfonline.com/doi/full/10.1080/00223891.2024.2346768#:~:text=,processes%20of%20grandiose%20narcissism)). This means NARQ is particularly useful when studying **social dynamics** (like conflict, admiration, leadership, manipulation in groups), whereas the FFNI’s richness is useful for linking narcissism to the broader personality taxonomy or clinical profiles. Additionally, practicality sets them apart: the full FFNI (148 items) is much longer than NARQ (18 items), so NARQ (especially NARQ-S with 6 items) is favored in large-scale surveys or multi-measure studies where brevity is important ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=ambulatory%20assessment%20in%20general%29,item%20version%20of%20the)) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=The%20most%20frequently%20used%20measure,and%20unreliability%20of%20scores)).

**Unique Insights of NARQ:** By comparing these models, we can see what the NARQ adds to narcissism research. The NARQ introduced a **dynamic, two-dimensional perspective** that clarified the internal structure of grandiose narcissism. Unlike the NPI, which left the “dark side” underrepresented, NARQ dedicates an entire scale to antagonistic tendencies, allowing researchers to study entitlement and aggression in narcissists with reliable measurement ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=Narcissistic%20Personality%20Inventory%20,of%20scores%20on%20some%20of)). Compared to the PNI and FFNI, which include vulnerability, NARQ’s narrower focus might seem a limitation, but it is actually a strength when specifically studying **grandiose narcissism in social contexts**. It cleanly separates how narcissists **strive for admiration** versus how they **react to threats**, providing insight into the push-pull of narcissistic behavior (e.g. why a person can be charming in one moment and vindictive the next). This has led to findings about the conditional nature of narcissists’ success: for instance, high Admiration can make narcissists effective leaders **until** Rivalry-driven behaviors alienate their peers ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=outcomes%20of%20narcissism%2C%20the%20agentic,whereas%20the%20antagonistic%20aspects%20are)) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=2004%3B%20dating%20success%2C%20Wurst%20et,also%20Campbell%20%26%20Campbell%2C%202009)). The NARQ framework has also spurred more nuanced hypotheses in research – such as exploring which aspect (Admiration or Rivalry) relates to creativity, or to ethical failures, etc., rather than asking only if “narcissism” as a whole does. 

In essence, NARQ carved out a **middle ground** in narcissism assessment: more differentiated than the one-factor NPI, but more concise and process-oriented than the PNI or FFNI. Its twofold model echoes longstanding theoretical distinctions (e.g. narcissistic grandiosity vs. vulnerability, narcissistic charm vs. nastiness) but packages them in an elegant way that directly ties to observable social behavior. This has enriched narcissism research by ensuring that we **examine both sides of narcissists** – the alluring “admiration-seeker” and the combative “rival” – rather than treating narcissism as a monolith. The result is a more complete understanding of how narcissistic personalities function and impact those around them ([Narcissistic admiration and rivalry: disentangling the bright and dark sides of narcissism - PubMed](https://pubmed.ncbi.nlm.nih.gov/24128186/#:~:text=differentially%20predicted%20general%20interpersonal%20orientations,conceptualization%20and%20measurement%20of%20narcissism)), with the NARQ now widely used as a go-to measure for studies exploring the intricate interplay of narcissistic self-love and aggression.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved narq dataset from a CSV file
dataset_path = '../Datasets/narq.csv'
try:
    narq_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'narq_df' is empty before processing
        if narq_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", narq_df.shape[0])
            print(f"📊 Processing {narq_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        narq_df['Embedding'] = narq_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/narq_embeddings.csv'
        narq_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = narq_df.shape[0]
        num_columns = narq_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in narq_df.columns and not narq_df['Embedding'].empty:
            embedding_length = len(narq_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(narq_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/narq_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/narq_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "narq_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/narq_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/narq_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"✅ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"❌ Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="narq Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "narq_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"📊 Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "narq_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"📊 Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "narq_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"🖼️ Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/narq_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"💾 Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"✅ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 8: Test and Evaluate Model** <a class="anchor" id="narq_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/narq_label_encoder.pkl"
model_path = "../Models/narq_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"✅ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"❌ Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"❌ An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"❌ Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Model testing (inference on new data) complete.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/narq_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="NARQ Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/narq_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "narq_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/narq_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of NARQ Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/narq_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of NARQ Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create NARQ Schema in Neo4j** <a class="anchor" id="NARQ_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("NARQ")

# Load the dataset
dataset_path = "../Embeddings/narq_embeddings.csv"
narq_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in narq_df.columns:
    narq_df['Embedding'] = narq_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the NARQ schema in Neo4j
def create_narq_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with NARQ labels
        graph.run("MERGE (f:NARQ_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:NARQ_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:NARQ_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:NARQ_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:NARQ_Noun {name: $noun})", noun=noun)
        
        # Create relationships with NARQ namespacing
        graph.run("""
            MATCH (f:NARQ_Factor {name: $factor}), (a:NARQ_Adjective {name: $adjective})
            MERGE (f)-[:NARQ_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:NARQ_Adjective {name: $adjective}), (s:NARQ_Synonym {name: $synonym})
            MERGE (a)-[:NARQ_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:NARQ_Adjective {name: $adjective}), (v:NARQ_Verb {name: $verb})
            MERGE (a)-[:NARQ_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:NARQ_Adjective {name: $adjective}), (n:NARQ_Noun {name: $noun})
            MERGE (a)-[:NARQ_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:NARQ_Factor {name: $factor}), (a:NARQ_Adjective {name: $adjective})
                MERGE (e:NARQ_Embedding {value: $embedding})
                MERGE (f)-[:NARQ_HAS_EMBEDDING]->(e)
                MERGE (a)-[:NARQ_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create NARQ Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create NARQ Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the NARQ schema
        create_narq_schema(graph, narq_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", narq_df.shape[0])
        mlflow.log_param("Number of Columns", narq_df.shape[1])
        
        # Print the DataFrame
        print(narq_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
