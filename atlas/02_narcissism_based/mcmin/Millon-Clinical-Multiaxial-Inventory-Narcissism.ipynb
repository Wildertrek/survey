# Millon Clinical Multiaxial Inventory (MCMI) – Narcissistic Personality Patterns

## Historical Development of the MCMI 
1. **1969:** Theodore Millon published *Modern Psychopathology*, proposing a comprehensive theory of personality pathology. This framework of personality “prototypes” laid the groundwork for his later tests ([Evolution of the Millon Clinical Multiaxial Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/26151730/#:~:text=Dr.%20Theodore%20Millon%20%281928,to%20its%20current%20status%20as)).  
2. **1977 – MCMI:** Millon released the **Millon Clinical Multiaxial Inventory (MCMI)**, aligned with the DSM-III. The original MCMI included 11 personality scales and 9 clinical syndrome scales ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=test%20construction%20of%20the%20MCMI,7)). It introduced a brief, theory-driven inventory to assess personality disorders and clinical syndromes on multiple axes.  
3. **1987 – MCMI-II:** Following DSM-III-R updates, MCMI-II expanded to 13 personality scales and 9 syndrome scales ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=With%20the%20publication%20of%20the,weighting%20system%20introduced)). Notable changes were splitting the Antisocial-Aggressive scale into separate **Antisocial** and **Sadistic** (Aggressive) scales, adding a **Self-Defeating** (Masochistic) scale, and introducing three **Modifying Indices** (validity checks) and a new item-weighting system ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=With%20the%20publication%20of%20the,weighting%20system%20introduced)). These refinements improved diagnostic specificity and response validity.  
4. **1994 – MCMI-III:** Updated for DSM-IV, MCMI-III featured 14 personality scales, 10 clinical syndrome scales, and 5 validity/correction scales ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,8)). This edition added a **Depressive** personality pattern and a **Post-Traumatic Stress Disorder** scale ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,8)). Some experimental scales (e.g. DSM-III-R Passive-Aggressive) were eliminated to reflect DSM-IV changes ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,8)). MCMI-III also introduced **Grossman Facet Scales** (more granular subscales for each personality pattern) and shifted to a simpler true/false item format (175 items with a 2-point weighting) ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,8)).  
5. **2015 – MCMI-IV:** The current MCMI-IV was revised for DSM-5. It contains 195 true/false items and a fully updated normative sample ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,of%20the%20Turbulent%20Personality%20Scale)). There are 15 personality pattern scales and 10 clinical syndrome scales, along with 5 validity indices ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,of%20the%20Turbulent%20Personality%20Scale)). MCMI-IV added the new **Turbulent** personality scale (reflecting an “ebullient–exuberant–turbulent” trait spectrum) and updated all test items to align with DSM-5 criteria ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,of%20the%20Turbulent%20Personality%20Scale)). Other changes included including ICD-10 codes and refining the Grossman Facet subscales ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,of%20the%20Turbulent%20Personality%20Scale)). Throughout its editions, the MCMI has remained relatively brief (under 30 minutes) and theory-grounded, evolving in parallel with DSM revisions ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,3)) ([The Millon® Clinical Multiaxial Inventory-IV (MCMI-IV)](https://millonpersonality.com/millon-inventories/multiaxial-inventory-iv-mcmi-iv/#:~:text=The%20MCMI%20,minutes%2C%20facilitating%20relatively%20simple%20and)).

## MCMI Structure and the Narcissistic Personality Scale 
**Overall Structure:** The MCMI is organized into multiple sets of scales. MCMI-IV, for example, has 30 scales: **25 clinical scales** (15 Personality and 10 Clinical Syndrome scales) plus **5 validity indices** ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,3%20Severe%20Personality%20Pathology%20scales)). The personality section is divided into **12 Clinical Personality Pattern scales** and **3 Severe Personality Pathology scales** ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=The%20MCMI,3%20Severe%20Personality%20Pathology%20scales)) ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=personality%20scales%20are%20further%20divided,3%20Severe%20Personality%20Pathology%20scales)). The 12 “clinical” patterns correspond to DSM-5 personality disorders (as refined by Millon’s theory), while the 3 “severe” scales capture more pervasive or severe disorders (Paranoid, Schizotypal, Borderline) ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=personality%20scales%20are%20further%20divided,3%20Severe%20Personality%20Pathology%20scales)) ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=have%20been%20broken%20down%20in,effectively%20in%20social%2C%20occupational%2C%20or)). Narcissistic personality features are assessed as one of the standard clinical pattern scales. 

**Narcissistic Scale:** The **Narcissistic Personality Pattern** is denoted as Scale 5 in the MCMI-IV ([Millon Clinical Multiaxial Inventory - Wikipedia](https://en.wikipedia.org/wiki/Millon_Clinical_Multiaxial_Inventory#:~:text=Abbreviation%20Description%20Clinical%20Personality%20Patterns,Turbulent%205%20Narcissistic%206A%20Antisocial)). It is designed to parallel the DSM-5 criteria for Narcissistic Personality Disorder (NPD), but with Millon’s theoretical enhancements ([The Millon® Clinical Multiaxial Inventory-IV (MCMI-IV)](https://millonpersonality.com/millon-inventories/multiaxial-inventory-iv-mcmi-iv/#:~:text=Validity%20scale,personality%20patterns%20or%20syndromal%20conditions)). In Millon’s model, narcissistic pathology is characterized by an **active–pleasure-seeking, self-focused** orientation, manifesting as grandiosity, entitlement, and lack of empathy. The MCMI Narcissistic scale’s items reflect these traits (e.g. feelings of superiority, deserving admiration, exploitative attitudes) in a clinical context. Importantly, each personality scale in the MCMI is scored on a **base rate** metric rather than simple raw scores – a base rate of 75 or 85 often indicates the presence of clinically significant traits, helping clinicians decide if the personality pattern likely meets diagnostic threshold.

**Facet Subscales:** For finer nuance, the Narcissistic scale (like all MCMI personality scales) is broken into three **Grossman Facet Scales**. These facets highlight core components of narcissistic personality functioning ([MCMI-IV comprehensive brochure](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=5%20Narcissistic%20%28CENarc%29%205,Image)). Specifically, the Narcissistic facets are: (1) **Interpersonally Exploitive** – indicating a tendency to exploit or take advantage of others in relationships; (2) **Cognitively Expansive** – reflecting an expansive, overconfident cognitive style (e.g. overestimating one’s abilities or importance); and (3) **Admirable Self-Image** – denoting an inflated self-regard or self-perception of being admirable and special ([MCMI-IV comprehensive brochure](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=5%20Narcissistic%20%28CENarc%29%205,Image)). These facet scores help clinicians see which aspects of narcissism are most prominent. For example, a high *Interpersonally Exploitive* facet might signal manipulative behaviors, whereas a high *Admirable Self-Image* facet underscores grandiose self-concept. Millon’s intent with such facets was to increase clinical utility by pinpointing specific trait domains (interpersonal conduct, cognitive style, self-image, etc.) within each personality pattern ([MCMI-IV comprehensive brochure](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=Turbulent%20Scale%E2%80%94New%21%20MCMI%C2%AE,IV%20can%20help%20you)) ([MCMI-IV comprehensive brochure](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=Dr,and%20an%20inability%20to%20maintain)).

**Relation to Theory:** The Narcissistic pattern in the MCMI exemplifies Millon’s personological theory. Millon considered narcissistic personalities as **“confident–egotistic”** types along a broader spectrum of normal to abnormal narcissism ([[PDF] MCMI-IV comprehensive brochure - Pearson Clinical Assessment](https://www.pearsonclinical.ca/content/dam/school/global/clinical/ca/assets/mcmi-iv/mcmi-iv-comprehensive-brochure.pdf#:~:text=%5BPDF%5D%20MCMI,broader%20range%20of%20personality)). Narcissism is thus seen as a continuum – in its moderate form, it may involve high self-confidence and ambition, but in its disordered form it becomes grandiose, entitled, and lacking reciprocity. MCMI scales are anchored in this idea of personality spectra. Notably, Millon’s theory posits that *clinical* narcissistic personalities often have only mild to moderate functional impairment (e.g. they may maintain work and relationships), in contrast to the *severe* personality disorders which cause more pervasive dysfunction ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=have%20been%20broken%20down%20in,effectively%20in%20social%2C%20occupational%2C%20or)). This theoretical stance is reflected in the MCMI: Narcissistic PD is a “clinical pattern” (not in the severe category), implying that while problematic, it is not as inherently disabling as disorders like Schizotypal or Borderline ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=histrionic%2C%20narcissistic%2C%20antisocial%2C%20sadistic%2C%20compulsive%2C,III%20breaks)). In sum, the MCMI’s structure encapsulates Narcissistic personality both as a distinct diagnostic scale and as a profile of facet-level traits, rooted in Millon’s blend of DSM criteria with evolutionary personality theory.

## Research on Narcissism in the MCMI 
The MCMI’s conceptualization of narcissism has been examined in various research studies, ranging from validation work to typology analyses:

- **Convergent Validity:** Early studies sought to verify that the MCMI’s Narcissistic scale indeed measures narcissism as understood in other contexts. For example, Auerbach (1984) compared the MCMI Narcissistic scale with the **Narcissistic Personality Inventory (NPI)** – a well-known narcissism measure in social/personality psychology. In a sample of undergraduates, the two scales showed a significant positive correlation (r ≈ .55) ([Validation of two scales for narcissistic personality disorder - PubMed](https://pubmed.ncbi.nlm.nih.gov/6520692/#:~:text=Undergraduate%20scores%20on%20the%20Narcissistic,presentation)), suggesting that MCMI’s narcissism scale converges well with trait narcissism as captured by the NPI. Notably, neither the NPI nor the MCMI Narcissistic scores correlated strongly with a social desirability measure ([Validation of two scales for narcissistic personality disorder - PubMed](https://pubmed.ncbi.nlm.nih.gov/6520692/#:~:text=Undergraduate%20scores%20on%20the%20Narcissistic,presentation)). This indicates that while narcissistic individuals often present in self-flattering ways, the test scores weren’t simply artifacts of “faking good” – rather, the Narcissistic scale taps genuine self-perceptions of superiority and entitlement, not just general self-favorability bias. Such findings support the **reliability and validity** of the MCMI Narcissistic pattern scale as a measure of the narcissism construct.

- **Clinical Correlates:** In clinical populations, researchers have studied how the MCMI Narcissistic pattern manifests and correlates with other psychopathology. Chatham et al. (1993), for instance, examined narcissism using both the MCMI and the MMPI in a psychiatric sample. They found that MCMI narcissism scores aligned with certain MMPI indicators of narcissistic and antisocial traits ([Validation of two scales for narcissistic personality disorder - PubMed](https://pubmed.ncbi.nlm.nih.gov/6520692/#:~:text=intercorrelated%20and%20were%20also%20correlated,presentation)) ([Validation of two scales for narcissistic personality disorder - PubMed](https://pubmed.ncbi.nlm.nih.gov/6520692/#:~:text=Undergraduate%20scores%20on%20the%20Narcissistic,presentation)). Generally, high MCMI narcissism tends to coincide with traits like low anxiety or depression (consistent with the idea that narcissists often lack insight into their flaws or distress) and higher aggression or antisocial tendencies in some cases ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=analysis%20and%20principal%20component%20analysis,and%20externalizing%20psychopathology%2Fantisocial%20acting%20out)) ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=syndromes%20scales%20%28third,Cluster)). Millon himself noted that narcissistic personalities can appear relatively well-adjusted on the surface – often **free of self-reported distress** – yet exhibit maladaptive interpersonal behaviors (e.g. exploitiveness, arrogance) that may only become evident through others’ feedback or in long-term outcomes.

- **Subtype Analyses:** Narcissistic personality disorder is heterogeneous, and research using the MCMI has identified **distinct subtypes** of narcissistic personalities. An influential study by DiGiuseppe et al. (1995) performed a cluster analysis on MCMI-II profiles of patients who scored very high on the Narcissistic scale. They discovered three reliable clusters of narcissistic personalities, labeled the **“true narcissist,” “compensating narcissist,” and “detached narcissist.”** Each subtype shared core narcissistic features (self-centeredness and a sense of entitlement) but differed in emotional experience and other traits ([(PDF) Cluster Analysis of Narcissistic Personality Disorders on the MCMI-II](https://www.researchgate.net/publication/274290809_Cluster_Analysis_of_Narcissistic_Personality_Disorders_on_the_MCMI-II#:~:text=best%20fit%20was%20a%20three,one%20was%20characterized%20by%20social)) ([(PDF) Cluster Analysis of Narcissistic Personality Disorders on the MCMI-II](https://www.researchgate.net/publication/274290809_Cluster_Analysis_of_Narcissistic_Personality_Disorders_on_the_MCMI-II#:~:text=were%20named%20the%20%27true%20narcissist%2C%27,that%20all%20narcissists%20experience%20negative)). The “true narcissist” cluster represented the classic grandiose type – individuals with high self-importance and little psychological distress. The “compensating narcissist” appeared to cover those who use grandiosity to offset underlying feelings of inferiority (similar to the concept of a covert narcissist who overcompensates); interestingly, this group also reported minimal distress, suggesting their compensatory strategy buffers them from conscious dysphoria ([(PDF) Cluster Analysis of Narcissistic Personality Disorders on the MCMI-II](https://www.researchgate.net/publication/274290809_Cluster_Analysis_of_Narcissistic_Personality_Disorders_on_the_MCMI-II#:~:text=including%20the%20General%20Health%20Questionnaire%2C,affectivity%20or%20are%20emotionally%20vulnerable)). The “detached narcissist” cluster, by contrast, was characterized by social aloofness or withdrawal on top of narcissistic traits – these individuals were narcissistic but also socially isolated and possibly prone to dysphoric affects like boredom or emptiness ([(PDF) Cluster Analysis of Narcissistic Personality Disorders on the MCMI-II](https://www.researchgate.net/publication/274290809_Cluster_Analysis_of_Narcissistic_Personality_Disorders_on_the_MCMI-II#:~:text=best%20fit%20was%20a%20three,one%20was%20characterized%20by%20social)). This subtype had more in common with the “vulnerable” narcissistic phenotype (sometimes called covert narcissism), except that in this study they still did not report severe negative affect, only interpersonal detachment. The identification of these clusters underscores that **not all narcissists are alike**: some are confident and extroverted, while others mask insecurity or remain socially avoidant. The authors concluded that researchers should “abandon the notion that all narcissists experience negative affectivity or are emotionally vulnerable” ([(PDF) Cluster Analysis of Narcissistic Personality Disorders on the MCMI-II](https://www.researchgate.net/publication/274290809_Cluster_Analysis_of_Narcissistic_Personality_Disorders_on_the_MCMI-II#:~:text=Scale%2C%20the%20Beck%20Depression%20Inventory,affectivity%20or%20are%20emotionally%20vulnerable)), highlighting the importance of subtype distinctions.

- **Millon’s Theoretical Subtypes:** Decades before such empirical subtyping, Millon had theorized multiple variants of narcissistic personalities in his writings. In *Disorders of Personality* (Millon & Davis, 1996), he outlined **five subtypes of NPD** ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=Subtype%20%20Features%20Unprincipled%20narcissist,esteem%3B%20offsets%20deficits%20by%20creating)) ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=results%20from%20self,positions%3B%20expecting%20recognition%20from%20others)):
  - *Unprincipled narcissist:* including antisocial features – unscrupulous, deceitful, arrogant, and exploitative ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=Subtype%20%20Features%20Unprincipled%20narcissist,cancel%20out%20deep%20feelings%20of)).
  - *Amorous narcissist:* including histrionic/sexualized features – seductive, charming, but lacking intimacy and fidelity ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=fraudulent%2C%20deceptive%2C%20arrogant%2C%20exploitive%3B%20a,worth)).
  - *Compensatory narcissist:* resembling a **covert narcissist** – driven by deep feelings of inferiority, they create illusions of superiority to counteract their shame or low self-esteem ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=affairs%2C%20often%20with%20exotic%20partners,status%20and%20advantages%20by%20association)).
  - *Elitist narcissist:* a narcissist who believes they have special status or privilege – they exaggerate their accomplishments and affiliation with “high-status” groups to feel important ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=results%20from%20self,positions%3B%20expecting%20recognition%20from%20others)).
  - *Normal narcissist:* a mild form closest to high self-confidence – competitive and self-centered but relatively functional and empathetic compared to other subtypes ([Narcissistic personality disorder - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_personality_disorder#:~:text=childhood%20status%20and%20pseudo,positions%3B%20expecting%20recognition%20from%20others)). 

  These subtypes, while not directly scored by the MCMI, provide a **rich conceptual context** for understanding the range of narcissistic personalities that might elevate the Narcissistic scale. For instance, a very high Narcissistic score combined with an elevated Antisocial scale might suggest an “unprincipled” narcissist (someone with both narcissistic and antisocial tendencies), whereas a high Narcissistic score with high Avoidant or Depressive features might hint at a compensatory/vulnerable narcissist using grandiosity to mask insecurity. Millon’s own research and case observations informed these prototypes, which remarkably align with later empirical findings (e.g. the “compensating narcissist” cluster above mirrors Millon’s compensatory subtype). Thus, the MCMI’s conceptualization of narcissism is not monolithic; it acknowledges diversity in expression, from flamboyant and aggressive to shy and self-inflating.

- **Other Research:** Additional studies have examined the psychometrics of the MCMI narcissism scale across cultures and in comparison to alternative instruments. For example, cross-cultural psychometric evaluations (e.g. Elion et al., 2021) confirm that the Narcissistic pattern scale, along with others, fits into a four-factor structure of personality pathology (with Narcissistic loading on a factor often interpreted as **Dominance or Acting-Out** pathology alongside antisocial and paranoid traits) ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=MCMI,and%20externalizing%20psychopathology%2Fantisocial%20acting%20out)) ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=syndromes%20scales%20%28third,Cluster)). This suggests narcissism, as measured by MCMI, aligns with an overarching dimension of hostile/aggrandizing personality tendencies. Also, researchers have noted that truly narcissistic individuals might under-report certain symptoms (due to ego-syntonic pride or lack of insight), making valid assessment challenging. The MCMI’s inclusion of validity indices (e.g. a high **Desirability** score Y) can alert clinicians if a test-taker is presenting an overly positive self-image – a pattern that might itself hint at narcissistic defensiveness. Overall, the body of research indicates the MCMI is a **useful tool for identifying narcissistic personality patterns**, especially in clinical settings, and its results correlate meaningfully with both external criteria and theoretical expectations of narcissism.

## Narcissism in MCMI vs. Other Narcissism Models 
Numerous models and instruments exist to assess narcissistic traits and pathology. The MCMI’s approach – a clinical diagnostic scale grounded in DSM criteria and Millon’s theory – differs in important ways from other narcissism measures used in psychology. Below is a comparison highlighting key similarities and differences:

- **Five-Factor Narcissism Inventory (FFNI):** The FFNI is a 148-item inventory that conceptualizes narcissism through **trait facets aligned with the five-factor model** (FFM) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Narcissism%20measures.%20Five,In%20the%20current)). It assesses 15 lower-level traits covering both **grandiose** and **vulnerable** aspects of narcissism ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Narcissism%20measures.%20Five,In%20the%20current)). For example, FFNI includes facets like reactive anger, shame (vulnerable side), as well as exhibitionism, authoritativeness (grandiose side). In contrast, the MCMI Narcissistic scale is a single composite reflecting narcissistic personality disorder as a whole. **Similarity:** Both FFNI and MCMI acknowledge that narcissism has multiple facets – indeed, many FFNI traits (e.g. exploitative, entitled, confident) parallel the content of MCMI’s narcissism items and facets. **Difference:** FFNI provides a **dimensional profile** of narcissistic traits (useful for research and nuance), whereas MCMI gives a **diagnostic-level assessment** – indicating if narcissistic personality *pattern* is present at clinically significant levels. The FFNI’s results can be mapped onto FFM personality dimensions (e.g. high antagonism, low agreeableness, etc.), offering a normative trait perspective. MCMI, however, places narcissism in a **clinical context**, often used to inform diagnoses per DSM. Notably, FFNI yields separate scores for Grandiose vs Vulnerable narcissism composites, something the MCMI does not explicitly separate (vulnerable narcissistic features might instead be reflected by elevations on other MCMI scales like Avoidant or Depressive alongside Narcissistic).

- **Pathological Narcissism Inventory (PNI):** The PNI is a self-report measure specifically designed to capture **pathological narcissism** in its two manifestations. It consists of 52 items across 7 subscales that map onto **Narcissistic Grandiosity** (e.g. Exploitativeness, Grandiose Fantasy, Entitlement Rage, Self-Sacrificing Self-enhancement) and **Narcissistic Vulnerability** (e.g. Contingent Self-Esteem, Hiding the Self, Devaluing others/self) ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=existing%20narcissism%20measures,The%20PNI)). This instrument directly addresses the “dual nature” of narcissism: the outward grandiosity and the inward vulnerability/shame. In comparison, the MCMI focuses on the *grandiose, DSM-defined NPD* phenotype – entitlement, admiration-seeking, arrogance – and does not have dedicated subscales for vulnerability (beyond what might be inferred from the facet patterns or other co-occurring scales). **Similarity:** Both PNI and MCMI view narcissism as a pathological construct and find that narcissism involves entitlement and exploitativeness ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=PNI%20is%20a%2052,The%20PNI)) (which MCMI Narcissistic facet 5.1 “Interpersonally Exploitive” mirrors). **Difference:** The **PNI explicitly measures internal pain and insecurity** that can accompany narcissistic pathology (e.g. feelings of emptiness, rage when disappointed ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=existing%20narcissism%20measures,The%20PNI))), whereas the MCMI Narcissistic scale, true to DSM, emphasizes the **observable criteria of NPD** (grandiosity, lack of empathy) and tends to capture the “thick-skinned” narcissist. A person who is highly **vulnerable narcissistic** (prone to shame and feelings of inferiority) might not score very high on MCMI’s Narcissistic scale if they do not endorse the boldly egotistical statements, whereas on the PNI they would register on the vulnerability subscales. In clinical practice, the MCMI might show such a person with a mixed profile (e.g. moderate Narcissistic score combined with Dependent or Avoidant elevations). Thus, PNI provides a **more fine-grained profile** of narcissistic pathology (differentiating fragile vs. grandiose states), complementing the MCMI’s broader diagnostic approach. Research often uses the PNI to study NPD in depth, whereas the MCMI might be used to **screen or diagnose** NPD in a clinical battery.

- **Narcissistic Personality Inventory (NPI):** The NPI is the most popular measure of **trait narcissism in the general (subclinical) population**. It is a forced-choice questionnaire (originally 40 items) where individuals choose between narcissistic statements and non-narcissistic alternatives. The NPI primarily captures **grandiose narcissism** — it was derived from DSM-III NPD criteria but framed for normals. It covers facets like authority, exhibitionism, superiority, vanity, entitlement, and exploitativeness ([
            Exploring network structure and central items of the Narcissistic Personality Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7051847/#:~:text=The%20main%20tool%20used%20to,being%20better%20than%20other%20people)) ([
            Exploring network structure and central items of the Narcissistic Personality Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7051847/#:~:text=statement,a%20strong%20will%20to%20power%E2%80%9D)). The NPI yields a total score and sometimes subscale scores (e.g. Leadership/Authority, Grandiose Exhibitionism, Entitlement/Exploitativeness). Compared to the MCMI, the NPI represents a **more normative and social/personality psychology perspective**. **Similarity:** Both the NPI and MCMI Narcissistic scale revolve around **grandiose self-perception and entitlement**. Endorsing items like “I am an extraordinary person” (NPI) or “I often feel I deserve special treatment” (hypothetical MCMI item) would reflect narcissistic beliefs in either instrument. Empirically, moderate correlations have been found between NPI scores and MCMI narcissism in nonclinical samples ([Validation of two scales for narcissistic personality disorder - PubMed](https://pubmed.ncbi.nlm.nih.gov/6520692/#:~:text=Undergraduate%20scores%20on%20the%20Narcissistic,presentation)). **Differences:** The **purpose and context differ greatly** – NPI is often used with high-functioning individuals to measure **individual differences** in narcissism where a certain level of narcissism might be normative or even adaptive (e.g. confidence, leadership). In fact, some NPI items reflect positive self-regard that isn’t overtly pathological (e.g. desire to be a leader). The MCMI Narcissistic pattern, on the other hand, is intended for clinical diagnosis of a personality disorder; it assumes that elevated narcissism is maladaptive. Another difference is response format: NPI’s forced-choice format can sometimes mask socially desirable responding, whereas MCMI’s true/false format, coupled with validity scales, can directly detect exaggeration or minimization. Moreover, the NPI does **not measure narcissistic vulnerability** at all and tends to focus on **“bright-side” narcissistic traits** (confidence, assertiveness) along with the “dark-side” (entitlement). MCMI’s Narcissistic scale items are more uniformly negative in connotation (e.g. conceit, exploitation), since they target a disorder. In summary, NPI is best for assessing **subclinical or “normal” narcissism**, while MCMI’s Narcissistic scale targets **clinical NPD**. The two overlap on grandiosity content, but MCMI won’t identify someone as narcissistic unless traits are extreme enough to resemble a disorder (often corresponding to a much higher NPI score percentile).

- **Narcissistic Admiration and Rivalry Questionnaire (NARQ):** The NARQ is a newer research instrument that splits grandiose narcissism into **two strategic components: Admiration and Rivalry**. **Admiration** reflects the self-promotional, charming, and assertive side of narcissism (agentic self-enhancement: “I am great, I deserve admiration”), whereas **Rivalry** reflects the antagonistic, defensive side (self-protection by denigrating others: “I want my rivals to fail”) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=The%20NARQ%20assesses%20grandiose%20narcissism,contains%20nine%20items%2C%20which%20participants)). Each dimension is measured by 9 items ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=The%20NARQ%20assesses%20grandiose%20narcissism,contains%20nine%20items%2C%20which%20participants)), and individuals can score high on one or both. The MCMI Narcissistic scale does not explicitly separate these dimensions, but conceptually it encompasses both – a narcissistic individual often both craves admiration *and* may react with hostility when challenged. **Similarity:** The core of narcissism – grandiose self-view and entitlement – underpins both MCMI and NARQ. Someone scoring high on MCMI narcissism would likely show high NARQ Admiration (high self-confidence and ambition for status) as well as some degree of Rivalry (lack of empathy, arrogance) ([
            Measurement Invariance of Three Narcissism Questionnaires Across the United States, the United Kingdom, and Germany - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7783690/#:~:text=The%20NARQ%20assesses%20grandiose%20narcissism,contains%20nine%20items%2C%20which%20participants)). Indeed, the MCMI facet “Interpersonally Exploitive” maps onto the antagonism captured by Rivalry, while “Admirable Self-Image” maps onto the self-enhancing positivity of Admiration. **Difference:** The NARQ provides a **more nuanced profile** by telling us *how* a narcissistic person pursues esteem – either mainly by emphasizing their own brilliance (Admiration) or by putting others down (Rivalry). In practice, two narcissistic individuals might both score high on MCMI Scale 5, but one could be primarily an Admiration-seeker (sociable, charismatic, less openly hostile) and another a Rivalry-focused narcissist (abrasive, competitive, prone to anger). The NARQ would distinguish these; the MCMI alone would not, as it yields one combined score. Additionally, the NARQ is used in research linking narcissism to interpersonal outcomes (e.g. relationship conflict, leadership styles), whereas the MCMI is used in **clinical assessment**. The NARQ’s differentiation is grounded in the idea of narcissistic maintenance strategies, something not addressed in DSM criteria. Thus, while MCMI tells us “this person has narcissistic personality traits at a clinical level,” the NARQ can elaborate on their **dominant narcissistic interpersonal style (charming vs. combative)**. Clinically, this kind of detail might emerge qualitatively or through the MCMI facet patterns, but the NARQ quantifies it directly.

- **Hypersensitive Narcissism Scale (HSNS):** The HSNS (Hendin & Cheek, 1997) is a brief 10-item scale that measures **covert or vulnerable narcissism** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Hypersensitive%20Narcissism%20Scale%20%28HSNS%29,5)). It was derived from Murray’s Narcissism scale and influenced by psychoanalytic descriptions of the “hypersensitive” (or covert) narcissist. The HSNS items tap feelings of inadequacy, sensitivity to criticism, and secret entitlement – for example, agreeing with statements like “I can become entirely absorbed in thinking about my personal affairs, my health, my cares or my relations to others” or “I dislike sharing the credit of an achievement with others” (these reflect an introverted self-absorption and entitlement). The **focus** is on the narcissist who is defensive, easily hurt, and often outwardly modest or shy (in contrast to the overt bragging narcissist). **Similarity:** Both HSNS and MCMI recognize entitlement and self-focus as features of narcissism. If an MCMI test-taker had a strong vein of hypersensitive narcissism, it might show up in a mixed profile – perhaps a moderate Narcissistic scale with co-elevations on Avoidant or Depressive (suggesting insecurity) and high scores on the Disclosure/Desirability indices (suggesting reluctance to admit flaws). **Differences:** The HSNS is **unidimensional for vulnerability** and intentionally does *not* measure grandiose behaviors. The MCMI Narcissistic pattern, conversely, is skewed towards **grandiose narcissism** (given its DSM origins). A high HSNS person might or might not score high on MCMI Narcissistic: they may not endorse arrogant statements on the MCMI (keeping that side hidden or unrecognized). In fact, such a person might even score higher on the MCMI Dependent or Negativistic scales, with only a mild narcissistic elevation – reflecting an inner narcissistic need that’s masked by avoidant tendencies. Another difference is application: HSNS is mostly used in research on personality (it’s an “economical” way to assess vulnerable narcissism ([[PDF] Millon Clinical Multiaxial Inventory-IV (MCMI-IV) Profile Patterns and ...](https://repository.fit.edu/cgi/viewcontent.cgi?article=1337&context=etd#:~:text=and%20,instruments%20such%20as%20the%20MMPI))), whereas the MCMI is a comprehensive clinical test. The HSNS’s concept of *hypersensitive narcissism* complements MCMI by covering the “thin-skinned” side of NPD which DSM-based tools tend to under-assess. Modern researchers often pair a grandiose measure (like NPI or NARQ Admiration) with a vulnerable measure (like HSNS or PNI Vulnerability) to get a full picture – something a single MCMI scale cannot directly provide. However, Millon’s inclusion of a **Depressive (Melancholic) personality scale** in MCMI-III/IV and recognition of **Compensatory narcissist** subtype show that the theory acknowledges these covert features, even if the Narcissistic scale itself centers on overt traits.

- **“Dark Tetrad” Narcissism (DT4):** In the realm of the “Dark Triad” or “Dark Tetrad” of personality (Machiavellianism, Narcissism, Psychopathy, and added Sadism for the tetrad), narcissism is viewed as one of four interrelated socially aversive traits. Specialized short measures such as the **Short Dark Tetrad (SD4)** have been developed, which include a Narcissism subscale typically labeled with a more euphemistic term like “Special” (to encourage honest responding) ([Correlates of the Dark Tetrad - PubMed](https://pubmed.ncbi.nlm.nih.gov/38502994/#:~:text=This%20study%20explored%20demographic%2C%20ideological%2C,third%20of%20the%20variance%20in)). In the SD4, Narcissism is measured by items reflecting a sense of specialness and superiority (e.g. “I insist on getting the respect I deserve,” akin to entitlement) alongside the other dark traits. **Similarity:** Both the MCMI and dark-triad measures treat narcissism as part of the cluster of antagonistic, egocentric traits. For instance, studies using the Dark Tetrad find that narcissism correlates with low agreeableness and high competitiveness, much like MCMI narcissism is often accompanied by aggressive, dominant traits ([Correlates of the Dark Tetrad - PubMed](https://pubmed.ncbi.nlm.nih.gov/38502994/#:~:text=This%20study%20explored%20demographic%2C%20ideological%2C,third%20of%20the%20variance%20in)) ([Correlates of the Dark Tetrad - PubMed](https://pubmed.ncbi.nlm.nih.gov/38502994/#:~:text=personality%20test%20,with%20all%20four%20dark%20traits)). **Difference:** The **Dark Triad/Tetrad approach frames narcissism as a “dark” trait** even in non-clinical populations, often focusing on its interpersonal toxicity (e.g. narcissists may exploit or manipulate, but to a lesser extent than psychopaths). The MCMI, on the other hand, frames narcissism as a clinical personality disorder, with richer description but used for diagnosis rather than labeling someone “dark.” Also, MCMI has separate scales for **Antisocial and Sadistic** personalities, which allows it to differentiate narcissistic traits from outright psychopathic or sadistic tendencies. In a Dark Triad inventory, someone scoring high in narcissism might also score high in Machiavellianism or psychopathy, but the test isn’t intended to diagnose each – it treats them as continuous traits. MCMI would separately measure those through Antisocial or Aggressive (Sadistic) scales. Essentially, **MCMI is more granular in the clinical domain**, whereas Dark Triad measures are broad but shallow, grouping all “dark” personalities together for research simplicity. Conceptually, narcissism in DT4 is often described as the **socially “brightest” of the dark traits** – narcissists want admiration and may still adhere to social norms to get praise, whereas psychopaths and sadists flagrantly violate norms. The MCMI captures this nuance by often showing narcissists with relatively intact social functioning (as Millon noted, narcissists can be successful professionally and aren’t necessarily flagged on severe pathology scales) ([Frontiers | Psychometric Properties of the Millon Clinical Multiaxial Inventory–III in an Arabic Clinical Sample Compared With American, Italian, and Dutch Cultures](https://www.frontiersin.org/journals/psychology/articles/10.3389/fpsyg.2021.562619/full#:~:text=histrionic%2C%20narcissistic%2C%20antisocial%2C%20sadistic%2C%20compulsive%2C,III%20breaks)). In sum, the Dark Tetrad perspective emphasizes narcissism’s overlap with other dark traits (like all being low in empathy), while the MCMI’s approach keeps the focus on narcissism as a distinct personality pattern that can, however, co-occur with other disorders.

**Summary of Approaches:** The MCMI’s narcissistic scale is rooted in a **clinical diagnostic tradition**, portraying narcissism as a personality disorder with defined criteria. Measures like the NPI and NARQ stem from **social-personality research**, treating narcissism as a trait continuum with both adaptive and maladaptive sides. The PNI and FFNI come from a **clinical/research hybrid perspective**, attempting to integrate trait models with the pathology of NPD (particularly emphasizing the **grandiose vs. vulnerable dichotomy**). The HSNS represents a targeted look at the covert side of narcissism, and Dark Triad measures place narcissism in a pantheon of antisocial traits. Despite their different lenses, there is substantial **conceptual overlap**: all recognize narcissism involves grandiose self-views and entitlement, and most acknowledge there are “two faces” of narcissism (even the MCMI theoretical writings do, via subtypes). The differences lie in **scope and usage**: MCMI is broad (covering many disorders) and used for **diagnosis and clinical formulation**, whereas instruments like NPI, NARQ, PNI, etc., are narrow tools used for **focused assessment or research** on narcissistic traits. In practice, a clinician might use the MCMI to determine if a patient meets criteria for Narcissistic Personality Disorder, and also consider something like the PNI to understand that patient’s particular narcissistic dynamics (e.g. are they fragile or purely grandiose?). Each model contributes to a fuller understanding. MCMI gives the **“big picture”** (is narcissism present, and how does it interact with other personality patterns?), while the other models zoom in on **specific nuances** (normal vs. pathological, grandiose vs. vulnerable, admiration vs. rivalry, etc.). Using findings from all these approaches can enrich both clinical practice and research—ensuring that Narcissistic Personality Patterns are recognized not only as a checklist of symptoms (as in DSM/MCMI) but as a complex trait domain spanning from normal personality into psychopathology ([Initial construction and validation of the Pathological Narcissism Inventory - PubMed](https://pubmed.ncbi.nlm.nih.gov/19719348/#:~:text=The%20construct%20of%20narcissism%20is,Entitlement%20Rage)) ([
            Exploring network structure and central items of the Narcissistic Personality Inventory - PMC
        ](https://pmc.ncbi.nlm.nih.gov/articles/PMC7051847/#:~:text=vulnerability%20are%20considered%20as%20the,by%20a%20core%20of%20entitlement)).

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("✅ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"📝 Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"✅ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"⚠️ Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("✅ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"❌ Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"❌ OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"❌ An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

# **Step 4: Create MCMIN Embeddings** <a class="anchor" id="mcmin_page_3"></a>

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("✅ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"❌ Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved mcmin dataset from a CSV file
dataset_path = '../Datasets/mcmin.csv'
try:
    mcmin_df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"❌ {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"❌ OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"❌ An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"⏳ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'mcmin_df' is empty before processing
        if mcmin_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", mcmin_df.shape[0])
            print(f"📊 Processing {mcmin_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        mcmin_df['Embedding'] = mcmin_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"✅ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/mcmin_embeddings.csv'
        mcmin_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"💾 Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = mcmin_df.shape[0]
        num_columns = mcmin_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in mcmin_df.columns and not mcmin_df['Embedding'].empty:
            embedding_length = len(mcmin_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"📏 Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("⚠️ Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(mcmin_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"❌ OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"❌ An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/mcmin_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/mcmin_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"💾 Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "mcmin_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"✅ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"✅ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"❌ An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n✅ Label encoding completed and encoder saved.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import joblib
import mlflow
from openai import OpenAI
import os
from dotenv import load_dotenv
import ast  # Import ast for literal_eval

# Load the dataset containing embeddings
embeddings_csv_path = '../Embeddings/mcmin_embeddings.csv'

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array using ast.literal_eval.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    try:
        return np.array(ast.literal_eval(embedding_str))
    except (ValueError, SyntaxError):
        return np.array()  # Or handle the error as appropriate

with mlflow.start_run(run_name="Step 9: Model Application, Visualization, and Analysis") as run:
    try:
        mlflow.log_param("Step", "Model Application, Visualization, and Analysis")

        # Define paths
        label_encoder_path = "../Models/mcmin_label_encoder.pkl"
        model_path = "../Models/mcmin_rf_model.pkl"

        mlflow.log_param("Embeddings CSV", embeddings_csv_path)
        mlflow.log_param("Label Encoder", label_encoder_path)
        mlflow.log_param("Model", model_path)

        # -------------------- Load Data and Artifacts --------------------

        # Load the dataset containing embeddings
        try:
            df = pd.read_csv(embeddings_csv_path)
            print(f"✅ Embeddings dataset loaded from: {embeddings_csv_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Embeddings dataset not found at: {embeddings_csv_path}"
            )
            print(f"❌ Error: Embeddings dataset not found at: {embeddings_csv_path}")
            raise

        # Apply the parsing function
        df['Embedding'] = df['Embedding'].apply(parse_embedding)

        # Print for debugging
        print("\nDataFrame with Embeddings:")
        print(df.head())
        print("\nShape of DataFrame:")
        print(df.shape)
        print("\nType of 'Embedding' column:")
        print(df['Embedding'].dtype)
        print("\nFirst Embedding:")
        if not df['Embedding'].empty:
            print(df['Embedding'].iloc[0])
        else:
            print("No embeddings found in DataFrame")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"✅ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"❌ Error: Label encoder not found at: {label_encoder_path}")
            raise

        # -------------------- PCA Visualization --------------------

        mlflow.log_param("Visualization", "PCA")

        try:
            # Stack embeddings
            X = np.stack(df['Embedding'].values)

            # Print shape of X
            print("\nShape of X:")
            print(X.shape)

            # Encode Factor
            y_encoded = label_encoder.transform(df['Factor'].values)

            # Use PCA for dimensionality reduction
            pca = PCA(n_components=2)
            X_pca = pca.fit_transform(X)

            # Create the visualization
            plt.figure(figsize=(12, 8))
            scatter = plt.scatter(
                X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis'
            )
            plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
            plt.xlabel("PCA Component 1")
            plt.ylabel("PCA Component 2")
            plt.title("PCA of MCMIN Embeddings")

            # Save and log the plot
            pca_plot_path = "mcmin_embeddings_pca.png"
            plt.savefig(pca_plot_path)
            mlflow.log_artifact(pca_plot_path, artifact_path="visualization")
            plt.close()  # Close the plot to release memory

            print(f"✅ PCA visualization saved to: {pca_plot_path}")

        except Exception as e:
            mlflow_error = f"An error occurred during PCA visualization: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ {mlflow_error}")
            raise

        # -------------------- K-Means Clustering --------------------

        mlflow.log_param("Clustering", "K-Means")

        try:
            n_clusters = 5  # You can parameterize this
            mlflow.log_param("n_clusters", n_clusters)

            # Stack embeddings (if not already done)
            if "X" not in locals():
                X = np.stack(df["Embedding"].values)

            # Cluster embeddings
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            clusters = kmeans.fit_predict(X)

            # Add cluster information to the DataFrame and save it
            df["Cluster"] = clusters
            clustered_csv_path = "mcmin_clustered_embeddings.csv"
            df.to_csv(clustered_csv_path, index=False)
            mlflow.log_artifact(clustered_csv_path, artifact_path="clustering")

            print(
                f"✅ K-Means clustering performed. Clustered data saved to: {clustered_csv_path}"
            )

        except Exception as e:
            mlflow_error = f"An error occurred during K-Means clustering: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ {mlflow_error}")
            raise

        # -------------------- Model Application/Inference --------------------

        mlflow.log_param("Model Application", "Inference")

        try:
            # Load the trained model
            loaded_clf = joblib.load(model_path)

            def predict_factor(new_text):
                """
                Predicts the personality factor for new text using the trained model
                and the saved label encoder.

                Args:
                    new_text: The text to predict the personality factor for.

                Returns:
                    The predicted personality factor.
                """
                try:
                    # Get the embedding for the new text
                    new_embedding = get_embedding(new_text)

                    # Predict the factor
                    predicted_factor_encoded = loaded_clf.predict([new_embedding])

                    # Decode the prediction using the label encoder
                    predicted_factor = label_encoder.inverse_transform(
                        predicted_factor_encoded
                    )[0]

                    return predicted_factor

                except Exception as e:
                    print(f"❌ An error occurred during prediction: {e}")
                    raise

            # Example usage:
            new_text = "I love going to parties and meeting new people."
            predicted_factor = predict_factor(new_text)
            print(f"Predicted Factor for '{new_text}': {predicted_factor}")
            mlflow.log_param("New Text", new_text)
            mlflow.log_param("Predicted Factor", predicted_factor)

        except FileNotFoundError as e:
            mlflow_error = f"Model not found at: {model_path}"
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ Error: {e}")
            raise

        except Exception as e:
            mlflow_error = f"An error occurred during prediction: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"❌ An error occurred during prediction: {e}")
            raise

    except Exception as e:
        # Catch any unexpected errors
        mlflow_error = f"An unexpected error occurred: {e}"
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", mlflow_error)
        print(f"❌ {mlflow_error}")
        raise

    finally:
        mlflow.end_run()

print("\n✅ Step 10: Save Model Visualization and Analysis complete.")
````

# **Step 13: Create MCMIN Schema in Neo4j** <a class="anchor" id="MCMIN_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("MCMIN")

# Load the dataset
dataset_path = "../Embeddings/mcmin_embeddings.csv"
mcmin_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in mcmin_df.columns:
    mcmin_df['Embedding'] = mcmin_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the MCMIN schema in Neo4j
def create_mcmin_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with MCMIN labels
        graph.run("MERGE (f:MCMIN_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:MCMIN_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:MCMIN_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:MCMIN_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:MCMIN_Noun {name: $noun})", noun=noun)
        
        # Create relationships with MCMIN namespacing
        graph.run("""
            MATCH (f:MCMIN_Factor {name: $factor}), (a:MCMIN_Adjective {name: $adjective})
            MERGE (f)-[:MCMIN_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:MCMIN_Adjective {name: $adjective}), (s:MCMIN_Synonym {name: $synonym})
            MERGE (a)-[:MCMIN_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:MCMIN_Adjective {name: $adjective}), (v:MCMIN_Verb {name: $verb})
            MERGE (a)-[:MCMIN_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:MCMIN_Adjective {name: $adjective}), (n:MCMIN_Noun {name: $noun})
            MERGE (a)-[:MCMIN_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:MCMIN_Factor {name: $factor}), (a:MCMIN_Adjective {name: $adjective})
                MERGE (e:MCMIN_Embedding {value: $embedding})
                MERGE (f)-[:MCMIN_HAS_EMBEDDING]->(e)
                MERGE (a)-[:MCMIN_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create MCMIN Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create MCMIN Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the MCMIN schema
        create_mcmin_schema(graph, mcmin_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", mcmin_df.shape[0])
        mlflow.log_param("Number of Columns", mcmin_df.shape[1])
        
        # Print the DataFrame
        print(mcmin_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"❌ Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
