# Narcissistic Personality Inventory (NPI)

## Original Development  
The **Narcissistic Personality Inventory (NPI)** was created by psychologists **Robert Raskin and Calvin S. Hall** in **1979** ([](http://www.sakkyndig.com/psykologi/artvit/emmons1987.pdf#:~:text=Raskin%20and%20Hall%20,expectation%20of%20special%20favors%20with%02out)). It was introduced as a **54-item, forced-choice questionnaire** designed to measure narcissism as a **personality trait** in the general (non-clinical) population ([](http://www.sakkyndig.com/psykologi/artvit/emmons1987.pdf#:~:text=Raskin%20and%20Hall%20,exhibitionistic%E2%80%94requires%20constant%20attention%20and)). The NPI‚Äôs construction was based on the newly defined **DSM-III criteria** for Narcissistic Personality Disorder (NPD) ([](http://www.sakkyndig.com/psykologi/artvit/emmons1987.pdf#:~:text=designed%20to%20measure%20individual%20differences,expectation%20of%20special%20favors%20with%02out)) ‚Äì criteria such as grandiose self-importance, fantasies of unlimited success/power, exhibitionism (need for attention), entitlement, and others. The initial purpose was to operationalize these narcissistic traits for research use, **not for clinical diagnosis** ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=The%20NPI%20is%20not%20intended,narcissism.%5B%201)). In other words, the NPI was intended to capture ‚Äúnormal‚Äù or **subclinical narcissism** levels in individuals ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=The%20NPI%20is%20not%20intended,narcissism.%5B%201)), providing a way to quantitatively study narcissistic tendencies outside of pathological contexts.

## Timeline of Key Publications and Revisions  
- **1979 ‚Äì NPI Introduction:** Raskin & Hall publish the first NPI scale in *Psychological Reports*, with 54 items derived from NPD diagnostic criteria ([](http://www.sakkyndig.com/psykologi/artvit/emmons1987.pdf#:~:text=Raskin%20and%20Hall%20,expectation%20of%20special%20favors%20with%02out)). This seminal paper established the NPI as the first reliable instrument for trait narcissism, filling the gap in empirical narcissism research.  
- **1984‚Äì1987 ‚Äì Factor Structure Identified:** Early analyses by Richard **Emmons** found that the NPI‚Äôs many items group into a set of core factors. Emmons (1984, 1987) identified a **four-factor structure** of narcissism: *Leadership/Authority*, *Self-Absorption/Self-Admiration*, *Superiority/Arrogance*, and *Exploitativeness/Entitlement* ([Narcissism: theory and measurement - PubMed](https://pubmed.ncbi.nlm.nih.gov/3820065/#:~:text=Lack%20of%20a%20suitable%20measuring,1982%29%20theory%20of%20self)). Notably, the *Exploitativeness/Entitlement* facet was linked to more pathological narcissism, foreshadowing later distinctions between ‚Äúadaptive‚Äù and ‚Äúmaladaptive‚Äù components of narcissism ([Narcissism: theory and measurement - PubMed](https://pubmed.ncbi.nlm.nih.gov/3820065/#:~:text=%28NPI%29,of%20pathological%20narcissism%20and%20affective)).  
- **1988 ‚Äì Refinement to 40 Items:** Raskin & Terry‚Äôs influential study refined the NPI to a **40-item version** (often referred to as the **NPI-40**). Using principal components analysis on over 1,000 participants, they found evidence of a general narcissism factor and **seven distinct subcomponents** ([](https://www.columbia.edu/~da358/npi16/raskin.pdf#:~:text=and%20found%20evidence%20for%20a,descriptions%2C%20and%20their)). They dropped 14 of the original items to improve clarity and homogeneity, yielding the more efficient NPI-40. This 1988 paper also reported those **7 first-order factors** (see ‚ÄúStructure and Factors‚Äù below) and provided evidence for the NPI‚Äôs construct validity ([](https://www.columbia.edu/~da358/npi16/raskin.pdf#:~:text=and%20found%20evidence%20for%20a,descriptions%2C%20and%20their)).  
- **1990s ‚Äì Research Boom:** The NPI became a widely used tool in personality and social psychology. For example, **Bushman & Baumeister (1998)** used the NPI to show that narcissistic individuals react more **aggressively** when their ego is threatened ([The Role of Narcissism in Aggression and Violence: A Systematic Review - Sinead Lambe, Catherine Hamilton-Giachritsis, Emily Garner, Julian Walker, 2018 ](https://journals.sagepub.com/doi/10.1177/1524838016650190#:~:text=Bushman%20B,and%20Social%20Psychology%2C%2075%2C%20219%E2%80%93229)). Such studies cemented the NPI‚Äôs reputation and sparked broader interest in narcissism‚Äôs role in outcomes like aggression, relationship behaviors, and leadership.  
- **2005 ‚Äì Reliability Findings:** By mid-2000s, researchers examined the NPI‚Äôs consistency. Del Rosario & White (2005) found the NPI has good test‚Äìretest stability and internal consistency over time ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=S2CID%C2%A011220300.%209.%20,8869)), supporting it as a reliable trait measure. However, debates were growing about what exactly the NPI was measuring (e.g. narcissism vs. self-confidence).  
- **2006 ‚Äì Short Form (NPI-16):** To facilitate easier use in surveys, **Ames, Rose, & Anderson (2006)** developed the **NPI-16**, a brief 16-item version of the inventory. This short form, also forced-choice, maintained **‚Äúmeaningful face, internal, discriminant, and predictive validity‚Äù** relative to the full NPI ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=grandiose%20%28%03%20%02%20,82)). The NPI-16 allowed researchers to measure narcissism with far fewer items while still capturing the essence of **grandiose narcissistic traits** ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=grandiose%20%28%03%20%02%20,82)).  
- **2008 ‚Äì ‚ÄúNarcissism Epidemic‚Äù Debate:** Using decades of college student data, Twenge et al. (2008) reported that average NPI scores had risen over time, suggesting a generational increase in narcissism. This work (and the 2010 book *The Narcissism Epidemic*) brought the NPI into public discourse, though it also ignited controversy over interpretation and methodology in measuring societal changes in narcissism.  
- **2009 ‚Äì Pathological Narcissism Inventory (PNI):** Recognizing that the NPI mainly taps *grandiose* narcissism, Pincus et al. introduced the **PNI** as a new 52-item measure spanning both **grandiose and vulnerable narcissistic dimensions**. The PNI was explicitly developed to address **‚Äúseveral limitations of the NPI, particularly the lack of content capturing vulnerable and pathological narcissism‚Äù** ([[PDF] The Unified Narcissism Scale: Moving Towards an Integrated ...](https://ps.psychopen.eu/index.php/ps/article/download/7417/7417.pdf#:~:text=,Pincus%20et%20al)). This marked a shift toward assessing narcissism‚Äôs more clinical and maladaptive features that the NPI tended to omit.  
- **2011 ‚Äì Factor Structure Revisited:** Ackerman et al. (2011) published *‚ÄúWhat Does the NPI Really Measure?‚Äù* in *Assessment*, clarifying the NPI‚Äôs structure and correlates. Across four studies, they supported a **three-factor model** for the NPI: *Leadership/Authority*, *Grandiose Exhibitionism*, and *Entitlement/Exploitativeness* ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=researchers%20proposing%20solutions%20ranging%20from,NPI%20should%20examine%20correlates%20at)). They noted that the **Leadership/Authority** aspect of NPI was relatively adaptive (associated with high self-confidence and leadership ability), whereas **Exhibitionism** and especially **Entitlement/Exploitativeness** corresponded to more negative, maladaptive outcomes ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=its%20nomological%20network,propose%20a%20hierarchical%20model%20for)). Ackerman‚Äôs work provided a hierarchical scoring scheme (a total score with three facet scores) to help researchers interpret NPI results more precisely ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=research%20aimed%20to%20clarify%20the,propose%20a%20hierarchical%20model%20for)).  
- **2012 ‚Äì Five-Factor Narcissism Inventory (FFNI):** In an effort to integrate narcissism with general personality theory, Glover, Miller et al. (2012) introduced the **FFNI**. This 148-item inventory measures 15 lower-order traits linked to narcissism, mapped onto the **Five-Factor Model** of personality ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)). The FFNI explicitly captures both grandiose and vulnerable narcissistic features (see comparison section). Its development was part of a broader movement to refine narcissism assessment using trait frameworks and to improve content coverage beyond what the NPI offers ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)).  
- **2014 ‚Äì Single-Item Narcissism Scale:** Sara Konrath and colleagues demonstrated an extreme brief measure: the **Single Item Narcissism Scale (SINS)**. This simply asks respondents to rate themselves on the statement *‚ÄúI am a narcissist.‚Äù* Surprisingly, the one-item SINS showed a significant correlation with the NPI and solid psychometric properties in large samples ([Development and Validation of the Single Item Narcissism Scale (SINS) | PLOS One](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0103469#:~:text=Results)). The authors recommended SINS only in special circumstances (e.g. severe time constraints), but its success underscored the robustness of the narcissism construct ‚Äì even a single self-admission of narcissism was revealing.  
- **2010s ‚Äì Continued Applications:** The NPI remains one of the most commonly used measures of narcissism in research. It has been applied in organizational settings (e.g. linking CEO narcissism to company outcomes), in social media studies, and in the study of the ‚ÄúDark Triad‚Äù traits. Meanwhile, ongoing research has addressed its limitations, spawned alternative measures (e.g. Narcissistic Admiration and Rivalry Questionnaire in 2013), and refined our understanding of narcissism‚Äôs multifaceted nature.

## Structure and Factors of the NPI  
**Length and Format:** The NPI is typically used in its **40-item form** (Raskin & Terry, 1988). It is a forced-choice instrument: for each of the 40 pairs of statements, respondents must choose the one that best reflects them. In each pair, one option is a narcissistic statement and the other is non-narcissistic (or less narcissistic). *Example:* ‚ÄúI have a natural talent for influencing people‚Äù vs. ‚ÄúI am not good at influencing people.‚Äù The forced-choice format was intended to reduce acquiescence and social desirability biases by **avoiding obvious rating scales** ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=Superiority%2C%20Exhibitionism%2C%20and%20Exploitativeness.,example%20of%20one%20is%20below)). Each chosen narcissistic answer counts toward the total narcissism score. (Short versions like the NPI-16 follow the same format with a subset of items.)

**Original Subscales:** Analyses of the NPI consistently find that its items cluster into several components or **facets of narcissism**. The full NPI-40 can be scored as a single composite or broken into subscale scores. Raskin and Terry (1988) identified **seven first-order factors** in the NPI ([](https://www.columbia.edu/~da358/npi16/raskin.pdf#:~:text=and%20found%20evidence%20for%20a,descriptions%2C%20and%20their)):

- **Authority** ‚Äì preference for leadership and authority roles  
- **Self-Sufficiency** ‚Äì seeing oneself as autonomous and independent  
- **Superiority** ‚Äì believing one is superior to others  
- **Exhibitionism** ‚Äì desire to be the center of attention, flamboyance  
- **Exploitativeness** ‚Äì willingness to exploit or take advantage of others  
- **Vanity** ‚Äì preoccupation with one‚Äôs physical appearance and attractiveness  
- **Entitlement** ‚Äì expectation of special favors and submissiveness from others  

These seven dimensions emerged from factor-analytic studies and were **conceptually meaningful**, aligning with traits described in clinical theory ([](https://www.columbia.edu/~da358/npi16/raskin.pdf#:~:text=and%20found%20evidence%20for%20a,descriptions%2C%20and%20their)). In practice, researchers sometimes aggregate these into broader indices (e.g. an ‚Äúadaptive narcissism‚Äù composite vs. a ‚Äúmaladaptive narcissism‚Äù composite) or examine specific subscale scores (for instance, **Entitlement** often stands out as the most maladaptive facet). 

**Alternative Factor Models:** The NPI‚Äôs factor structure has been debated. Emmons‚Äôs early work (1987) collapsed the traits into **four factors**, naming them *Leadership/Authority*, *Self-Absorption/Self-Admiration*, *Superiority/Arrogance*, and *Exploitativeness/Entitlement*, as noted above ([Narcissism: theory and measurement - PubMed](https://pubmed.ncbi.nlm.nih.gov/3820065/#:~:text=Lack%20of%20a%20suitable%20measuring,1982%29%20theory%20of%20self)). Later researchers found even simpler structures: for example, some analyses yield a **two-factor solution** ‚Äì one factor capturing an ambitious, assertive dimension (combining authority, superiority, etc.) and another capturing a self-promoting, boastful dimension (exhibitionism and entitlement) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=reflect%20the%20central%20dogma%20of,1%20%5D%20Generally%2C%20variations)). In one study, these were labeled ‚ÄúPower/Leadership‚Äù vs. ‚ÄúExhibitionism/Entitlement‚Äù ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=research%20conducted%20by%20Emmons%2C%20four,1%20%5D%20Generally%2C%20variations)). Another team proposed **two factors** they termed *‚Äúpower‚Äù* and *‚Äúexhibitionism‚Äù* to summarize NPI content ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=research%20conducted%20by%20Emmons%2C%20four,1%20%5D%20Generally%2C%20variations)). 

Most recently, Ackerman et al. (2011) argued for a **hierarchical three-factor model** (with facets that correspond closely to Emmons‚Äô original four). Their three dimensions were **Leadership/Authority (LA)**, **Grandiose Exhibitionism (GE)**, and **Entitlement/Exploitativeness (EE)** ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=researchers%20proposing%20solutions%20ranging%20from,NPI%20should%20examine%20correlates%20at)). In this model, LA encompasses the more adaptive, self-confident aspects of narcissism, whereas EE represents the most toxic aspects (entitlement and exploitation), and GE covers vanity and exhibitionistic tendencies. Notably, Ackerman‚Äôs analyses showed **LA facet correlating with adaptive outcomes** (e.g. higher self-esteem, charisma) while **EE facet correlates with maladaptive outcomes** like aggression and low agreeableness ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=its%20nomological%20network,propose%20a%20hierarchical%20model%20for)). This reinforces the idea that not all components of the NPI are equally ‚Äúpathological.‚Äù Researchers are advised to look at NPI **facet-level relationships**, not just the total score ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=consisting%20of%20the%20dimensions%20of,propose%20a%20hierarchical%20model%20for)).

**Psychometric Properties:** Overall, the NPI has shown decent reliability (Cronbach‚Äôs Œ± often in the .80 range for the 40-item version) and stability over time ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=S2CID%C2%A011220300.%209.%20,8869)). Its **convergent validity** is solid with other narcissism measures of the *grandiose* type. For instance, NPI scores correlate positively with expert ratings of prototypical narcissistic personality disorder based on the five-factor model (high extraversion, low agreeableness, etc.) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=Rhodewalt%20%26%20Morf.,1)). The NPI also correlates as expected with behaviors like self-enhancement, vanity, and aggression. However, critics have noted **weak convergent validity with some other narcissism constructs** and certain NPI items that may not reflect core narcissism (e.g. *‚ÄúI see myself as a good leader‚Äù* might tap general confidence more than narcissism) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=The%20NPI%20has%20weak%20convergent,recently%2C%20research%20by%20Kubarych%2C%20Deary)). Internal consistency can be lower if one tries to use all seven subscales separately (some subscales have few items). In sum, the NPI is multidimensional, and its factor structure can shift slightly depending on sample and analysis method, which has led to ongoing refinement and occasional short-form versions to improve coherence.

## Theoretical Basis of the NPI  
The NPI was **grounded in clinical theory**, but designed for personality-trait research. It was directly **derived from the DSM-III definition of Narcissistic Personality Disorder** ([](http://www.sakkyndig.com/psykologi/artvit/emmons1987.pdf#:~:text=designed%20to%20measure%20individual%20differences,expectation%20of%20special%20favors%20with%02out)), meaning its content reflects the traits clinicians associated with narcissism: grandiose self-concept, fantasies of success, demand for admiration, arrogance, entitlement, and lack of empathy. By translating those diagnostic features into questionnaire items, Raskin and Hall effectively bridged clinical **psychoanalytic concepts of narcissism** (via DSM) into a measurable trait. Narcissism in DSM-III and later DSM-IV was heavily influenced by psychoanalytic and object-relations theories (e.g. Freud‚Äôs writings on narcissism, Kohut‚Äôs and Kernberg‚Äôs theories), emphasizing grandiosity and self-focus. The NPI‚Äôs theoretical basis is thus a **personality disorder prototype** that can be expressed on a continuum in normal populations. It captures what is often called **‚Äúgrandiose narcissism‚Äù**, characterized by high self-confidence, assertiveness, and a sense of superiority.

**Trait Model Alignment:** In contemporary personality theory, NPI-defined narcissism maps onto well-known trait dimensions. Studies have found that NPI scores show strong positive associations with **Extraversion** (especially the assertiveness, dominance aspect) and inversely with **Agreeableness** (reflecting antagonism, arrogance) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=indicating%20high%20reliability%20pertaining%20to,1)). For example, in expert ratings of personality profiles, the NPI correlates with a trait profile of high Extraversion, low Agreeableness, and some aspects of high *Psychoticism* (a factor in Eysenck‚Äôs model related to tough-mindedness and impulsivity) ([Narcissistic Personality Inventory - Wikipedia](https://en.wikipedia.org/wiki/Narcissistic_Personality_Inventory#:~:text=indicating%20high%20reliability%20pertaining%20to,1)). This suggests the NPI is tapping a combination of **agentic extraversion** (social boldness, leadership) and **antagonism** (egocentrism, lack of empathy), which aligns with many theoretical models of narcissism as a blend of **high agency** and **low communion**. Importantly, the NPI was never intended to capture the more self-effacing or emotionally insecure form of narcissism often termed ‚Äúvulnerable narcissism‚Äù ‚Äì its conceptual focus is on the **overt, grandiose form** ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=there%20was%20no%20expectation%20that,This%20expectation%20was%20confirmed)). Consistent with this, the NPI correlates only weakly or not at all with measures of narcissistic vulnerability (e.g. hypersensitive insecurity) and instead correlates with measures of dominance and self-enhancement ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=with%20the%20other%20measures%20of,Entitlement%2FExploitativeness%20subscale%20of%20the%20NPI)).

**Grandiose vs. Vulnerable Narcissism:** Theoretical work distinguishes two broad manifestations of narcissism ‚Äì **grandiose** (overt) and **vulnerable** (covert) ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=with%20the%20other%20measures%20of,Entitlement%2FExploitativeness%20subscale%20of%20the%20NPI)). Grandiose narcissism entails an overt sense of superiority, exhibitionism, and exploitative attitudes (often with high extraversion and low empathy). Vulnerable narcissism, by contrast, involves defensiveness, insecurity, introversion, and underlying shame or hypersensitivity (often accompanied by anxiety or depression). The NPI was theorized and validated primarily as a **measure of grandiose narcissism** in normal populations ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=there%20was%20no%20expectation%20that,This%20expectation%20was%20confirmed)). It aligns with theories that narcissism can be an extreme on normal trait dimensions ‚Äì for instance, **Paulhus‚Äôs ‚ÄúDark Triad‚Äù framework**, where narcissism (as measured by NPI) is one leg of a trio of socially aversive traits (alongside Machiavellianism and psychopathy), representing an interpersonally exploitative yet psychologically **high self-esteem** profile. The NPI‚Äôs focus means it captures the **socially bold, confident** narcissistic persona that fits the grandiose profile, and does not directly assess vulnerable dynamics like shame, fragility, or introverted self-absorbed fantasy.

At the same time, **not all NPI content is uniformly pathological**. Theoretical interpretations (supported by factor analyses) often separate **‚Äúadaptive‚Äù vs. ‚Äúmaladaptive‚Äù** narcissism within the NPI. For example, having high authority and self-confidence might be seen as adaptive or ‚Äúhealthy‚Äù narcissism (sometimes called **leadership/authority** or even ‚Äúhealthy self-esteem‚Äù aspects), whereas entitlement and exploitativeness are unequivocally maladaptive. Ackerman et al. (2011) explicitly showed that **Leadership/Authority facet relates to higher self-esteem and lower neuroticism (adaptive)**, while **Entitlement/Exploitativeness facet relates to aggression, low agreeableness, and other negative outcomes (maladaptive)** ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=its%20nomological%20network,propose%20a%20hierarchical%20model%20for)). This aligns with theories that a certain **degree of narcissism can be functional** in leadership or competitive environments (confidence, charisma), but the **toxic elements** of narcissism (entitlement, lack of empathy) are what lead to interpersonal problems. In short, the NPI is grounded in a **trait theory of narcissism as a continuous distribution** in the population, reflecting the grandiose features described by clinical theory. It has been invaluable in testing theories about narcissism‚Äôs role in self-regulation (e.g. the **Morf & Rhodewalt (2001)** social-cognitive model of narcissism uses NPI to show how narcissists alternate between self-aggrandizement and self-protection strategies) and in interpersonal outcomes (e.g. how narcissists handle feedback, form relationships, etc.). The NPI‚Äôs widespread use has helped bridge clinical concepts of narcissism with empirical personality psychology.

## Comparison of NPI with the Five-Factor Narcissism Inventory (FFNI)  
The **Five-Factor Narcissism Inventory (FFNI)** is a newer assessment instrument (developed by Glover, Miller, Lynam, et al., 2012) that provides a more nuanced, trait-based profile of narcissism. There are several key differences between the traditional NPI and the FFNI in terms of **measurement approach** and **conceptualization of narcissism**:

- **Theoretical Framework:** The NPI was built top-down from clinical criteria (DSM NPD), whereas the FFNI was built bottom-up from general personality traits. The FFNI explicitly uses the **Five-Factor Model (FFM)** as its backbone ‚Äì it assesses narcissism through 15 trait scales, each representing a maladaptive extreme of an FFM facet relevant to narcissism ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,93%2C%20respectively)). For example, FFNI ‚ÄúEntitlement‚Äù corresponds to an extreme of (low) Agreeableness, ‚ÄúExhibitionism‚Äù corresponds to extreme Extraversion, ‚ÄúReactive Anger‚Äù corresponds to high Neuroticism, etc. In contrast, the NPI does not map neatly onto the FFM facets in its design (though post-hoc one can interpret NPI results in FFM terms). The FFNI‚Äôs conceptualization sees narcissism as a **constellation of trait elevations** ‚Äì primarily high antagonism (low agreeableness), high agentic extraversion, and in some cases high neuroticism (for vulnerable aspects) ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,of%20an%20FFM%20trait%20found)). The NPI, on the other hand, treats narcissism more as a **unitary construct with several correlated facets** of grandiosity. 

- **Content Coverage ‚Äì Grandiose vs. Vulnerable:** A major conceptual difference is that **FFNI measures both grandiose and vulnerable narcissism**, while the NPI largely measures grandiose aspects. The FFNI includes specific subscales for **vulnerable narcissistic traits** (e.g., *‚ÄúShame‚Äù*, *‚ÄúReactive Anger‚Äù*, *‚ÄúNeed for Admiration‚Äù*, *‚ÄúDistrust‚Äù*) that capture the insecure, hypersensitive side of narcissism ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Vulnerable%20Narcissism%20%3D%20sum%20of,Manipulativeness%2C%20Exploitativeness%2C%20Entitlement%2C%20Lack%20of)). It also has subscales for **grandiose traits** (e.g., *‚ÄúAuthoritativeness‚Äù*, *‚ÄúExhibitionism‚Äù*, *‚ÄúManipulativeness‚Äù*, *‚ÄúEntitlement‚Äù*, etc.) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Vulnerable%20Narcissism%20%3D%20sum%20of,Manipulativeness%2C%20Exploitativeness%2C%20Entitlement%2C%20Lack%20of)). This dual focus means the FFNI can yield **two higher-order scores** ‚Äì one for **Grandiose Narcissism** and one for **Vulnerable Narcissism** ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Vulnerable%20Narcissism%20%3D%20sum%20of,Manipulativeness%2C%20Exploitativeness%2C%20Entitlement%2C%20Lack%20of)) ‚Äì in addition to a total score. By contrast, the NPI was intentionally not designed to capture vulnerability (e.g., it has no items about feeling anxious, inadequate, or depressed). It primarily assesses **overt narcissistic attitudes** like superiority and self-importance. In fact, research shows that NPI scores correlate strongly with the FFNI‚Äôs **Grandiose dimension**, but minimally with the FFNI‚Äôs **Vulnerable dimension** ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=with%20the%20other%20measures%20of,Entitlement%2FExploitativeness%20subscale%20of%20the%20NPI)). Specifically, the FFNI grandiose narcissism composite correlates with **all facets of the NPI**, whereas the FFNI vulnerable narcissism composite correlates significantly **only with the NPI‚Äôs Entitlement/Exploitativeness facet** and not with NPI‚Äôs leadership or exhibitionism content ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=with%20the%20other%20measures%20of,Entitlement%2FExploitativeness%20subscale%20of%20the%20NPI)). This indicates the NPI captures mostly the grandiose side (with entitlement being the one facet that bridges to vulnerability a bit), whereas the FFNI captures a much broader spectrum of narcissistic pathology.

- **Measurement Format:** The NPI uses a **forced-choice format** (binary choices for each item pair), which is somewhat unusual compared to typical personality inventories. The FFNI, however, uses a **standard Likert rating scale** (participants rate their agreement with each statement on a scale). The FFNI has 148 items (usually rated 1‚Äì5 or 1‚Äì4 from disagree to agree), making it a fairly lengthy inventory. The NPI‚Äôs forced-choice format can make it **trickier to score and analyze subcomponents**, and it might mask some information (since it doesn‚Äôt allow intensity ratings ‚Äì you either endorse the narcissistic statement or not). The FFNI‚Äôs Likert format provides more granular data and is in line with most modern personality measures. Because of its length and detail, the FFNI provides **facet-level scores** for 15 traits (e.g. separate scores for **Manipulativeness, Grandiose Fantasies, Arrogance, Indifference**, etc., on top of the higher-order grandiose/vulnerable scores) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Need%20for%20Admiration%3D%204%20%2B,R)) ([](https://psychology.uga.edu/sites/default/files/CVs/FFNI_Long.pdf#:~:text=Total%20Score%20%3D%20sum%20of,Acclaim%20Seeking%2C%20and%20Thrill%20Seeking)). The NPI typically provides either a total score or a few factor scores at most. Thus, **FFNI offers a more fine-grained profile** of an individual‚Äôs narcissistic tendencies, whereas NPI offers a broad stroke measure of narcissism.

- **Adaptive vs. Maladaptive Elements:** The NPI, as noted, includes some content that could be seen as relatively **adaptive (or at least socially desirable)** ‚Äì for example, items about leadership ability or confidence. The FFNI‚Äôs approach tends to frame each trait in its **maladaptive extreme** (it was designed to focus on narcissistic personality *disorder* traits). For instance, FFNI *Authoritativeness* corresponds to a confident leadership tendency, but phrased in a way that reflects overconfidence or dictatorial inclination (maladaptive end of **Extraversion**), and FFNI *Attention Seeking* (akin to Exhibitionism) reflects extreme extraverted behavior craving admiration. The **NPI‚Äôs Leadership/Authority items can sound positive or neutral** (e.g. ‚ÄúI am a born leader‚Äù); indeed, Ackerman et al. (2011) found that this NPI facet was linked with high self-esteem and not with the aggression or interpersonal problems associated with narcissism ([What does the narcissistic personality inventory really measure? - PubMed](https://pubmed.ncbi.nlm.nih.gov/20876550/#:~:text=its%20nomological%20network,propose%20a%20hierarchical%20model%20for)). The FFNI, by design, **emphasizes the pathological aspects** even of leadership (its ‚ÄúAuthoritativeness‚Äù facet would capture a *domineering* leadership style, for example). Thus, people with certain narcissistic traits might score somewhat differently on NPI vs FFNI. A person who is high in narcissistic leadership and exhibitionism but low in entitlement might get a high NPI score, but on the FFNI their profile would show high scores on some grandiose facets and low on others (and perhaps low overall pathological narcissism if entitlement and antagonism facets are low). In contrast, someone high in narcissistic vulnerability (e.g. shame, contingent self-esteem) would show up clearly on the FFNI as high vulnerable narcissism, but might not score high on the NPI at all (because they wouldn‚Äôt endorse the overtly grandiose statements). In short, **the NPI skews toward measuring the confident, outgoing narcissist**, whereas the FFNI can identify both that *and* the defensive, internally troubled narcissist.

- **Use and Research Context:** The NPI has been a **popular tool in social/personality psychology** for decades ‚Äì its brevity and focus on subclinical narcissism made it convenient for large samples (even the 40-item version is moderate in length, and the 16-item version is very quick). It‚Äôs often used in studies on college populations, leadership research, relationship studies, etc. The FFNI, being longer and more detailed, has seen most use in **clinical and forensic research or specialized personality research**. For example, studies comparing narcissistic personality disorder patients vs. controls might prefer the FFNI (to capture subtle pathology), whereas a social psychology study on narcissism and selfie-posting might stick with the NPI for efficiency. The FFNI‚Äôs development by Miller and colleagues was partly motivated by findings that **different narcissism measures capture different aspects** ‚Äì they wanted a **comprehensive measure** that integrates those aspects. Indeed, when comparing NPI with other measures, researchers found NPI was **strongly related to FFNI‚Äôs grandiose (antagonism/extraversion-based) components**, but **missing the neurotic, vulnerable components** present in instruments like the PNI or FFNI ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=with%20the%20other%20measures%20of,Entitlement%2FExploitativeness%20subscale%20of%20the%20NPI)).

**Summary of NPI vs FFNI:** The NPI is **shorter, broader, and focused on overt narcissism**, originating from clinical criteria but used as a personality trait measure. The FFNI is **longer, in-depth, and grounded in the five-factor model**, explicitly capturing both flamboyant and fragile sides of narcissism. NPI results in a single narcissism score (with optional facet scores) and is best at identifying **grandiose narcissistic tendencies**. FFNI results in a rich profile of scores, distinguishing **narcissistic grandiosity and vulnerability** and various sub-facets, giving a more diagnostic picture. Both have their uses: NPI is well-suited for quick assessment of ‚Äúhow narcissistic‚Äù someone is in general terms ([HJPA_A_670680_O](https://people.uncw.edu/hakanr/documents/narcissism.pdf#:~:text=there%20was%20no%20expectation%20that,This%20expectation%20was%20confirmed)), while FFNI is suited for understanding the **nuanced makeup of someone‚Äôs narcissism** ‚Äì e.g., are they entitled and exploitative? Or predominantly attention-seeking and boastful? Do they also secretly feel inadequate? In essence, the **FFNI broadens the conceptualization** of narcissism by anchoring it in fundamental personality traits ([](https://psychology.uga.edu/sites/default/files/CVs/Miller%20et%20al._2014b_Psychological%20Assessment.pdf#:~:text=Five,93%2C%20respectively)), whereas the NPI represents a more **classical social/personality approach** to narcissism as a single (grandiose) trait with a mix of egoistic facets.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

# **Step 4: Create NPI Embeddings** <a class="anchor" id="npi_page_3"></a>

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved npi dataset from a CSV file
dataset_path = '../Datasets/npi.csv'
try:
    npi_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'npi_df' is empty before processing
        if npi_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", npi_df.shape[0])
            print(f"üìä Processing {npi_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        npi_df['Embedding'] = npi_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/npi_embeddings.csv'
        npi_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = npi_df.shape[0]
        num_columns = npi_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in npi_df.columns and not npi_df['Embedding'].empty:
            embedding_length = len(npi_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(npi_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/npi_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="NPI Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/npi_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "npi_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/npi_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of NPI Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/npi_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of NPI Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create NPI Schema in Neo4j** <a class="anchor" id="NPI_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("NPI")

# Load the dataset
dataset_path = "../Embeddings/npi_embeddings.csv"
npi_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in npi_df.columns:
    npi_df['Embedding'] = npi_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the NPI schema in Neo4j
def create_npi_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with NPI labels
        graph.run("MERGE (f:NPI_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:NPI_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:NPI_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:NPI_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:NPI_Noun {name: $noun})", noun=noun)
        
        # Create relationships with NPI namespacing
        graph.run("""
            MATCH (f:NPI_Factor {name: $factor}), (a:NPI_Adjective {name: $adjective})
            MERGE (f)-[:NPI_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:NPI_Adjective {name: $adjective}), (s:NPI_Synonym {name: $synonym})
            MERGE (a)-[:NPI_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:NPI_Adjective {name: $adjective}), (v:NPI_Verb {name: $verb})
            MERGE (a)-[:NPI_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:NPI_Adjective {name: $adjective}), (n:NPI_Noun {name: $noun})
            MERGE (a)-[:NPI_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:NPI_Factor {name: $factor}), (a:NPI_Adjective {name: $adjective})
                MERGE (e:NPI_Embedding {value: $embedding})
                MERGE (f)-[:NPI_HAS_EMBEDDING]->(e)
                MERGE (a)-[:NPI_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create NPI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create NPI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the NPI schema
        create_npi_schema(graph, npi_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", npi_df.shape[0])
        mlflow.log_param("Number of Columns", npi_df.shape[1])
        
        # Print the DataFrame
        print(npi_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
