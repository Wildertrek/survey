# Five-Factor Narcissism Inventory ‚Äì Short Form (FFNI-SF)

## Development History and Structure  
The **Five-Factor Narcissism Inventory (FFNI)** was originally developed by Glover, Miller, Lynam, Crego, & Widiger in 2012 as a comprehensive measure of narcissistic traits grounded in the Five-Factor Model (FFM) of personality ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=The%20Five,Miller%2C%20Lynam%2C%20Crego%2C%20%26%20Widiger)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=%28FFM%29,PDs)). The full FFNI contains 148 items assessing 15 trait facets that span **grandiose** and **vulnerable** dimensions of narcissism, as well as Narcissistic Personality Disorder (NPD), from a five-factor perspective ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=The%20Five,Miller%2C%20Lynam%2C%20Crego%2C%20%26%20Widiger)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). The FFNI was designed based on expert surveys and literature reviews to capture maladaptive extremes of FFM traits relevant to narcissism ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=the%20creation%20of%20the%20FFNI%2C,empirical%20reviews%20were%20examined%20to)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=3)). For example, facets were drawn from across all Big Five domains (e.g. low Agreeableness: arrogance, exploitativeness; high Extraversion: exhibitionism, assertiveness; high Neuroticism: shame, reactive anger) that correspond to narcissistic behavior patterns ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=3)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Upon%20review%20of%20these%20sources,15%20facets%20were%20necessary%20to)). Ultimately, 15 trait facets were included to represent the most common grandiose and vulnerable narcissistic presentations within the FFM framework ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Upon%20review%20of%20these%20sources,15%20facets%20were%20necessary%20to)). 

The **FFNI‚ÄìShort Form (FFNI-SF)** was later developed by Sherman, Miller, Few, Campbell, Widiger, Crego, & Lynam (published 2015) to provide a briefer assessment while retaining the original‚Äôs content and structure ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). The FFNI-SF contains 60 items (4 per facet) covering the same 15 trait subscales as the full FFNI ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). Item selection for the short form was based on psychometric analyses to ensure each facet‚Äôs reliability and the overall inventory‚Äôs validity remain high. Indeed, Sherman et al. (2015) reported a very high correspondence between the FFNI-SF and the full FFNI (intraclass correlations indicating strong compatibility of scores) ([](https://files.eric.ed.gov/fulltext/EJ1115015.pdf#:~:text=et%20al,SF%E2%80%99s%20original%20form%20explained)) ([](https://files.eric.ed.gov/fulltext/EJ1115015.pdf#:~:text=59,Data%20from)). The FFNI-SF maintains the ability to score **two higher-order dimensions** ‚Äì **Grandiose Narcissism** and **Vulnerable Narcissism** ‚Äì as in the original FFNI ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). It also reproduces an empirically derived **three-factor structure** (antagonism, neuroticism, and agentic extraversion) mapping onto FFM traits ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=two%20rationally%20created%20dimensions%20commonly,narcissism%20literature%20%E2%80%93%20grandiose%20and)). In sum, the FFNI-SF (2015) was developed to mirror the FFNI (2012) in content and theoretical basis, but with fewer items for practical efficiency, without sacrificing the nuanced trait coverage of narcissistic personality features.

## Comparisons with Other Narcissism Models  
**FFNI-SF vs. Full FFNI:** The FFNI-SF is essentially a condensed version of the full FFNI. It retains the same 15 narcissism-related facets and two primary dimensions (grandiose and vulnerable) as the original measure ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). The main difference is length: 60 items in the short form versus 148 in the full FFNI. Psychometric evaluations show the FFNI-SF has very similar trait profiles and validity as the full FFNI, with only minimal loss of information ([](https://files.eric.ed.gov/fulltext/EJ1115015.pdf#:~:text=et%20al,SF%E2%80%99s%20original%20form%20explained)). Thus, conceptually the FFNI-SF does not differ from the full FFNI, but it offers a more efficient tool while preserving the FFNI‚Äôs five-factor trait approach to narcissism.

**FFNI-SF vs. Narcissistic Personality Inventory (NPI):** The NPI (Raskin & Hall, 1979; 40 items) is a classic measure of **grandiose narcissism**, focusing on traits like vanity, entitlement, and leadership/authority. Unlike the FFNI-SF, the NPI was not built from the five-factor trait model and largely omits **vulnerable** aspects of narcissism (e.g. insecurity or shame). In fact, research shows the NPI aligns well with grandiose narcissism but fails to capture narcissistic vulnerability ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expected%20by%20experts%20for%20measures,toward%20the%20selection%20of%20narcissism)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Personality%20Inventory%20,These)). The FFNI-SF, by contrast, explicitly includes vulnerable facets (e.g. shame, reactive anger) in addition to grandiose traits, giving it a broader conceptualization. The NPI uses a forced-choice format emphasizing socially bold and self-enhancing tendencies, whereas the FFNI-SF uses a Likert format and frames narcissism as maladaptive extremes of normal traits. In sum, the NPI views narcissism in terms of overt grandiosity/vanity, while the FFNI-SF conceptualizes it as a combination of antagonism, neuroticism, and extraversion-related traits (capturing both the ‚Äúbright‚Äù and ‚Äúdark‚Äù sides).

**FFNI-SF vs. Pathological Narcissism Inventory (PNI):** The PNI (Pincus et al., 2009) is a 52-item measure designed to assess pathological narcissism with **two broad factors**: Narcissistic Grandiosity and Narcissistic Vulnerability. Both the PNI and FFNI-SF recognize narcissism as a heterogeneous construct with grandiose and vulnerable manifestations. The key difference is in approach: the PNI‚Äôs content is derived from clinical theory (e.g. facets like contingent self-esteem, hiding the self, entitlement rage), focusing on self-esteem regulation failures and distress. The FFNI-SF, on the other hand, ties narcissistic behaviors to general personality traits (e.g. lack of empathy, thrill-seeking) within the FFM. Empirically, the PNI‚Äôs vulnerable scales (e.g. shame, need for others‚Äô approval) overlap with the FFNI-SF‚Äôs vulnerable facet scores, and both have been shown to capture the core of narcissistic vulnerability well ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Personality%20Inventory%20,These)). However, the PNI‚Äôs grandiosity scales (which include aspects like exploitativeness and grandiose fantasy) are somewhat less correlated with expert ratings of grandiose narcissism compared to the FFNI‚Äôs grandiose scale ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expected%20by%20experts%20for%20measures,toward%20the%20selection%20of%20narcissism)). In short, both inventories measure grandiose and vulnerable narcissism, but the FFNI-SF does so through the lens of fundamental personality traits, whereas the PNI emphasizes the psychopathological processes of narcissism (e.g. dysregulated self-esteem, fantasy vs. devaluation cycles).

**FFNI-SF vs. Narcissistic Admiration and Rivalry Questionnaire (NARQ):** The NARQ (Back et al., 2013) is an 18-item measure based on the **two-dimensional ‚ÄúAdmiration‚ÄìRivalry‚Äù model** of grandiose narcissism ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=for%20inventories%20that%20can%20be,Back%20et%20al)) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=and%20self,S%20as%20a%20short%20and)). It divides grandiose narcissism into **Admiration** (self-promoting, assertive, charm-driven aspects) and **Rivalry** (self-protective, antagonistic, other-derogating aspects). The FFNI-SF and NARQ share the idea that narcissism has an agentic (admiration/extraversion) component and an antagonistic component. In fact, the NARQ‚Äôs admiration scale corresponds to the assertive, status-seeking side of grandiosity, while its rivalry scale corresponds to hostile and arrogant tendencies ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=for%20inventories%20that%20can%20be,Back%20et%20al)) ([Validation of the Narcissistic Admiration and Rivalry Questionnaire short scale (NARQ-S) in convenience and representative samples](https://www.diw.de/documents/publikationen/73/diw_01.c.550208.de/diw_sp0884.pdf#:~:text=respectively,scale%2C%20representative%20sample%2C%20validation%2C%20personality)). The difference is that NARQ is narrowly focused on **grandiose narcissism** only ‚Äì it does not measure vulnerable feelings like shame or insecurity. The FFNI-SF provides a more granular breakdown (15 facets) and includes those vulnerable facets. Conceptually, NARQ treats narcissism as a **self-regulatory strategy** (boosting oneself versus derogating others), whereas the FFNI-SF treats it as a constellation of trait dispositions (e.g. callousness, need for admiration, etc.). Both approaches find that narcissistic antagonism overlaps with disagreeableness and hostility, but FFNI-SF extends to measuring internal distress (via neuroticism-related facets) which NARQ omits. Thus, FFNI-SF offers a broader conceptualization (grandiose + vulnerable), whereas NARQ hones in on the **‚Äúbright‚Äù vs ‚Äúdark‚Äù sides** of grandiose narcissism.

**FFNI-SF vs. Hypersensitive Narcissism Scale (FFNI-SF):** The FFNI-SF (Hendin & Cheek, 1997) is a brief 10-item measure of **covert or vulnerable narcissism**, often termed ‚Äúhypersensitive‚Äù narcissism. It was derived to capture narcissistic traits like insecurity, defensiveness, and sensitivity to criticism (in contrast to the overt narcissism measured by instruments like the NPI). The FFNI-SF‚Äôs vulnerable narcissism dimension overlaps considerably with what the FFNI-SF measures ‚Äì traits such as distrust, shame, and reactive anger stemming from wounded self-esteem. Indeed, studies show the FFNI-SF correlates strongly with expert ratings of vulnerable narcissism ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Personality%20Inventory%20,These)), similar to the FFNI‚Äôs vulnerable scale. The difference is scope: the FFNI-SF is narrowly focused on the shy, sensitive aspect of narcissism (e.g. feeling misunderstood or unappreciated), whereas the FFNI-SF covers that domain *and* additionally covers grandiose narcissism. In practical terms, FFNI-SF offers a quick gauge of narcissistic vulnerability (covert narcissism), while FFNI-SF provides a fuller profile including entitlement, exploitativeness, and admiration-seeking. The FFNI-SF conceptualizes narcissism as including hypersensitivity **and** grandiosity within one framework, whereas FFNI-SF treats hypersensitivity as a singular narcissistic trait dimension. 

## FFNI-SF Structure and Descriptive Components  
**Factors and Subscales:** The FFNI-SF assesses **15 narcissistic traits** (subscales), each corresponding to maladaptive extremes of FFM facets. These can be grouped into two higher-order factors: **Grandiose narcissism** (11 subscales) and **Vulnerable narcissism** (4 subscales) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=expressions%20of%20the%20FFM%20facets,The%20full%20FFN%20I)). Below is the hierarchical structure of the FFNI-SF, with all facet-level scales:

- **Grandiose Narcissism** ‚Äì composed of 11 trait facets related to arrogant, exploitative, and self-enhancing tendencies ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Exploitativeness%20,or%20take%20advantages%20of%20others)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Grandiose%20Fantasies%20,of%20glory%2C%20success%2C%20and%20status)). These **subscales** are: *Acclaim-Seeking*, *Arrogance*, *Authoritativeness*, *Entitlement*, *Exhibitionism*, *Exploitativeness*, *Indifference*, *Lack of Empathy*, *Manipulativeness*, *Grandiose Fantasies*, and *Thrill-Seeking*. Each facet reflects a distinct flavor of grandiose narcissism (e.g. Entitlement = expectation of special treatment ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Lack%20of%20Empathy%20,of%20the%20feelings%20of%20others)); Lack of Empathy = inability to appreciate others‚Äô feelings ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Exploitativeness%20,or%20take%20advantages%20of%20others))).  
  - **Descriptive adjectives (with synonyms, verbs, and nouns):** To characterize the grandiose narcissism factor, one can use traits such as:  
    - **Arrogant** ‚Äì (synonyms: *haughty*, *conceited*; verbs: *brag*, *condescend*; nouns: *arrogance*, *egotism*)  
    - **Entitled** ‚Äì (synonyms: *privileged*, *self-important*; verbs: *demand*, *insist*; nouns: *entitlement*, *privilege*)  
    - **Manipulative** ‚Äì (synonyms: *scheming*, *calculating*; verbs: *exploit*, *deceive*; nouns: *manipulation*, *exploitation*)  
    - **Exhibitionistic** ‚Äì (synonyms: *showy*, *flamboyant*; verbs: *show off*, *flaunt*; nouns: *exhibitionism*, *showmanship*)  
    - **Dominant** ‚Äì (synonyms: *overbearing*, *controlling*; verbs: *dictate*, *boss* (around); nouns: *dominance*, *authority*)  
    - **Callous** ‚Äì (synonyms: *cold-hearted*, *unempathetic*; verbs: *belittle*, *disregard*; nouns: *callousness*, *apathy*)  

- **Vulnerable Narcissism** ‚Äì composed of 4 trait facets reflecting defensiveness, emotional volatility, and self-esteem fragility ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Reactive%20Anger%20,response%20to%20criticism%20or%20rebuke)) ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Indifference%20,response%20to%20criticism%20or%20rebuke)). These **subscales** are: *Reactive Anger*, *Shame*, *Need for Admiration*, and *Distrust*. They represent the inward-facing, self-critical side of narcissism (e.g. Shame = sensitivity to criticism leading to feelings of inferiority ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Reactive%20Anger%20,response%20to%20criticism%20or%20rebuke)); Reactive Anger = hostile outbursts in response to ego threats ([(PDF) Five-Factor Narcissism Inventory](https://www.researchgate.net/publication/318605521_Five-Factor_Narcissism_Inventory#:~:text=Reactive%20Anger%20,response%20to%20criticism%20or%20rebuke))).  
  - **Descriptive adjectives (with synonyms, verbs, and nouns):** Words that describe the vulnerable narcissism factor include:  
    - **Insecure** ‚Äì (synonyms: *self-doubting*, *unconfident*; verbs: *worry*, *second-guess*; nouns: *insecurity*, *self-doubt*)  
    - **Oversensitive** ‚Äì (synonyms: *touchy*, *thin-skinned*; verbs: *take offense*, *overreact*; nouns: *sensitivity*, *defensiveness*)  
    - **Defensive** ‚Äì (synonyms: *guarded*, *self-protective*; verbs: *justify* (oneself), *deny* (fault); nouns: *defensiveness*, *self-protection*)  
    - **Ashamed** ‚Äì (synonyms: *humiliated*, *mortified*; verbs: *withdraw*, *hide* (oneself); nouns: *shame*, *humiliation*)  
    - **Distrustful** ‚Äì (synonyms: *suspicious*, *wary*; verbs: *doubt*, *suspect*; nouns: *distrust*, *suspicion*)  
    - **Resentful** ‚Äì (synonyms: *bitter*, *grudging*; verbs: *begrudge*, *blame*; nouns: *resentment*, *bitterness*)  

Each of these adjectives (and their related words) captures an aspect of the FFNI-SF‚Äôs factors. **Grandiose narcissism** adjectives emphasize a proud, egocentric, and domineering style, whereas **vulnerable narcissism** adjectives emphasize an anxious, defensive, and reactive sensitivity. This reflects the FFNI-SF‚Äôs nuanced approach: it measures narcissism not as a single trait, but as a profile of traits ranging from brazen self-importance to brittle self-esteem. The structured breakdown above aligns the FFNI-SF with comparable datasets for instruments like the NPI, PNI, and NARQ, detailing the content domains (factors and subscales) and providing descriptive terminology for each facet of narcissism it assesses.

The **FFNI Short Form (FFNI-SF)** is essentially a condensed, streamlined version of the full **Five-Factor Narcissism Inventory (FFNI)**. It retains the core structure of the original FFNI but reduces its length from **148 items** down to **60 items**, making it quicker and more practical for research and clinical use while maintaining psychometric robustness.

### Comparison by Factors:

Both versions retain identical **two-factor structures**:

- **Grandiose Narcissism**
  - Authority
  - Exhibitionism
  - Exploitativeness
  - Grandiose Fantasy
  - Self-Sufficiency
  - Superiority
  - Reactive Anger
  - Need for Admiration
  - Self-Absorption
  - Overconfidence
  - Assertiveness

- **Vulnerable Narcissism**
  - Entitlement
  - Contingent Self-Esteem
  - Hiding the Self
  - Devaluing
  - Shame or Insecurity

**However**, the **FFNI-SF** reduces the number of items per facet, preserving the overall factor structure but potentially sacrificing some granularity.

---

### Practical Considerations:

- **FFNI (Full)**
  - Provides a **detailed personality profile**, suitable for deep, individualized clinical or psychological research.
  - Can assess nuanced differences between individuals within facets.

- **FFNI-SF**
  - Provides a **faster assessment** suited for large-scale studies, screenings, or when time is limited.
  - Slightly reduces detail within each facet, but maintains high validity and reliability.

---

### Recommendation:

- **Use the full FFNI** for clinical diagnostics, in-depth personality profiling, and comprehensive research where nuanced differences are important.
- **Use the FFNI-SF** for quick screening, large-sample surveys, and research that prioritizes efficiency without sacrificing substantial validity.

2.  **Step 2: API Key Handling and Initialization**

    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved ffni_sf dataset from a CSV file
dataset_path = '../Datasets/ffni_sf.csv'
try:
    ffni_sf_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'ffni_sf_df' is empty before processing
        if ffni_sf_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", ffni_sf_df.shape[0])
            print(f"üìä Processing {ffni_sf_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        ffni_sf_df['Embedding'] = ffni_sf_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/ffni_sf_embeddings.csv'
        ffni_sf_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = ffni_sf_df.shape[0]
        num_columns = ffni_sf_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in ffni_sf_df.columns and not ffni_sf_df['Embedding'].empty:
            embedding_length = len(ffni_sf_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(ffni_sf_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 13: Create FFNI-SF Schema in Neo4j** <a class="anchor" id="FFNI-SF_page_4"></a>


[Back to Top](#FFNI-SF_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("FFNI-SF")

# Load the dataset
dataset_path = "../Embeddings/ffni_sf_embeddings.csv"
ffni_sf_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in ffni_sf_df.columns:
    ffni_sf_df['Embedding'] = ffni_sf_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the FFNI-SF schema in Neo4j
def create_ffni_sf_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with FFNI-SF labels
        graph.run("MERGE (f:FFNISF_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:FFNISF_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:FFNISF_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:FFNISF_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:FFNISF_Noun {name: $noun})", noun=noun)
        
        # Create relationships with FFNI-SF namespacing
        graph.run("""
            MATCH (f:FFNISF_Factor {name: $factor}), (a:FFNISF_Adjective {name: $adjective})
            MERGE (f)-[:FFNISF_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:FFNISF_Adjective {name: $adjective}), (s:FFNISF_Synonym {name: $synonym})
            MERGE (a)-[:FFNISF_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:FFNISF_Adjective {name: $adjective}), (v:FFNISF_Verb {name: $verb})
            MERGE (a)-[:FFNISF_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:FFNISF_Adjective {name: $adjective}), (n:FFNISF_Noun {name: $noun})
            MERGE (a)-[:FFNISF_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:FFNISF_Factor {name: $factor}), (a:FFNISF_Adjective {name: $adjective})
                MERGE (e:FFNISF_Embedding {value: $embedding})
                MERGE (f)-[:FFNISF_HAS_EMBEDDING]->(e)
                MERGE (a)-[:FFNISF_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create FFNISF Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create FFNISF Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the FFNISF schema
        create_ffni_sf_schema(graph, ffni_sf_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", ffni_sf_df.shape[0])
        mlflow.log_param("Number of Columns", ffni_sf_df.shape[1])
        
        # Print the DataFrame
        print(ffni_sf_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
