## Beck Depression Inventory (BDI)<a class="anchor" id="PTMD_page_20.2"></a>

[Back to Top](#PTMD_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">


The Beck Depression Inventory (BDI) is a widely used self-report assessment tool designed to measure the severity of depressive symptoms in individuals. It was developed by Dr. Aaron T. Beck, a prominent psychiatrist, and has become a standard instrument in clinical psychology and psychiatry. Here's a summary of the BDI:

**Purpose**: The primary purpose of the BDI is to assess the presence and severity of depressive symptoms in individuals, helping clinicians and researchers diagnose and monitor depression.

**Format**: The BDI consists of a series of 21 multiple-choice questions, each corresponding to a specific symptom of depression. Respondents choose one of four statements that best describes their experience over the past two weeks, including feelings of sadness, guilt, changes in sleep and appetite, and loss of interest in activities.

**Scoring**: Each response is assigned a score ranging from 0 to 3, with higher scores indicating greater severity of depressive symptoms. The total score is calculated by summing the individual item scores. The overall score categorizes the level of depression as minimal, mild, moderate, or severe.

**Administration**: The BDI is typically self-administered, but it can also be administered by a trained clinician. It is suitable for adults and adolescents.

**Applications**:
- Clinical Diagnosis: BDI is used by mental health professionals to assist in diagnosing depression and determining the severity of the condition.
- Treatment Monitoring: It helps track changes in depressive symptoms over the course of treatment and assess treatment effectiveness.
- Research: The BDI is frequently used in research studies to measure depression in various populations and settings.

**Considerations**: The BDI is a valuable tool, but it should be interpreted in conjunction with clinical judgment. It provides insights into an individual's self-reported depressive symptoms but does not substitute for a comprehensive clinical assessment. It's essential for the results to be interpreted by a trained clinician.

The Beck Depression Inventory has undergone several revisions (BDI-II being one of the most common) and has been translated into multiple languages, making it a versatile and widely adopted instrument for assessing depression.

The **BDI-II (1996 version)** is the most commonly used form today, and its **21 factors** are as follows:

### **Official BDI-II Factors**
1. Sadness  
2. Pessimism  
3. Past Failure  
4. Loss of Pleasure  
5. Guilty Feelings  
6. Punishment Feelings  
7. Self-Dislike  
8. Self-Criticalness  
9. Suicidal Thoughts or Wishes  
10. Crying  
11. Agitation  
12. Loss of Interest  
13. Indecisiveness  
14. Worthlessness  
15. Loss of Energy  
16. Changes in Sleeping Pattern  
17. Irritability  
18. Changes in Appetite  
19. Difficulty Concentrating  
20. Tiredness or Fatigue  
21. Loss of Interest in Sex  

These questions cover a wide range of emotional, cognitive, and physical symptoms commonly associated with depression. Individuals taking the BDI respond to each question by selecting the statement that best describes their feelings or experiences over a specific time period, usually the past two weeks. The responses are then scored to assess the severity of depression, with higher scores indicating greater depression symptomatology.

##  Steps for BDI Personality Modeling Workflow 

1.  **Step 1: Create the Personality Dataset**

    * **Purpose:** This initial step involves defining and generating the dataset that represents the Eysenck Personality Model (BDI).
    * **Actions:**
        * Define the BDI dataset structure (e.g., using a dictionary to represent factors, adjectives, synonyms, etc.).
        * Generate the dataset by organizing the data into a Pandas DataFrame.
        * Save the dataset to a CSV file.
        * Log dataset metadata (e.g., number of rows, factors, data schema) and the dataset file itself as an artifact in MLflow.
    * **Importance:** This step creates the raw data that will be used for embedding generation and model training.

2.  **Step 2: API Key Handling and Initialization**

    * **Purpose:** This critical step ensures that your OpenAI API key is securely loaded and the OpenAI client is initialized. This sets the foundation for using the OpenAI API in subsequent steps.
    * **Actions:**
        * Load the OpenAI API key from a secure location (e.g., a file in the user's home directory).
        * Validate the API key (e.g., check for existence, emptiness, and potentially a basic API call).
        * Initialize the OpenAI client (`client`).
        * Log the API key handling process and its outcome in MLflow.
    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

3.  **Step 3: Test Embedding API**

    * **Purpose:** This step verifies that the OpenAI Embedding API is accessible and functioning correctly.
    * **Actions:**
        * Use the initialized OpenAI client (`client`) to make a test call to the Embedding API (e.g., by embedding a sample text).
        * Check the API response for validity.
        * Log the API call details and the outcome (success or failure) in MLflow.
    * **Importance:** This step ensures that you can successfully generate embeddings before proceeding to the next step.

4.  **Step 4: Create Embeddings for the Dataset**

    * **Purpose:** This step generates numerical representations (embeddings) for the text data in the BDI dataset using the OpenAI Embedding API.
    * **Actions:**
        * Load the BDI dataset (created in Step 2).
        * Use the OpenAI client (`client`) to generate embeddings for the relevant text fields (e.g., combining factor, adjective, synonym, verb, noun).
        * Add the generated embeddings as a new column in the Pandas DataFrame.
        * Save the DataFrame with embeddings to a new CSV file.
        * Log embedding generation parameters (e.g., embedding model used), statistics (e.g., embedding length), and the embeddings file as an artifact in MLflow.
    * **Importance:** This step transforms the text data into a numerical format that can be used for machine learning models.

5.  **Step 5: Create and Visualize a Label Encoder**

    * **Purpose:** This step prepares the categorical labels (personality factors) for model training by encoding them into numerical values and provides a visualization of this encoding.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Initialize a `LabelEncoder` from scikit-learn.
        * Fit the `LabelEncoder` to the 'Factor' column to create the mapping between personality factors and numerical codes.
        * Transform the 'Factor' column using the fitted `LabelEncoder` to create a new 'Factor_Encoded' column.
        * Save the fitted `LabelEncoder` object.
        * Generate a visualization (e.g., a bar chart) to show the mapping between original factors and encoded values.
        * Save the visualization as an image file.
        * Log the label encoder object and the visualization as artifacts in MLflow.
        * Log the mapping between original factors and encoded values as a dictionary in MLflow.
    * **Importance:** This step prepares the target variable for model training and provides a clear representation of the encoding.

6.  **Step 6: Create our BDI Model (Model Training and Evaluation)**

    * **Purpose:** This step trains a machine learning model on the generated embeddings to predict personality factors and evaluates its performance.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Prepare the data for model training:
            * Extract the embeddings as features (`X`).
            * Encode the 'Factor' column using the loaded `LabelEncoder` to get the target variable (`y`).
            * Split the data into training and testing sets.
        * Initialize a machine learning model (e.g., `RandomForestClassifier`).
        * Train the model on the training data.
        * Make predictions on the test data.
        * Evaluate the model's performance using appropriate metrics (e.g., accuracy, classification report, confusion matrix).
        * Generate visualizations of the evaluation results (e.g., confusion matrix plot).
        * Save the trained model.
        * Log model training parameters (e.g., hyperparameters), evaluation metrics, visualizations, and the trained model as artifacts in MLflow.
    * **Importance:** This step is the core of the machine learning process, where the model learns to predict personality factors from the embeddings.

7.  **Step 7: Model Testing (Inference on New Data)**

    * **Purpose:** This step demonstrates how to use the trained model to predict personality factors for new, unseen text inputs.
    * **Actions:**
        * Load the trained model (saved in Step 6).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Define a function that:
            * Takes new text as input.
            * Generates an embedding for the new text using the OpenAI API.
            * Uses the loaded model to predict the personality factor.
            * Uses the loaded `LabelEncoder` to decode the numerical prediction back to the original factor name.
        * Provide example new text inputs.
        * Use the function to predict personality factors for the example texts.
        * Print the predictions.
        * Log the test inputs and predictions in MLflow.
    * **Importance:** This step demonstrates the practical application of the trained model for making predictions on new data.

8.  **Step 8: Model Application, Visualization, and Analysis**

    * **Purpose:** This step provides additional visualization and analysis of the data and model.

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

# **Step 4: Create BDI Embeddings** <a class="anchor" id="bdi_page_3"></a>

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved bdi dataset from a CSV file
dataset_path = '../Datasets/bdi.csv'
try:
    bdi_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'bdi_df' is empty before processing
        if bdi_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", bdi_df.shape[0])
            print(f"üìä Processing {bdi_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        bdi_df['Embedding'] = bdi_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/bdi_embeddings.csv'
        bdi_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = bdi_df.shape[0]
        num_columns = bdi_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in bdi_df.columns and not bdi_df['Embedding'].empty:
            embedding_length = len(bdi_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(bdi_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/bdi_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="BDI Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/bdi_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "bdi_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/bdi_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of BDI Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/bdi_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of BDI Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create BDI Schema in Neo4j** <a class="anchor" id="BDI_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("BDI")

# Load the dataset
dataset_path = "../Embeddings/bdi_embeddings.csv"
bdi_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in bdi_df.columns:
    bdi_df['Embedding'] = bdi_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the BDI schema in Neo4j
def create_bdi_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with BDI labels
        graph.run("MERGE (f:BDI_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:BDI_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:BDI_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:BDI_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:BDI_Noun {name: $noun})", noun=noun)
        
        # Create relationships with BDI namespacing
        graph.run("""
            MATCH (f:BDI_Factor {name: $factor}), (a:BDI_Adjective {name: $adjective})
            MERGE (f)-[:BDI_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:BDI_Adjective {name: $adjective}), (s:BDI_Synonym {name: $synonym})
            MERGE (a)-[:BDI_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:BDI_Adjective {name: $adjective}), (v:BDI_Verb {name: $verb})
            MERGE (a)-[:BDI_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:BDI_Adjective {name: $adjective}), (n:BDI_Noun {name: $noun})
            MERGE (a)-[:BDI_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:BDI_Factor {name: $factor}), (a:BDI_Adjective {name: $adjective})
                MERGE (e:BDI_Embedding {value: $embedding})
                MERGE (f)-[:BDI_HAS_EMBEDDING]->(e)
                MERGE (a)-[:BDI_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create BDI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create BDI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the BDI schema
        create_bdi_schema(graph, bdi_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", bdi_df.shape[0])
        mlflow.log_param("Number of Columns", bdi_df.shape[1])
        
        # Print the DataFrame
        print(bdi_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
