The Millon Clinical Multiaxial Inventory (MCMI) was created by Dr. Theodore Millon, a prominent American psychologist known for his groundbreaking work in the field of personality theory and assessment. Dr. Millon was born in 1928 and dedicated his career to advancing our understanding of personality and psychopathology.

The development of the MCMI can be traced back to the mid-20th century when Dr. Millon began his work on personality theory and assessment. His research and clinical experience led to the formulation of a comprehensive theory of personality that emphasized the role of underlying personality structures in the development and manifestation of psychological disorders.

In 1982, Dr. Theodore Millon, along with his colleague Dr. Roger D. Davis, introduced the first edition of the Millon Clinical Multiaxial Inventory (MCMI). This assessment tool was designed to align with Dr. Millon's personality theory and provide clinicians with a standardized way to evaluate personality traits, cognitive styles, and psychopathology.

The MCMI was groundbreaking in its approach, as it incorporated a "base rate" scoring system that compared an individual's responses to those of a normative sample. This approach allowed for a more comprehensive assessment of both personality characteristics and potential clinical disorders. The inclusion of a multiaxial system, similar to the one used in the Diagnostic and Statistical Manual of Mental Disorders (DSM), further enhanced its clinical utility.

Over the years, the MCMI underwent revisions and updates to align with changes in the field of psychology and advancements in personality assessment. These revisions have ensured that the MCMI remains a valuable tool for mental health professionals in diagnosing and planning treatment for various psychological conditions.

Today, the MCMI is widely used by clinicians, including clinical psychologists and psychiatrists, in both clinical and research settings. It plays a vital role in understanding the complex relationship between personality factors and mental health, contributing to more accurate diagnoses and more effective treatment strategies for individuals seeking mental health support.

Dr. Theodore Millon's significant contributions to the field of psychology, particularly in the development of personality assessment tools like the MCMI, have left a lasting legacy, and his work continues to influence the practice of clinical psychology and the understanding of personality and psychopathology.

The MCMI is structured around several clinical scales, each of which assesses specific personality traits, cognitive styles, and psychological disorders. These scales are organized into three main categories: Personality Disorders, Clinical Syndromes, and Severe Personality Pathology. The assessment is designed to provide insights into an individual's mental health and help clinicians make more accurate diagnoses.

One of the notable features of the MCMI is its use of a "base rate" scoring system, which compares an individual's responses to those of a normative sample. This approach allows for a comprehensive assessment of both personality traits and potential clinical disorders. Additionally, the MCMI provides a Multiaxial Form, which aligns with the multiaxial system used in the Diagnostic and Statistical Manual of Mental Disorders (DSM). This form enables clinicians to consider multiple dimensions of an individual's mental health.

The MCMI is frequently used by mental health professionals, including clinical psychologists and psychiatrists, to assist in the diagnosis and treatment planning for a range of psychological conditions, such as personality disorders, mood disorders, anxiety disorders, and psychotic disorders. It can also be valuable in forensic assessments and in guiding therapeutic interventions.

Overall, the Millon Clinical Multiaxial Inventory is a valuable tool in the field of clinical psychology, aiding clinicians in understanding the complex interplay of personality factors and psychological disorders, which in turn helps inform more effective treatment strategies for individuals seeking mental health support.

1. **Purpose of the MCMI:** 
    - The MCMI is a psychological tool designed to assess both personality patterns and clinical syndromes. Its structure allows professionals to gauge the degree to which individuals exhibit specific traits, behaviors, or sets of symptoms.

2. **Hierarchical Structure:** 
    - **Type of Scale:**
        - Personality Patterns
        - Clinical Syndromes
    - **Specific Scale or Subscale:** 
        - E.g., Alcohol Dependence Scale, Dysthymia Scale, Depression Subscale.
    - **Descriptors:** 
        - Adjectives, factors, and other terms that further detail and provide a nuanced understanding of each scale or subscale.

3. **Scales and Subscales:** 
    - **Scales:** 
        - Major categorizations that assess broader psychological dimensions.
        - Examples: Alcohol Dependence Scale, PTSD Scale, Anxiety Disorder Scale, and more.
    - **Subscales:** 
        - More nuanced breakdowns of the broader scales, focusing on specific facets or nuances of a psychological dimension.
        - Examples: Depression Subscale, Social Alienation Subscale, Schizotypal Features Subscale, and so forth.

4. **Personality Patterns vs. Clinical Syndromes:** 
    - **Personality Patterns:** 
        - Assess enduring patterns of behavior, cognition, and emotion. They are often pervasive and persistent across various situations and over time.
        - Examples: Schizotypal Features, Schizoid Features, Borderline Features.
    - **Clinical Syndromes:** 
        - Measure transient, symptomatic dimensions of psychological disturbances. They represent the current state of an individual, which may change with time or treatment.
        - Examples: Alcohol Dependence, PTSD, Dysthymia.

5. **Noteworthy Features:** 
    - The MCMI does not just quantify the presence or absence of a trait or syndrome but gauges the degree to which it is present. This allows for a gradient assessment rather than a binary one.
    - While some scales and subscales may sound similar, each has been designed to capture distinct psychological phenomena.
  

The Millon Clinical Multiaxial Inventory-IV (MCMI-IV), the current version of the test, consists of 195 items. These items are designed to assess a range of personality disorders and clinical syndromes across the following scales:

1. **Personality Pattern Scales:** Assess 15 distinct personality patterns.
2. **Clinical Syndrome Scales:** Measure 10 clinical syndromes.
3. **Modifying Indices:** Evaluate response tendencies like exaggerating or minimizing symptoms.

This inventory helps clinicians evaluate underlying personality traits and potential clinical disorders in patients.
  
The Millon Clinical Multiaxial Inventory-IV (MCMI-IV) is designed to assess complex psychological conditions, categorizing personality and clinical syndromes across three main groups of scales:

   - **Description:** These scales measure enduring and pervasive personality patterns that influence an individual's thoughts, behaviors, and interpersonal relationships. They are based on the concept of personality disorders outlined by Theodore Millon.
   - **Patterns Assessed:**

2. **Clinical Syndrome Scales:**
   - **Description:** These scales identify symptomatic clinical syndromes, reflecting the current or recent onset of emotional disturbances that may require clinical intervention.
   - **Syndromes Assessed:** 
     - Generalized Anxiety
     - Somatic Symptom Disorder
     - Bipolar Spectrum Disorder
     - Persistent Depression
     - Alcohol Use Disorder
     - Drug Use Disorder
     - Post-Traumatic Stress Disorder (PTSD)
     - Schizophrenic Spectrum Disorder
     - Major Depression
     - Delusional Disorder

3. **Modifying Indices:**
   - **Description:** These scales provide insight into the test-taking attitudes and response biases of the examinee, helping to evaluate the accuracy and validity of the responses.
   - **Indices Assessed:**
     - **Disclosure Index:** Detects under-reporting or exaggeration of symptoms.
     - **Desirability Index:** Identifies tendencies to present oneself in a favorable light.
     - **Debasement Index:** Assesses the inclination to present oneself in an overly negative manner.

Each of these scales contributes to a comprehensive assessment by identifying both long-standing personality patterns and acute clinical symptoms.

##  Steps for MCMI Personality Modeling Workflow 

1.  **Step 1: Create the Personality Dataset**

    * **Purpose:** This initial step involves defining and generating the dataset that represents the Eysenck Personality Model (MCMI).
    * **Actions:**
        * Define the MCMI dataset structure (e.g., using a dictionary to represent factors, adjectives, synonyms, etc.).
        * Generate the dataset by organizing the data into a Pandas DataFrame.
        * Save the dataset to a CSV file.
        * Log dataset metadata (e.g., number of rows, factors, data schema) and the dataset file itself as an artifact in MLflow.
    * **Importance:** This step creates the raw data that will be used for embedding generation and model training.

2.  **Step 2: API Key Handling and Initialization**

    * **Purpose:** This critical step ensures that your OpenAI API key is securely loaded and the OpenAI client is initialized. This sets the foundation for using the OpenAI API in subsequent steps.
    * **Actions:**
        * Load the OpenAI API key from a secure location (e.g., a file in the user's home directory).
        * Validate the API key (e.g., check for existence, emptiness, and potentially a basic API call).
        * Initialize the OpenAI client (`client`).
        * Log the API key handling process and its outcome in MLflow.
    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

3.  **Step 3: Test Embedding API**

    * **Purpose:** This step verifies that the OpenAI Embedding API is accessible and functioning correctly.
    * **Actions:**
        * Use the initialized OpenAI client (`client`) to make a test call to the Embedding API (e.g., by embedding a sample text).
        * Check the API response for validity.
        * Log the API call details and the outcome (success or failure) in MLflow.
    * **Importance:** This step ensures that you can successfully generate embeddings before proceeding to the next step.

4.  **Step 4: Create Embeddings for the Dataset**

    * **Purpose:** This step generates numerical representations (embeddings) for the text data in the MCMI dataset using the OpenAI Embedding API.
    * **Actions:**
        * Load the MCMI dataset (created in Step 2).
        * Use the OpenAI client (`client`) to generate embeddings for the relevant text fields (e.g., combining factor, adjective, synonym, verb, noun).
        * Add the generated embeddings as a new column in the Pandas DataFrame.
        * Save the DataFrame with embeddings to a new CSV file.
        * Log embedding generation parameters (e.g., embedding model used), statistics (e.g., embedding length), and the embeddings file as an artifact in MLflow.
    * **Importance:** This step transforms the text data into a numerical format that can be used for machine learning models.

5.  **Step 5: Create and Visualize a Label Encoder**

    * **Purpose:** This step prepares the categorical labels (personality factors) for model training by encoding them into numerical values and provides a visualization of this encoding.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Initialize a `LabelEncoder` from scikit-learn.
        * Fit the `LabelEncoder` to the 'Factor' column to create the mapping between personality factors and numerical codes.
        * Transform the 'Factor' column using the fitted `LabelEncoder` to create a new 'Factor_Encoded' column.
        * Save the fitted `LabelEncoder` object.
        * Generate a visualization (e.g., a bar chart) to show the mapping between original factors and encoded values.
        * Save the visualization as an image file.
        * Log the label encoder object and the visualization as artifacts in MLflow.
        * Log the mapping between original factors and encoded values as a dictionary in MLflow.
    * **Importance:** This step prepares the target variable for model training and provides a clear representation of the encoding.

6.  **Step 6: Create our MCMI Model (Model Training and Evaluation)**

    * **Purpose:** This step trains a machine learning model on the generated embeddings to predict personality factors and evaluates its performance.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Load the saved `LabelEncoder` (created in Step 5).
        * Prepare the data for model training:
            * Extract the embeddings as features (`X`).
            * Encode the 'Factor' column using the loaded `LabelEncoder` to get the target variable (`y`).
            * Split the data into training and testing sets.
        * Initialize a machine learning model (e.g., `RandomForestClassifier`).
        * Train the model on the training data.
        * Make predictions on the test data.
        * Evaluate the model's performance using appropriate metrics (e.g., accuracy, classification report, confusion matrix).
        * Generate visualizations of the evaluation results (e.g., confusion matrix plot).
        * Save the trained model.
        * Log model training parameters (e.g., hyperparameters), evaluation metrics, visualizations, and the trained model as artifacts in MLflow.
    * **Importance:** This step is the core of the machine learning process, where the model learns to predict personality factors from the embeddings.

7.  **Step 7: Model Testing (Inference on New Data)**

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key Status", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"‚úÖ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key Status", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key Status", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("üîë API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("ü§ñ OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("‚úÖ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"‚ö†Ô∏è Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"‚ö†Ô∏è Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/mcmi_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/mcmi_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"üíæ Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "mcmi_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"‚úÖ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"‚úÖ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"‚ùå An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n‚úÖ Label encoding completed and encoder saved.")
````

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/mcmi_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/mcmi_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="mcmi Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "mcmi_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"üìä Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "mcmi_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"üìä Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "mcmi_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"üñºÔ∏è Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/mcmi_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"üíæ Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"‚úÖ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 13: Create MCMI Schema in Neo4j** <a class="anchor" id="MCMI_page_4"></a>


[Back to Top](#MCMI_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("MCMI")

# Load the dataset
dataset_path = "../Embeddings/mcmi_embeddings.csv"
mcmi_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in mcmi_df.columns:
    mcmi_df['Embedding'] = mcmi_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the MCMI schema in Neo4j
def create_mcmi_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with MCMI labels
        graph.run("MERGE (f:MCMI_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:MCMI_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:MCMI_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:MCMI_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:MCMI_Noun {name: $noun})", noun=noun)
        
        # Create relationships with MCMI namespacing
        graph.run("""
            MATCH (f:MCMI_Factor {name: $factor}), (a:MCMI_Adjective {name: $adjective})
            MERGE (f)-[:MCMI_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:MCMI_Adjective {name: $adjective}), (s:MCMI_Synonym {name: $synonym})
            MERGE (a)-[:MCMI_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:MCMI_Adjective {name: $adjective}), (v:MCMI_Verb {name: $verb})
            MERGE (a)-[:MCMI_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:MCMI_Adjective {name: $adjective}), (n:MCMI_Noun {name: $noun})
            MERGE (a)-[:MCMI_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:MCMI_Factor {name: $factor}), (a:MCMI_Adjective {name: $adjective})
                MERGE (e:MCMI_Embedding {value: $embedding})
                MERGE (f)-[:MCMI_HAS_EMBEDDING]->(e)
                MERGE (a)-[:MCMI_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create MCMI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create MCMI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the MCMI schema
        create_mcmi_schema(graph, mcmi_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", mcmi_df.shape[0])
        mlflow.log_param("Number of Columns", mcmi_df.shape[1])
        
        # Print the DataFrame
        print(mcmi_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
