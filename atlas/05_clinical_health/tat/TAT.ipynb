## Thematic Apperception Test (TAT)<a class="anchor" id="PTMD_page_20.7"></a>

[Back to Top](#PTMD_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">


The **Thematic Apperception Test (TAT)** is a projective psychological assessment tool that explores a person's patterns of thought, attitudes, emotional responses, and observational abilities. Developed by Henry A. Murray and Christiana D. Morgan in the 1930s, the test involves showing ambiguous images (usually human figures in different situations) to the subject and asking them to narrate a story about what is happening in the image. The narratives reveal themes related to the subject's psychological state, interpersonal relationships, and underlying motivations.

### Thematic Apperception Test Dataset Structure

To reflect the key elements assessed through TAT narratives, the following structure can be used:

- **Characters**: Key figures present in the story, including their actions, relationships, and emotions.
  - Examples: Hero, Villain, Authotaty Figure, Supportive Figure, Peer, etc.

- **Themes**: Recurrent ideas that drive the narrative and reflect the individual's concerns or motivations.
  - Examples: Conflict, Achievement, Intimacy, Power, Failure, Rejection, etc.

- **Conflicts**: Interpersonal or intrapersonal struggles between characters.
  - Examples: Betrayal, Rivalry, Dominance, Reconciliation, Support vs. Isolation, etc.

- **Outcomes**: The resolutions or consequences of the story, revealing the individual's perspective on coping or problem-solving.
  - Examples: Success, Failure, Compromise, Transformation, Stagnation, etc.

- **Scoring Information**: Interpretation guidelines related to the psychological dimensions measured (e.g., defense mechanisms, attachment style, self-concept).

````python
import openai
import os
import mlflow

# Start an MLFlow run for API key setup
with mlflow.start_run(run_name="API Key Handling and Initialization") as run:
    try:
        # Log the environment setup process
        mlflow.log_param("Step", "API Key Handling and Initialization")

        # Define the path to your API key file
        api_key_file_path = os.path.expanduser('~/openai_api_key.txt')
        mlflow.log_param("API Key File Path", api_key_file_path)

        # Evaluation: Check if the API key file exists
        if not os.path.exists(api_key_file_path):
            error_message = f"API key file not found at: {api_key_file_path}. Please ensure the file exists."
            mlflow.log_param("API Key Status", "Error: File not found")
            mlflow.log_param("Error", error_message)
            raise FileNotFoundError(error_message)
        else:
            mlflow.log_param("API Key File Existence", "Confirmed")
            print(f"‚úÖ API key file found at: {api_key_file_path}")

        # Read the API key from the file
        with open(api_key_file_path, 'r') as file:
            api_key = file.read().strip()

        # Evaluation: Check if the read API key is empty
        if not api_key:
            error_message = f"API key file at: {api_key_file_path} is empty. Please ensure your API key is in the file."
            mlflow.log_param("API Key Status", "Error: Empty file")
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("API Key Status", "Read successfully")
            mlflow.log_param("API Key Length", len(api_key)) # Log the length as a basic sanity check
            print("üîë API key read successfully.")

        # Set up your OpenAI API key
        openai.api_key = api_key
        client = openai.OpenAI(api_key=api_key)
        mlflow.log_param("OpenAI API Client", "Initialized successfully")
        print("ü§ñ OpenAI API client initialized.")

        # Evaluation: Attempt a basic API call to verify the key (optional, but recommended for immediate feedback)
        try:
            response = client.models.list()  # Removed the 'limit' argument
            mlflow.log_param("API Key Verification", "Successful (models list)")
            print("‚úÖ OpenAI API key verified successfully with a basic API call.")
        except openai.AuthenticationError as auth_error:
            mlflow.log_param("API Key Verification", "Failed (AuthenticationError)")
            mlflow.log_param("Error", str(auth_error))
            raise openai.AuthenticationError(f"OpenAI API key authentication failed: {auth_error}")
        except openai.OpenAIError as general_error:
            mlflow.log_param("API Key Verification", "Failed (OpenAIError)")
            mlflow.log_param("Error", str(general_error))
            print(f"‚ö†Ô∏è Warning: OpenAI API client initialized, but a test call failed with: {general_error}. Further API calls might fail.")
            mlflow.log_param("API Key Verification Warning", str(general_error))

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.AuthenticationError as e:
        mlflow.log_param("Error Type", "AuthenticationError")
        print(f"‚ùå Error: {e}")
        raise
    except openai.OpenAIError as e:
        mlflow.log_param("Error Type", "OpenAIError")
        print(f"‚ö†Ô∏è Warning during API initialization: {e}")
        raise
    except Exception as e:
        # Log the error if any other unexpected issue occurs
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

# **Step 3: Test Embedding** <a class="anchor" id="tat_page_2"></a>

[Back to Top](#tat_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
import openai
import mlflow

# Function to test the OpenAI Embedding API
def test_openai_embedding_api():
    with mlflow.start_run(run_name="Test Embedding API") as run:
        try:
            # Log the step name
            mlflow.log_param("Step", "Test Embedding API")

            # Evaluation: Check if the OpenAI client is initialized
            if 'client' not in globals() or not isinstance(client, openai.OpenAI):
                error_message = "OpenAI client is not initialized. Ensure the API key setup step was executed successfully."
                mlflow.log_param("API Client Status", "Not Initialized")
                mlflow.log_param("Error", error_message)
                raise RuntimeError(error_message)
            else:
                mlflow.log_param("API Client Status", "Initialized")
                print("‚úÖ OpenAI client found and ready.")

            # Example text to embed
            text = "The quick brown fox jumps over the lazy dog."
            mlflow.log_param("Test Text", text)
            print(f"üìù Testing embedding for text: '{text}'")

            # Evaluation: Check if the specified embedding model is available (optional, but good practice)
            embedding_model = "text-embedding-3-small"
            mlflow.log_param("Embedding Model", embedding_model)
            try:
                model_info = client.models.retrieve(embedding_model)
                mlflow.log_param("Embedding Model Availability", "Confirmed")
                print(f"‚úÖ Embedding model '{embedding_model}' is available.")
            except openai.NotFoundError:
                error_message = f"Embedding model '{embedding_model}' not found. Please check the model name."
                mlflow.log_param("Embedding Model Availability", "Not Found")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            except Exception as e:
                error_message = f"Error checking embedding model availability: {e}"
                mlflow.log_param("Embedding Model Availability Check Error", str(e))
                mlflow.log_param("Error", error_message)
                print(f"‚ö†Ô∏è Warning: Error checking model availability: {e}. Proceeding with embedding request.")

            # Request to generate embeddings
            response = client.embeddings.create(
                input=[text],  # The input should be a list of strings
                model=embedding_model
            )

            # Evaluation: Check if the embedding response contains data
            if not response.data:
                error_message = "Embedding API response does not contain any data."
                mlflow.log_param("Embedding API Response Status", "No Data")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding API Response Status", "Data Received")

            # Extract the embedding
            embedding = response.data[0].embedding

            # Evaluation: Check if the extracted embedding is not empty
            if not embedding:
                error_message = "Extracted embedding is empty."
                mlflow.log_param("Embedding Extraction Status", "Empty Embedding")
                mlflow.log_param("Error", error_message)
                raise ValueError(error_message)
            else:
                mlflow.log_param("Embedding Extraction Status", "Success")

            # Log the embedding length and a snippet
            embedding_length = len(embedding)
            mlflow.log_param("Embedding Length", embedding_length)
            mlflow.log_param("Embedding Snippet", embedding[:10])

            # Print the embedding length and a snippet
            print(f"Embedding length: {embedding_length}")
            print(f"Embedding snippet: {embedding[:10]}")  # Print the first 10 elements of the embedding

            print("‚úÖ Embedding API test successful.")

        except RuntimeError as e:
            mlflow.log_param("Error Type", "RuntimeError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except ValueError as e:
            mlflow.log_param("Error Type", "ValueError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.NotFoundError as e:
            mlflow.log_param("Error Type", "NotFoundError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå Error: {e}")
            raise
        except openai.OpenAIError as e:
            mlflow.log_param("Error Type", "OpenAIError")
            mlflow.log_param("Error", str(e))
            print(f"‚ùå OpenAI API error: {e}")
            raise
        except Exception as e:
            # Log the error if any other unexpected issue occurs
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", str(e))
            print(f"‚ùå An unexpected error occurred: {e}")
            raise

        finally:
            # End the MLFlow run
            mlflow.end_run()

# Test the OpenAI Embedding API
test_openai_embedding_api()
````

# **Step 4: Create TAT Embeddings** <a class="anchor" id="tat_page_3"></a>

[Back to Top](#tat_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved tat dataset from a CSV file
dataset_path = '../Datasets/tat.csv'
try:
    tat_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'tat_df' is empty before processing
        if tat_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", tat_df.shape[0])
            print(f"üìä Processing {tat_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        tat_df['Embedding'] = tat_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/tat_embeddings.csv'
        tat_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = tat_df.shape[0]
        num_columns = tat_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in tat_df.columns and not tat_df['Embedding'].empty:
            embedding_length = len(tat_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(tat_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 5: Create Label Embeddings** <a class="anchor" id="tat_page_3"></a>

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/tat_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/tat_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"üíæ Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "tat_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"‚úÖ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"‚úÖ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"‚ùå An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n‚úÖ Label encoding completed and encoder saved.")
````

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/tat_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/tat_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.4, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="tat Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.4)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "tat_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"üìä Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "tat_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"üìä Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "tat_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"üñºÔ∏è Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/tat_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"üíæ Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"‚úÖ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import joblib
import mlflow
from openai import OpenAI
import os
from dotenv import load_dotenv
import ast  # Import ast for literal_eval

# Load the dataset containing embeddings
embeddings_csv_path = '../Embeddings/tat_embeddings.csv'

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array using ast.literal_eval.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    try:
        return np.array(ast.literal_eval(embedding_str))
    except (ValueError, SyntaxError):
        return np.array()  # Or handle the error as appropriate

with mlflow.start_run(run_name="Step 9: Model Application, Visualization, and Analysis") as run:
    try:
        mlflow.log_param("Step", "Model Application, Visualization, and Analysis")

        # Define paths
        label_encoder_path = "../Models/tat_label_encoder.pkl"
        model_path = "../Models/tat_rf_model.pkl"

        mlflow.log_param("Embeddings CSV", embeddings_csv_path)
        mlflow.log_param("Label Encoder", label_encoder_path)
        mlflow.log_param("Model", model_path)

        # -------------------- Load Data and Artifacts --------------------

        # Load the dataset containing embeddings
        try:
            df = pd.read_csv(embeddings_csv_path)
            print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Embeddings dataset not found at: {embeddings_csv_path}"
            )
            print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}")
            raise

        # Apply the parsing function
        df['Embedding'] = df['Embedding'].apply(parse_embedding)

        # Print for debugging
        print("\nDataFrame with Embeddings:")
        print(df.head())
        print("\nShape of DataFrame:")
        print(df.shape)
        print("\nType of 'Embedding' column:")
        print(df['Embedding'].dtype)
        print("\nFirst Embedding:")
        if not df['Embedding'].empty:
            print(df['Embedding'].iloc[0])
        else:
            print("No embeddings found in DataFrame")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}")
            raise

        # -------------------- PCA Visualization --------------------

        mlflow.log_param("Visualization", "PCA")

        try:
            # Stack embeddings
            X = np.stack(df['Embedding'].values)

            # Print shape of X
            print("\nShape of X:")
            print(X.shape)

            # Encode Factor
            y_encoded = label_encoder.transform(df['Factor'].values)

            # Use PCA for dimensionality reduction
            pca = PCA(n_components=2)
            X_pca = pca.fit_transform(X)

            # Create the visualization
            plt.figure(figsize=(12, 8))
            scatter = plt.scatter(
                X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis'
            )
            plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
            plt.xlabel("PCA Component 1")
            plt.ylabel("PCA Component 2")
            plt.title("PCA of TAT Embeddings")

            # Save and log the plot
            pca_plot_path = "tat_embeddings_pca.png"
            plt.savefig(pca_plot_path)
            mlflow.log_artifact(pca_plot_path, artifact_path="visualization")
            plt.close()  # Close the plot to release memory

            print(f"‚úÖ PCA visualization saved to: {pca_plot_path}")

        except Exception as e:
            mlflow_error = f"An error occurred during PCA visualization: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"‚ùå {mlflow_error}")
            raise

        # -------------------- K-Means Clustering --------------------

        mlflow.log_param("Clustering", "K-Means")

        try:
            n_clusters = 5  # You can parameterize this
            mlflow.log_param("n_clusters", n_clusters)

            # Stack embeddings (if not already done)
            if "X" not in locals():
                X = np.stack(df["Embedding"].values)

            # Cluster embeddings
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            clusters = kmeans.fit_predict(X)

            # Add cluster information to the DataFrame and save it
            df["Cluster"] = clusters
            clustered_csv_path = "tat_clustered_embeddings.csv"
            df.to_csv(clustered_csv_path, index=False)
            mlflow.log_artifact(clustered_csv_path, artifact_path="clustering")

            print(
                f"‚úÖ K-Means clustering performed. Clustered data saved to: {clustered_csv_path}"
            )

        except Exception as e:
            mlflow_error = f"An error occurred during K-Means clustering: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"‚ùå {mlflow_error}")
            raise

        # -------------------- Model Application/Inference --------------------

        mlflow.log_param("Model Application", "Inference")

        try:
            # Load the trained model
            loaded_clf = joblib.load(model_path)

            def predict_factor(new_text):
                """
                Predicts the personality factor for new text using the trained model
                and the saved label encoder.

                Args:
                    new_text: The text to predict the personality factor for.

                Returns:
                    The predicted personality factor.
                """
                try:
                    # Get the embedding for the new text
                    new_embedding = get_embedding(new_text)

                    # Predict the factor
                    predicted_factor_encoded = loaded_clf.predict([new_embedding])

                    # Decode the prediction using the label encoder
                    predicted_factor = label_encoder.inverse_transform(
                        predicted_factor_encoded
                    )[0]

                    return predicted_factor

                except Exception as e:
                    print(f"‚ùå An error occurred during prediction: {e}")
                    raise

            # Example usage:
            new_text = "I love going to parties and meeting new people."
            predicted_factor = predict_factor(new_text)
            print(f"Predicted Factor for '{new_text}': {predicted_factor}")
            mlflow.log_param("New Text", new_text)
            mlflow.log_param("Predicted Factor", predicted_factor)

        except FileNotFoundError as e:
            mlflow_error = f"Model not found at: {model_path}"
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", mlflow_error)
            print(f"‚ùå Error: {e}")
            raise

        except Exception as e:
            mlflow_error = f"An error occurred during prediction: {e}"
            mlflow.log_param("Error Type", type(e).__name__)
            mlflow.log_param("Error", mlflow_error)
            print(f"‚ùå An error occurred during prediction: {e}")
            raise

    except Exception as e:
        # Catch any unexpected errors
        mlflow_error = f"An unexpected error occurred: {e}"
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", mlflow_error)
        print(f"‚ùå {mlflow_error}")
        raise

    finally:
        mlflow.end_run()

print("\n‚úÖ Step 10: Save Model Visualization and Analysis complete.")
````

````python
import os
import ast
import joblib
import mlflow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from dotenv import load_dotenv
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.model_selection import StratifiedKFold, cross_val_score
from openai import OpenAI

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Load environment (if needed) ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
load_dotenv(override=True)

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Paths & Helpers ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
EMBEDDINGS_CSV = '../Embeddings/tat_embeddings.csv'
LABEL_ENCODER_PKL = '../Models/tat_label_encoder.pkl'
RF_MODEL_PKL      = '../Models/tat_rf_model.pkl'

def parse_embedding(embedding_str):
    try:
        return np.array(ast.literal_eval(embedding_str))
    except Exception:
        return np.array([])

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Main Run ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
with mlflow.start_run(run_name="Step 9: Model Application, Visualization, and Analysis"):

    # Log inputs
    mlflow.log_param("embeddings_csv", EMBEDDINGS_CSV)
    mlflow.log_param("label_encoder_pkl",  LABEL_ENCODER_PKL)
    mlflow.log_param("rf_model_pkl",       RF_MODEL_PKL)

    # ---- 1. Load embeddings DataFrame ----
    df = pd.read_csv(EMBEDDINGS_CSV)
    df['Embedding'] = df['Embedding'].apply(parse_embedding)

    # Stack into X, extract raw labels y
    X = np.stack(df['Embedding'].values)
    raw_labels = df['Factor'].values

    # ---- 2. Load artifacts ----
    label_encoder = joblib.load(LABEL_ENCODER_PKL)
    rf_clf         = joblib.load(RF_MODEL_PKL)

    # Encode labels
    y = label_encoder.transform(raw_labels)

    # ---- 3. PCA Visualization ----
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
    plt.colorbar(scatter, ticks=range(len(label_encoder.classes_)))
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.title("PCA of TAT Embeddings")

    PCA_PLOT = "tat_embeddings_pca.png"
    plt.savefig(PCA_PLOT)
    plt.close()
    mlflow.log_artifact(PCA_PLOT, artifact_path="visualization")

    # ---- 4. K‚ÄëMeans Clustering ----
    N_CLUSTERS = 5
    mlflow.log_param("n_clusters", N_CLUSTERS)

    kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42)
    df['Cluster'] = kmeans.fit_predict(X)

    CLUSTER_CSV = "tat_clustered_embeddings.csv"
    df.to_csv(CLUSTER_CSV, index=False)
    mlflow.log_artifact(CLUSTER_CSV, artifact_path="clustering")

    # ---- 5. (Optional) Safe Stratified CV ----
    # if you ever need cross_val_score, pick n_splits ‚â§ min_class_count
    min_count  = df['Factor'].value_counts().min()
    safe_splits = min(5, min_count)
    mlflow.log_param("safe_n_splits", safe_splits)

    cv = StratifiedKFold(n_splits=safe_splits, shuffle=True, random_state=42)
    cv_scores = cross_val_score(rf_clf, X, y, cv=cv, scoring='accuracy')
    print(f"‚úÖ Cross‚Äëval scores ({safe_splits}‚Äëfold):", cv_scores)
    mlflow.log_metric("cv_mean_accuracy", np.mean(cv_scores))

    # ---- 6. Inference Function ----
    def predict_factor(text: str) -> str:
        # your existing get_embedding() must be in scope
        emb = get_embedding(text)
        pred = rf_clf.predict([emb])
        return label_encoder.inverse_transform(pred)[0]

    example = "I love going to parties and meeting new people."
    pred    = predict_factor(example)
    print(f"Predicted Factor for ‚Äú{example}‚Äù: {pred}")
    mlflow.log_param("example_text", example)
    mlflow.log_param("example_prediction", pred)

# End run
print("‚úÖ Step 9 complete.")
````

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("TAT")

# Load the dataset
dataset_path = "../Embeddings/tat_embeddings.csv"
tat_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in tat_df.columns:
    tat_df['Embedding'] = tat_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the TAT schema in Neo4j
def create_tat_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any ctatical value is missing

        # Create nodes with TAT labels
        graph.run("MERGE (f:TAT_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:TAT_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:TAT_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:TAT_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:TAT_Noun {name: $noun})", noun=noun)
        
        # Create relationships with TAT namespacing
        graph.run("""
            MATCH (f:TAT_Factor {name: $factor}), (a:TAT_Adjective {name: $adjective})
            MERGE (f)-[:TAT_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:TAT_Adjective {name: $adjective}), (s:TAT_Synonym {name: $synonym})
            MERGE (a)-[:TAT_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:TAT_Adjective {name: $adjective}), (v:TAT_Verb {name: $verb})
            MERGE (a)-[:TAT_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:TAT_Adjective {name: $adjective}), (n:TAT_Noun {name: $noun})
            MERGE (a)-[:TAT_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:TAT_Factor {name: $factor}), (a:TAT_Adjective {name: $adjective})
                MERGE (e:TAT_Embedding {value: $embedding})
                MERGE (f)-[:TAT_HAS_EMBEDDING]->(e)
                MERGE (a)-[:TAT_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create TAT Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create TAT Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the TAT schema
        create_tat_schema(graph, tat_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", tat_df.shape[0])
        mlflow.log_param("Number of Columns", tat_df.shape[1])
        
        # Print the DataFrame
        print(tat_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast

#‚îÄ‚îÄ 1. Env & MLflow setup ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
load_dotenv(override=True)
NEO4J_URI      = os.getenv("NEO4J_URI")
NEO4J_USER     = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI"))
mlflow.set_experiment("TAT")

#‚îÄ‚îÄ 2. Load CSV ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
tat_df = pd.read_csv("../Embeddings/tat_embeddings.csv")
if "Embedding" in tat_df.columns:
    tat_df["Embedding"] = tat_df["Embedding"].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else x
    )

#‚îÄ‚îÄ 3. Connect & create constraints ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
for stmt in [
    "CREATE CONSTRAINT IF NOT EXISTS FOR (f:TAT_Factor)    REQUIRE f.name IS UNIQUE",
    "CREATE CONSTRAINT IF NOT EXISTS FOR (a:TAT_Adjective) REQUIRE a.name IS UNIQUE",
    "CREATE CONSTRAINT IF NOT EXISTS FOR (s:TAT_Synonym)   REQUIRE s.name IS UNIQUE",
    "CREATE CONSTRAINT IF NOT EXISTS FOR (v:TAT_Verb)      REQUIRE v.name IS UNIQUE",
    "CREATE CONSTRAINT IF NOT EXISTS FOR (n:TAT_Noun)      REQUIRE n.name IS UNIQUE"
]:
    graph.run(stmt)

#‚îÄ‚îÄ 4. Bulk import ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
rows = tat_df.to_dict("records")
graph.run("""
UNWIND $rows AS row
MERGE (f:TAT_Factor    {name: row.Factor})
MERGE (a:TAT_Adjective {name: row.Adjective})
MERGE (s:TAT_Synonym   {name: row.Synonym})
MERGE (v:TAT_Verb      {name: row.Verb})
MERGE (n:TAT_Noun      {name: row.Noun})
MERGE (f)-[:TAT_HAS_ADJECTIVE]->(a)
MERGE (a)-[:TAT_HAS_SYNONYM]->(s)
MERGE (a)-[:TAT_HAS_VERB]->(v)
MERGE (a)-[:TAT_HAS_NOUN]->(n)
FOREACH (_ IN CASE WHEN row.Embedding IS NOT NULL AND size(row.Embedding)>0 THEN [1] ELSE [] END |
  MERGE (e:TAT_Embedding {value: row.Embedding})
  MERGE (f)-[:TAT_HAS_EMBEDDING]->(e)
  MERGE (a)-[:TAT_HAS_EMBEDDING]->(e)
)
""", rows=rows)

#‚îÄ‚îÄ 5. Show the query for the Browser ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
layout_query = """
MATCH (f:TAT_Factor)-[r1:TAT_HAS_ADJECTIVE]->(a:TAT_Adjective)
OPTIONAL MATCH (a)-[r2:TAT_HAS_SYNONYM]->(s:TAT_Synonym)
OPTIONAL MATCH (a)-[r3:TAT_HAS_VERB]->(v:TAT_Verb)
OPTIONAL MATCH (a)-[r4:TAT_HAS_NOUN]->(n:TAT_Noun)
RETURN f,r1,a,r2,s,r3,v,r4,n
"""
print("‚úÖ Data loaded into Neo4j.")
print("‚ñ∂Ô∏é Copy & paste this into Neo4j Browser and run it:")
print(layout_query)
print("\nüîß In the Browser‚Äôs graph result, click the ‚öôÔ∏è ‚ÄòTree‚Äô layout icon (top‚Äëright) to root the view at your TAT_Factor nodes.")
````
