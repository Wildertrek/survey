## **Approach/Avoidance Motivation: Navigating the Dual Forces of Human Drive**

In the intricate tapestry of human motivation, the Approach/Avoidance Motivation theory serves as a compass, guiding us through the dynamic interplay of our desires and fears. This theory, rooted in psychology, explores the dual nature of human motivation, where individuals are driven either toward rewarding goals (approach) or away from potentially aversive outcomes (avoidance). It unveils the multifaceted factors that influence our behavior, shape our decisions, and ultimately define our paths in life.

**Dissecting the Approach/Avoidance Motivation Dataset:**

The Approach/Avoidance Motivation theory is divided into two primary categories:

**Approach Motivation**: This facet of motivation revolves around the pursuit of success, achievement, and positive outcomes. It embodies qualities such as ambition, optimism, challenge-seeking, reward sensitivity, and persistence. Individuals driven by approach motivation are characterized by their enthusiasm for reaching goals and their ability to persevere in the face of challenges.

**Avoidance Motivation**: In contrast, avoidance motivation is fueled by the desire to evade failure, negative consequences, and potential threats. It encompasses attributes such as fear of failure, pessimism, risk avoidance, loss aversion, and procrastination. Individuals driven by avoidance motivation may exhibit caution and hesitation, often motivated by a desire to prevent unfavorable outcomes.

Within each of these categories, we delve deeper into specific factors that influence human behavior, exploring a rich tapestry of adjectives and synonyms:

**Approach Motivation Factors**:
- **Desire for Success**: This factor embodies ambition, motivation, determination, and goal-oriented behavior.
- **Optimism**: Reflecting a positive and confident outlook on life, optimism fosters enthusiasm and energy.
- **Challenge-seeking**: Individuals seeking challenges are adventurous, curious, and unafraid of risk.
- **Reward Sensitivity**: Reward-sensitive individuals are responsive, attentive, and constantly seeking excitement.
- **Persistence**: Persistence is characterized by tenacity, resilience, and unwavering dedication.

**Avoidance Motivation Factors**:
- **Fear of Failure**: This factor encompasses anxiety, avoidance, insecurity, and worry.
- **Pessimism**: Pessimism is characterized by negativity, doubt, cynicism, and defeatist thinking.
- **Risk Avoidance**: Those practicing risk avoidance are cautious, hesitant, and seek security.
- **Loss Aversion**: Loss-averse individuals are conservative, protective, and avoid risks to prevent regret.
- **Procrastination**: Procrastination involves indecisiveness, delaying, avoidance, and putting off tasks.

````python
import os
import pandas as pd
import mlflow

# Set the MLflow tracking URI
mlflow.set_tracking_uri("http://127.0.0.1:5000/")

# Set up AAM experiment
experiment_name = "AAM"
mlflow.set_experiment(experiment_name)

aam_dataset = {
    'Approach Motivation': {
        'Desire for Success': {
            'Synonyms': ['Driven', 'Passionate', 'Focused', 'High-achieving'],
            'Verbs': ['Drive', 'Passion', 'Focus', 'Achieve'],
            'Nouns': ['Drive', 'Passion', 'Focus', 'Achievement']
        },
        'Optimism': {
            'Synonyms': ['Upbeat', 'Cheerful', 'Energetic', 'Optimistic'],
            'Verbs': ['Uplift', 'Cheer', 'Energize', 'Hope'],
            'Nouns': ['Optimism', 'Cheerfulness', 'Energy', 'Hopefulness']
        },
        'Challenge-seeking': {
            'Synonyms': ['Daring', 'Inquisitive', 'Bold', 'Venturesome'],
            'Verbs': ['Dare', 'Inquire', 'Confront', 'Venture'],
            'Nouns': ['Daring', 'Inquiry', 'Boldness', 'Venture']
        },
        'Reward Sensitivity': {
            'Synonyms': ['Sensory', 'Attentive', 'Vigilant', 'Aroused'],
            'Verbs': ['Sense', 'Attend', 'Vigil', 'Arouse'],
            'Nouns': ['Sensitivity', 'Attention', 'Vigilance', 'Arousal']
        },
        'Persistence': {
            'Synonyms': ['Steadfast', 'Unyielding', 'Indomitable', 'Committed'],
            'Verbs': ['Persist', 'Endure', 'Overcome', 'Commit'],
            'Nouns': ['Persistence', 'Endurance', 'Resilience', 'Commitment']
        }
    },
    'Avoidance Motivation': {
        'Fear of Failure': {
            'Synonyms': ['Nervous', 'Apprehensive', 'Uneasy', 'Tense'],
            'Verbs': ['Fear', 'Worry', 'Tremble', 'Tense up'],
            'Nouns': ['Fear', 'Apprehension', 'Unease', 'Tension']
        },
        'Pessimism': {
            'Synonyms': ['Skeptical', 'Disheartened', 'Despondent', 'Discouraged'],
            'Verbs': ['Doubt', 'Despair', 'Dismay', 'Discourage'],
            'Nouns': ['Skepticism', 'Despair', 'Despondency', 'Discouragement']
        },
        'Risk Avoidance': {
            'Synonyms': ['Prudent', 'Wary', 'Reserved', 'Conservative'],
            'Verbs': ['Avoid', 'Hesitate', 'Secure', 'Conserve'],
            'Nouns': ['Caution', 'Wariness', 'Reservation', 'Conservatism']
        },
        'Loss Aversion': {
            'Synonyms': ['Defensive', 'Guarded', 'Cautious', 'Fearful'],
            'Verbs': ['Defend', 'Guard', 'Prevent', 'Fear'],
            'Nouns': ['Defense', 'Guardedness', 'Caution', 'Fear']
        },
        'Procrastination': {
            'Synonyms': ['Hesitant', 'Neglectful', 'Postponing', 'Sluggish'],
            'Verbs': ['Delay', 'Neglect', 'Postpone', 'Stall'],
            'Nouns': ['Delay', 'Neglect', 'Postponement', 'Stalling']
        }
    }
}

#Create a list to store the dataset
data = []
for factor, attributes in aam_dataset.items():
    for adjective, details in attributes.items():
        synonyms = details['Synonyms']
        verbs = details['Verbs']
        nouns = details['Nouns']
        # Iterate through each corresponding synonym, verb, and noun
        for synonym, verb, noun in zip(synonyms, verbs, nouns):
            data.append((factor, adjective, synonym, verb, noun))

# Create a DataFrame from the data list
aam_df = pd.DataFrame(data, columns=['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun'])

# Define the dataset path
dataset_path = "../Datasets/aam.csv"

# Start an MLflow run for dataset creation
with mlflow.start_run(run_name="AAM Dataset Creation"):
    # Log dataset metadata
    mlflow.log_param("Experiment", "AAM")
    mlflow.log_param("Total Rows", len(aam_df))
    mlflow.log_param("Total Factors", len(aam_dataset))
    mlflow.log_param("Data Schema", str(aam_df.columns.tolist()))
    
    # Save dataset to CSV
    aam_df.to_csv(dataset_path, index=False)
    mlflow.log_artifact(dataset_path, artifact_path="data")  # Logs the dataset to MLflow

    # Evaluation: Basic dataset inspection
    print(f"‚úÖ AAM dataset created and saved to {dataset_path}")
    print("First 5 rows of the dataset:")
    print(aam_df.head())
    print("\nDataset shape:", aam_df.shape)
    print("\nNumber of unique Factors:", aam_df['Factor'].nunique())
    print("Unique Factors:", aam_df['Factor'].unique())

    # Confirm MLflow tracking
    experiment = mlflow.get_experiment_by_name(experiment_name)
    if experiment:
        print(f"üìå MLflow tracking active under experiment '{experiment_name}' (ID: {experiment.experiment_id})")
    else:
        print("‚ùå Failed to track experiment in MLflow.")
````

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved aam dataset from a CSV file
dataset_path = '../Datasets/aam.csv'
try:
    aam_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'aam_df' is empty before processing
        if aam_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", aam_df.shape[0])
            print(f"üìä Processing {aam_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        aam_df['Embedding'] = aam_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/aam_embeddings.csv'
        aam_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = aam_df.shape[0]
        num_columns = aam_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in aam_df.columns and not aam_df['Embedding'].empty:
            embedding_length = len(aam_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(aam_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns
import os  # Import the os module

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/aam_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the previous embedding generation step was successful.")
    exit()

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder to the 'Factor' column and transform it
df['Factor_Encoded'] = label_encoder.fit_transform(df['Factor'])

# Save the fitted LabelEncoder
label_encoder_path = "../Models/aam_label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print(f"üíæ Label encoder saved to: {label_encoder_path}")

# Define the visualization function here (no import needed)
def visualize_label_encoder(label_encoder, artifact_path="visualization"):
    """
    Visualizes the label encoder mapping using a bar chart.

    Args:
        label_encoder: The fitted LabelEncoder object.
        artifact_path: Path within MLflow to save the visualization.
    """
    try:
        # mlflow.log_param("Step", "Visualize Label Encoder")  # REMOVE THIS LINE

        classes = label_encoder.classes_
        encoded_values = label_encoder.transform(classes)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=classes, y=encoded_values)
        plt.xlabel("Original Factor")
        plt.ylabel("Encoded Value")
        plt.title("Label Encoder Mapping")
        plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels

        # Save and log the plot with the desired filename
        label_encoder_plot_path = "aam_label_encoder_mapping.png"
        plt.savefig(label_encoder_plot_path)
        print(f"‚úÖ Label encoder visualization saved to: {os.path.abspath(label_encoder_plot_path)}")
        mlflow.log_artifact(
            label_encoder_plot_path, artifact_path=artifact_path
        )
        plt.close()

        print(f"‚úÖ Label encoder visualization saved to: {label_encoder_plot_path}")

    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param(
            "Error", f"An error occurred during Label Encoder visualization: {e}"
        )
        print(f"‚ùå An error occurred during Label Encoder visualization: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Create and Visualize Label Encoder") as run:
    mlflow.log_param("Step", "Create and Visualize Label Encoder")  # Log the overall step
    mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")

    # Evaluation: Inspect the encoded labels
    print("\nFirst 5 rows with encoded labels:")
    print(df[['Factor', 'Factor_Encoded']].head())
    print("\nMapping of original labels to encoded labels:")
    for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_))):
        print(f"{original}: {encoded}")

    # Log the class mapping as a dictionary
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    mlflow.log_dict(label_mapping, "label_encoder/mapping.json")

    # Visualize the label encoder
    visualize_label_encoder(label_encoder)

print("\n‚úÖ Label encoding completed and encoder saved.")
````

# **Step 6: Create Model** <a class="anchor" id="aam_page_3"></a>

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/aam_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/aam_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="aam Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "aam_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"üìä Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "aam_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"üìä Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "aam_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"üñºÔ∏è Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/aam_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"üíæ Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"‚úÖ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 13: Create AAM Schema in Neo4j** <a class="anchor" id="AAM_page_4"></a>


[Back to Top](#AAM_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("AAM")

# Load the dataset
dataset_path = "../Embeddings/aam_embeddings.csv"
aam_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in aam_df.columns:
    aam_df['Embedding'] = aam_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the AAM schema in Neo4j
def create_aam_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with AAM labels
        graph.run("MERGE (f:AAM_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:AAM_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:AAM_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:AAM_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:AAM_Noun {name: $noun})", noun=noun)
        
        # Create relationships with AAM namespacing
        graph.run("""
            MATCH (f:AAM_Factor {name: $factor}), (a:AAM_Adjective {name: $adjective})
            MERGE (f)-[:AAM_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:AAM_Adjective {name: $adjective}), (s:AAM_Synonym {name: $synonym})
            MERGE (a)-[:AAM_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:AAM_Adjective {name: $adjective}), (v:AAM_Verb {name: $verb})
            MERGE (a)-[:AAM_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:AAM_Adjective {name: $adjective}), (n:AAM_Noun {name: $noun})
            MERGE (a)-[:AAM_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:AAM_Factor {name: $factor}), (a:AAM_Adjective {name: $adjective})
                MERGE (e:AAM_Embedding {value: $embedding})
                MERGE (f)-[:AAM_HAS_EMBEDDING]->(e)
                MERGE (a)-[:AAM_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create AAM Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create AAM Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the AAM schema
        create_aam_schema(graph, aam_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", aam_df.shape[0])
        mlflow.log_param("Number of Columns", aam_df.shape[1])
        
        # Print the DataFrame
        print(aam_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
