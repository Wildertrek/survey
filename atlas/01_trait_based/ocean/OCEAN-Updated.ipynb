## **The Five Factor or OCEAN Model: Navigating the Seas of Personality**

### Description

In the intricate realm of understanding human personality, the OCEAN model stands as a guiding compass. Also known as the Five-Factor Model (FFM), it provides a structured approach to comprehending the fundamental dimensions that make each person unique. These five factors, encapsulated by the acronym OCEAN, represent different facets of personality: Openness, Conscientiousness, Extraversion, Agreeableness, and Neuroticism.

This model emerged to fill a critical gap in psychological research, offering a standardized taxonomy for personality traits that is empirically robust and applicable across various cultures and contexts. By delineating personality into these five core dimensions, it allows for a nuanced assessment of individual differences, facilitating systematic studies into complex human behaviors and traits.

The foundational work that gave rise to the Big Five Model can be traced back to the early 1960s with key researchers such as:

1. **Ernest Tupes and Raymond Christal (1961):** Their research identified several primary factors of personality, some aligning with the Big Five dimensions, laying essential groundwork for the model.

2. **Warren Norman (1963):** Norman's work further developed these factors, using different terms such as Surgency (similar to Extraversion) and Culture (related to Openness), instrumental in identifying the Big Five.

3. **Lewis Goldberg (1981):** Goldberg's research in the 1980s played a pivotal role in refining and popularizing the Big Five Model, introducing the terms that constitute OCEAN.

As the OCEAN model gained recognition and acceptance, it became invaluable in clinical, occupational, and social psychology, as well as a cornerstone in research linking personality with life outcomes, including job performance, academic success, and interpersonal relationships.

However, despite its widespread use, the OCEAN model has been subject to criticism. Some psychologists argue that it oversimplifies the rich tapestry of human personality and does not adequately account for cultural differences or situational variables. There's also a viewpoint that the Big Five factors may not capture the full spectrum of personality traits, suggesting additional factors such as honesty-humility to provide a more complete picture.

These factors can be visualized as currents within the vast ocean of personality:

1. **Openness**: A current flowing through the creative and imaginative realms, depicting the adventurous spirit in seeking novel experiences and unconventional ideas.
   
2. **Conscientiousness**: A well-charted course within the ocean, symbolizing an organized and reliable journey through life.
   
3. **Extraversion**: The lively waves on the ocean's surface, representing the energetic and social nature of extroverts.
   
4. **Agreeableness**: A warm and inviting current, embodying the kindness and cooperative spirit of agreeable individuals.
   
5. **Neuroticism**: The churning waters reflecting the emotional stability, where high neuroticism suggests stormier seas.

Acknowledging both the strengths and limitations of the OCEAN model is crucial. It remains an essential tool within the psychological toolkit, yet it underscores the importance of continued research to refine our understanding of personality. This balanced perspective is vital as we navigate through the complexities of human traits and their applications in various domains of life.


### Applications

The OCEAN model has profoundly influenced various domains by providing a nuanced understanding of personality traits:

- **Psychology:** In psychological research, the OCEAN model is pivotal in exploring the nexus between personality traits and various life outcomes. For example, studies have shown that high conscientiousness predicts better job performance and academic success, while high neuroticism can be a risk factor for mental health issues. The model serves as a foundational framework for both theoretical exploration and practical assessment in understanding individual differences and behaviors.

- **Education:** The educational sector utilizes the OCEAN model to adapt teaching methods to students' personality profiles. A case in point is the use of collaborative learning techniques for students high in agreeableness, or providing structured learning environments for those high in conscientiousness, thereby fostering effective learning and engagement. The model's insights assist educators in creating diverse instructional strategies that cater to the varied personalities within a classroom.

- **Workplace:** In organizational settings, the OCEAN model informs HR practices such as recruitment, where traits like conscientiousness and agreeableness are often predictive of job performance. It aids in constructing teams with complementary personality traits, ensuring a balance of roles and enhancing team dynamics. Leadership development programs also use the model to cultivate leaders who are self-aware and can effectively manage diverse personalities.

- **Marketing:** Marketers apply the OCEAN model to predict consumer behavior and tailor their strategies accordingly. For instance, individuals high in openness might be targeted with innovative and novel products, while those high in extraversion might respond better to social media marketing campaigns. This approach allows for more personalized marketing, which can lead to increased customer engagement and loyalty.

- **Therapy:** In therapeutic settings, the OCEAN model provides therapists with a framework to understand clients' predispositions, such as a tendency towards anxiety or depression in individuals with high neuroticism. This understanding enables the development of tailored intervention strategies that align with clients' personality structures, leading to more effective treatment outcomes.

### Cross-Cultural Implications

The applicability of the OCEAN model in cross-cultural contexts has been a subject of extensive research. While the model is largely robust across cultures, variations in the expression of certain traits have been observed. For example, the manifestation of agreeableness may differ between collectivist and individualist societies. Cross-cultural studies enhance the model's global relevance, showing that while the core traits are universal, cultural nuances influence trait expression and should be considered in practical applications.

### RecSys Usage with the OCEAN Model

The integration of the OCEAN model into recommendation systems (RecSys) has led to innovative applications and research exploring the intersection of personality traits with user preferences and behaviors:

- **Content Recommendations:** A notable example is music streaming services that use personality profiles to suggest playlists. For instance, users with high openness may receive recommendations for eclectic and less mainstream music, while extroverted users might be suggested playlists with upbeat and energetic songs. This tailoring of content to personality traits seeks to improve user satisfaction and service stickiness.

- **Personalized Experiences:** Online learning platforms have begun to incorporate personality traits into their course recommendation algorithms. A user high in conscientiousness might be recommended courses with structured learning paths and clear objectives, enhancing the learning experience and fostering course completion.

- **Tailored Marketing:** E-commerce platforms implement RecSys that utilize the OCEAN model to suggest products. Research has shown that users with certain personality traits are more likely to respond to specific types of products and marketing messages, leading to more effective targeting and increased sales.

- **Enhanced User Engagement:** Social media platforms use personality trait predictions to tailor users' feeds and advertisements. For example, individuals with high agreeableness might see more community-focused content, while those with high extraversion might see content that encourages social interaction.

### Challenges and Considerations

Applying the OCEAN model to RecSys is not without challenges. One of the primary considerations is the accuracy of personality trait measurement, often reliant on self-reported data that can be subjective. There's also the ethical concern regarding privacy and the potential for manipulation when using psychological profiles in marketing and content recommendation.

Another consideration is the cultural applicability of personality-based RecSys. The interpretation of traits can vary across different cultures, which can impact the effectiveness of the RecSys in global or multicultural markets.

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/ffm_embeddings.csv'
df = pd.read_csv(embeddings_csv_path)

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values  # Assuming we want to predict the Factor

# Encode the categorical labels as numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Start MLFlow run
with mlflow.start_run(run_name="OCEAN Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        
        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)
        
        # Make predictions on the test set
        y_pred = clf.predict(X_test)
        
        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)
        
        mlflow.log_metric("Accuracy", accuracy)
        
        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        
        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        
        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        
        # Save the model
        model_path = "../Models/ocean_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        
        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])
        
        print(f"Model training complete. Accuracy: {accuracy}")
        
    except Exception as e:
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/ffm_embeddings_gpt.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="OCEAN FFM Model Creation and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Full Cycle Model Application Workflow")
        mlflow.log_param("Dataset", "../Embeddings/ffm_embeddings_gpt.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of FFM Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/ffm_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of FFM Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USER")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("OCEAN")

# Load the dataset
dataset_path = "../Embeddings/ffm_embeddings.csv"
ffm_df = pd.read_csv(dataset_path)

# Function to create the OCEAN schema in Neo4j
def create_ocean_schema(graph, dataframe):
    # Clear the database
    graph.run("MATCH (n) DETACH DELETE n")
    
    # Create the schema
    for index, row in dataframe.iterrows():
        factor = row['Factor']
        adjective = row['Adjective']
        synonym = row['Synonym']
        verb = row['Verb']
        noun = row['Noun']
        embedding = row['Embedding']
        
        # Create nodes
        graph.run("MERGE (f:Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:Noun {name: $noun})", noun=noun)
        
        # Create relationships
        graph.run("""
            MATCH (f:Factor {name: $factor}), (a:Adjective {name: $adjective})
            MERGE (f)-[:HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (s:Synonym {name: $synonym})
            MERGE (a)-[:HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (v:Verb {name: $verb})
            MERGE (a)-[:HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (n:Noun {name: $noun})
            MERGE (a)-[:HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding
        graph.run("""
            MATCH (f:Factor {name: $factor}), (a:Adjective {name: $adjective})
            MERGE (e:Embedding {value: $embedding})
            MERGE (f)-[:HAS_EMBEDDING]->(e)
            MERGE (a)-[:HAS_EMBEDDING]->(e)
        """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create OCEAN Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create OCEAN Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the OCEAN schema
        create_ocean_schema(graph, ffm_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", ffm_df.shape[0])
        mlflow.log_param("Number of Columns", ffm_df.shape[1])
        
        # Print the DataFrame
        print(ffm_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
# specific OCEAN Schema
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("OCEAN")

# Load the dataset
dataset_path = "../Embeddings/ffm_embeddings.csv"
ffm_df = pd.read_csv(dataset_path)

# Function to create the OCEAN schema in Neo4j
def create_ocean_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        factor = row['Factor']
        adjective = row['Adjective']
        synonym = row['Synonym']
        verb = row['Verb']
        noun = row['Noun']
        embedding = row['Embedding']
        
        # Create nodes with OCEAN labels
        graph.run("MERGE (f:OCEAN_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:OCEAN_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:OCEAN_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:OCEAN_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:OCEAN_Noun {name: $noun})", noun=noun)
        
        # Create relationships with OCEAN namespacing
        graph.run("""
            MATCH (f:OCEAN_Factor {name: $factor}), (a:OCEAN_Adjective {name: $adjective})
            MERGE (f)-[:OCEAN_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:OCEAN_Adjective {name: $adjective}), (s:OCEAN_Synonym {name: $synonym})
            MERGE (a)-[:OCEAN_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:OCEAN_Adjective {name: $adjective}), (v:OCEAN_Verb {name: $verb})
            MERGE (a)-[:OCEAN_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:OCEAN_Adjective {name: $adjective}), (n:OCEAN_Noun {name: $noun})
            MERGE (a)-[:OCEAN_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding
        graph.run("""
            MATCH (f:OCEAN_Factor {name: $factor}), (a:OCEAN_Adjective {name: $adjective})
            MERGE (e:OCEAN_Embedding {value: $embedding})
            MERGE (f)-[:OCEAN_HAS_EMBEDDING]->(e)
            MERGE (a)-[:OCEAN_HAS_EMBEDDING]->(e)
        """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create OCEAN Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create OCEAN Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the OCEAN schema
        create_ocean_schema(graph, ffm_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", ffm_df.shape[0])
        mlflow.log_param("Number of Columns", ffm_df.shape[1])
        
        # Print the DataFrame
        print(ffm_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````

````python
from py2neo import Graph
from dotenv import load_dotenv
import os
import mlflow

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("OCEAN")

# Function to get node statistics
def get_node_stats(graph):
    query_nodes = "MATCH (n) RETURN labels(n) AS labels, count(n) AS count"
    node_stats = graph.run(query_nodes).data()
    return node_stats

# Function to get relationship statistics
def get_relationship_stats(graph):
    query_relationships = "MATCH ()-[r]->() RETURN type(r) AS type, count(r) AS count"
    relationship_stats = graph.run(query_relationships).data()
    return relationship_stats

# Start an MLFlow run
with mlflow.start_run(run_name="OCEAN Schema Statistics"):
    try:
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Log the test start
        mlflow.log_param("Step", "Retrieve OCEAN Schema Statistics")
        
        # Retrieve node statistics
        node_stats = get_node_stats(graph)
        
        # Retrieve relationship statistics
        relationship_stats = get_relationship_stats(graph)
        
        # Log node statistics
        for record in node_stats:
            label = ', '.join(record['labels'])
            mlflow.log_metric(f"Node_Count_{label}", record['count'])
        
        # Log relationship statistics
        for record in relationship_stats:
            mlflow.log_metric(f"Relationship_Count_{record['type']}", record['count'])
        
        # Print node statistics
        print("Node labels and counts:")
        for record in node_stats:
            print(f"Label: {record['labels']}, Count: {record['count']}")
        
        # Print relationship statistics
        print("\nRelationship types and counts:")
        for record in relationship_stats:
            print(f"Type: {record['type']}, Count: {record['count']}")
    
    except Exception as e:
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e
    
    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

````python
## Delete old schema with embeddings
import os
from dotenv import load_dotenv
from py2neo import Graph
import mlflow

# Load environment variables
load_dotenv(override=True)

# Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("Delete_OCEAN_Schema_With_Embeddings")

# Function to delete the old schema
def delete_old_schema(graph):
    tx = graph.begin()
    tx.run("MATCH (n)-[r:HAS_EMBEDDING]->() DELETE r")
    tx.run("MATCH (n:Embedding) DELETE n")
    tx.commit()

# Start an MLFlow run to delete the old schema
with mlflow.start_run(run_name="Delete OCEAN Schema With Embeddings"):
    try:
        # Establish connection to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Log the start of the deletion process
        mlflow.log_param("Step", "Delete OCEAN Schema With Embeddings")
        
        # Delete the old schema
        delete_old_schema(graph)
        
        # Log success
        mlflow.log_param("Status", "Old schema with embeddings deleted successfully")
    
    except Exception as e:
        print(f"An error occurred: {e}")
        mlflow.log_param("Error", str(e))
        raise e
    
    finally:
        # Ensure the MLFlow run is properly closed even if an error occurs
        mlflow.end_run()
````
