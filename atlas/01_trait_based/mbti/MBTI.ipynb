## **Navigating Personality with the Myers-Briggs Type Indicator (MBTI)**

In the vast and complex realm of human personality, the Myers-Briggs Type Indicator (MBTI) serves as a trusted compass. Developed by Katharine Cook Briggs and her daughter Isabel Briggs Myers, this model offers a structured framework for understanding and categorizing human personality based on four dichotomies: Extraversion vs. Introversion, Sensing vs. Intuition, Thinking vs. Feeling, and Judging vs. Perceiving.

Imagine personality as a diverse landscape, and MBTI as a map that helps us navigate and appreciate its intricacies:

1. **Extraversion vs. Introversion (E/I)**: This dimension represents the social orientation of an individual. Extraverts are like the bright stars in the night sky, drawing energy from social interactions. They are outgoing, talkative, and assertive. In contrast, introverts are like the serene moonlight, finding solace in introspection. They are reserved, reflective, and independent.

2. **Sensing vs. Intuition (S/N)**: This dichotomy reflects how people gather information. Sensors are like skilled cartographers, relying on tangible, concrete details. They are practical, detail-oriented, and observant. Intuitives, on the other hand, are like visionary explorers, seeking patterns and possibilities beyond the visible. They are innovative, imaginative, and future-oriented.

3. **Thinking vs. Feeling (T/F)**: This dimension delves into decision-making. Thinkers are like logical analysts, making choices based on rationality and objectivity. They are decisive, questioning, and rational. Feelers are like compassionate diplomats, weighing decisions based on empathy and harmony. They are sensitive, understanding, and warm.

4. **Judging vs. Perceiving (J/P)**: This aspect explores how individuals approach the external world. Judgers are like efficient organizers, preferring structure and planning. They are organized, disciplined, and purposeful. Perceivers are like flexible adventurers, embracing spontaneity and open-mindedness. They are adaptable, curious, and flexible.

Now, let's set our sights on the `mbti_dataset`, a valuable treasure trove for studying and understanding MBTI. This dataset is akin to a collection of tools and references for explorers of personality. Within the "Extraversion" category, for instance, you'll find adjectives like "Outgoing," "Energetic," and "Assertive," each with its set of synonyms. "Outgoing" can be described as "Sociable," "Friendly," and "Convivial."

This dataset acts as a compass, guiding researchers, psychologists, and enthusiasts through the intricate landscape of MBTI. By analyzing these adjectives and synonyms, one can gain a profound understanding of how individuals express their personality preferences and tendencies.

Just as seafarers rely on maps and charts to navigate the mbtis, those seeking to understand human nature turn to the MBTI model and its accompanying dataset. It is a trusted guide that helps us sail through the diverse waters of personality, illuminating the unique qualities that make each person who they are.

**MBTI Development and Research Timeline:**

- **1921:** Carl Jung publishes "Psychological Types," introducing the theory of psychological types, which lays the foundational ideas for the MBTI.
- **1943-1944:** Isabel Briggs Myers and her mother, Katharine Cook Briggs, begin the development of the MBTI as a tool to help individuals identify their personality type, influenced by Jung's theory.
- **1962:** The first MBTI Manual is published, formalizing the assessment tool for broader use.
- **1975:** The Educational Testing Service (ETS) begins publishing the MBTI, marking its transition to a more widely accessible instrument.
- **1980:** Isabel Briggs Myers publishes "Gifts Differing: Understanding Personality Type," providing a comprehensive overview of the MBTI and its applications in understanding personality differences.
- **1985:** The second edition/publication of the "MBTI Manual: A Guide to the Development and Use of the Myers-Briggs Type Indicator" by Isabel Briggs Myers and Mary H. McCaulley, offering a detailed guide to the MBTI.
- **1989:** Robert R. McCrae and Paul T. Costa publish "A Confirmatory Factor Analysis of the Myers-Briggs Type Indicator," examining the structure of the MBTI in relation to the Big Five personality traits.
- **1990:** Mary H. McCaulley's "MBTI: From Theory to Practice" in the "Journal of Counseling & Development," discusses the development and practical applications of MBTI.
- **1993:** David J. Pittenger's "The Myers-Briggs Type Indicator: A Critical Review and Practical Guide" offers a critical assessment of the MBTI's reliability and validity.
- **1995:** Widiger and Mullins investigate the relationship between MBTI types and personality disorders in "Relations Between Myers-Briggs Psychological Types and Personality Disorders."
- **1998:** The "MBTI Manual: A Guide to the Development and Use of the Myers-Briggs Type Indicator, 3rd Edition" is published, including updated research and applications.
- **2000s-2020s:** Continued research and debate on the psychometric properties of the MBTI, its applications in various fields, and its comparison to other personality models such as the Big Five.

This timeline encapsulates key moments in the development and critical examination of the MBTI, highlighting its evolution from a theory-based tool to a widely used personality assessment, as well as the ongoing academic discourse surrounding its validity and application.

### Cross-Cultural Implications of the Myers-Briggs Type Indicator (MBTI)

The cross-cultural applicability of the Myers-Briggs Type Indicator (MBTI) has been explored in various studies, though it has also faced criticism and challenges. Here are some key points regarding the MBTI's cross-cultural implications:

1. **Universality vs. Cultural Specificity:** While MBTI is based on Jung's theory which claims a universal applicability of personality types, research has shown that culture can significantly influence the distribution and expression of MBTI types. For instance, the prevalence of certain types may vary across cultures due to different societal values and norms.

2. **Cultural Bias in MBTI Items:** There is a concern that MBTI, originally developed in the United States, may reflect a Western bias in its questions and interpretations. This can potentially lead to inaccuracies or misinterpretations when used in non-Western cultures.

3. **Adaptation and Translation:** Efforts have been made to adapt and translate the MBTI for use in different cultures. However, the challenge remains to ensure that these adaptations accurately reflect the nuances of local cultures while maintaining the integrity of the original instrument.

4. **Use in Multicultural Settings:** In increasingly multicultural workplaces and educational settings, understanding the cultural background of individuals becomes crucial when interpreting MBTI results. What might be considered a sign of introversion in one culture could be seen as respect or politeness in another.

5. **Research on Cross-Cultural Validity:** Ongoing research is needed to explore the cross-cultural validity of the MBTI. Studies comparing MBTI results across cultures can provide valuable insights into the universal versus culture-specific aspects of the instrument.

6. **Training and Awareness:** Professionals using the MBTI in cross-cultural contexts are encouraged to receive training in cultural competence. This includes being aware of potential cultural biases and adapting their interpretation and feedback accordingly.

In summary, while the MBTI is a popular tool for understanding personality preferences, its application in cross-cultural contexts requires careful consideration of cultural influences and potential biases. Continuous research and adaptation are essential to enhance the tool's effectiveness and relevance across different cultural settings.

## Step 6: Create Embeddings

````python
import numpy as np
from datetime import datetime

# Load the previously saved mbti dataset from a CSV file
mbti_df = pd.read_csv('../Datasets/mbti.csv')

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    response = client.embeddings.create(
        input=[text],  # The input should be a list of strings
        model=model
    )
    embedding = response.data[0].embedding
    return embedding

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", "../Datasets/mbti.csv")
        
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        
        # Embedding Generation
        mbti_df['Embedding'] = mbti_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )
        
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        
        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/mbti_embeddings.csv'
        mbti_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        
        # Log statistical information
        num_rows = mbti_df.shape[0]
        num_columns = mbti_df.shape[1]
        embedding_length = len(mbti_df['Embedding'][0])
        
        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)
        mlflow.log_param("Embedding Length", embedding_length)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        
        # Log the location of the new embeddings dataset
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        
        # Print the DataFrame with embeddings
        print(mbti_df.head())

    except Exception as e:
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 7: Create mbti Model** <a class="anchor" id="mbti_page_3"></a>

[Back to Top](#mbti_toc)

<hr style="height:3px;border-width:0;color:Blue;background-color:Blue">

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/mbti_embeddings.csv'
df = pd.read_csv(embeddings_csv_path)

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values  # Assuming we want to predict the Factor

# Encode the categorical labels as numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Start MLFlow run
with mlflow.start_run(run_name="mbti Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        
        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)
        
        # Make predictions on the test set
        y_pred = clf.predict(X_test)
        
        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)
        
        mlflow.log_metric("Accuracy", accuracy)
        
        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        
        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        
        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        
        # Save the model
        model_path = "../Models/mbti_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        
        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])
        
        print(f"Model training complete. Accuracy: {accuracy}")
        
    except Exception as e:
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # End MLFlow run
        mlflow.end_run()
````

This block of code is a comprehensive step-by-step process focusing specifically on:
1. **Data Loading and Preprocessing**: Parsing and preparing embeddings from a CSV file for machine learning.
2. **Model Training**: Using a RandomForestClassifier to train on the embeddings.
3. **Model Evaluation**: Calculating and logging metrics such as accuracy, alongside detailed classification reports and confusion matrices.
4. **Visualization and Logging**: Visualizing the confusion matrix and logging both the visual representation and numerical data as artifacts in MLflow.
5. **Model Persistence**: Saving the trained model for future use or deployment.

# **Create MBTI Schema in Neo4j** <a class="anchor" id="MBTI_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USER")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("MBTI")

# Load the dataset
dataset_path = "../Embeddings/mbti_embeddings.csv"
mbti_df = pd.read_csv(dataset_path)

# Function to create the MBTI schema in Neo4j
def create_mbti_schema(graph, dataframe):
    # Clear the database
    graph.run("MATCH (n) DETACH DELETE n")
    
    # Create the schema
    for index, row in dataframe.iterrows():
        factor = row['Factor']
        adjective = row['Adjective']
        synonym = row['Synonym']
        verb = row['Verb']
        noun = row['Noun']
        embedding = row['Embedding']
        
        # Create nodes
        graph.run("MERGE (f:Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:Noun {name: $noun})", noun=noun)
        
        # Create relationships
        graph.run("""
            MATCH (f:Factor {name: $factor}), (a:Adjective {name: $adjective})
            MERGE (f)-[:HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (s:Synonym {name: $synonym})
            MERGE (a)-[:HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (v:Verb {name: $verb})
            MERGE (a)-[:HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:Adjective {name: $adjective}), (n:Noun {name: $noun})
            MERGE (a)-[:HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding
        graph.run("""
            MATCH (f:Factor {name: $factor}), (a:Adjective {name: $adjective})
            MERGE (e:Embedding {value: $embedding})
            MERGE (f)-[:HAS_EMBEDDING]->(e)
            MERGE (a)-[:HAS_EMBEDDING]->(e)
        """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create MBTI Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create MBTI Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the MBTI schema
        create_mbti_schema(graph, mbti_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", mbti_df.shape[0])
        mlflow.log_param("Number of Columns", mbti_df.shape[1])
        
        # Print the DataFrame
        print(mbti_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
