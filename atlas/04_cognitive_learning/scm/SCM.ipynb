## **Title: Deciphering Human Behavior: A Journey Through the Social-Cognitive Model (SCM)**

The Social-Cognitive Model (SCM) is a broad theoretical framework that has evolved over time through the contributions of many researchers from various fields. While it doesn't have a single set of creators, several prominent psychologists and scholars have made significant contributions to the development of social-cognitive theories and concepts. Some of these researchers include:

1. **Albert Bandura:** Bandura is perhaps one of the most influential figures associated with social-cognitive theory. His work on observational learning, self-efficacy, and social modeling laid the foundation for many aspects of the SCM.

2. **Julian Rotter:** Rotter is known for his research on social learning theory and locus of control, which are central to social-cognitive models.

3. **Walter Mischel:** Mischel's research on the ability of individuals to delay gratification and his work on cognitive-affective personality systems contributed to our understanding of self-regulation within the SCM.

4. **Leon Festinger:** Festinger's cognitive dissonance theory has connections to the SCM, particularly in how individuals perceive and respond to discrepancies between beliefs and behaviors.

5. **Jerome Bruner:** Bruner's work in cognitive psychology and his emphasis on the role of culture and social interaction in cognitive development align with some aspects of the SCM.

6. **John Bargh:** Bargh's research on automaticity and the influence of unconscious processes on behavior has connections to the SCM's exploration of cognition and social behavior.

7. **Elliot Aronson:** Aronson's work on cognitive dissonance, persuasion, and social influence contributes to our understanding of how individuals perceive and adapt to social situations.

8. **Richard Lazarus:** Lazarus's research on emotion and appraisal theory has implications for understanding the affective components of the SCM.

9. **Daniel Kahneman:** While primarily known for his work in behavioral economics, Kahneman's research on dual-system thinking and cognitive biases has relevance to the SCM's exploration of cognition and decision-making.

It's important to note that the SCM is an integrative framework that draws from various disciplines and researchers. It doesn't have a single "creator" in the traditional sense, but rather it has been shaped by the collective efforts of scholars who have explored the interplay between social factors and cognitive processes in understanding human behavior.

The Social-Cognitive Model (SCM) provides a robust framework for exploring the complex interactions between individual behavior and cognitive processes within diverse environmental contexts. This model delineates how individuals perceive, relate to, and act within their social environments, offering valuable insights for psychological research.

### **Behavioral Dynamics**

The SCM categorizes human behavior into distinct modes, each reflecting different ways individuals engage with their environment:

- **Active Behavior**: Characterized by high energy and dynamism, individuals exhibiting active behaviors are continually engaging with their environment, driven by a zest for action and interaction.
  
- **Passive Behavior**: In stark contrast, passive behaviors are marked by low energy and minimal engagement, where individuals exhibit lethargy and a preference for minimal social interaction.

- **Aggressive Behavior**: This mode involves forceful actions and assertive, sometimes confrontational, interactions with others. It represents a spectrum of behaviors where individuals assert dominance and control.

- **Submissive Behavior**: Submissive individuals typically exhibit compliance and a willingness to acquiesce to others, prioritizing harmony and avoidance of conflict in their interactions.

### **Cognitive Processes**

SCM also emphasizes the role of cognitive processes in shaping how individuals interpret and respond to their world:

- **Analytical Cognition**: This involves a logical and systematic approach to processing information, where individuals rely on rationality and structured thinking to make decisions.

- **Intuitive Cognition**: Intuitive processes are less about logical analysis and more about gut feelings and instinctive responses. This cognition style is spontaneous and often driven by internal cues rather than external data.

- **Abstract Cognition**: Here, thinking transcends specific instances and focuses on broader concepts and theoretical frameworks. This style is characterized by a preference for conceptual over concrete details.

- **Concrete Cognition**: In contrast to abstract thinking, concrete cognition focuses on real, tangible, and specific details. It is pragmatic and rooted in the immediate and observable aspects of the environment.

### **Interpersonal and Affective Dimensions**

The SCM further explores how these behavioral and cognitive components interact with affective states and interpersonal dynamics:

- **Affective Responses**: Individuals' emotional responses can range from positive (optimistic and uplifting) to negative (pessimistic and depressing), significantly influencing their behavioral outputs and cognitive styles.

- **Interpersonal Relations**: The model examines how individuals navigate social interactions, whether through extroverted engagement or introverted withdrawal, and cooperative or competitive dynamics.

### **Conclusion: Integrating SCM Insights**

Integrating these elements, the SCM offers a comprehensive lens through which to view the interplay between an individual's internal states and their external actions. For researchers and practitioners in psychology, this model serves as a foundational tool to understand and predict behaviors in various social and environmental contexts, ultimately aiming to enhance interventions and support individual well-being in a nuanced manner.

    * **Purpose:** This initial step involves defining and generating the dataset that represents the Eysenck Personality Model (SCM).
    * **Actions:**
        * Define the SCM dataset structure (e.g., using a dictionary to represent factors, adjectives, synonyms, etc.).
        * Generate the dataset by organizing the data into a Pandas DataFrame.
        * Save the dataset to a CSV file.
        * Log dataset metadata (e.g., number of rows, factors, data schema) and the dataset file itself as an artifact in MLflow.
    * **Importance:** This step creates the raw data that will be used for embedding generation and model training.

2.  **Step 2: API Key Handling and Initialization**

    * **Purpose:** This critical step ensures that your OpenAI API key is securely loaded and the OpenAI client is initialized. This sets the foundation for using the OpenAI API in subsequent steps.
    * **Actions:**
        * Load the OpenAI API key from a secure location (e.g., a file in the user's home directory).
        * Validate the API key (e.g., check for existence, emptiness, and potentially a basic API call).
        * Initialize the OpenAI client (`client`).
        * Log the API key handling process and its outcome in MLflow.
    * **Importance:** This step must succeed for the rest of the workflow that utilizes the OpenAI API (like embedding generation) to function. It's essential to handle potential errors (e.g., file not found, invalid key) gracefully.

3.  **Step 3: Test Embedding API**

    * **Purpose:** This step verifies that the OpenAI Embedding API is accessible and functioning correctly.
    * **Actions:**
        * Use the initialized OpenAI client (`client`) to make a test call to the Embedding API (e.g., by embedding a sample text).
        * Check the API response for validity.
        * Log the API call details and the outcome (success or failure) in MLflow.
    * **Importance:** This step ensures that you can successfully generate embeddings before proceeding to the next step.

4.  **Step 4: Create Embeddings for the Dataset**

    * **Purpose:** This step generates numerical representations (embeddings) for the text data in the SCM dataset using the OpenAI Embedding API.
    * **Actions:**
        * Load the SCM dataset (created in Step 2).
        * Use the OpenAI client (`client`) to generate embeddings for the relevant text fields (e.g., combining factor, adjective, synonym, verb, noun).
        * Add the generated embeddings as a new column in the Pandas DataFrame.
        * Save the DataFrame with embeddings to a new CSV file.
        * Log embedding generation parameters (e.g., embedding model used), statistics (e.g., embedding length), and the embeddings file as an artifact in MLflow.
    * **Importance:** This step transforms the text data into a numerical format that can be used for machine learning models.

5.  **Step 5: Create and Visualize a Label Encoder**

    * **Purpose:** This step prepares the categorical labels (personality factors) for model training by encoding them into numerical values and provides a visualization of this encoding.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Initialize a `LabelEncoder` from scikit-learn.
        * Fit the `LabelEncoder` to the 'Factor' column to create the mapping between personality factors and numerical codes.
        * Transform the 'Factor' column using the fitted `LabelEncoder` to create a new 'Factor_Encoded' column.
        * Save the fitted `LabelEncoder` object.
        * Generate a visualization (e.g., a bar chart) to show the mapping between original factors and encoded values.
        * Save the visualization as an image file.
        * Log the label encoder object and the visualization as artifacts in MLflow.
        * Log the mapping between original factors and encoded values as a dictionary in MLflow.
    * **Importance:** This step prepares the target variable for model training and provides a clear representation of the encoding.

    * **Purpose:** This step trains a machine learning model on the generated embeddings to predict personality factors and evaluates its performance.
    * **Actions:**
        * Load the dataset with embeddings (created in Step 4).
        * Load the saved `LabelEncoder` (created in Step 5).

            * Extract the embeddings as features (`X`).
            * Encode the 'Factor' column using the loaded `LabelEncoder` to get the target variable (`y`).
            * Split the data into training and testing sets.
        * Initialize a machine learning model (e.g., `RandomForestClassifier`).
        * Train the model on the training data.

````python
import numpy as np
import pandas as pd
from datetime import datetime
from openai import OpenAI
import os
import mlflow
from dotenv import load_dotenv
import joblib

load_dotenv(override=True)

# Function to get the OpenAI API key from the saved file
def get_openai_api_key_from_file(filepath='~/openai_api_key.txt'):
    filepath = os.path.expanduser(filepath)
    try:
        with open(filepath, 'r') as f:
            api_key = f.read().strip()
        if not api_key:
            raise ValueError(f"API key file at '{filepath}' is empty.")
        return api_key
    except FileNotFoundError:
        raise FileNotFoundError(f"API key file not found at '{filepath}'.")
    except Exception as e:
        raise RuntimeError(f"Error reading API key from file '{filepath}': {e}")

# Initialize OpenAI client using the function
try:
    openai_api_key = get_openai_api_key_from_file()
    client = OpenAI(api_key=openai_api_key)
    print("‚úÖ OpenAI client initialized using API key from file.")
except (FileNotFoundError, ValueError, RuntimeError) as e:
    print(f"‚ùå Error initializing OpenAI client: {e}")
    # It's crucial to handle this error, perhaps by exiting or using a default behavior
    exit()

# Load the previously saved scm dataset from a CSV file
dataset_path = '../Datasets/scm.csv'
try:
    scm_df = pd.read_csv(dataset_path)
    print(f"‚úÖ Dataset loaded successfully from: {dataset_path}")
except FileNotFoundError:
    error_message = f"Error: Dataset not found at: {dataset_path}. Ensure the previous step (dataset creation) was successful."
    print(f"‚ùå {error_message}")
    raise FileNotFoundError(error_message)

# Function to generate embeddings using the latest OpenAI API
def get_embedding(text, model="text-embedding-3-small"):
    try:
        response = client.embeddings.create(
            input=[text],  # The input should be a list of strings
            model=model
        )
        embedding = response.data[0].embedding
        return embedding
    except openai.APIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise
    except Exception as e:
        print(f"‚ùå An unexpected error occurred during embedding generation: {e}")
        raise

with mlflow.start_run(run_name="Generate Embeddings") as run:
    try:
        mlflow.log_param("Step", "Generate Embeddings")
        mlflow.log_param("Model", "text-embedding-3-small")
        mlflow.log_param("Dataset", dataset_path)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Embedding generation started at: {start_time}")

        # Evaluation: Check if the 'scm_df' is empty before processing
        if scm_df.empty:
            error_message = "Error: Loaded dataset is empty. Check the dataset file."
            mlflow.log_param("Error", error_message)
            raise ValueError(error_message)
        else:
            mlflow.log_param("Dataset Row Count", scm_df.shape[0])
            print(f"üìä Processing {scm_df.shape[0]} rows for embedding generation.")

        # Embedding Generation
        scm_df['Embedding'] = scm_df.apply(
            lambda row: get_embedding(f"{row['Factor']} {row['Adjective']} {row['Synonym']} {row['Verb']} {row['Noun']}"),
            axis=1
        )

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Embedding generation finished at: {end_time} (Duration: {end_time - start_time})")

        # Save the embeddings to a new CSV file
        embeddings_csv_path = '../Embeddings/scm_embeddings.csv'
        scm_df[['Factor', 'Adjective', 'Synonym', 'Verb', 'Noun', 'Embedding']].to_csv(embeddings_csv_path, index=False)
        mlflow.log_artifact(embeddings_csv_path, artifact_path="embeddings")
        mlflow.log_param("Embeddings CSV Path", embeddings_csv_path)
        print(f"üíæ Embeddings saved to: {embeddings_csv_path}")

        # Log statistical information
        num_rows = scm_df.shape[0]
        num_columns = scm_df.shape[1]
        # Evaluation: Check if the 'Embedding' column was successfully created and has at least one embedding
        if 'Embedding' in scm_df.columns and not scm_df['Embedding'].empty:
            embedding_length = len(scm_df['Embedding'].iloc[0])
            mlflow.log_param("Embedding Length", embedding_length)
            print(f"üìè Length of the first embedding: {embedding_length}")
        else:
            mlflow.log_param("Embedding Length", "N/A (Embedding column missing or empty)")
            print("‚ö†Ô∏è Warning: Embedding column is missing or empty. Check for errors during embedding generation.")

        mlflow.log_param("Number of Rows", num_rows)
        mlflow.log_param("Number of Columns", num_columns)

        # Evaluation: Print the first few rows with the new 'Embedding' column
        print("\nFirst 5 rows with embeddings:")
        print(scm_df.head())

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except ValueError as e:
        mlflow.log_param("Error Type", "ValueError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except openai.APIError as e:
        mlflow.log_param("Error Type", "OpenAI APIError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå OpenAI API error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An unexpected error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

````python
import pandas as pd
import numpy as np
import mlflow
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset with embeddings
embeddings_csv_path = '../Embeddings/scm_embeddings.csv'
try:
    df = pd.read_csv(embeddings_csv_path)
    print(f"‚úÖ Embeddings dataset loaded from: {embeddings_csv_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Embeddings dataset not found at: {embeddings_csv_path}. Ensure the embedding generation step was successful.")
    exit()

# Function to convert embedding strings back to lists of floats
def parse_embedding(embedding_str):
    return np.array(eval(embedding_str))

# Convert the embedding column from strings to lists of floats
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Load the pre-fitted label encoder
label_encoder_path = "../Models/scm_label_encoder.pkl"
try:
    label_encoder = joblib.load(label_encoder_path)
    print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
except FileNotFoundError:
    print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}. Ensure the label encoding step was successful.")
    exit()

# Extract embeddings and labels
X = np.stack(df['Embedding'].values)
y = df['Factor'].values

# Encode the categorical labels using the loaded label encoder
y_encoded = label_encoder.transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Start MLFlow run
with mlflow.start_run(run_name="scm Model Training and Evaluation Workflow") as run:
    try:
        mlflow.log_param("Step", "Detailed Model Training and Evaluation Workflow")
        mlflow.log_param("Model Type", "RandomForestClassifier")
        mlflow.log_param("Dataset", embeddings_csv_path)
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_param("Test Size", 0.2)
        mlflow.log_param("Random State", 42)
        mlflow.log_param("Stratify", True)

        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))
        print(f"‚è≥ Model training started at: {start_time}")

        # Train a RandomForestClassifier on the training data
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X_train, y_train)

        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))
        print(f"‚úÖ Model training finished at: {end_time} (Duration: {end_time - start_time})")

        # Make predictions on the test set
        y_pred = clf.predict(X_test)

        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
        conf_matrix = confusion_matrix(y_test, y_pred)

        mlflow.log_metric("Accuracy", accuracy)

        # Log classification report and confusion matrix as artifacts
        class_report_df = pd.DataFrame(class_report).transpose()
        class_report_path = "scm_classification_report.csv"
        class_report_df.to_csv(class_report_path, index=True)
        mlflow.log_artifact(class_report_path, artifact_path="metrics")
        print(f"üìä Classification report saved to: {class_report_path}")

        conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
        conf_matrix_path = "scm_confusion_matrix.csv"
        conf_matrix_df.to_csv(conf_matrix_path, index=True)
        mlflow.log_artifact(conf_matrix_path, artifact_path="metrics")
        print(f"üìä Confusion matrix saved to: {conf_matrix_path}")

        # Visualize and log confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        conf_matrix_img_path = "scm_confusion_matrix.png"
        plt.savefig(conf_matrix_img_path)
        mlflow.log_artifact(conf_matrix_img_path, artifact_path="metrics")
        plt.close()
        print(f"üñºÔ∏è Confusion matrix plot saved to: {conf_matrix_img_path}")

        # Save the model
        model_path = "../Models/scm_rf_model.pkl"
        joblib.dump(clf, model_path)
        mlflow.log_artifact(model_path, artifact_path="models")
        print(f"üíæ Trained model saved to: {model_path}")

        # Log additional parameters
        mlflow.log_param("Train Set Size", X_train.shape[0])
        mlflow.log_param("Test Set Size", X_test.shape[0])
        mlflow.log_param("Number of Features", X_train.shape[1])

        print(f"‚úÖ Model training complete. Accuracy: {accuracy}")
        print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
        print("\nConfusion Matrix:\n", conf_matrix_df)

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred: {e}")
        raise

    finally:
        # End MLFlow run
        mlflow.end_run()
````

# **Step 8: Test and Evaluate Model** <a class="anchor" id="scm_page_3"></a>

````python
import pandas as pd
import numpy as np
import mlflow
import joblib
from openai import OpenAI  # Keep this import for type hinting or other openai functions
import os

# Define paths
label_encoder_path = "../Models/scm_label_encoder.pkl"
model_path = "../Models/scm_rf_model.pkl"

# -------------------- Utility Function --------------------

def parse_embedding(embedding_str):
    """
    Converts a string representation of an embedding (from the CSV)
    into a NumPy array.

    Args:
        embedding_str: The string representation of the embedding.

    Returns:
        A NumPy array representing the embedding.
    """
    return np.fromstring(embedding_str.strip(""), sep=", ")

# -------------------- Embedding Function --------------------

def get_embedding(text, model="text-embedding-3-small"):
    """
    Generates an embedding for the given text using the OpenAI API.

    Args:
        text: The text to embed.
        model: The name of the embedding model to use.

    Returns:
        The generated embedding as a list.
    """
    try:
        response = client.embeddings.create(input=[text], model=model)
        embedding = response.data[0].embedding
        return embedding
    except openai.OpenAIError as e:
        print(f"‚ùå OpenAI API error during embedding generation: {e}")
        raise

# Start MLflow run
with mlflow.start_run(run_name="Step 7: Model Testing (Inference on New Data)") as run:
    try:
        mlflow.log_param("Step", "Model Testing (Inference on New Data)")
        mlflow.log_artifact(label_encoder_path, artifact_path="label_encoder")
        mlflow.log_artifact(model_path, artifact_path="model")

        print("\n--- Model Testing (Inference on New Data) ---")

        # Load the pre-fitted label encoder
        try:
            label_encoder = joblib.load(label_encoder_path)
            print(f"‚úÖ Label encoder loaded from: {label_encoder_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param(
                "Error", f"Label encoder not found at: {label_encoder_path}"
            )
            print(f"‚ùå Error: Label encoder not found at: {label_encoder_path}")
            raise

        # Load the trained model
        try:
            loaded_clf = joblib.load(model_path)
            print(f"‚úÖ Trained model loaded from: {model_path}")
        except FileNotFoundError:
            mlflow.log_param("Error Type", "FileNotFoundError")
            mlflow.log_param("Error", f"Model not found at: {model_path}")
            print(f"‚ùå Error: {e}")
            raise

        # -------------------- Model Application/Inference --------------------

        def predict_factor(new_text):
            """
            Predicts the personality factor for new text using the trained model
            and the saved label encoder.

            Args:
                new_text: The text to predict the personality factor for.

            Returns:
                The predicted personality factor.
            """
            try:
                # Get the embedding for the new text
                new_embedding = get_embedding(new_text)

                # Predict the factor
                predicted_factor_encoded = loaded_clf.predict([new_embedding])

                # Decode the prediction using the label encoder
                predicted_factor = label_encoder.inverse_transform(
                    predicted_factor_encoded
                )[0]

                return predicted_factor

            except Exception as e:
                print(f"‚ùå An error occurred during prediction: {e}")
                raise

        # Example usage:
        test_texts = [
            "I love going to parties and meeting new people.",
            "I prefer staying home with a good book.",
            "I often feel anxious and worried.",
            "I am generally calm and relaxed.",
            "I enjoy taking risks and trying new things."
        ]
        mlflow.log_param("Number of Test Inputs", len(test_texts))

        print("\n--- Testing Model on New Data ---")
        for i, text in enumerate(test_texts):
            try:
                predicted_factor = predict_factor(text)
                print(f"Predicted Factor for '{text[:50]}...': {predicted_factor}")
                mlflow.log_param(f"New Text {i+1}", text)
                mlflow.log_param(f"Predicted Factor {i+1}", predicted_factor)
            except Exception as e:
                mlflow.log_param("Error Type", type(e).__name__)
                mlflow.log_param(
                    "Error", f"Error predicting for text: {text[:50]}...: {e}"
                )
                print(f"‚ùå Error predicting for text: {text[:50]}...: {e}")

    except FileNotFoundError as e:
        mlflow.log_param("Error Type", "FileNotFoundError")
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise
    except Exception as e:
        mlflow.log_param("Error Type", type(e).__name__)
        mlflow.log_param("Error", str(e))
        print(f"‚ùå An error occurred during model testing: {e}")
        raise

    finally:
        mlflow.end_run()

print("\n‚úÖ Model testing (inference on new data) complete.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import mlflow
import mlflow.sklearn
from datetime import datetime
import openai

# Load the dataset containing embeddings
df = pd.read_csv('../Embeddings/scm_embeddings.csv')

# Define a function to parse string representations of embeddings into numpy arrays
def parse_embedding(embedding_str):
    # Remove brackets and split the string by commas, then convert to a numpy array
    return np.fromstring(embedding_str.strip('[]'), sep=', ')

# Apply the function to each row in the 'Embedding' column to convert all data
df['Embedding'] = df['Embedding'].apply(parse_embedding)

# Stack arrays (embeddings) vertically into a matrix for model input (features)
X = np.stack(df['Embedding'].values)
# Extract labels for prediction from the dataframe
y = df['Factor'].values  

# Convert categorical labels into integers for model processing
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Start tracking the experiment with MLflow
with mlflow.start_run(run_name="SCM Model Visualization and Evaluation") as run:
    try:
        # Log basic parameters like the step and dataset being used
        mlflow.log_param("Step", "Visualize model and evaluate")
        mlflow.log_param("Dataset", "../Embeddings/scm_embeddings.csv")

        # Record the start time of the model run
        start_time = datetime.now()
        mlflow.log_param("Start Time", str(start_time))

        # Initialize and train a RandomForestClassifier
        clf = RandomForestClassifier(n_estimators=100, random_state=42)
        clf.fit(X, y_encoded)

        # Log the model to MLflow for later access or deployment
        mlflow.sklearn.log_model(clf, "scm_random_forest_model")

        # Predict on the training set to evaluate the model
        y_pred = clf.predict(X)
        # Generate a confusion matrix to visualize classification accuracy
        conf_matrix = confusion_matrix(y_encoded, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.show()

        # Calculate and log accuracy to MLflow
        accuracy = accuracy_score(y_encoded, y_pred)
        mlflow.log_metric("Accuracy", accuracy)

        # Generate and log a detailed classification report
        class_report = classification_report(y_encoded, y_pred, output_dict=True)
        for label, metrics in class_report.items():
            if isinstance(metrics, dict):
                for metric, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric}", value)
            else:
                mlflow.log_metric(f"{label}", metrics)

        # Save and log the confusion matrix as an artifact for record-keeping
        conf_matrix_path = "../Embeddings/scm_confusion_matrix.csv"
        np.savetxt(conf_matrix_path, conf_matrix, delimiter=",")
        mlflow.log_artifact(conf_matrix_path, artifact_path="evaluation")

        # Use PCA for dimensionality reduction to visualize the embedding distribution
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
        plt.colorbar(scatter, ticks=range(len(set(y_encoded))))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('PCA of SCM Embeddings')
        plt.show()

        # Cluster embeddings to identify groups with similar traits
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(X)

        # Add cluster information to the DataFrame and save it
        df['Cluster'] = clusters
        clustered_csv_path = '../Embeddings/scm_clustered_embeddings.csv'
        df.to_csv(clustered_csv_path, index=False)
        mlflow.log_artifact(clustered_csv_path, artifact_path="clustered_data")

        # Visualize clusters to understand data segmentation
        plt.figure(figsize=(12, 8))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10')
        plt.colorbar(scatter, ticks=range(5))
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title('K-Means Clusters of SCM Embeddings')
        plt.show()

        # Record the end time of the experiment
        end_time = datetime.now()
        mlflow.log_param("End Time", str(end_time))

        # Function to predict the personality trait based on new text embeddings
        def predict_factor(new_text):
            # Assume `get_embedding` is a function to convert text to embedding
            new_embedding = get_embedding(new_text)
            predicted_factor = clf.predict([new_embedding])
            return label_encoder.inverse_transform(predicted_factor)[0]

        # Demonstrate the model's application on new text data
        test_texts = [
            "Active Energetic Activate Activator",
            "Calm Relaxed Soothing Pacify",
            "Curious Inquisitive Explore Investigator",
            "Organized Meticulous Plan Planner",
            "Moody Anxious Worry Worrier"
        ]

        for text in test_texts:
            predicted_factor = predict_factor(text)
            print(f"Predicted Factor for '{text}': {predicted_factor}")
            param_name = f"Predicted_Factor_for_{text.replace(' ', '_')}"
            mlflow.log_param(param_name, predicted_factor)

    except Exception as e:
        # Log any errors that occur during the model run
        mlflow.log_param("Error", str(e))
        print(f"An error occurred: {e}")
        raise e

    finally:
        # Ensure the MLflow run is properly closed even if an error occurs
        mlflow.end_run()
````

# **Step 13: Create SCM Schema in Neo4j** <a class="anchor" id="SCM_page_4"></a>

````python
import os
from dotenv import load_dotenv
from py2neo import Graph
import pandas as pd
import mlflow
import ast  # Convert string embeddings to lists

# Load environment variables
load_dotenv(override=True)

# MLflow and Neo4j connection settings
NEO4J_URI = os.getenv("NEO4J_URI")
NEO4J_USER = os.getenv("NEO4J_USERNAME") 
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
MLFLOW_TRACKING_URI = os.getenv("MLFLOW_TRACKING_URI")

# Set the MLflow tracking URI
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Set an experiment name
mlflow.set_experiment("SCM")

# Load the dataset
dataset_path = "../Embeddings/scm_embeddings.csv"
scm_df = pd.read_csv(dataset_path)

# Convert Embeddings (if stored as strings)
if 'Embedding' in scm_df.columns:
    scm_df['Embedding'] = scm_df['Embedding'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to create the SCM schema in Neo4j
def create_scm_schema(graph, dataframe):
    for index, row in dataframe.iterrows():
        # Extract data safely
        factor = row.get('Factor', 'Unknown')
        adjective = row.get('Adjective', 'Unknown')
        synonym = row.get('Synonym', 'Unknown')
        verb = row.get('Verb', 'Unknown')
        noun = row.get('Noun', 'Unknown')
        embedding = row.get('Embedding', [])

        # Ensure values exist before inserting
        if not factor or not adjective or not synonym or not verb or not noun:
            continue  # Skip if any critical value is missing

        # Create nodes with SCM labels
        graph.run("MERGE (f:SCM_Factor {name: $factor})", factor=factor)
        graph.run("MERGE (a:SCM_Adjective {name: $adjective})", adjective=adjective)
        graph.run("MERGE (s:SCM_Synonym {name: $synonym})", synonym=synonym)
        graph.run("MERGE (v:SCM_Verb {name: $verb})", verb=verb)
        graph.run("MERGE (n:SCM_Noun {name: $noun})", noun=noun)
        
        # Create relationships with SCM namespacing
        graph.run("""
            MATCH (f:SCM_Factor {name: $factor}), (a:SCM_Adjective {name: $adjective})
            MERGE (f)-[:SCM_HAS_ADJECTIVE]->(a)
        """, factor=factor, adjective=adjective)
        
        graph.run("""
            MATCH (a:SCM_Adjective {name: $adjective}), (s:SCM_Synonym {name: $synonym})
            MERGE (a)-[:SCM_HAS_SYNONYM]->(s)
        """, adjective=adjective, synonym=synonym)
        
        graph.run("""
            MATCH (a:SCM_Adjective {name: $adjective}), (v:SCM_Verb {name: $verb})
            MERGE (a)-[:SCM_HAS_VERB]->(v)
        """, adjective=adjective, verb=verb)
        
        graph.run("""
            MATCH (a:SCM_Adjective {name: $adjective}), (n:SCM_Noun {name: $noun})
            MERGE (a)-[:SCM_HAS_NOUN]->(n)
        """, adjective=adjective, noun=noun)
        
        # Log the embedding if present
        if embedding:
            graph.run("""
                MATCH (f:SCM_Factor {name: $factor}), (a:SCM_Adjective {name: $adjective})
                MERGE (e:SCM_Embedding {value: $embedding})
                MERGE (f)-[:SCM_HAS_EMBEDDING]->(e)
                MERGE (a)-[:SCM_HAS_EMBEDDING]->(e)
            """, factor=factor, adjective=adjective, embedding=embedding)

# Start an MLFlow run
with mlflow.start_run(run_name="Create SCM Schema in Neo4j"):
    try:
        # Log the dataset creation process
        mlflow.log_param("Step", "Create SCM Schema in Neo4j")
        
        # Connect to Neo4j
        graph = Graph(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))
        
        # Create the SCM schema
        create_scm_schema(graph, scm_df)
        
        # Log the dataset as an artifact
        mlflow.log_artifact(dataset_path, artifact_path="datasets")

        # Log the number of rows and columns in the dataset
        mlflow.log_param("Number of Rows", scm_df.shape[0])
        mlflow.log_param("Number of Columns", scm_df.shape[1])
        
        # Print the DataFrame
        print(scm_df.head())

    except Exception as e:
        # Log the error if any occurs
        mlflow.log_param("Error", str(e))
        print(f"‚ùå Error: {e}")
        raise e

    finally:
        # End the MLFlow run
        mlflow.end_run()
````
